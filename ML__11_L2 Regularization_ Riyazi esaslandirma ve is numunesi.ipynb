## Distributions

## **1Ô∏è‚É£ Uniform (B…ôrab…ôr) Paylanma**  
üìå **D√ºstur:**  
- **Diskret halda:**  
  $$
  P(X = x) = \frac{1}{b - a + 1}, \quad a \leq x \leq b
  $$
- **Davamlƒ± halda:**  
  $$
  f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
  $$

üìå **D√ºsturda:**  
- $X$ ‚Üí T…ôsad√ºfi d…ôyi≈ü…ôn  
- $a, b$ ‚Üí Paylanmanƒ±n a≈üaƒüƒ± v…ô yuxarƒ± h…ôddi  
- $P(X = x)$ ‚Üí Hadis…ônin ba≈ü verm…ô ehtimalƒ±  
- $f(x)$ ‚Üí Ehtimal sƒ±xlƒ±q funksiyasƒ± (PDF)  

üìå **ƒ∞zah:** H…ôr bir hadis…ônin eyni ehtimalla ba≈ü verm…ôsini ifad…ô edir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Random …ôd…ôd generasiyasƒ±, Monte Carlo simulyasiyasƒ±.  

---

## **2Ô∏è‚É£ Bernoulli Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0,1\}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí 0 v…ô ya 1 ala bil…ôn t…ôsad√ºfi d…ôyi≈ü…ôn  
- $p$ ‚Üí Uƒüur ehtimalƒ± ($P(X = 1) = p$)  
- $1 - p$ ‚Üí Uƒüursuzluq ehtimalƒ± ($P(X = 0) = 1 - p$)  

üìå **ƒ∞zah:** Bir hadis…ô yalnƒ±z iki m√ºmk√ºn n…ôtic…ôy…ô malik olduqda (ba≈üarƒ± v…ô uƒüursuzluq) istifad…ô olunur.  

üìå **ƒ∞stifad…ô sah…ôsi:** T…ôk bir sƒ±naqda uƒüur/q…ôbul q…ôrarlarƒ± (m…ôs., sikk…ô atƒ±≈üƒ±).  


---

## **3Ô∏è‚É£ Binomial Paylanma**  
üìå **D√ºstur:**  
$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Uƒüurlarƒ±n sayƒ±  
- $n$ ‚Üí T…ôcr√ºb…ôl…ôrin √ºmumi sayƒ±  
- $k$ ‚Üí Uƒüurlarƒ±n sayƒ±  
- $p$ ‚Üí Uƒüur ehtimalƒ±  
- $\binom{n}{k}$ ‚Üí Kombinasiyalar sayƒ± ($\frac{n!}{k!(n-k)!}$)  

üìå **ƒ∞zah:** Bir hadis…ônin m√º…ôyy…ôn sayda ($ n $) sƒ±naqdan sonra ne√ß…ô d…ôf…ô ba≈ü ver…ôc…ôyini g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Anketl…ôr, testl…ôrd…ô d√ºzg√ºn cavablarƒ±n sayƒ±, istehsalatda defektli m…ôhsullarƒ±n hesablanmasƒ±. 

---

## **4Ô∏è‚É£ Geometrik Paylanma**  
üìå **D√ºstur:**  
$$
P(X = k) = (1 - p)^{k-1} p
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí ƒ∞lk uƒüura q…ôd…ôr olan sƒ±naq sayƒ±  
- $k$ ‚Üí ƒ∞lk uƒüura q…ôd…ôr edil…ôn sƒ±naqlarƒ±n sayƒ±  
- $p$ ‚Üí Uƒüur ehtimalƒ±  

üìå **ƒ∞zah:** ƒ∞lk uƒüura q…ôd…ôr olan uƒüursuzluq sayƒ±nƒ± g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** M√º≈üt…ôri tapana q…ôd…ôr edil…ôn z…ôngl…ôrin sayƒ±, ma≈üƒ±nƒ±n sƒ±radan √ßƒ±xana q…ôd…ôr i≈ül…ôm…ô d√∂vrl…ôri.  

---

## **5Ô∏è‚É£ Poisson Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Hadis…ôl…ôrin sayƒ±  
- $\lambda$ ‚Üí Hadis…ônin ba≈üverm…ô orta sayƒ±  
- $k!$ ‚Üí Faktorial ($k$ …ôd…ôdinin faktorialƒ±)  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  

üìå **ƒ∞zah:** M√º…ôyy…ôn zaman aralƒ±ƒüƒ±nda nadir hadis…ôl…ôrin sayƒ±nƒ± g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** √áaƒürƒ± m…ôrk…ôzin…ô daxil olan z…ôngl…ôrin sayƒ±, s…ôhiyy…ôd…ô x…ôst…ôxanaya q…ôbul olunma sayƒ±.  

---

## **6Ô∏è‚É£ Exponential Paylanma**  
üìå **D√ºstur:**  
$$
f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Hadis…ônin ba≈üverm…ô vaxtƒ±  
- $\lambda$ ‚Üí Hadis…ônin orta ba≈ü verm…ô s√ºr…ôti  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  

üìå **ƒ∞zah:** Hadis…ônin ba≈ü verm…ô intervallarƒ±nƒ± g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Ma≈üƒ±nƒ±n sƒ±radan √ßƒ±xma vaxtƒ±, serverd…ô sorƒüularƒ±n g…ôlm…ô intervallarƒ±.

---

## **7Ô∏è‚É£ Normal (Qaus) Paylanma**  
üìå **D√ºstur:**  
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Normal paylanmƒ±≈ü t…ôsad√ºfi d…ôyi≈ü…ôn  
- $\mu$ ‚Üí Ortalamadƒ±r  
- $\sigma$ ‚Üí Standart sapmadƒ±r  
- $\pi$ ‚Üí T…ôqrib…ôn 3.1416 olan riyazi sabit  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  

üìå **ƒ∞zah:** T…ôbii hadis…ôl…ôrin …ôks…ôriyy…ôti normal paylanmaya malikdir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Boy uzunluqlarƒ±, IQ paylanmasƒ±, s…ôhml…ôrin qiym…ôt d…ôyi≈üiklikl…ôri.  


---

## **8Ô∏è‚É£ Polinomial Paylanma**  
üìå **D√ºstur:**  
$$
P(X_1 = x_1, ..., X_k = x_k) = \frac{n!}{x_1! x_2! ... x_k!} p_1^{x_1} p_2^{x_2} ... p_k^{x_k}
$$

üìå **D√ºsturda:**  
- $X_i$ ‚Üí H…ôr bir f…ôrqli n…ôtic…ônin t…ôkrarlanma sayƒ±  
- $n$ ‚Üí √úmumi sƒ±naq sayƒ±  
- $p_i$ ‚Üí H…ôr n…ôtic…ônin ba≈üverm…ô ehtimalƒ±  
- $x_i!$ ‚Üí H…ôr n…ôtic…ônin faktorialƒ±  

üìå **ƒ∞zah:** Polinomial paylanma, bir hadis…ônin m√º…ôyy…ôn sayda (ùëõ) sƒ±naqdan sonra m√ºxt…ôlif n…ôtic…ôl…ôrin (m…ôs…ôl…ôn, h…ôr sƒ±naqda bir ne√ß…ô f…ôrqli n…ôtic…ô ola bil…ôr) ne√ß…ô d…ôf…ô ba≈ü ver…ôc…ôyini g√∂st…ôrir. Bu paylanma, h…ôr bir n…ôtic…ô √º√ß√ºn ehtimallarƒ± n…ôz…ôr…ô alaraq, h…ôr bir n…ôtic…ônin m√ºmk√ºn sayda ba≈ü verm…ô ehtimallarƒ±nƒ± hesablamaƒüa imkan verir.

üìå **ƒ∞stifad…ô sah…ôsi:** Se√ßkil…ôrd…ô m√ºxt…ôlif partiyalara s…ôs paylanmasƒ±, m√º≈üt…ôril…ôrin m√ºxt…ôlif m…ôhsullarƒ± se√ßm…ô ehtimallarƒ±.

## **9Ô∏è‚É£ T-student Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
f(t) = \frac{\Gamma(\frac{\nu+1}{2})}{\sqrt{\nu\pi} \Gamma(\frac{\nu}{2})} (1 + \frac{t^2}{\nu})^{-\frac{\nu+1}{2}}
$$

üìå **D√ºsturda:**  
- $t$ ‚Üí T-statistikasƒ±  
- $\nu$ ‚Üí Azadlƒ±q d…ôr…ôc…ôl…ôri  
- $\Gamma(x)$ ‚Üí Gamma funksiyasƒ±  


üìå **ƒ∞zah:** Ki√ßik n√ºmun…ôl…ôrd…ô ortalamanƒ±n normallƒ±ƒüƒ±nƒ± qiym…ôtl…ôndirm…ôk √º√ß√ºn istifad…ô olunur.  

üìå **ƒ∞stifad…ô sah…ôsi:** A/B testl…ôri, hipotez yoxlamalarƒ±.  

---

## **üîü Chi-Square Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
f(x) = \frac{x^{\frac{k}{2} - 1} e^{-x/2}}{2^{k/2} \Gamma(k/2)}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Chi-square t…ôsad√ºfi d…ôyi≈ü…ôni  
- $k$ ‚Üí Azadlƒ±q d…ôr…ôc…ôl…ôri  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  
- $\Gamma(x)$ ‚Üí Gamma funksiyasƒ±  

üìå **ƒ∞zah:** Variansƒ±n paylanmasƒ±nƒ± qiym…ôtl…ôndirir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Qiym…ôtl…ôndirm…ô testl…ôri, baƒüƒ±msƒ±zlƒ±q analizl…ôri.  

---

## **1Ô∏è‚É£1Ô∏è‚É£ F Paylanmasƒ± (Fisher's Distribution)**  
üìå **D√ºstur:**  
$$
f(x) = \frac{\Gamma(\frac{d_1 + d_2}{2})}{\Gamma(\frac{d_1}{2}) \Gamma(\frac{d_2}{2})} \left( \frac{d_1}{d_2} \right)^{\frac{d_1}{2}} \frac{x^{\frac{d_1}{2} - 1}}{\left( 1 + \frac{d_1}{d_2}x \right)^{\frac{d_1 + d_2}{2}}}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí F-paylanmƒ±≈ü t…ôsad√ºfi d…ôyi≈ü…ôn  
- $d_1, d_2$ ‚Üí Azadlƒ±q d…ôr…ôc…ôl…ôri  
- $\Gamma(x)$ ‚Üí Gamma funksiyasƒ±  

üìå **ƒ∞zah:** ƒ∞ki variansƒ±n nisb…ôtini yoxlamaq √º√ß√ºn istifad…ô olunur.  

üìå **ƒ∞stifad…ô sah…ôsi:** ANOVA testl…ôri.

## Poisson v…ô Normal paylanmalar arasƒ±nda m√º…ôyy…ôn …ôlaq…ô var

Poisson v…ô Normal paylanmalar arasƒ±nda m√º…ôyy…ôn …ôlaq…ô var v…ô Poisson paylanmasƒ± b…ôzi hallarda Normal paylanmaya b…ônz…ôy…ô bil…ôr. Bunun s…ôb…ôbl…ôrini a≈üaƒüƒ±dakƒ± kimi izah ed…ô bil…ôrik:  

### **1Ô∏è‚É£ M…ôrk…ôzi Limit Teoremi**  
Poisson paylanmasƒ± diskret bir paylanmadƒ±r, lakin **Œª** (orta ba≈üverm…ô sayƒ±) b√∂y√ºd√ºkc…ô, onun qrafiki Normal paylanmaya daha √ßox ox≈üamaƒüa ba≈ülayƒ±r. Bu, **M…ôrk…ôzi Limit Teoremi** il…ô izah olunur. ∆èg…ôr **Œª b√∂y√ºkd√ºrs…ô (Œª > 30)**, Poisson paylanmasƒ± **$ N(\lambda, \sqrt{\lambda}) $** Normal paylanmasƒ±na yaxƒ±nla≈üƒ±r.  

### **2Ô∏è‚É£ Simmetriya Artƒ±mƒ±**  
Ki√ßik **Œª** d…ôy…ôrl…ôrind…ô Poisson paylanmasƒ±nƒ±n qrafiki saƒüa …ôyilmi≈ü olur (√ßarpƒ±qlƒ±q var). Lakin **Œª b√∂y√ºd√ºkc…ô**, paylanma daha simmetrik olur v…ô Normal paylanmaya daha √ßox b…ônz…ôyir.  

### **3Ô∏è‚É£ T…ôbi…ôtd…ô Real T…ôtbiql…ôri**  
- Poisson paylanmasƒ± nadir hadis…ôl…ôrin ba≈ü verm…ô sayƒ±nƒ± √∂l√ß√ºr (m…ôs…ôl…ôn, √ßaƒürƒ± m…ôrk…ôzin…ô daxil olan z…ôngl…ôr).  
- Lakin, …ôg…ôr bu tip nadir hadis…ôl…ôr √ßox tez-tez ba≈ü verirs…ô (y…ôni **Œª** b√∂y√ºkd√ºrs…ô), onda bu hadis…ôl…ôrin sayƒ± t…ôxmin…ôn Normal paylanmaya uyƒüun g…ôlir.  

**Qrafikl…ôrd…ô bunu m√º≈üahid…ô etm…ôk √º√ß√ºn Poisson paylanmasƒ±nƒ±n Œª d…ôy…ôrini artƒ±raraq yoxlaya bil…ôrs…ôn.**  

∆èg…ôr qrafikd…ô f…ôrqi a√ßƒ±q g√∂rm…ôk ist…ôyirs…ôns…ô, **ki√ßik Œª d…ôy…ôrl…ôri (m…ôs…ôl…ôn, Œª=5) √º√ß√ºn Poisson paylanmasƒ±nƒ±n qrafikin…ô bax**, sonra is…ô **b√∂y√ºk Œª d…ôy…ôrl…ôri (m…ôs…ôl…ôn, Œª=50) √º√ß√ºn yoxla**. Œª b√∂y√ºd√ºkc…ô Poisson paylanmasƒ± Normal paylanmaya daha √ßox ox≈üayacaq. üöÄ

## `np.polyfit(x_train, y_train, 3)` n…ôdir?

`np.polyfit(x_train, y_train, 3)` funksiyasƒ± **NumPy** kitabxanasƒ±ndan istifad…ô ed…ôr…ôk veril…ôn `x_train` v…ô `y_train` veril…ônl…ôr d…ôstin…ô uyƒüun olaraq 3-c√º d…ôr…ôc…ôli bir polinomu yerin…ô yetirir. Bu funksiya a≈üaƒüƒ±dakƒ±larƒ± h…ôyata ke√ßirir:

1. **Polinom ∆èmsallarƒ±nƒ± Hesablamaq**: `np.polyfit()` veril…ônl…ôri …ôn yax≈üƒ± ≈ü…ôkild…ô uyƒüunla≈üdƒ±rmaq √º√ß√ºn polinomial regresiya yerin…ô yetirir. `3` d…ôr…ôc…ôsi il…ô uyƒüunla≈üdƒ±rma aparƒ±lacaq, y…ôni √º√ß√ºnc√º d…ôr…ôc…ôli (kubik) polinom qurulacaq.

2. **Yalnƒ±z Polinomun ∆èmsallarƒ±nƒ± Qaytarƒ±r**: Bu funksiya, …ôn yax≈üƒ± uyƒüunla≈üan polinomun …ômsallarƒ±nƒ± (koeffisentl…ôrini) geri qaytaracaq. N…ôtic…ô bir siyahƒ± (v…ô ya numpy array) olacaq ki, bu da 3-c√º d…ôr…ôc…ôli polinomun …ômsallarƒ±nƒ± ehtiva edir. Polinom a≈üaƒüƒ±dakƒ± formadadƒ±r:

$$
P(x) = a_3x^3 + a_2x^2 + a_1x + a_0
$$

Burada $a_3$, $a_2$, $a_1$, $a_0$ polinomun …ômsallarƒ± olacaq v…ô bu …ômsallar `np.polyfit()` t…ôr…ôfind…ôn hesablanacaq.

## Python dilind…ô  izah

````python
# NumPy kitabxanasƒ±nƒ± √ßaƒüƒ±rƒ±r
import numpy as np

# Verilmi≈ü x d…ôy…ôri √º√ß√ºn polinomial ifad…ôni hesablayƒ±r
# res = k0*x^n + k1*x^(n-1) + ... + kn*x^0
def predict_poly(x, koeff):
    res = 0
    xx = [x ** (len(koeff) - n - 1) for n in range(len(koeff))]  # H…ôr d…ôr…ôc…ô √º√ß√ºn x^n hesablayƒ±r

    for i, k in enumerate(koeff):  # H…ôr …ômsalƒ± uyƒüun x d…ôr…ôc…ôsi il…ô vurur
        res += k * xx[i]

    return res

# 0-dan 10-a q…ôd…ôr 0.1 addƒ±mlƒ± x d…ôy…ôrl…ôri yaradƒ±r
x = np.arange(0, 10.1, 0.1)
# Funksiya: y = 1 / (1 + 10 * x^2) riyazi d√ºsturu il…ô y d…ôy…ôrl…ôrini hesablayƒ±r
y = 1 / (1 + 10 * np.square(x))

# H…ôr ikinci elementi se√ß…ôr…ôk t…ôlim (train) √º√ß√ºn x v…ô y d…ôy…ôrl…ôrini ayƒ±rƒ±r
x_train, y_train = x[::2], y[::2]

# M…ôlumatƒ±n √ºmumi sayƒ±nƒ± m√º…ôyy…ôn edir
N = len(x)

# T…ôlim m…ôlumatlarƒ±na uyƒüun 10-cu d…ôr…ôc…ôli polinomial funksiyanƒ±n …ômsallarƒ±nƒ± tapƒ±r
z_train = np.polyfit(x_train, y_train, 3)  # Min kvadrat √ºsulu il…ô polinom uyƒüunla≈üdƒ±rƒ±r
# ∆èld…ô olunan …ômsallarƒ± √ßap edir
print(z_train)
````

## Yeni Proqnoz bir …ôd…ôd data √º√ß√ºn def() function

````python
def predict_poly(x=15, koeff=z_train):
    res = 0
    xx = [x ** (len(koeff) - n - 1) for n in range(len(koeff))]  # H…ôr d…ôr…ôc…ô √º√ß√ºn x^n hesablayƒ±r

    for i, k in enumerate(koeff):  # H…ôr …ômsalƒ± uyƒüun x d…ôr…ôc…ôsi il…ô vurur
        res += k * xx[i]

    return res

predict_poly()
````
