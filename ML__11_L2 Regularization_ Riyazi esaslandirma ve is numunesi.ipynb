## Distributions

## **1Ô∏è‚É£ Uniform (B…ôrab…ôr) Paylanma**  
üìå **D√ºstur:**  
- **Diskret halda:**  
  $$
  P(X = x) = \frac{1}{b - a + 1}, \quad a \leq x \leq b
  $$
- **Davamlƒ± halda:**  
  $$
  f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
  $$

üìå **D√ºsturda:**  
- $X$ ‚Üí T…ôsad√ºfi d…ôyi≈ü…ôn  
- $a, b$ ‚Üí Paylanmanƒ±n a≈üaƒüƒ± v…ô yuxarƒ± h…ôddi  
- $P(X = x)$ ‚Üí Hadis…ônin ba≈ü verm…ô ehtimalƒ±  
- $f(x)$ ‚Üí Ehtimal sƒ±xlƒ±q funksiyasƒ± (PDF)  

üìå **ƒ∞zah:** H…ôr bir hadis…ônin eyni ehtimalla ba≈ü verm…ôsini ifad…ô edir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Random …ôd…ôd generasiyasƒ±, Monte Carlo simulyasiyasƒ±.  

---

## **2Ô∏è‚É£ Bernoulli Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0,1\}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí 0 v…ô ya 1 ala bil…ôn t…ôsad√ºfi d…ôyi≈ü…ôn  
- $p$ ‚Üí Uƒüur ehtimalƒ± ($P(X = 1) = p$)  
- $1 - p$ ‚Üí Uƒüursuzluq ehtimalƒ± ($P(X = 0) = 1 - p$)  

üìå **ƒ∞zah:** Bir hadis…ô yalnƒ±z iki m√ºmk√ºn n…ôtic…ôy…ô malik olduqda (ba≈üarƒ± v…ô uƒüursuzluq) istifad…ô olunur.  

üìå **ƒ∞stifad…ô sah…ôsi:** T…ôk bir sƒ±naqda uƒüur/q…ôbul q…ôrarlarƒ± (m…ôs., sikk…ô atƒ±≈üƒ±).  


---

## **3Ô∏è‚É£ Binomial Paylanma**  
üìå **D√ºstur:**  
$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Uƒüurlarƒ±n sayƒ±  
- $n$ ‚Üí T…ôcr√ºb…ôl…ôrin √ºmumi sayƒ±  
- $k$ ‚Üí Uƒüurlarƒ±n sayƒ±  
- $p$ ‚Üí Uƒüur ehtimalƒ±  
- $\binom{n}{k}$ ‚Üí Kombinasiyalar sayƒ± ($\frac{n!}{k!(n-k)!}$)  

üìå **ƒ∞zah:** Bir hadis…ônin m√º…ôyy…ôn sayda ($ n $) sƒ±naqdan sonra ne√ß…ô d…ôf…ô ba≈ü ver…ôc…ôyini g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Anketl…ôr, testl…ôrd…ô d√ºzg√ºn cavablarƒ±n sayƒ±, istehsalatda defektli m…ôhsullarƒ±n hesablanmasƒ±. 

---

## **4Ô∏è‚É£ Geometrik Paylanma**  
üìå **D√ºstur:**  
$$
P(X = k) = (1 - p)^{k-1} p
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí ƒ∞lk uƒüura q…ôd…ôr olan sƒ±naq sayƒ±  
- $k$ ‚Üí ƒ∞lk uƒüura q…ôd…ôr edil…ôn sƒ±naqlarƒ±n sayƒ±  
- $p$ ‚Üí Uƒüur ehtimalƒ±  

üìå **ƒ∞zah:** ƒ∞lk uƒüura q…ôd…ôr olan uƒüursuzluq sayƒ±nƒ± g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** M√º≈üt…ôri tapana q…ôd…ôr edil…ôn z…ôngl…ôrin sayƒ±, ma≈üƒ±nƒ±n sƒ±radan √ßƒ±xana q…ôd…ôr i≈ül…ôm…ô d√∂vrl…ôri.  

---

## **5Ô∏è‚É£ Poisson Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Hadis…ôl…ôrin sayƒ±  
- $\lambda$ ‚Üí Hadis…ônin ba≈üverm…ô orta sayƒ±  
- $k!$ ‚Üí Faktorial ($k$ …ôd…ôdinin faktorialƒ±)  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  

üìå **ƒ∞zah:** M√º…ôyy…ôn zaman aralƒ±ƒüƒ±nda nadir hadis…ôl…ôrin sayƒ±nƒ± g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** √áaƒürƒ± m…ôrk…ôzin…ô daxil olan z…ôngl…ôrin sayƒ±, s…ôhiyy…ôd…ô x…ôst…ôxanaya q…ôbul olunma sayƒ±.  

---

## **6Ô∏è‚É£ Exponential Paylanma**  
üìå **D√ºstur:**  
$$
f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Hadis…ônin ba≈üverm…ô vaxtƒ±  
- $\lambda$ ‚Üí Hadis…ônin orta ba≈ü verm…ô s√ºr…ôti  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  

üìå **ƒ∞zah:** Hadis…ônin ba≈ü verm…ô intervallarƒ±nƒ± g√∂st…ôrir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Ma≈üƒ±nƒ±n sƒ±radan √ßƒ±xma vaxtƒ±, serverd…ô sorƒüularƒ±n g…ôlm…ô intervallarƒ±.

---

## **7Ô∏è‚É£ Normal (Qaus) Paylanma**  
üìå **D√ºstur:**  
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Normal paylanmƒ±≈ü t…ôsad√ºfi d…ôyi≈ü…ôn  
- $\mu$ ‚Üí Ortalamadƒ±r  
- $\sigma$ ‚Üí Standart sapmadƒ±r  
- $\pi$ ‚Üí T…ôqrib…ôn 3.1416 olan riyazi sabit  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  

üìå **ƒ∞zah:** T…ôbii hadis…ôl…ôrin …ôks…ôriyy…ôti normal paylanmaya malikdir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Boy uzunluqlarƒ±, IQ paylanmasƒ±, s…ôhml…ôrin qiym…ôt d…ôyi≈üiklikl…ôri.  


---

## **8Ô∏è‚É£ Polinomial Paylanma**  
üìå **D√ºstur:**  
$$
P(X_1 = x_1, ..., X_k = x_k) = \frac{n!}{x_1! x_2! ... x_k!} p_1^{x_1} p_2^{x_2} ... p_k^{x_k}
$$

üìå **D√ºsturda:**  
- $X_i$ ‚Üí H…ôr bir f…ôrqli n…ôtic…ônin t…ôkrarlanma sayƒ±  
- $n$ ‚Üí √úmumi sƒ±naq sayƒ±  
- $p_i$ ‚Üí H…ôr n…ôtic…ônin ba≈üverm…ô ehtimalƒ±  
- $x_i!$ ‚Üí H…ôr n…ôtic…ônin faktorialƒ±  

üìå **ƒ∞zah:** Polinomial paylanma, bir hadis…ônin m√º…ôyy…ôn sayda (ùëõ) sƒ±naqdan sonra m√ºxt…ôlif n…ôtic…ôl…ôrin (m…ôs…ôl…ôn, h…ôr sƒ±naqda bir ne√ß…ô f…ôrqli n…ôtic…ô ola bil…ôr) ne√ß…ô d…ôf…ô ba≈ü ver…ôc…ôyini g√∂st…ôrir. Bu paylanma, h…ôr bir n…ôtic…ô √º√ß√ºn ehtimallarƒ± n…ôz…ôr…ô alaraq, h…ôr bir n…ôtic…ônin m√ºmk√ºn sayda ba≈ü verm…ô ehtimallarƒ±nƒ± hesablamaƒüa imkan verir.

üìå **ƒ∞stifad…ô sah…ôsi:** Se√ßkil…ôrd…ô m√ºxt…ôlif partiyalara s…ôs paylanmasƒ±, m√º≈üt…ôril…ôrin m√ºxt…ôlif m…ôhsullarƒ± se√ßm…ô ehtimallarƒ±.

## **9Ô∏è‚É£ T-student Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
f(t) = \frac{\Gamma(\frac{\nu+1}{2})}{\sqrt{\nu\pi} \Gamma(\frac{\nu}{2})} (1 + \frac{t^2}{\nu})^{-\frac{\nu+1}{2}}
$$

üìå **D√ºsturda:**  
- $t$ ‚Üí T-statistikasƒ±  
- $\nu$ ‚Üí Azadlƒ±q d…ôr…ôc…ôl…ôri  
- $\Gamma(x)$ ‚Üí Gamma funksiyasƒ±  


üìå **ƒ∞zah:** Ki√ßik n√ºmun…ôl…ôrd…ô ortalamanƒ±n normallƒ±ƒüƒ±nƒ± qiym…ôtl…ôndirm…ôk √º√ß√ºn istifad…ô olunur.  

üìå **ƒ∞stifad…ô sah…ôsi:** A/B testl…ôri, hipotez yoxlamalarƒ±.  

---

## **üîü Chi-Square Paylanmasƒ±**  
üìå **D√ºstur:**  
$$
f(x) = \frac{x^{\frac{k}{2} - 1} e^{-x/2}}{2^{k/2} \Gamma(k/2)}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí Chi-square t…ôsad√ºfi d…ôyi≈ü…ôni  
- $k$ ‚Üí Azadlƒ±q d…ôr…ôc…ôl…ôri  
- $e$ ‚Üí T…ôqrib…ôn 2.718 olan riyazi sabit  
- $\Gamma(x)$ ‚Üí Gamma funksiyasƒ±  

üìå **ƒ∞zah:** Variansƒ±n paylanmasƒ±nƒ± qiym…ôtl…ôndirir.  

üìå **ƒ∞stifad…ô sah…ôsi:** Qiym…ôtl…ôndirm…ô testl…ôri, baƒüƒ±msƒ±zlƒ±q analizl…ôri.  

---

## **1Ô∏è‚É£1Ô∏è‚É£ F Paylanmasƒ± (Fisher's Distribution)**  
üìå **D√ºstur:**  
$$
f(x) = \frac{\Gamma(\frac{d_1 + d_2}{2})}{\Gamma(\frac{d_1}{2}) \Gamma(\frac{d_2}{2})} \left( \frac{d_1}{d_2} \right)^{\frac{d_1}{2}} \frac{x^{\frac{d_1}{2} - 1}}{\left( 1 + \frac{d_1}{d_2}x \right)^{\frac{d_1 + d_2}{2}}}
$$

üìå **D√ºsturda:**  
- $X$ ‚Üí F-paylanmƒ±≈ü t…ôsad√ºfi d…ôyi≈ü…ôn  
- $d_1, d_2$ ‚Üí Azadlƒ±q d…ôr…ôc…ôl…ôri  
- $\Gamma(x)$ ‚Üí Gamma funksiyasƒ±  

üìå **ƒ∞zah:** ƒ∞ki variansƒ±n nisb…ôtini yoxlamaq √º√ß√ºn istifad…ô olunur.  

üìå **ƒ∞stifad…ô sah…ôsi:** ANOVA testl…ôri.

## Poisson v…ô Normal paylanmalar arasƒ±nda m√º…ôyy…ôn …ôlaq…ô var

Poisson v…ô Normal paylanmalar arasƒ±nda m√º…ôyy…ôn …ôlaq…ô var v…ô Poisson paylanmasƒ± b…ôzi hallarda Normal paylanmaya b…ônz…ôy…ô bil…ôr. Bunun s…ôb…ôbl…ôrini a≈üaƒüƒ±dakƒ± kimi izah ed…ô bil…ôrik:  

### **1Ô∏è‚É£ M…ôrk…ôzi Limit Teoremi**  
Poisson paylanmasƒ± diskret bir paylanmadƒ±r, lakin **Œª** (orta ba≈üverm…ô sayƒ±) b√∂y√ºd√ºkc…ô, onun qrafiki Normal paylanmaya daha √ßox ox≈üamaƒüa ba≈ülayƒ±r. Bu, **M…ôrk…ôzi Limit Teoremi** il…ô izah olunur. ∆èg…ôr **Œª b√∂y√ºkd√ºrs…ô (Œª > 30)**, Poisson paylanmasƒ± **$ N(\lambda, \sqrt{\lambda}) $** Normal paylanmasƒ±na yaxƒ±nla≈üƒ±r.  

### **2Ô∏è‚É£ Simmetriya Artƒ±mƒ±**  
Ki√ßik **Œª** d…ôy…ôrl…ôrind…ô Poisson paylanmasƒ±nƒ±n qrafiki saƒüa …ôyilmi≈ü olur (√ßarpƒ±qlƒ±q var). Lakin **Œª b√∂y√ºd√ºkc…ô**, paylanma daha simmetrik olur v…ô Normal paylanmaya daha √ßox b…ônz…ôyir.  

### **3Ô∏è‚É£ T…ôbi…ôtd…ô Real T…ôtbiql…ôri**  
- Poisson paylanmasƒ± nadir hadis…ôl…ôrin ba≈ü verm…ô sayƒ±nƒ± √∂l√ß√ºr (m…ôs…ôl…ôn, √ßaƒürƒ± m…ôrk…ôzin…ô daxil olan z…ôngl…ôr).  
- Lakin, …ôg…ôr bu tip nadir hadis…ôl…ôr √ßox tez-tez ba≈ü verirs…ô (y…ôni **Œª** b√∂y√ºkd√ºrs…ô), onda bu hadis…ôl…ôrin sayƒ± t…ôxmin…ôn Normal paylanmaya uyƒüun g…ôlir.  

**Qrafikl…ôrd…ô bunu m√º≈üahid…ô etm…ôk √º√ß√ºn Poisson paylanmasƒ±nƒ±n Œª d…ôy…ôrini artƒ±raraq yoxlaya bil…ôrs…ôn.**  

∆èg…ôr qrafikd…ô f…ôrqi a√ßƒ±q g√∂rm…ôk ist…ôyirs…ôns…ô, **ki√ßik Œª d…ôy…ôrl…ôri (m…ôs…ôl…ôn, Œª=5) √º√ß√ºn Poisson paylanmasƒ±nƒ±n qrafikin…ô bax**, sonra is…ô **b√∂y√ºk Œª d…ôy…ôrl…ôri (m…ôs…ôl…ôn, Œª=50) √º√ß√ºn yoxla**. Œª b√∂y√ºd√ºkc…ô Poisson paylanmasƒ± Normal paylanmaya daha √ßox ox≈üayacaq. üöÄ

## `np.polyfit(x_train, y_train, 3)` n…ôdir?

`np.polyfit(x_train, y_train, 3)` funksiyasƒ± **NumPy** kitabxanasƒ±ndan istifad…ô ed…ôr…ôk veril…ôn `x_train` v…ô `y_train` veril…ônl…ôr d…ôstin…ô uyƒüun olaraq 3-c√º d…ôr…ôc…ôli bir polinomu yerin…ô yetirir. Bu funksiya a≈üaƒüƒ±dakƒ±larƒ± h…ôyata ke√ßirir:

1. **Polinom ∆èmsallarƒ±nƒ± Hesablamaq**: `np.polyfit()` veril…ônl…ôri …ôn yax≈üƒ± ≈ü…ôkild…ô uyƒüunla≈üdƒ±rmaq √º√ß√ºn polinomial regresiya yerin…ô yetirir. `3` d…ôr…ôc…ôsi il…ô uyƒüunla≈üdƒ±rma aparƒ±lacaq, y…ôni √º√ß√ºnc√º d…ôr…ôc…ôli (kubik) polinom qurulacaq.

2. **Yalnƒ±z Polinomun ∆èmsallarƒ±nƒ± Qaytarƒ±r**: Bu funksiya, …ôn yax≈üƒ± uyƒüunla≈üan polinomun …ômsallarƒ±nƒ± (koeffisentl…ôrini) geri qaytaracaq. N…ôtic…ô bir siyahƒ± (v…ô ya numpy array) olacaq ki, bu da 3-c√º d…ôr…ôc…ôli polinomun …ômsallarƒ±nƒ± ehtiva edir. Polinom a≈üaƒüƒ±dakƒ± formadadƒ±r:

Burada $a_3$, $a_2$, $a_1$, $a_0$ polinomun …ômsallarƒ± olacaq v…ô bu …ômsallar `np.polyfit()` t…ôr…ôfind…ôn hesablanacaq.

### Misal:

### N…ôtic…ô:

## L2 c…ôrim…ôl…ôndirm…ô (L2 regularization) v…ô ya Ridge Regression

Bu d√ºstur **L2 c…ôrim…ôl…ôndirm…ô (L2 regularization) v…ô ya Ridge Regression** √º√ß√ºn obyektiv funksiyanƒ± ifad…ô edir. Bu metod **overfitting-in qar≈üƒ±sƒ±nƒ± almaq** v…ô **modelin √ºmumil…ô≈üdirm…ô qabiliyy…ôtini artƒ±rmaq** √º√ß√ºn istifad…ô olunur.

---

### **D√ºsturun a√ßƒ±lƒ±≈üƒ± v…ô izahƒ±:**
$$
\tilde{L}_i(\omega) = L_i(\omega) + \frac{\lambda}{2} \|\omega\|^2 = L_i(\omega) + \frac{\lambda}{2} \sum_{j=1}^{n} \omega_j^2 \to \min_{\omega}
$$

#### **ƒ∞zah:**
Bu ifad…ô **orijinal itki funksiyasƒ±na …ôlav…ô olaraq c…ôrim…ôl…ôndirm…ô termini …ôlav…ô ed…ôn** optimalla≈üdƒ±rma problemidir. 

- **$L_i(\omega)$** ‚Äì Orijinal itki funksiyasƒ±dƒ±r, m…ôs…ôl…ôn, **Mean Squared Error (MSE)** kimi ola bil…ôr:
  $$
  L_i(\omega) = \sum_{i=1}^{m} (y_i - f(x_i, \omega))^2
  $$
- **$\frac{\lambda}{2} \|\omega\|^2$** ‚Äì Bu, **c…ôrim…ôl…ôndirm…ô (regularization) termini**dir. Burada:
  - **$\lambda$** ‚Äì **Regularization (c…ôrim…ôl…ôndirm…ô) parametri**dir. Modelin n…ô q…ôd…ôr c…ôrim…ôl…ôn…ôc…ôyini m√º…ôyy…ôn edir.
  - **$\|\omega\|^2 = \sum_{j=1}^{n} \omega_j^2$** ‚Äì Modelin b√ºt√ºn …ômsallarƒ±nƒ±n kvadratlarƒ±nƒ±n c…ômi. Bu, **L2 normasƒ±dƒ±r**.
- **$\to \min_{\omega}$** ‚Äì M…ôqs…ôdimiz bu funksiyanƒ± minimalla≈üdƒ±rmaqdƒ±r.

---

### **Niy…ô bunu edirik?**
L2 regularization modelin √ßox b√∂y√ºk …ômsallara sahib olmasƒ±nƒ±n qar≈üƒ±sƒ±nƒ± alƒ±r v…ô bu da **overfitting-i azaldƒ±r**. Bunun …ôsas s…ôb…ôbi:

1. **B√∂y√ºk …ômsallarƒ± c…ôzalandƒ±rƒ±r:** ∆èg…ôr bir modelin …ômsallarƒ± √ßox b√∂y√ºkd√ºrs…ô, bu, veril…ônl…ôr…ô h…ôddind…ôn artƒ±q uyƒüunla≈ümaq (overfitting) riskini artƒ±rƒ±r.
2. **Daha sabit model yaradƒ±r:** Ki√ßik …ômsallar modelin **ki√ßik d…ôyi≈üiklikl…ôr…ô h…ôssaslƒ±ƒüƒ±nƒ± azaldƒ±r** v…ô **√ºmumil…ô≈üdirm…ô qabiliyy…ôtini artƒ±rƒ±r**.
3. **Kovariasiya problemini h…ôll edir:** Multikolinearlƒ±q (kovariasiya y√ºks…ôk olduqda) olduqda Ridge Regression bu problemi yum≈üaldƒ±r v…ô **daha stabill…ô≈ümi≈ü model** qurmaƒüa k√∂m…ôk edir.

---

### **$\lambda$ (regularization parametri) nec…ô se√ßilir?**
$\lambda$-nƒ±n se√ßimi **√ßox ki√ßik v…ô ya √ßox b√∂y√ºk olmamalƒ±dƒ±r**. ∆èn √ßox istifad…ô olunan d…ôy…ôrl…ôr:  
- **Ki√ßik d…ôy…ôrl…ôr ($\lambda \approx 0.01, 0.1$)** ‚Üí ∆èg…ôr c…ôrim…ôl…ôndirm…ô azdƒ±rsa, model h…ôl…ô d…ô **overfitting** ed…ô bil…ôr.  
- **Orta d…ôy…ôrl…ôr ($\lambda \approx 1$)** ‚Üí Modeli balanslƒ± saxlayƒ±r.  
- **B√∂y√ºk d…ôy…ôrl…ôr ($\lambda \approx 10, 100$)** ‚Üí Modeli √ßox c…ôrim…ôl…ôndirir v…ô …ômsallar **sƒ±fƒ±ra yaxƒ±nla≈üƒ±r** ‚Üí **underfitting** problemi yarana bil…ôr.  

**Optimal $\lambda$ d…ôy…ôrini t…ôyin etm…ôk √º√ß√ºn Cross-Validation (√áarpaz Doƒürulama) v…ô ya Grid Search istifad…ô olunur.**

---

### **Ridge Regression t…ôtbiqi (Python n√ºmun…ôsi):**
```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
import numpy as np

# M…ôlumat n√ºmun…ôsi

y = np.array([2.2, 2.8, 3.6, 4.5])

# Lambda √º√ß√ºn namiz…ôd d…ôy…ôrl…ôr
lambda_values = [0.01, 0.1, 1, 10, 100]

# GridSearchCV il…ô optimal lambda tapmaq
ridge = Ridge()
grid = GridSearchCV(ridge, {'alpha': lambda_values}, cv=5)
grid.fit(X, y)

# ∆èn yax≈üƒ± lambda d…ôy…ôri
print("Optimal lambda:", grid.best_params_['alpha'])
```

Burada `alpha` **c…ôrim…ôl…ôndirm…ô …ômsalƒ±nƒ±** t…ôyin edir. ∆èg…ôr `alpha` √ßox b√∂y√ºk olarsa, modelin …ômsallarƒ± sƒ±fƒ±ra yaxƒ±nla≈üacaq. ∆èg…ôr `alpha = 0` olarsa, adi **Linear Regression** modelin…ô qayƒ±dƒ±rƒ±q.

---

### **N…ôtic…ô:**
Bu d√ºstur **L2 Regularization (Ridge Regression)** √º√ß√ºn obyektiv funksiyanƒ± ifad…ô edir. **∆èsas m…ôqs…ôd modeli stabil saxlamaq v…ô overfitting-in qar≈üƒ±sƒ±nƒ± almaqdƒ±r.** Python-da GridSearchCV istifad…ô ed…ôr…ôk **optimal $\lambda$ d…ôy…ôrini avtomatik tapmaq m√ºmk√ºnd√ºr.**

## T…ôrs matrisin tapƒ±lmasƒ± √º√ß√ºn ≈ü…ôrt

∆èslind…ô, **t…ôrs matrisin tapƒ±lmasƒ±** √º√ß√ºn **≈ü…ôrt** odur ki, matris **kvadrat** v…ô **determinantƒ± sƒ±fƒ±rdan f…ôrqli** olsun.  

S…ônin dediyin v…ôziyy…ôtd…ô **iki f…ôrqli hal** var:  

1. **S…ôtir sayƒ± s√ºtun sayƒ±ndan azdƒ±r** (Under-determined system):  
   - Bu halda **$X$** matrisi **kvadrat deyil**, ona g√∂r…ô d…ô t…ôrsini tapmaq m√ºmk√ºn deyil.  
   - Bundan …ôlav…ô, **$X^T X$** matrisini hesablasaq, o da tam r√ºtb…ôli olmaya bil…ôr (rank-deficient), bu da onun t…ôrsini tapmaq problemini yaradƒ±r.  
   
2. **S√ºtun sayƒ± s…ôtir sayƒ±ndan azdƒ±r** (Over-determined system):  
   - Burada da **$X$** matrisi kvadrat deyil, ona g√∂r…ô d…ô t…ôrsi birba≈üa tapƒ±la bilmir.  
   - Amma **$X^T X$** ad…ôt…ôn tam r√ºtb…ôli olur v…ô t…ôrsi tapƒ±la bil…ôr. Buna g√∂r…ô d…ô, normal t…ônlikl…ôrd…ô **$(X^T X)^{-1} X^T y$** istifad…ô ed…ôr…ôk h…ôll tapmaq m√ºmk√ºnd√ºr.  

### N…ôtic…ô:  
- ∆èg…ôr **s…ôtir sayƒ± azdƒ±rsa $(m < n)$**, t…ôrs tapmaq √ßox vaxt m√ºmk√ºn olmur, √ß√ºnki **$X^T X$** singulyar olur.  
- ∆èg…ôr **s√ºtun sayƒ± azdƒ±rsa $(m > n)$**, t…ôrs birba≈üa tapƒ±lmasa da, **normal t…ônlikl…ôr vasit…ôsil…ô** h…ôll tapmaq m√ºmk√ºnd√ºr.  

Ona g√∂r…ô d…ô "s…ôtir sayƒ± az olduqda t…ôrs tapmaq olmur" ifad…ôsi daha √ßox praktikada yaranan probleml…ôrl…ô baƒülƒ±dƒ±r. ∆èslind…ô, **kvadrat olmayan he√ß bir matrisin t…ôrsi m√∂vcud deyil**, amma b…ôzi hallarda alternativ √ºsullarla (m…ôs…ôl…ôn, pseudoinverse) h…ôll tapmaq m√ºmk√ºnd√ºr.

## Kvadrat olmayan matris √º√ß√ºn s√ºtun sayƒ± azdƒ±rsa birba≈üa t…ôrs matrisi tapmaq m√ºmk√ºn olmasa da, ancaq normal t…ônlikl…ôr v…ô dig…ôr metodlar vasit…ôsil…ô h…ôll tapmaq m√ºmk√ºnd√ºr.

∆èg…ôr **s√ºtun sayƒ± azdƒ±rsa** ($m > n$, y…ôni m√º≈üahid…ôl…ôrin sayƒ± x√ºsusiyy…ôtl…ôrin sayƒ±ndan √ßoxdursa), o zaman **kvadrat matris olmadƒ±ƒüƒ± √º√ß√ºn birba≈üa t…ôrs matrisi tapmaq m√ºmk√ºn deyil**. Ancaq **normal t…ônlikl…ôr** v…ô dig…ôr metodlar vasit…ôsil…ô h…ôll tapmaq m√ºmk√ºnd√ºr.  

### **1. Normal T…ônlikl…ôr Metodu**
Bu metod **OLS (Ordinary Least Squares)** √º√ß√ºn istifad…ô olunur:  

$$
X^T X \beta = X^T y
$$

Burada **$X^T X$** kvadrat matrisdir v…ô √ßox vaxt tam r√ºtb…ôli olur, y…ôni t…ôrsi tapƒ±la bil…ôr:

$$
\beta = (X^T X)^{-1} X^T y
$$

Bu √ºsul yalnƒ±z **$X^T X$** t…ôrs ala bil…ôn (nonsingular) olduqda i≈ül…ôyir. ∆èg…ôr **$X$** tam r√ºtb…ôli deyils…ô, bu metod i≈ül…ôm…ôy…ôc…ôk.  

---

### **2. Pseudoinverse (Moore-Penrose Inverse)**
∆èg…ôr **$X^T X$** t…ôrs ala bilmirs…ô, onda **Moore-Penrose pseudoinverse** istifad…ô olunur:

$$
\beta = X^+ y
$$

Burada **$X^+$** matrisin **pseudo-t…ôrsidir (pseudoinverse)** v…ô **SVD (Singular Value Decomposition)** vasit…ôsil…ô hesablanƒ±r. Python-da bunu **`np.linalg.pinv(X)`** il…ô hesablamaq m√ºmk√ºnd√ºr.  

---

### **3. QR Decomposition**
QR par√ßalanmasƒ± il…ô **OLS h…ôlli** daha stabil tapƒ±la bil…ôr. Burada **$X$** matrisini **QR formasƒ±nda** yazƒ±rƒ±q:

$$
X = QR
$$

Burada:
- **$Q$** ortonormal matrisdir ($Q^T Q = I$)
- **$R$** yuxarƒ± √º√ßbucaq matrisidir.

OLS h…ôlli bu olur:

$$
\beta = R^{-1} Q^T y
$$

Python-da `np.linalg.qr(X)` istifad…ô etm…ôkl…ô QR par√ßalanmasƒ±nƒ± hesablamaq m√ºmk√ºnd√ºr.

---

### **4. Ridge Regression (L2 Regularization)**
∆èg…ôr **$X^T X$** t…ôrs ala bilmirs…ô v…ô ya √ßox ki√ßik determinantlƒ±dƒ±rsa (multicollinearity problemi varsa), **Ridge Regression** (L2 c…ôrim…ôl…ôndirm…ô) istifad…ô olunur:

$$
\beta = (X^T X + \lambda I)^{-1} X^T y
$$

Burada **$\lambda$** c…ôrim…ôl…ôndirm…ô parametri ki√ßik d…ôy…ôrl…ôri stabill…ô≈üdirir v…ô t…ôrs matris problemini h…ôll edir. **Python-da `Ridge` modelind…ôn (`sklearn.linear_model.Ridge`) istifad…ô etm…ôk m√ºmk√ºnd√ºr.**

---

### **N…ôtic…ô**
∆èg…ôr **s√ºtun sayƒ± azdƒ±rsa ($m > n$)**, t…ôrs matris birba≈üa tapƒ±lmasa da, h…ôll √º√ß√ºn a≈üaƒüƒ±dakƒ± metodlar istifad…ô olunur:
1. **Normal t…ônlikl…ôr** ‚Üí $ \beta = (X^T X)^{-1} X^T y $ (…ôg…ôr $X^T X$ tam r√ºtb…ôlidirs…ô)
2. **Pseudoinverse (Moore-Penrose)** ‚Üí $ \beta = X^+ y $ (…ôg…ôr $X^T X$ t…ôrs ala bilmirs…ô)
3. **QR Decomposition** ‚Üí $ \beta = R^{-1} Q^T y $
4. **Ridge Regression** ‚Üí $ \beta = (X^T X + \lambda I)^{-1} X^T y $ (t…ôrs matris problemi varsa)

∆èg…ôr **$X$** matrisind…ô x…ôtti asƒ±lƒ± s√ºtunlar varsa, …ôn etibarlƒ± √ºsul **pseudoinverse v…ô ya Ridge Regression** istifad…ô etm…ôkdir.

## **Ridge Regression** (L2 c…ôrim…ôl…ônmi≈ü Linear Regression) modelinin **…ômsallarƒ±nƒ±** $ \omega $ hesablamaq

Bu ≈ü…ôkild…ôki formul **Ridge Regression** (L2 c…ôrim…ôl…ônmi≈ü Linear Regression) modelinin **…ômsallarƒ±nƒ±** $ \omega $ hesablamaq √º√ß√ºn istifad…ô olunan **qapalƒ± forma** (closed-form solution) ifad…ôsidir.

**Formullar**:

$$
\tilde{L}_i(\omega) = L_i(\omega) + \frac{\lambda}{2} \|\omega\|^2
$$

$$
L_i(\omega) = (\omega^T x_i - y_i)^2 + \frac{\lambda}{2} \omega^T \omega
$$

  
$$
Q(\omega, X^T) = \frac{1}{2}(X \omega - Y)^T (X \omega - Y) + \frac{\lambda}{2} \omega^T \omega
$$

**T√∂r…ôm…ôsini tapƒ±b sƒ±fƒ±ra b…ôrab…ôr edib  $ \omega $ -nƒ± tapƒ±rƒ±q:**

$$
\omega = (X^T X + \lambda I)^{-1} X^T Y
$$

### ƒ∞ndi d…ôyi≈ü…ônl…ôri izah ed…ôk:

- **œâ** ‚Äì Axtardƒ±ƒüƒ±mƒ±z …ômsallar vektorudur. Bu vektor modelin √ß…ôkil…ôrini (weights) t…ômsil edir.  
- **X** ‚Äì Giri≈ü (input) d…ôyi≈ü…ônl…ôrinin matrisi. √ñl√ß√ºs√º **m √ó n**-dir, burada *m* n√ºmun…ôl…ôrin (observation) sayƒ±, *n* is…ô x√ºsusiyy…ôtl…ôrin (feature) sayƒ±dƒ±r.  
- **X·µÄ** ‚Äì X matrisinin transpozesi. √ñl√ß√ºs√º **n √ó m** olur.  
- **Y** ‚Äì H…ôd…ôf (target) d…ôyi≈ü…ôn vektorudur. √ñl√ß√ºs√º **m √ó 1**-dir.  
- **Œª (lambda)** ‚Äì C…ôrim…ô parametridir. Ridge regresiyasƒ±nda bu parametr overfitting-i (h…ôddind…ôn artƒ±q √∂yr…ônm…ôni) azaltmaq √º√ß√ºn istifad…ô olunur.  
- **I** ‚Äì ƒ∞dentiklik matrisidir. √ñl√ß√ºs√º **n √ó n** olan kvadrat matrisdir v…ô diaqonalƒ±nda 1-l…ôr olur.  
- **(X·µÄX + ŒªI)‚Åª¬π** ‚Äì Bu hiss…ô matrisin t…ôrsidir (inverse). Ridge regresiyada bu ifad…ô modelin stabil olmasƒ±na k√∂m…ôk edir, x√ºsusil…ô d…ô X·µÄX t…ôrs alƒ±nmayan v…ôziyy…ôtl…ôrd…ô.  


### ∆èsas f…ôrq:

- ∆èg…ôr $ \lambda = 0 $ olsa, bu zaman formul **adi Linear Regression**-ƒ±n qapalƒ± formasƒ± olur:
  $$
  \omega = (X^T X)^{-1} X^T Y
  $$

- ∆èg…ôr $ \lambda > 0 $ olsa, bu **Ridge Regression** olur v…ô bu, modelin **kompleksliyini c…ôrim…ôl…ôyir**, bel…ôlikl…ô overfitting azaldƒ±lƒ±r.

ƒ∞st…ôs…ôn, bu formulun Python kodla t…ôtbiqini d…ô g√∂st…ôr…ô bil…ôr…ôm.

## T…ôrs Matrisin olmamasƒ± (determinant = 0, x…ôtti asƒ±lƒ±lƒ±q) niy…ô overfitting-…ô s…ôb…ôb olur



√áox g√∂z…ôl sualdƒ±r. Bu hiss…ôni **riyazi v…ô intuitiv** baxƒ±mdan ba≈üa d√º≈üm…ôk √ßox vacibdir. G…ôlin, **niy…ô t…ôrs matrisin olmamasƒ± (determinant = 0, x…ôtti asƒ±lƒ±lƒ±q)** **overfitting-…ô** s…ôb…ôb olur ‚Äî bunu **addƒ±m-addƒ±m izah ed…ôk:**

---

### üîπ 1. X…ôtti asƒ±lƒ±lƒ±q n…ôdir?

∆èg…ôr **bir ne√ß…ô x√ºsusiyy…ôt bir-birind…ôn asƒ±lƒ±dƒ±rsa**, y…ôni:
$$
x_3 = 2 \cdot x_1 + 5 \cdot x_2
$$
kimi bir …ôlaq…ô varsa, onda dizayner matrisi $ X $ x…ôtti asƒ±lƒ± s√ºtunlar da≈üƒ±yƒ±r.

N…ôtic…ô:
- $ X^T X $ matrisinin **determinantƒ± sƒ±fƒ±r olur** ‚Üí **t…ôrsi yoxdur**.
> - Y…ôni, …ômsallar $ \omega $ √º√ß√ºn **yegan…ô cavab yoxdur** ‚Üí **sonsuz sayda h…ôll var**.

---

### üîπ 2. Niy…ô bu overfitting-…ô s…ôb…ôb olur?

Bu v…ôziyy…ôtd…ô:
> - Model **h…ôr n√∂qt…ôy…ô tam uyƒüun g…ôl…ôn** √ßoxlu **…ômsal kombinasiyalarƒ±** se√ß…ô bil…ôr.
- Bu …ômsallar √ßox **b√∂y√ºk v…ô qeyri-real** olur (√ß√ºnki sistem qeyri-sabitdir),
- Ki√ßik d…ôyi≈üiklikl…ôr bel…ô (yeni data, s…ôs-k√ºy) modelin √ßƒ±xƒ±≈üƒ±nƒ± **dramatik ≈ü…ôkild…ô d…ôyi≈ü…ô bil…ôr**.

N…ôtic…ô:
- Model **train set-d…ô √ßox d…ôqiq n…ôtic…ô verir**, amma
- **Test set-d…ô √ßox z…ôifl…ôdir** ‚Üí bu **overfitting**-dir.

---

### üîπ 3. Riyazi baxƒ±≈ü: $ \omega = (X^T X)^{-1} X^T Y $

∆èg…ôr $ X^T X $ t…ôrsi yoxdursa:
- Bu o dem…ôkdir ki, $ \omega $ …ômsallarƒ±nƒ± **d…ôqiq v…ô sabit** tapmaq olmur,
- √á√ºnki sistem **d…ôqiq m√º…ôyy…ôn olunmayƒ±b** (ill-posed problem).

---

### üîπ 4. H…ôll: Regularization (Ridge)

Ridge regression bel…ô edir:
$$
\omega = (X^T X + \lambda I)^{-1} X^T Y
$$

- Burada $ \lambda I $ …ôlav…ô olunduƒüuna g√∂r…ô,
- Matris artƒ±q **pozitiv-definit** olur (t…ôrsi var),
- V…ô …ômsallar **ki√ßik, sabit, daha d√ºzg√ºn** olur ‚Üí **overfitting azalƒ±r**.

---

### üí° Yekun:

| Problem             | N…ôtic…ô                                      |
|---------------------|----------------------------------------------|
| X…ôtti asƒ±lƒ±lƒ±q      | $X^T X$ singular olur (determinant = 0)   |
| T…ôrs yoxdur         | ∆èmsallar qeyri-sabit v…ô √ßox b√∂y√ºk olur       |
| Bu da g…ôtirir ki... | Model s…ôs-k√ºy…ô bel…ô uyƒüunla≈üƒ±r (overfit)     |

ƒ∞st…ôyirs…ôns…ô bu hiss…ôni qrafik v…ô ya n√ºmun…ô il…ô d…ô g√∂st…ôr…ô bil…ôr…ôm. Yax≈üƒ± sualdƒ±r, bel…ô d√º≈ü√ºnm…ôk s…ôni daha da g√ºcl…ôndir…ôc…ôk!

## T…ôrs Matrisin olmamasƒ± (determinant = 0, x…ôtti asƒ±lƒ±lƒ±q) probleminin h…ôll yolu L2(Ridge Regression) c…ôrim…ôl…ôndirm…ô metodudur.



B…ôli, **bu problemin (x…ôtti asƒ±lƒ±lƒ±q ‚Üí t…ôrs matris yoxdur ‚Üí overfitting)** qar≈üƒ±sƒ±nƒ± almaƒüƒ±n …ôn m…ô≈ühur h…ôll yollarƒ±ndan biri **L2 c…ôrim…ôl…ôndirm…ô**, y…ôni **Ridge Regression** metodudur.

---

## üß† Problem: X…ôtti Asƒ±lƒ±lƒ±q v…ô Overfitting

∆èsas m…ôs…ôl…ô:
- $ X^T X $ **t…ôrs alƒ±nmƒ±r** ‚Üí sistem sabit deyil ‚Üí **model overfit edir**.
- H…ôll yollarƒ± arasƒ±nda **L2 c…ôrim…ôl…ôndirm…ô** √ßox effektivdir.

---

## ‚úÖ H…ôll: L2 C…ôrim…ôsi (Ridge Regression)

Ridge Regression-d…ô m…ôqs…ôd funksiyasƒ± budur:

$$
J(\omega) = \| X\omega - Y \|^2 + \lambda \| \omega \|^2
$$

Burada:
- $ \| X\omega - Y \|^2 $: adi Mean Squared Error (modelin s…ôhvi),
- $ \lambda \| \omega \|^2 $: **L2 c…ôrim…ô** (…ômsallarƒ±n b√∂y√ºkl√ºy√ºn√º c…ôzalandƒ±rƒ±r),
- $ \lambda $: **regularization parametri** (t…ônziml…ôm…ô g√ºc√º).

---

### üìå Yeni ∆èmsallar Formulu:

Ridge Regression-d…ô …ômsal vektoru bel…ô hesablanƒ±r:

$$
\omega = (X^T X + \lambda I)^{-1} X^T Y
$$

> **∆èlav…ô olunan $ \lambda I $ matrisi** say…ôsind…ô:
> - $ X^T X + \lambda I $ artƒ±q **t…ôrs alƒ±nƒ±r** (h…ôtta x…ôtti asƒ±lƒ±lƒ±q olsa bel…ô),
> - Bu da **sistemi sabitl…ô≈üdirir**, …ômsallar **y√ºks…ôklikd…ôn qorunur**,
> - Model **overfitting etmir**, test n…ôtic…ôl…ôri d…ô yax≈üƒ±la≈üƒ±r.

---

## üéØ N…ôtic…ô:

| L2 Regularization (Ridge Regression) n…ô edir? | A√ßƒ±qlama |
|----------------------------------------------|----------|
| T…ôrs matris problemini aradan qaldƒ±rƒ±r       | $ \lambda I $ …ôlav…ô olunur |
| ∆èmsallarƒ±n b√∂y√ºm…ôsini c…ôzalandƒ±rƒ±r           | Daha sabit v…ô sad…ô model |
| Overfitting-in qar≈üƒ±sƒ±nƒ± alƒ±r                | Yax≈üƒ± generalizasiya |

---

ƒ∞st…ôyirs…ôns…ô, bunu vizual v…ô ya real n√ºmun…ô il…ô izah ed…ô bil…ôr…ôm (m…ôs…ôl…ôn, Python-da n√ºmun…ô). Amma qƒ±sa cavab budur: **L2 regularization = t…ôrs matris problemini sabitl…ô≈üdir…ôr…ôk overfitting-in qar≈üƒ±sƒ±nƒ± alƒ±r.**

## Python dilind…ô  izah

````python
# NumPy kitabxanasƒ±nƒ± √ßaƒüƒ±rƒ±r
import numpy as np

# Verilmi≈ü x d…ôy…ôri √º√ß√ºn polinomial ifad…ôni hesablayƒ±r
# res = k0*x^n + k1*x^(n-1) + ... + kn*x^0
def predict_poly(x, koeff):
    res = 0
    xx = [x ** (len(koeff) - n - 1) for n in range(len(koeff))]  # H…ôr d…ôr…ôc…ô √º√ß√ºn x^n hesablayƒ±r

    for i, k in enumerate(koeff):  # H…ôr …ômsalƒ± uyƒüun x d…ôr…ôc…ôsi il…ô vurur
        res += k * xx[i]

    return res

# 0-dan 10-a q…ôd…ôr 0.1 addƒ±mlƒ± x d…ôy…ôrl…ôri yaradƒ±r
x = np.arange(0, 10.1, 0.1)
# Funksiya: y = 1 / (1 + 10 * x^2) riyazi d√ºsturu il…ô y d…ôy…ôrl…ôrini hesablayƒ±r
y = 1 / (1 + 10 * np.square(x))

# H…ôr ikinci elementi se√ß…ôr…ôk t…ôlim (train) √º√ß√ºn x v…ô y d…ôy…ôrl…ôrini ayƒ±rƒ±r
x_train, y_train = x[::2], y[::2]

# M…ôlumatƒ±n √ºmumi sayƒ±nƒ± m√º…ôyy…ôn edir
N = len(x)

# T…ôlim m…ôlumatlarƒ±na uyƒüun 10-cu d…ôr…ôc…ôli polinomial funksiyanƒ±n …ômsallarƒ±nƒ± tapƒ±r
z_train = np.polyfit(x_train, y_train, 3)  # Min kvadrat √ºsulu il…ô polinom uyƒüunla≈üdƒ±rƒ±r
# ∆èld…ô olunan …ômsallarƒ± √ßap edir
print(z_train)
````

## L2 Regularization olmadan Proqnoz

````python
# NumPy kitabxanasƒ±nƒ± Numpy array v…ô riyazi …ôm…ôliyyatlar √º√ß√ºn idxal edirik
import numpy as np

# Matplotlib kitabxanasƒ±nƒ± qrafikl…ôri vizualla≈üdƒ±rmaq √º√ß√ºn idxal edirik
import matplotlib.pyplot as plt

# x oxu √ºzr…ô 0-dan 10-a q…ôd…ôr 0.1 aralƒ±qla n√∂qt…ôl…ôr yaradƒ±lƒ±r
x = np.arange(0, 10.1, 0.1)
# y d…ôy…ôrl…ôri: verilmi≈ü polinom funksiyasƒ±na …ôsas…ôn hesablanƒ±r (y = x¬≥ - 10x¬≤ + 3x + 500)
y = np.array([a ** 3 - 10 * a ** 2 + 3 * a + 500 for a in x])
# M…ô≈üq (train) √º√ß√ºn veril…ônl…ôr: h…ôr ikinci n√∂qt…ô se√ßilir
x_train, y_train = x[::2], y[::2]
# Polinomun maksimal d…ôr…ôc…ôsi (N-1). N = 13 o dem…ôkdir ki, x‚Å∞-dan x¬π¬≤-…ô q…ôd…ôr istifad…ô olunacaq
N = 13
# Regularizasiya g√ºc√º: L2 c…ôzasƒ± √º√ß√ºn lambda (hiperparametr)
L = 0

# X matrisini yaradƒ±rƒ±q: h…ôr bir x √º√ß√ºn x‚Å∞, x¬π, ..., x‚Åø‚Åª¬π s√ºtunlarƒ±
X = np.array([[a ** n for n in range(N)] for a in x])
# Œª * I regularizasiya matrisi (diagonalda yalnƒ±z lambda d…ôy…ôrl…ôri, qalanƒ± 0)
IL = np.array([[L if i == j else 0 for j in range(N)] for i in range(N)])
# Bias (intercept) √º√ß√ºn …ômsal (x‚Å∞) regularizasiya olunmur, buna g√∂r…ô onu 0 edirik
IL[0][0] = 0
# X_train: m…ô≈üq √º√ß√ºn uyƒüun olan (x‚Å∞-dan x‚Åø-…ô q…ôd…ôr) veril…ônl…ôr matrisind…ôn h…ôr ikinci s…ôtir
X_train = X[::2]
# Y: m…ô≈üq √º√ß√ºn target d…ôy…ôrl…ôr (y_train)
Y = y_train

# Ridge Regression formulu: w = (X·µóX + ŒªI)^-1 X·µóY
A = np.linalg.inv(X_train.T @ X_train + IL)  # (X·µóX + ŒªI)‚Åª¬π hiss…ôsi
w = A @ X_train.T @ Y  # tam …ômsal vektoru: w = A * X·µóY
print(w)  # ∆èld…ô olunan …ômsallarƒ± √ßap et

# Yeni modelin proqnoz d…ôy…ôrl…ôrini hesablamaq: h…ôr x √º√ß√ºn w·µóx
yy = [np.dot(w, x) for x in X]
# Proqnoz qrafiki (mavi x…ôtt)
plt.plot(x, yy, label='Proqnoz')
# ∆èsas funksiyanƒ±n …ôsl qrafiki (sarƒ± x…ôtt)
plt.plot(x, y, label='∆èsas funksiya')
# Leqenda g√∂st…ôrilir
plt.legend()  
# ≈û…ôb…ôk…ôni g√∂st…ôr
plt.grid(True)
# Qrafiki g√∂st…ôr
plt.show()
````
