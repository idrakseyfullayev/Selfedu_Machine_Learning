## Distributions

## **1️⃣ Uniform (Bərabər) Paylanma**  
📌 **Düstur:**  
- **Diskret halda:**  
  $$
  P(X = x) = \frac{1}{b - a + 1}, \quad a \leq x \leq b
  $$
- **Davamlı halda:**  
  $$
  f(x) = \frac{1}{b - a}, \quad a \leq x \leq b
  $$

📌 **Düsturda:**  
- $X$ → Təsadüfi dəyişən  
- $a, b$ → Paylanmanın aşağı və yuxarı həddi  
- $P(X = x)$ → Hadisənin baş vermə ehtimalı  
- $f(x)$ → Ehtimal sıxlıq funksiyası (PDF)  

📌 **İzah:** Hər bir hadisənin eyni ehtimalla baş verməsini ifadə edir.  

📌 **İstifadə sahəsi:** Random ədəd generasiyası, Monte Carlo simulyasiyası.  

---

## **2️⃣ Bernoulli Paylanması**  
📌 **Düstur:**  
$$
P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0,1\}
$$

📌 **Düsturda:**  
- $X$ → 0 və ya 1 ala bilən təsadüfi dəyişən  
- $p$ → Uğur ehtimalı ($P(X = 1) = p$)  
- $1 - p$ → Uğursuzluq ehtimalı ($P(X = 0) = 1 - p$)  

📌 **İzah:** Bir hadisə yalnız iki mümkün nəticəyə malik olduqda (başarı və uğursuzluq) istifadə olunur.  

📌 **İstifadə sahəsi:** Tək bir sınaqda uğur/qəbul qərarları (məs., sikkə atışı).  


---

## **3️⃣ Binomial Paylanma**  
📌 **Düstur:**  
$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}
$$

📌 **Düsturda:**  
- $X$ → Uğurların sayı  
- $n$ → Təcrübələrin ümumi sayı  
- $k$ → Uğurların sayı  
- $p$ → Uğur ehtimalı  
- $\binom{n}{k}$ → Kombinasiyalar sayı ($\frac{n!}{k!(n-k)!}$)  

📌 **İzah:** Bir hadisənin müəyyən sayda ($ n $) sınaqdan sonra neçə dəfə baş verəcəyini göstərir.  

📌 **İstifadə sahəsi:** Anketlər, testlərdə düzgün cavabların sayı, istehsalatda defektli məhsulların hesablanması. 

---

## **4️⃣ Geometrik Paylanma**  
📌 **Düstur:**  
$$
P(X = k) = (1 - p)^{k-1} p
$$

📌 **Düsturda:**  
- $X$ → İlk uğura qədər olan sınaq sayı  
- $k$ → İlk uğura qədər edilən sınaqların sayı  
- $p$ → Uğur ehtimalı  

📌 **İzah:** İlk uğura qədər olan uğursuzluq sayını göstərir.  

📌 **İstifadə sahəsi:** Müştəri tapana qədər edilən zənglərin sayı, maşının sıradan çıxana qədər işləmə dövrləri.  

---

## **5️⃣ Poisson Paylanması**  
📌 **Düstur:**  
$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

📌 **Düsturda:**  
- $X$ → Hadisələrin sayı  
- $\lambda$ → Hadisənin başvermə orta sayı  
- $k!$ → Faktorial ($k$ ədədinin faktorialı)  
- $e$ → Təqribən 2.718 olan riyazi sabit  

📌 **İzah:** Müəyyən zaman aralığında nadir hadisələrin sayını göstərir.  

📌 **İstifadə sahəsi:** Çağrı mərkəzinə daxil olan zənglərin sayı, səhiyyədə xəstəxanaya qəbul olunma sayı.  

---

## **6️⃣ Exponential Paylanma**  
📌 **Düstur:**  
$$
f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
$$

📌 **Düsturda:**  
- $X$ → Hadisənin başvermə vaxtı  
- $\lambda$ → Hadisənin orta baş vermə sürəti  
- $e$ → Təqribən 2.718 olan riyazi sabit  

📌 **İzah:** Hadisənin baş vermə intervallarını göstərir.  

📌 **İstifadə sahəsi:** Maşının sıradan çıxma vaxtı, serverdə sorğuların gəlmə intervalları.

---

## **7️⃣ Normal (Qaus) Paylanma**  
📌 **Düstur:**  
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

📌 **Düsturda:**  
- $X$ → Normal paylanmış təsadüfi dəyişən  
- $\mu$ → Ortalamadır  
- $\sigma$ → Standart sapmadır  
- $\pi$ → Təqribən 3.1416 olan riyazi sabit  
- $e$ → Təqribən 2.718 olan riyazi sabit  

📌 **İzah:** Təbii hadisələrin əksəriyyəti normal paylanmaya malikdir.  

📌 **İstifadə sahəsi:** Boy uzunluqları, IQ paylanması, səhmlərin qiymət dəyişiklikləri.  


---

## **8️⃣ Polinomial Paylanma**  
📌 **Düstur:**  
$$
P(X_1 = x_1, ..., X_k = x_k) = \frac{n!}{x_1! x_2! ... x_k!} p_1^{x_1} p_2^{x_2} ... p_k^{x_k}
$$

📌 **Düsturda:**  
- $X_i$ → Hər bir fərqli nəticənin təkrarlanma sayı  
- $n$ → Ümumi sınaq sayı  
- $p_i$ → Hər nəticənin başvermə ehtimalı  
- $x_i!$ → Hər nəticənin faktorialı  

📌 **İzah:** Polinomial paylanma, bir hadisənin müəyyən sayda (𝑛) sınaqdan sonra müxtəlif nəticələrin (məsələn, hər sınaqda bir neçə fərqli nəticə ola bilər) neçə dəfə baş verəcəyini göstərir. Bu paylanma, hər bir nəticə üçün ehtimalları nəzərə alaraq, hər bir nəticənin mümkün sayda baş vermə ehtimallarını hesablamağa imkan verir.

📌 **İstifadə sahəsi:** Seçkilərdə müxtəlif partiyalara səs paylanması, müştərilərin müxtəlif məhsulları seçmə ehtimalları. 

---

## **9️⃣ T-student Paylanması**  
📌 **Düstur:**  
$$
f(t) = \frac{\Gamma(\frac{\nu+1}{2})}{\sqrt{\nu\pi} \Gamma(\frac{\nu}{2})} (1 + \frac{t^2}{\nu})^{-\frac{\nu+1}{2}}
$$

📌 **Düsturda:**  
- $t$ → T-statistikası  
- $\nu$ → Azadlıq dərəcələri  
- $\Gamma(x)$ → Gamma funksiyası  


📌 **İzah:** Kiçik nümunələrdə ortalamanın normallığını qiymətləndirmək üçün istifadə olunur.  

📌 **İstifadə sahəsi:** A/B testləri, hipotez yoxlamaları.  

---

## **🔟 Chi-Square Paylanması**  
📌 **Düstur:**  
$$
f(x) = \frac{x^{\frac{k}{2} - 1} e^{-x/2}}{2^{k/2} \Gamma(k/2)}
$$

📌 **Düsturda:**  
- $X$ → Chi-square təsadüfi dəyişəni  
- $k$ → Azadlıq dərəcələri  
- $e$ → Təqribən 2.718 olan riyazi sabit  
- $\Gamma(x)$ → Gamma funksiyası  

📌 **İzah:** Variansın paylanmasını qiymətləndirir.  

📌 **İstifadə sahəsi:** Qiymətləndirmə testləri, bağımsızlıq analizləri.  

---

## **1️⃣1️⃣ F Paylanması (Fisher's Distribution)**  
📌 **Düstur:**  
$$
f(x) = \frac{\Gamma(\frac{d_1 + d_2}{2})}{\Gamma(\frac{d_1}{2}) \Gamma(\frac{d_2}{2})} \left( \frac{d_1}{d_2} \right)^{\frac{d_1}{2}} \frac{x^{\frac{d_1}{2} - 1}}{\left( 1 + \frac{d_1}{d_2}x \right)^{\frac{d_1 + d_2}{2}}}
$$

📌 **Düsturda:**  
- $X$ → F-paylanmış təsadüfi dəyişən  
- $d_1, d_2$ → Azadlıq dərəcələri  
- $\Gamma(x)$ → Gamma funksiyası  

📌 **İzah:** İki variansın nisbətini yoxlamaq üçün istifadə olunur.  

📌 **İstifadə sahəsi:** ANOVA testləri.

## Binomial paylanma çox sayda sınaq ($ n $) və kiçik uğur ehtimalı ($ p $) olduqda, Normal paylanmaya yaxınlaşır

Binomial paylanma çox sayda sınaq ($ n $) və kiçik uğur ehtimalı ($ p $) olduqda, Normal paylanmaya yaxınlaşır. **Mərkəzi Limit Teoremi**nə görə, əgər $ n $ böyükdürsə, Binomial paylanma təqribən **Normal paylanma** ilə ifadə oluna bilər:  

$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

burada:  
- $ \mu = np $ — Binomial paylanmanın gözlənilən dəyəri  
- $ \sigma^2 = np(1-p) $ — Dispersiyası  

Bu səbəbdən, əgər **n kifayət qədər böyükdürsə**, Binomial paylanmanın qrafiki Normal paylanmaya bənzəyir.  

Əgər fərqli görmək istəyirsənsə, **n** və **p** dəyərlərini dəyişdirərək Binomial paylanmanın daha fərqli formasını çəkmək olar. Məsələn, **n=10, p=0.5** götürsək, daha diskret bir qrafik alınacaq. Onu istəyirsən?

## Poisson və Normal paylanmalar arasında müəyyən əlaqə var

Poisson və Normal paylanmalar arasında müəyyən əlaqə var və Poisson paylanması bəzi hallarda Normal paylanmaya bənzəyə bilər. Bunun səbəblərini aşağıdakı kimi izah edə bilərik:  

### **1️⃣ Mərkəzi Limit Teoremi**  
Poisson paylanması diskret bir paylanmadır, lakin **λ** (orta başvermə sayı) böyüdükcə, onun qrafiki Normal paylanmaya daha çox oxşamağa başlayır. Bu, **Mərkəzi Limit Teoremi** ilə izah olunur. Əgər **λ böyükdürsə (λ > 30)**, Poisson paylanması **$ N(\lambda, \sqrt{\lambda}) $** Normal paylanmasına yaxınlaşır.  

### **2️⃣ Simmetriya Artımı**  
Kiçik **λ** dəyərlərində Poisson paylanmasının qrafiki sağa əyilmiş olur (çarpıqlıq var). Lakin **λ böyüdükcə**, paylanma daha simmetrik olur və Normal paylanmaya daha çox bənzəyir.  

### **3️⃣ Təbiətdə Real Tətbiqləri**  
- Poisson paylanması nadir hadisələrin baş vermə sayını ölçür (məsələn, çağrı mərkəzinə daxil olan zənglər).  
- Lakin, əgər bu tip nadir hadisələr çox tez-tez baş verirsə (yəni **λ** böyükdürsə), onda bu hadisələrin sayı təxminən Normal paylanmaya uyğun gəlir.  

**Qrafiklərdə bunu müşahidə etmək üçün Poisson paylanmasının λ dəyərini artıraraq yoxlaya bilərsən.**  

Əgər qrafikdə fərqi açıq görmək istəyirsənsə, **kiçik λ dəyərləri (məsələn, λ=5) üçün Poisson paylanmasının qrafikinə bax**, sonra isə **böyük λ dəyərləri (məsələn, λ=50) üçün yoxla**. λ böyüdükcə Poisson paylanması Normal paylanmaya daha çox oxşayacaq. 🚀

## Diskret Paylanma və Davamlı (Kəsilməz) Paylanma nədir

### **Diskret Paylanma Nədir?**  

📌 **Diskret paylanma** — diskret təsadüfi dəyişənin mümkün olan qiymətlərini və onların ehtimallarını göstərən paylanmadır. **Diskret təsadüfi dəyişən** yalnız **ayrıq (diskret) qiymətlər** ala bilər. Bu qiymətlər **müəyyən bir siyahı** şəklində olur (məsələn, 0, 1, 2, 3 və s.).  

---  

### **🔹 Diskret Paylanmaya Misallar**  

#### **1️⃣ Binom Paylanması (Binomial Distribution)**  
- **Təsviri:** Təcrübə bir neçə dəfə təkrarlanır və hər dəfə yalnız **iki mümkün nəticə** olur (Məsələn, uğur və uğursuzluq).  
- **Misal:** Bir sikkəni 10 dəfə atırıq və neçə dəfə geriyə düşdüyünü sayırıq.  

#### **2️⃣ Poisson Paylanması (Poisson Distribution)**  
- **Təsviri:** Müəyyən bir zaman aralığında və ya sahədə baş verən hadisələrin sayını təsvir edir.  
- **Misal:** Bir bankın kassasında 1 saat ərzində neçə müştərinin gəlməsini hesablamaq.  

#### **3️⃣ Geometrik Paylanma (Geometric Distribution)**  
- **Təsviri:** İlk uğurun neçə-ci sınaqda baş verəcəyini göstərir.  
- **Misal:** Zər atan bir şəxs ilk dəfə 6 rəqəmini nə vaxt atacaq?  

#### **4️⃣ Poisson-Binomial Paylanması**  
- **Təsviri:** Bir neçə müstəqil, lakin eyni ehtimala malik olmayan binom paylanmalı təsadüfi dəyişənlərin cəmi.  

#### **5️⃣ Hipergeometrik Paylanma (Hypergeometric Distribution)**  
- **Təsviri:** Seçmə **geri qaytarılmadan** edildikdə uğur sayını göstərir.  
- **Misal:** 10 qırmızı və 5 mavi top olan bir qutudan 4 top seçirik, neçə qırmızı top seçdiyimizi hesablamaq.

#### **6️⃣ Polinomial Paylanma (Multinomial Distribution)**  
- **Təsviri:** Binom paylanmasının **birdən çox nəticəsi** olduğu ümumiləşdirilmiş formasıdır.  
- **Misal:** Bir zar 10 dəfə atıldıqda hər bir ədədin neçə dəfə düşdüyünü saymaq.  

**Nəticə:**  
**Diskret paylanmalar** əsasən **sayma** ilə əlaqəli hadisələri modelləşdirir, yəni **nəticələr məhdud və ayrıqdır**.

---

### **Davamlı (Kəsilməz) Paylanma Nədir?**

📌 **Davamlı (kəsilməz) paylanma** — davamlı təsadüfi dəyişənin mümkün olan qiymətlərini və onların ehtimallarını göstərən paylanmadır. **Davamlı təsadüfi dəyişən** hər hansı bir aralıqda, hətta sonsuz sayda fərqli qiymət ala bilər. **Yəni, müəyyən bir nöqtənin ehtimalı sıfırdır, lakin bir aralıq daxilində ehtimal hesablanır.**

---

### **🔹 Davamlı Paylanmaya Misallar**

#### **1️⃣ Normal Paylanma (Normal Distribution)**
- **Təsviri:** Ən çox istifadə olunan kəsilməz paylanmalardan biridir və bir çox təbii hadisəni təsvir edir. Ehtimal sıxlığı simmetrikdir və “bell curve” şəklində göstərilir.
- **Misal:** Bir məktəbin şagirdlərinin boyunun paylanması.

#### **2️⃣ Ekspontensial Paylanma (Exponential Distribution)**
- **Təsviri:** Müəyyən bir hadisənin baş verməsi üçün keçən vaxtın paylanmasını təsvir edir.
- **Misal:** Bir telefon zənginin gəlmə vaxtı və ya maşınların arızalanma müddəti.

#### **3️⃣ Uniform Paylanma (Uniform Distribution)**
- **Təsviri:** Bütün dəyərlərin bərabər ehtimal ilə baş verməsi vəziyyətini təsvir edir.
- **Misal:** Bir zarın üzərindəki qiymətlərin paylanması.

#### **4️⃣ T-Student Paylanması (Student’s t Distribution)**
- **Təsviri:** Kiçik nümunə ölçüləri üçün istifadə edilən paylanmadır, xüsusilə ortalamaların müqayisəsi üçün.
- **Misal:** Kiçik nümunə əsasında bir populyasiyanın ortalamasını təxmin etmək.

#### **5️⃣ Beta Paylanması (Beta Distribution)**
- **Təsviri:** [0, 1] aralığında olan dəyişənlər üçün istifadə edilən paylanmadır və xüsusilə ehtimal nəzəriyyəsi və statistikasında geniş tətbiq olunur.
- **Misal:** Bir layihənin tamamlama ehtimalı.

📌 **Qeyd:** **Kəsilməz paylanmalar**da ehtimal bir nöqtə üçün deyil, bir aralıq üçün hesablanır. Məsələn, **Normal paylanma** üçün bir interval daxilindəki ehtimalı hesablamaq mümkündür.

---

### **Nəticə:**  
**Davamlı paylanmalar** əsasən **ölçmə** ilə əlaqəli hadisələri modelləşdirir, yəni **nəticələr aralıqdan alınan dəyərlərdir**. Məsələn, **boy, çəki, temperatur, zaman** kimi dəyişənlər kəsilməz paylanmalara uyğun gəlir.

🔎 Daha ətraflı başa düşmək üçün hansı kəsilməz paylanma səni daha çox maraqlandırır? 😊



### **📊 Diskret və Davamlı (Kəsilməz) Paylanmaların Fərqi**  

| **Xüsusiyyət**      | **Diskret Paylanma** | **Davamlı Paylanma** |
|---------------------|--------------------|--------------------|
| **Qiymətlər**      | Ayrıqdır (0, 1, 2, ...) | İstənilən real ədədi ala bilər |
| **Misallar**       | Binom, Poisson, Geometrik | Normal, Ekspontensial, Uniform |
| **Ehtimal ($P(X=x)$)** | Xüsusi bir nöqtə üçün **0-dan fərqli olur** | Xüsusi bir nöqtə üçün **həmişə 0-dır** |

📌 **Qeyd:** Əgər bir paylanma **"kəsilməz" (continuous)** olarsa, artıq diskret olmur. Məsələn, **Normal paylanma (Gaussian Distribution)** davamlı paylanmadır.  


## Polinom Funksiya

**1️⃣ Polinomial Paylanma**  

📌 **Ehtimal paylanmasının düsturu:**  
$$
P(X_1 = x_1, X_2 = x_2, ..., X_k = x_k) = \frac{n!}{x_1! x_2! ... x_k!} p_1^{x_1} p_2^{x_2} ... p_k^{x_k}
$$  

📌 **Düsturda istifadə olunan elementlər:**  
- $X_i$ → Hər bir mümkün nəticənin başvermə sayı  
- $n$ → Ümumi sınaq (təcrübə) sayı  
- $k$ → Mövcud mümkün nəticələrin sayı  
- $p_i$ → Hər bir nəticənin ehtimalı ($p_1 + p_2 + ... + p_k = 1$)  
- $x_i$ → Müəyyən bir nəticənin neçə dəfə baş verdiyi  
- $x_1 + x_2 + ... + x_k = n$ şərti ödənməlidir  
- $\frac{n!}{x_1! x_2! ... x_k!}$ → Mövcud halların sayını göstərən kombinasiyalar düsturu  

✅ **İstifadə sahələri:**  
Polinomial paylanma, bir neçə mümkün nəticə olan təcrübələrin modelləşdirilməsində istifadə olunur. Məsələn:  
- Siyasi seçkilərdə bir neçə namizədə səs verilmə ehtimallarının hesablanması  
- Bir fabrikdə istehsal olunan müxtəlif növ məhsulların faiz nisbətinin təhlili  
- Mağazalarda müştərilərin fərqli brendləri seçmə ehtimallarının hesablanması  

---

**2️⃣ Polinom Funksiyası** 

**Polinom, dəyişənlər və əmsalların cəmi şəklində ifadə olunan riyazi funksiyadır**

📌 **İzah:** Polinomial funksiya, dəyişənin müxtəlif güclərinin cəmi ilə təyin olunan və hər bir gücün özünə məxsus bir koeffisientə sahib olan funksiyadır.

📌 **Funksiya düsturu:**  
$$
P(x) = a_n x^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0
$$  

📌 **Düsturda istifadə olunan elementlər:**  
- $P(x)$ → Polinom funksiyası  
- $a_n, a_{n-1}, ..., a_0$ → Polinomun əmsalları  
- $x$ → Müstəqil dəyişən  
- $n$ → Polinomun dərəcəsi (ən yüksək qüvvət)

Məsələn, **3-cü dərəcəli polinom** belə ola bilər:  

$$
P(x) = 2x^3 - 4x^2 + 3x - 7
$$

Polinomlar maşın öyrənmədə və statistik modelləşdirmədə, xüsusilə **polinomial reqressiya** kimi sahələrdə geniş istifadə olunur. 🚀


✅ **Bu funksiya haradan alınır və harada istifadə olunur?**  
Polinom funksiyaları riyaziyyatın və tətbiqi sahələrin müxtəlif bölmələrində geniş istifadə olunur.  
- **Cəbr və analiz:** Funksiyaların yaxınlaşması və diferensial tənliklərin həlli  
- **Fizika:** Hərəkət tənlikləri və elektrik dövrələrinin modelləşdirilməsi  
- **İqtisadiyyat və maliyyə:** Trend analizi və proqnozlaşdırma modelləri  
- **Maşın öyrənməsi və süni intellekt:** Regresyon modelləri  

Polinom funksiyaları interpolasiya, inteqrasiya və differensial hesablama kimi bir çox riyazi problemlərin həllində istifadə olunur.  

Bu iki anlayışın fərqini anlamaq üçün polinomial paylanmanın ehtimallarla əlaqəli olduğunu, polinom funksiyasının isə cəbri funksiyalardan biri olduğunu unutma! 🚀

## Polinom funksiyası necə yaranır

### **1️⃣ Polinom funksiyası necə yaranır?**  

Polinom funksiyası, **hər hansı bir verilmiş verilənlər toplusuna uyğun funksional ifadə qurmaq üçün** istifadə olunan əsas riyazi modeldir.  

#### **A) Funksiya anlayışı və polinomun yaranması**  

Ən ümumi funksiyalardan biri **x dəyişəni üzrə bir neçə qüvvət üzərində qurulan** ifadədir. Məsələn:  
- **Sabit funksiya:** $ f(x) = c $ (bütün $ x $ üçün eyni dəyər alır)  
- **Xətti funksiya:** $ f(x) = ax + b $  
- **Kvadrat funksiya:** $ f(x) = ax^2 + bx + c $  
- **Kubik funksiya:** $ f(x) = ax^3 + bx^2 + cx + d $  
- **Ümumi polinom:**  
  $$
  P(x) = a_n x^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0
  $$  

Bu tip funksiyalar **daha yüksək dərəcəli dəyişənlərdən ibarət olan cəbri ifadələrdir**.  

#### **B) Polinomların yaranma mənbəyi**  

Polinomlar əsasən aşağıdakı metodlarla əldə edilir:  
1. **Datalara uyğun interpolasiya və aproksimasiya:**  
   - Əgər bir neçə nöqtə verilirsə və bu nöqtələri keçən bir funksiya tapmaq lazımdırsa, polinom interpolasiyası tətbiq olunur.  
   - Məsələn, **Lagrange interpolasiyası** və ya **Newton interpolasiyası** istifadə olunur.  

2. **Diferensial tənliklərin həlli:**  
   - Bir çox diferensial tənliklərdə həll funksiyası polinom şəklində yazıla bilər.  
   - Məsələn, **Taylor seriyası** ilə istənilən funksiyanı polinomla yaxınlaşdırmaq olar.  

3. **Alqoritmik model qurulması:**  
   - Maşın öyrənməsində **Polinomial Regression** polinom funksiyası ilə modelləşdirilir.  

#### **C) Polinom düsturunun əsas hissələri**  

- $a_n, a_{n-1}, ..., a_0$ — polinomun əmsallarıdır və onlar verilənlərə uyğun təyin olunur.  
- $ x^n, x^{n-1}, ... $ — polinomun dərəcələridir, **hər bir termin bir qüvvətə malikdir**.  
- **Polinomun əsas prinsipi:** **x dəyişəndən asılı olan və tədricən artan və ya azalan ifadə yaratmaqdır**.  

---

### **2️⃣ Bu funksiyalar harada istifadə olunur?**  
Polinom funksiyaları **müxtəlif sahələrdə** tətbiq olunur:  
- **Fizikada:** Hərəkət tənlikləri və dalğa tənlikləri üçün  
- **İqtisadiyyatda:** Qiymət dəyişiklikləri və trend analizi  
- **Data Science-də:** Regresyon modelləri və maşın öyrənməsi  
- **İnşaat mühəndisliyində:** Strukturların dayanıqlılığının hesablanması  

Əgər xüsusi bir tətbiq haqqında daha çox bilmək istəsən, deyə bilərsən! 🚀

## Polinomial funksiyası daha geniş izah

**Polinomial funksiya** riyaziyyatda çox geniş istifadə olunan və dəyişənin müxtəlif güclərinin cəmindən ibarət olan bir funksiyadır. Polinomial funksiyalar analitik cəhətdən sadə, lakin müxtəlif sahələrdə güclü tətbiq sahələrinə malikdir. İndi polinomial funksiyanın təfərrüatlarına daha ətraflı baxaq:

### 1. **Polinomial funksiyanın tərifi:**
Polinomial funksiya aşağıdakı şəkildə təyin olunur:

$$
f(x) = a_n x^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0
$$

Burada:
- **$x$** dəyişən (input)dir.
- **$a_n, a_{n-1}, \dots, a_1, a_0$** koeffisentlərdir və onlar polinomialın sabit qiymətləridir.
- **$n$** polinomialın dərəcəsidir (yəni, $x$-in ən yüksək gücü).
- **$a_n$**, **$a_{n-1}$** və digər koeffisentlər, polinomialın "formasını" müəyyən edir.

**Misal:**  
$f(x) = 3x^4 - 2x^3 + x^2 - 7x + 5$

Burada polinomialın dərəcəsi 4-dür, çünki ən yüksək güc $x^4$-dü.

### 2. **Polinomial funksiyanın xüsusiyyətləri:**

- **Dərəcə (Degree):** Polinomialın dərəcəsi onun ən yüksək gücünü göstərir. Məsələn, $f(x) = 2x^3 + 4x^2 + 3x + 1$ polinomialı 3-cü dərəcəli polinomdur.
  
- **Koeffisentlər (Coefficients):** Polinomialın müxtəlif komponentlərindəki koeffisentlər, yəni $a_n, a_{n-1}, \dots, a_0$, funksiyanın davranışını müəyyən edir. Koeffisentlər həm müsbət, həm də mənfi ola bilər.

- **Dağılma və Çoxalma:** Polinomial funksiyalar toplama və çıxarmaya qarşı müsbət reaksiya göstərir. Məsələn, iki polinomialın cəmi və ya fərqi yenə bir polinomial funksiyasıdır. Eyni şəkildə, bir polinomialın başqa bir polinomial ilə çarpılmasi yenə polinomialdır.

### 3. **Polinomial funksiyanın tətbiqləri:**

Polinomial funksiyalar müxtəlif sahələrdə çox geniş istifadə olunur. Bəzi sahələrdə tətbiqlərinə nəzər salaq:

- **Verilənlərə uyğunlaşma (Curve Fitting):** Polinomial funksiyalar, verilənlərə ən yaxşı **uyğunlaşma (fitting)** tapmaq üçün tez-tez istifadə olunur. Məsələn, **bir verilənlər dəstəsinin qrafiki üzərində polinomial bir xətt çəkmək, verilənlərə ən yaxşı uyğun gələn funksiyanı tapmaq** üçün istifadə olunur.

- **Optimallaşdırma:** Polinomial funksiyalar müxtəlif optimallaşdırma problemlərində, məsələn, minimizasiya və ya maksimizasiya problemlərində istifadə oluna bilər.

- **Hesablama:** Polinomial funksiyalar diferensial və inteqrasiya əməliyyatlarında çox faydalıdır. Polinomiallar həmçinin Taylor seriyaları və digər yaxınlaşma üsullarında geniş istifadə olunur.

- **Fizika və mühəndislik:** Fizikada və mühəndislik sahələrində bir çox təbiət qanunları polinomial tənliklərlə təmsil oluna bilər. Məsələn, əgər bir cismin hərəkəti tərtib olursa, bu hərəkətin qanunları polinomial funksiyalarla təyin oluna bilər.

### 4. **Polinomialın davranışı:**

- **Qrafik (Graph):** Polinomial funksiyasının qrafiki polinomialın dərəcəsinə görə fərqlənir. Məsələn, 2-ci dərəcəli (kvadrat) polinomialın qrafiki parabola formasındadır. Yüksək dərəcəli polinomiaların isə daha kompleks qrafikləri ola bilər.

- **Maksimum və minimum nöqtələr (Extrema):** Polinomial funksiyalar çox vaxt maksimum və minimum nöqtələrə sahib olur. Bu nöqtələri tapmaq üçün diferensial hesablamalar istifadə olunur.

- **Kök (Roots) və sıfırlar:** Polinomialın kökləri, onun sıfır olduğu nöqtələrdir. Bu köklər çox zaman faktorizasiya və ya kök tapma metodları ilə tapılır.

### 5. **Polinomial funksiyanın növləri:**

- **Kvadrat polinomial (2-ci dərəcə):** Ən sadə polinomial funksiyalardan biridir və forması belədir:
  $$
  f(x) = ax^2 + bx + c
  $$
  Burada $a$, $b$ və $c$ sabitlərdir. Bu cür polinomiaların qrafiki parabola şəklində olur.

- **Kubik polinomial (3-cü dərəcə):** 3-cü dərəcəli polinomialar daha mürəkkəb qrafiklər göstərir və daha çox maksimum və minimum nöqtələrə sahib ola bilər.

- **Yüksək dərəcəli polinomialar:** 4-cü dərəcə və daha yüksək dərəcəli polinomialar daha mürəkkəb formalar alır və bir neçə kökə (sıfıra) sahib ola bilər.

### 6. **Polinomial funksiyalar və təhlil:**

Polinomial funksiyaların təhlili, əksər hallarda, aşağıdakı addımları əhatə edir:
- **Derivativ:** Polinomialın törəməsi onun əyrisinin meylini təhlil etmək üçün istifadə olunur.
- **İnteqrasiya:** Polinomialın inteqralı sahə və ya ümumi məbləğ tapmaq üçün istifadə oluna bilər.
- **Köklərin tapılması:** Polinomialın sıfırlarını tapmaq (yəni köklərini tapmaq) üçün faktorizasiya və ya kvadrat formuldan istifadə olunur.

### Nəticə:
Polinomial funksiyalar riyaziyyatda və tətbiqi sahələrdə çox geniş istifadə olunur. Onlar həm sadə, həm də mürəkkəb problemləri həll etmək üçün güclü alətlərdir. Həmçinin, verilənlər analizi, optimallaşdırma, fizika və mühəndislik kimi sahələrdə əsas yer tutur.

## Polinomial funksiyalar niyə tətbiq olunur geniş izah

**Polinomial funksiyalar** müxtəlif sahələrdə geniş tətbiq olunur və bir çox mühüm funksional məqsədləri yerinə yetirir. Aşağıda polinomial funksiyalarının tətbiq sahələri və nə iş gördüyü barədə ətraflı izah verəcəyəm:

### 1. **Verilənlərə Uyğunlaşma (Curve Fitting):**
Polinomial funksiyalar, verilənlərə ən yaxşı uyğunlaşma tapmaq üçün tez-tez istifadə olunur. Məsələn:
- **Data fitting:** Təsadüfi məlumatlar (data points) üzərində polinomial bir xətt çəkmək verilənlərdəki ümumi meyli və əlaqəni tapmağa kömək edir. Bu, xüsusi ilə statistik analizlərdə, təhlillərdə, və model qurma proseslərində istifadə edilir.
- **Polinomial regresiya:** Polinomial regresiya, verilənlərdəki qeyri-xətti əlaqələri modelləşdirmək üçün istifadə olunur. Məsələn, 2-ci dərəcəli bir polinomial ($ f(x) = ax^2 + bx + c $) verilənlərə uyğun gələn ən yaxşı cədvəli tapmaq üçün istifadə edilə bilər.

### 2. **Optimallaşdırma və Maksimum/Minum Tapma:**
Polinomial funksiyalar optimallaşdırma problemlərində istifadə olunur. Məsələn:
- **Maksimum və ya minimum nöqtələrin tapılması:** Polinomial funksiyalar, bir funksiyanın maksimum və ya minimum nöqtələrini tapmaq üçün diferensial üsullardan istifadə edilərək təhlil edilir. Bu cür optimallaşdırma məsələləri mühəndislik, iqtisadiyyat və maliyyə sahələrində geniş istifadə olunur.
- **Ən yaxşı həllərin tapılması:** Məsələn, bir istehsal prosesinin optimallaşdırılması və ya maliyyə risklərinin idarə edilməsi kimi məsələlərdə polinomial funksiyalar optimallaşdırma məqsədilə istifadə olunur.

### 3. **Fizika və Mexanika:**
Polinomial funksiyalar təbii hadisələrin modelləşdirilməsi və fiziki qanunların izahı üçün istifadə olunur. Məsələn:
- **Hərəkət qanunları:** Fizikada, cismin hərəkətinin müəyyən edilməsi üçün polinomial funksiyalar istifadə oluna bilər. Məsələn, bir cismin sürətinin zamanla necə dəyişdiyini təyin etmək üçün polinomial funksiyalar istifadə edilə bilər.
- **Qüvvə və enerji:** Mexanika və elektromaqnetizm kimi sahələrdə polinomiallar təzyiq, sürət, temperatur və digər parametrlər arasında əlaqəni müəyyən etmək üçün istifadə edilə bilər.

### 4. **Komputer Qrafikası və Şəkil Emalı:**
Polinomial funksiyalar komputer qrafikasında və şəkil emalında geniş istifadə olunur. Məsələn:
- **Ekranlar və cizgilərin düzülməsi:** Polinomiallar çox vaxt şəkil və qrafiklərdə konturların və əyrilərin təyin edilməsi üçün istifadə olunur. Məsələn, Bézier əyriləri və ya splines kimi polinomial əsaslı əyrilər istifadə olunur.
- **Şəkil və video analizi:** Polinomiallar şəkil və video emalı sahələrində də istifadə olunur, məsələn, şəkil təkmilləşdirmə, noise azaldılması və ya təbii hadisələrin simulyasiyası üçün.

### 5. **Sahə Hesablamaları və İntegrasiya:**
Polinomial funksiyalar sahə hesablama və inteqrasiya məsələlərində istifadə olunur. Məsələn:
- **Sahə altında inteqrasiya:** Polinomial funksiyanın inteqralı, həmin polinomialın yaratdığı sahəni tapmağa kömək edir. Məsələn, bir əyrinin altındakı sahə, bir polinomialın inteqralı vasitəsilə hesablanır.
- **Yaxınlaşma və təxmini hesablamalar:** Polinomial funksiyalar, xüsusilə daha mürəkkəb funksiyaların yaxınlaşma təhlilində istifadə olunur. Məsələn, **Taylor seriyası** kimi metodlar polinomial ifadələr vasitəsilə kompleks funksiyaların yaxınlaşmalarını tapmaq üçün istifadə edilir.

### 6. **Kompleks Proseslərin Modelləşdirilməsi:**
Bəzi mürəkkəb və çoxfaktorlu proseslər polinomial funksiyalar vasitəsilə modelləşdirilir. Məsələn:
- **İqtisadiyyat və maliyyə:** İqtisadi göstəricilərin modelləşdirilməsi üçün polinomial funksiyalar istifadə olunur. Məsələn, maliyyə bazarlarının analizində müəyyən göstəricilərin (məsələn, inflyasiya, faiz dərəcələri) polinomial modelləri istifadə edilir.
- **Sosial elm və davranış analizi:** Polinomial funksiyalar insan davranışlarının və sosial hadisələrin təhlili üçün istifadə oluna bilər.

### 7. **Biologiya və Tibb:**
Polinomial funksiyalar bəzən bioloji və tibbi təhlillərdə istifadə olunur:
- **Genetik analiz:** Biologiyada, genetik məlumatların təhlilində polinomial funksiyalar, xüsusilə xəstəliklərin və genetik xüsusiyyətlərin modelləşdirilməsi üçün istifadə edilə bilər.
- **Tibbi görüntüləmə:** Polinomiallar tibbi şəkillərin analizində və təhlilində, məsələn, xəstəliklərin təhlili və müalicə planlarının hazırlanmasında istifadə edilə bilər.

### Nəticə:
**Polinomial funksiyalar** çoxsaylı sahələrdə geniş tətbiq olunur və mühüm funksiyalar yerinə yetirir. Onlar verilənlərə uyğunlaşma, optimallaşdırma, fiziki və tibbi modelləşdirmə, qrafik və şəkil emalı kimi müxtəlif sahələrdə istifadə olunur. Polinomialların təbiətini və necə işlədiyini başa düşmək, mürəkkəb problemləri həll etmək və təbii hadisələri modelləşdirmək üçün mühüm bir alətdir.

## `np.polyfit(x_train, y_train, 3)` nədir?

`np.polyfit(x_train, y_train, 3)` funksiyası **NumPy** kitabxanasından istifadə edərək verilən `x_train` və `y_train` verilənlər dəstinə uyğun olaraq 3-cü dərəcəli bir polinomu yerinə yetirir. Bu funksiya aşağıdakıları həyata keçirir:

1. **Polinom Əmsallarını Hesablamaq**: `np.polyfit()` verilənləri ən yaxşı şəkildə uyğunlaşdırmaq üçün polinomial regresiya yerinə yetirir. `3` dərəcəsi ilə uyğunlaşdırma aparılacaq, yəni üçüncü dərəcəli (kubik) polinom qurulacaq.

2. **Yalnız Polinomun Əmsallarını Qaytarır**: Bu funksiya, ən yaxşı uyğunlaşan polinomun əmsallarını (koeffisentlərini) geri qaytaracaq. Nəticə bir siyahı (və ya numpy array) olacaq ki, bu da 3-cü dərəcəli polinomun əmsallarını ehtiva edir. Polinom aşağıdakı formadadır:

$$
P(x) = a_3x^3 + a_2x^2 + a_1x + a_0
$$

Burada $a_3$, $a_2$, $a_1$, $a_0$ polinomun əmsalları olacaq və bu əmsallar `np.polyfit()` tərəfindən hesablanacaq.

### Misal:

Deyək ki, `x_train` və `y_train` verilənləri belədir:

```python
import numpy as np

# Məsələn verilənlər
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([2, 3, 5, 7, 11])

# 3-cü dərəcəli polinomun əmsallarını hesablayırıq
coefficients = np.polyfit(x_train, y_train, 3)

print(coefficients)
```

Bu kodun nəticəsi belə bir polinomun əmsallarını verə bilər:

```
[ 0.1 -0.3  2.5  1.2]
```

Bu o deməkdir ki, uyğunlaşdırılmış polinom belə olacaq:

$$
P(x) = 0.1x^3 - 0.3x^2 + 2.5x + 1.2
$$

### Nəticə:
- **Polinom Əmsalları**: `np.polyfit(x_train, y_train, 3)` funksiyası 3-cü dərəcəli polinomun əmsallarını (koeffisentlərini) geri qaytarır.
- **Dəqiqlik və Uyğunlaşma**: Bu polinom ən yaxşı şəkildə verilənlərə uyğunlaşdırılmış olur, yəni verilənlərə ən yaxın olan 3-cü dərəcəli polinom tapılır.

---

Bu nəticəni daha sonra `np.polyval()` funksiyası ilə istifadə edib, verilən $x$ dəyərləri üçün polinomun nəticələrini hesablaya bilərsiniz.

### Misal:
Əgər biz $x_{\text{new}} = \text{np.array}([6, 7, 8])$ dəyərləri üçün polinomun nəticələrini hesablamaq istəyiriksə:

```python
import numpy as np

# Əmsallar
coefficients = [0.1, -0.3, 2.5, 1.2]

# Yeni x dəyərləri üçün nəticələr
x_new = np.array([6, 7, 8])
y_new = np.polyval(coefficients, x_new)

print(y_new)
```

Bu kodun nəticəsi belə olacaq:

```
[26.  38.3 53.2]
```

Əgər polinomun formulu $ P(x) = 0.1x^3 - 0.3x^2 + 2.5x + 1.2 $ ilə əl ilə hesablasanız, məsələn, $ x = 6 $, $ x = 7 $, və $ x = 8 $ üçün nəticələr belə olacaq:

$$
P(6) = 0.1 \times 6^3 - 0.3 \times 6^2 + 2.5 \times 6 + 1.2 = 26
$$
$$
P(7) = 0.1 \times 7^3 - 0.3 \times 7^2 + 2.5 \times 7 + 1.2 = 38.3
$$
$$

````python
# NumPy kitabxanasını çağırır
import numpy as np

# Verilmiş x dəyəri üçün polinomial ifadəni hesablayır
# res = k0*x^n + k1*x^(n-1) + ... + kn*x^0
def predict_poly(x, koeff):
    res = 0
    xx = [x ** (len(koeff) - n - 1) for n in range(len(koeff))]  # Hər dərəcə üçün x^n hesablayır

    for i, k in enumerate(koeff):  # Hər əmsalı uyğun x dərəcəsi ilə vurur
        res += k * xx[i]

    return res

# 0-dan 10-a qədər 0.1 addımlı x dəyərləri yaradır
x = np.arange(0, 10.1, 0.1)
# Funksiya: y = 1 / (1 + 10 * x^2) riyazi düsturu ilə y dəyərlərini hesablayır
y = 1 / (1 + 10 * np.square(x))

# Hər ikinci elementi seçərək təlim (train) üçün x və y dəyərlərini ayırır
x_train, y_train = x[::2], y[::2]

# Məlumatın ümumi sayını müəyyən edir
N = len(x)

# Təlim məlumatlarına uyğun 10-cu dərəcəli polinomial funksiyanın əmsallarını tapır
z_train = np.polyfit(x_train, y_train, 3)  # Min kvadrat üsulu ilə polinom uyğunlaşdırır
# Əldə olunan əmsalları çap edir
print(z_train)
````
