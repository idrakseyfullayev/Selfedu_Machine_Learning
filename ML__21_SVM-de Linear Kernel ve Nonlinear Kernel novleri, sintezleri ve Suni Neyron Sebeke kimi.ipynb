#### SVM-in iş prinsipi - İxtiyari 𝑥 nöqtəsinin Support Vectorlarla skalyar hasillərinin cəmi ilə Sinfin müəyyyən edilməsi

#### 1 **Support Vectorlar ilə SVM qərar funksiyasının alınması**.


1. $ a(x) = \text{sign}((\omega , x) - b) $ - **SVM-in qərar funksiyasıdır** — obyektin hansı sinfə aid olduğunu təyin edir.

2. $ \omega = \sum\limits_{i=1}^{l} \lambda_i y_i x_i $ - **SVM-də Çəkilər vektorunun $ ( \omega ) $ ifadəsidir** — o, **Support Vectorların xətti kombinasiyası** kimi yazılır.
$$$$

- İndi **2-ci düsturu 1-ci düsturda ω yerinə qoyuruq**.

$$
a(x) = \text{sign}((\sum_{i=1}^{l}\lambda_i y_i x_i , x) - b)
$$

- **Skalyar hasilin xətti xüsusiyyətinə görə:**

$$
(\sum_{i=1}^{l}\lambda_i y_i x_i , x) = \sum_{i=1}^{l}\lambda_i y_i (x_i , x)
$$


- **$ \omega $ ifadəsini birinci düstura qoymaq və skalyar hasilin xətti xüsusiyyətindən istifadə etməklə Düstur alınır.**

$$
a(x)=sign\left(\sum_{i=1}^{h}\lambda_i y_i \langle x_i,x\rangle - b\right)
$$

- **İzah:**

* $x$ — yeni obyekt
* $x_i$ — **support vectorlar**
* $y_i$ — sinif etiketi (+1 və ya −1)
* $\lambda_i$ — həmin support vectorun təsir gücü
* $\langle x_i,x\rangle$ — **skalyar hasil**

📌 **Qrafik məntiqi:**
- Şəkildə yeni **$x$** nöqtəsi bütün support vectorlarla skalyar hasilə girir.
- Bu nəticələr çəkilər ilə toplanır və **işarəyə görə sinif müəyyən olunur**.

> Bu **SVM qərar funksiyasıdır**. **Yeni $x$ obyektinin sinfi support vectorlarla skalyar hasil hesablanaraq müəyyən olunur.**

---

#### 2 Bu yazılış **bütün support vectorların çoxluğunu** göstərir.

$$
\{{x_i}\}_{i=1}^{h}
$$

**İzah:**


Qrafikdə onlar:

* $x_1,x_2$ → **$(+1)$ sinfi**
* $x_3,x_4$ → **$(−1)$ sinfi**

---

#### 3 Şəkildə sadə göstərmək üçün **bias sıfır götürülüb**.

$$
b=0
$$

**İzah:**


**Belə olduqda qərar funksiyası sadələşir və yalnız skalyar hasillər qalır.**

---

#### 4 Burada **ixtiyari** **$x$** nöqtəsinin **Support Vectorlarla skalyar hasillərinin cəmi** hesablanır.
-Bu ifadə **iki sinfin təsirini ayrıca göstərir**.
$$
\lambda_1\langle x_1,x\rangle
+
\lambda_2\langle x_2,x\rangle
-
\lambda_3\langle x_3,x\rangle
-
\lambda_4\langle x_4,x\rangle
$$

**İzah:**


* $x_1,x_2$ → **$(+1)$ sinifinin support vectorları**
* $x_3,x_4$ → **$(−1)$ sinifinin support vectorları**

📌 Qrafikdə mavi və sarı istiqamətlər həmin təsiri göstərir.

Bu ifadə əslində **iki sinfin təsirini müqayisə edir**.

- ✔ Əgər nəticə **müsbət olarsa → $(+1)$ sinfi**
- ✔ Əgər nəticə **mənfi olarsa → $(−1)$ sinfi**

Başqa cür desək:

> $x$ nöqtəsinin **proyeksiyası hansı sinfin support vectorları istiqamətində daha böyükdürsə**, obyekt **o sinfə aid edilir**.

---

#### 5 **Sinfin ümumi vektorları**
$$$$
#### Bu **müsbət sinfin ümumi vektorudur**.

$$
\omega_+ = \lambda_1 x_1 + \lambda_2 x_2
$$


####  Bu isə **mənfi sinfin ümumi vektorudur**.

$$
\omega_- = \lambda_3 x_3 + \lambda_4 x_4
$$

**İzah:**

> Yəni $(+1)$ və ya $(-1)$ sinifinə aid support vectorların **çəki ilə toplanmış vektoru**.

---

#### 6 **Skalyar hasillərin fərqi**

$$
\langle \omega_+,x\rangle - \langle \omega_-,x\rangle
$$

**İzah:**
**Yeni $x$ nöqtəsinin**

* **(+) sinifinə yaxınlığı**, yəni **(+) sinfin istiqamətinə proyeksiyası**
* **(−) sinifinə yaxınlığı**, yəni **(−) sinfin istiqamətinə proyeksiyası**

**skalyar hasil vasitəsilə müqayisə olunur.**

📌 Qrafikdə yaşıl vektor $x$-i göstərir.

---

#### 7 Son **Qərar funksiyasıdır.**

$$
a(x)=sign\left(\langle \omega_+,x\rangle-\langle \omega_-,x\rangle\right)
$$

**İzah:**


* nəticə **müsbət → $(+1)$ sinfi**
* nəticə **mənfi → $(−1)$ sinfi**

---

✅ **Qrafikin əsas ideyası:**

Yeni obyekt **$x$**

- 1️⃣ bütün **support vectorlarla skalyar hasilə** girir
- 2️⃣ bu nəticələr **λ ilə çəkilir**
- 3️⃣ **iki sinfin təsiri müqayisə olunur**
- 4️⃣ **sign funksiyası sinfi müəyyən edir**

$$$$

💡 **Bir cümləlik ən düzgün məntiq:**

> **SVM-də ixtiyari $x$ nöqtəsinin sinfi, onun support vectorlarla skalyar hasilinin çəkili cəminin işarəsinə görə müəyyən olunur.**



### SVM-də Kernel - Qeyri-xətti Nüvə-lərin izahı

> **SVM Qeyri-Xətti ayrılan verilənlər çoxluğunu emal etmək üçün **Nüvə fəndindən (Kernel trick)** istifadə edir.**

#### 1. **Təsnifat funksiyası (Decision Function)**

$$a(x) = \text{sign} \left( \sum_{i=1}^{h} \lambda_i y_i \langle x_i, x \rangle - b \right)$$

* **Mənası:** 
> **Yeni gələn $x$ obyektinin hansı sinfə (müsbət yoxsa mənfi) aid olduğunu təyin edən düsturdur.**

* **$\langle x_i, x \rangle$:** — **Support Vectot-lar ilə yeni obyekt arasındakı skalyar hasildir. SVM-in bütün gücü bu hasilə söykənir.**
* **$\lambda_i$** — **Laqranj vuruqları**
* **$y_i$** — **sinif etiketləri**

> **Bu düstur yeni obyektin ($x$) təsnifatını təyin edir.**

---

#### 2. **Skalyar Hasilin Modifikasiyası - Xətti və Kvadratik münasibətlər**

- **Biz məlumatı sadəcə xətti $\langle x_i, x \rangle$ deyil, həm də kvadratik $\langle x_i, x \rangle^2$ şəkildə hesablaya bilərik. Bu, düz xətlə ayrıla bilməyən nöqtələri daha yüksək ölçülü sahədə ayırmağa kömək edir.**

> **Aşağıdakı münasibətlər xətti fəzadan qeyri-xətti fəzaya keçidi ifadə edir:**

* $\langle x_i, x \rangle = x_i^T \cdot x, \quad i=1,2,...,h$ (**Xətti asılılıq**)
* $\langle x_i, x \rangle^2 = (x_i^T \cdot x)^2, \quad i=1,2,...,h$ (**Kvadratik asılılıq**)

> **Sistemin Laqranj funksiyasından istifadə etməklə **Karuş-Kuna-Takker (KKT)** şərtləri əsasında həlli dəyişməz qalır, yalnız skalyar hasil nüvə funksiyası ilə əvəzlənir.**

---

#### 3. **Nüvə Funksiyası və Mercer Şərti**

$$K(x, x')$$

$$$$

- **Məqsəd:** 
> **Düz xətlə bölünməyən datanı əyri xətlə bölmək.**

- **$K(x, x')$ mənası?** 
> **İki nöqtənin yüksək ölçülü fəzada "nə qədər yaxın" olduğunu göstərən ölçüdür. İki obyekt arasındakı "oxşarlığı" ölçür.**

- **Niyə $K$-dan istifadə edirik?** 
> **Ağır hesablama aparmadan, sadəcə bir düsturla qeyri-xətti sahəyə keçid etmək üçün.**

- **Məntiqi:** 
> **Biz nöqtələri fiziki olaraq 3D və ya 4D-yə köçürmürük (bu çox çətindir). Bunun əvəzinə, sanki onları köçürmüşük kimi aralarındakı skalyar hasili hesablayan bir "sehrli" funksiya ($K$) istifadə edirik.**


$$$$

> **$K(x, x')$ funksiyasının SVM məsələsi üçün yararlı bir nüvə (kernel) sayılması üçün o, simmetrik $K(x, x') = K(x', x)$ və müsbət müəyyən olunmuş olmalıdır, yəni bu funksiya ilə qurulan matris heç vaxt mənfi nəticə verməməlidir**

- **Funksiyanın Nüvə olması üçün zəruri və kafi şərt - Mercer teoremi aşağdakı kimidir**:
- **İstənilən $g : X \to \mathbb{R}$ funksiyası üçün aşağıdakı şərt ödənilməlidir:**


$$\int_X \int_X K(x, x') \ g(x) \ g(x') \ dx \ dx' \geq 0$$

* **Niyə bu formaya düşdü?** 
> **Çünki inteqralın $\geq 0$ olması funksiyanın riyazi olaraq "skalyar hasil" kimi davranacağını zəmanət altına alır.**

**İzahı**
* $X$ — **inteqralın götürüldüyü sahə**
* $x, x'$ — **iki dəyişən (ikiqat inteqral üçün)**
* $K(x,x')$ — **kernel funksiyası, obyektlər arasındakı əlaqəni göstərir**
* $g(x), g(x')$ — **istənilən funksiyalar**
* $dx,dx'$ — **inteqralın ölçü vahidləri**
* $\ge 0$ — **nəticənin həmişə qeyri-mənfi olduğunu göstərir (kernel **müsbət yarı-müəyyən** olduğu üçün)**


> **Bu inteqral şərti təmin edir ki, seçilmiş $K$ funksiyası riyazi olaraq yüksək ölçülü hansısa fəzadakı skalyar hasilə ekvivalentdir.**

---

#### 4. **Nümunə Nüvə Funksiyası**

$$K(x, x') = \langle x, x' \rangle^2$$


> **Bu, **Polinominal Nüvənin** xüsusi halıdır. O, giriş verilənlərini birbaşa hesablama aparmadan daha yüksək dərəcəli fəzada təsvir etməyə və beləliklə, xətti ayrılması mümkün olmayan nöqtələr arasında əyri xətt (hiper-səth) çəkməyə imkan verir. Nöqtələri əyri xətlərlə (məsələn, çevrə və ya ellips) ayırmaq üçün istifadə olunur.**

---

#### 5. **Nüvə Fəndinin (Kernel Trick) Riyazi Sübutu**

- **"Nüvə fəndi"nin necə işlədiyini addım-addım sübut edək. Biz ikiölçülü ($2D$) fəzadakı nöqtələri birbaşa üçölçülü ($3D$) fəzaya köçürmədən, sanki oradaymışıq kimi hesablama aparırıq.**

$$K(x, x') = \langle x, x' \rangle^2$$

**Verilənlər:**

* $u = [u_1, u_2]^T$ və $v = [v_1, v_2]^T$ (**İkiölçülü giriş vektorları**)
* $K(u, v) = \langle u, v \rangle^2$ (**Kvadratik nüvə funksiyası**)

**Hesablama mərhələləri:**

1. **Skalyar hasilin kvadratı:**
- **İki vektorun skalyar hasilini tapıb kvadrata yüksəldirik:**

$$\langle u, v \rangle^2 = (u_1v_1 + u_2v_2)^2$$


2. **Cəbri açılış:**
- **Müxtəsər vurma düsturu ilə ifadəni açırıq:**

$$(u_1v_1 + u_2v_2)^2 = u_1^2v_1^2 + u_2^2v_2^2 + 2u_1v_1u_2v_2$$


3. **Vektorların ayrılması (Mapping function $\psi$):**
- **Bu ifadəni iki yeni vektorun skalyar hasili şəklində qruplaşdıra bilərik:**

$$u_1^2v_1^2 + u_2^2v_2^2 + (\sqrt{2}u_1u_2)(\sqrt{2}v_1v_2) = \left\langle \begin{bmatrix} u_1^2 \\ u_2^2 \\ \sqrt{2}u_1u_2 \end{bmatrix}, \begin{bmatrix} v_1^2 \\ v_2^2 \\ \sqrt{2}v_1v_2 \end{bmatrix} \right\rangle$$

- **Nəticə:**

> **Bu riyazi çevrilmə bizə göstərir ki, $2D$ fəzadakı $K(u, v)$ hesabı, əslində $3D$ fəzadakı $\psi(u)$ və $\psi(v)$ vektorlarının skalyar hasilinə bərabərdir:**

- **$\psi(u) = [u_1^2, \ u_2^2, \ \sqrt{2}u_1u_2]^T$** $ \ $ - $ \ $ **3 Ölçülü Vector**
- **$\psi(v) = [v_1^2, \ v_2^2, \ \sqrt{2}v_1v_2]^T$** $ \ $ - $ \ $ **3 Ölçülü Vector**

- **Əsas ideya:** 
> **Bizə bu $3D$ vektorları (sağ tərəf) bir-bir hesablamaq lazım deyil. Biz sadəcə sol tərəfdəki sadə düsturu ($K$) hesablayırıq və "Nüvə fəndi" sayəsində avtomatik olaraq yüksək ölçülü fəzada qeyri-xətti ayrılmanı (məsələn, ellips və ya hiperbola ilə) əldə edirik.**

- **Üstünlüyü:**  
> **Sürət: Milyonlarla ölçülü fəzalarda belə hesablama yükü artmır**.

- **Effektivlik:** 
> **Yaddaşa qənaət edilir, çünki yeni vektorları saxlamağa ehtiyac qalmır.**

----

#### 6 **"Nüvə fəndi"nin (Kernel Trick) ən kritik məqamı.** **İfadəni iki tərəfə bölərək tam aydınlaşdıraq:**

$$$$

#### **Sol tərəf (Kernel funksiyası - Sadə yol)**

**Bərabərliyin ən solunda dayanan və bizim əslində hesabladığımız hissədir:**


$$K(u, v) = (u_1v_1 + u_2v_2)^2$$

* **Mənası** 
> **Bu, sadəcə ikiölçülü ($2D$) vektorların elementlərini bir-birinə vurub kvadratını almaqdır.**
* **Üstünlüyü:** 
> **Çox sürətlidir. Burada heç bir yeni fəzaya keçid yoxdur, sadəcə 2 dənə rəqəmlə ( $u_1, u_2$ ) işləyirik.**

$$$$

#### **Sağ tərəf (Feature Mapping - Ağır yol)**

**Bərabərliyin ən sağında, mötərizə daxilində göstərilən uzun vektorlardır:**


$$\left\langle \begin{bmatrix} u_1^2 \\ u_2^2 \\ \sqrt{2}u_1u_2 \end{bmatrix}, \begin{bmatrix} v_1^2 \\ v_2^2 \\ \sqrt{2}v_1v_2 \end{bmatrix} \right\rangle$$

* **Mənası** 
> **Bu, bizim $2D$ nöqtələrimizi $3D$ fəzasına (3 komponentli yeni vektorlara) köçürülmüş halıdır.**
* **Çətinliyi:** 
> **Əgər biz hər bir nöqtəni əvvəlcə bu formaya gətirib ($u \to \psi(u)$), sonra onların skalyar hasilini hesablasaq, bu çox vaxt və yaddaş aparacaq. Xüsusilə minlərlə ölçülü fəzalarda bu, kompüteri "dondura" bilər.**

$$$$

**Xülasə: Sehr nədədir?**

- **Riyazi olaraq sübut etdik ki:**

> **Sol tərəf (Sadə kvadrat düsturu) $=$ Sağ tərəf (Yüksək ölçülü skalyar hasil)**

> **Yəni biz sol tərəfdəki sadə hesablama aparmaqla, əslində sağ tərəfdəki kimi mürəkkəb $3D$ fəzada nöqtələri bir-birindən ayırmış oluruq. Biz heç vaxt o $3D$ vektorları fiziki olaraq yaratmırıq, amma onların arasındakı "əlaqəni" (skalyar hasili) tapmış oluruq.**

**Nüvə fəndi** elə budur: **Mürəkkəb fəzaya getmədən, o fəzanın nəticələrini əldə etmək.**

---

#### 7 **Nümunə**
**Biz 3D fəzaya fiziki olaraq keçmirik, sadəcə 2D-də qalıb 3D-nin "effektini" alırıq.**

Gəlin bunu konkret rəqəmlərlə, çox sadə şəkildə müqayisə edək:

Tutaq ki, iki nöqtəmiz var: $u = [1, 2]^T$ və $v = [3, 4]^T$.
<br><br>
####  **Sol tərəf (Kernel fəndi - BİZİM ETDİYİMİZ)**

Biz heç bir yeni vektor yaratmırıq. Sadəcə əlimizdəki rəqəmləri düstura qoyuruq:

* Əvvəlcə skalyar hasil: $(1 \cdot 3) + (2 \cdot 4) = 3 + 8 = 11$
* Sonra kvadratı: $11^2 = \mathbf{121}$

**Nəticə:** Cəmi 2-3 saniyəyə **121** rəqəmini aldıq.
<br><br>
####  **Sağ tərəf (Fiziki keçid - BİZİM ETMƏDİYİMİZ)**

Əgər biz "fənddən" istifadə etməsəydik, hər iki nöqtəni əvvəlcə 3D-yə köçürməli idik:

* $\psi(u) = [1^2, 2^2, \sqrt{2} \cdot 1 \cdot 2] = [1, 4, 2.82]$
* $\psi(v) = [3^2, 4^2, \sqrt{2} \cdot 3 \cdot 4] = [9, 16, 16.97]$

İndi isə bu yeni 3D vektorların skalyar hasilini tapaq:

* $(1 \cdot 9) + (4 \cdot 16) + (2.82 \cdot 16.97) = 9 + 64 + 47.85 \approx \mathbf{121}$


#### **Fərq nədədir?**

* **Sol tərəfdə:** Biz ancaq **11** və **121** rəqəmləri ilə işlədik. Yaddaşda yeni yer açmadıq, kökaltı ($\sqrt{2}$) hesablamadıq, yeni vektorlar qurmadıq.
* **Sağ tərəfdə:** Biz hər nöqtə üçün 3 yeni komponent hesabladıq, onları yaddaşda saxladıq və sonra daha uzun bir vurma əməliyyatı etdik.

**İndi təsəvvür edin:** Reallıqda biz $2D$-dən $3D$-yə yox, $100$ ölçüdən **milyonlarla ölçülü** fəzalara keçirik. O milyonlarla yeni komponenti (sağ tərəfi) hesablamaq kompüter üçün qeyri-mümkündür. Amma sol tərəfdəki sadə düstur ($K$) həmişə eyni sürətlə işləyir.

**Qısa cavab:** Şəkildəki sağ tərəf (3D ifadə) sadəcə bir **isbatdır**. Kompüter o isbatı icra etmir, o sadəcə sol tərəfdəki kvadratı hesablayır və "sehrli" şəkildə 3D-nin nəticəsini əldə edir.


### SVM-də əsas Kernel növləri (funksiyaları)


> **Kernel funksiyasının əsas məqsədi, xətti olaraq ayrıla bilməyən məlumatları daha yüksək ölçülü bir fəzaya köçürməklə onları xətti ayrıla bilən hala gətirməkdir. Buna "Kernel Trick" deyilir.**
<br><br>
#### 1. **Linear Kernel (Xətti Kernel)**
- **Düstur:**
$$K(x, x') = \langle x, x' \rangle$$

- $x$ – datasetdə olan **bir obyektin xüsusiyyət vektoru**
- $x′$ – datasetdə olan **başqa (və ya eyni) obyektin xüsusiyyət vektoru** *(SVM-də çox vaxt bu vektor **support vector** olur, amma mütləq deyil)*
* **$⟨x, x′⟩$** — **bir obyektin xüsusiyyət vektorunun digər obyektin xüsusiyyət vektoru (və ya support vector-u) ilə skalyar hasilidir (dot product). İki obyektin xüsusiyyət vektorlarının skalyar hasilini hesablayan lineer kernelidir; nəticə onların oxşarlığını ölçür və SVM-də xətti qərar sərhədləri yaratmaq üçün istifadə olunur.**

<br>

* **İzahı:** 
> **Ən sadə kernel növüdür. Məlumatlar arasında birbaşa skalyar hasili hesablayır.**
* **Nə zaman istifadə olunur:** 
> **Məlumatlar artıq xətti olaraq ayrıla biləndirsə və ya xüsusiyyətlərin (features) sayı çox lakin daxilolmaların (observations) sayı az olduqda (məsələn, mətn təsnifatı) effektivdir.**

- **Sadə skalyar hasil (dot product).**
- **Heç bir transformasiya etmir → data olduğu kimi qalır.**
- **Nəticə: xətti ayrılan (linear separable) problemlər üçün.**
<br><br>
#### 2. **Polynomial Kernel (Polinomial Kernel)**
- **Düstur:** 
$$K(x, x') = \langle x, x' \rangle^d$$ və ya $$K(x, x') = (\langle x, x' \rangle + 1)^d$$

* **$x$** – datasetdə olan **bir obyektin xüsusiyyət vektoru**
* **$x′$** – datasetdə olan **başqa (və ya eyni) obyektin xüsusiyyət vektoru** *(SVM-də çox vaxt **support vector** olur, amma mütləq deyil)*
* **$⟨x, x′⟩$** — **bir obyektin xüsusiyyət vektorunun digər obyektin xüsusiyyət vektoru (və ya support vector-u) ilə skalyar hasilidir (dot product)**.
* **$d$** – **polinomun dərəcəsi (modelin mürəkkəbliyini müəyyən edir)**
* **$1$** – **sabit ədəd (bias kimi çıxış edir)**. Yalnız skalyar hasil yox, həm də **sabit terminlər (intercept)** nəzərə alınsın. **Model daha çevik olsun (sıfır olmayan hallarda da işləsin)**. $+1$ → **modeli zənginləşdirir (daha güclü edir)**
* **$⟨x, x′⟩ᵈ$** — **iki obyektin xüsusiyyət vektorlarının skalyar hasilini $d$-ci dərəcə qüvvətləndirərək polinomial (polynomial) kernel yaradır, SVM-də qeyri-xətti qərar sərhədləri üçün istifadə olunur.**
* **$(⟨x, x′⟩ + 1)ᵈ$** — **skalyar hasilə 1 əlavə edərək və $d$-ci qüvvətləndirərək daha geniş polinomial kernel təmin edir, bu da SVM-də əlavə fleksibil qərar sərhədləri yaratmağa imkan verir. Nəticə qeyri-xətti oxşarlığı ölçür və SVM-də mürəkkəb qərar sərhədləri yaratmaq üçün istifadə olunur.**


<br>

* **İzahı:**
> **Məlumatları $d$ dərəcəli polinom fəzasına köçürür. "Homogen olmayan" polinom kernelidir.**
* **Nə zaman istifadə olunur:**
> **Görüntülərin emalında və məlumatlar arasında qeyri-xətti əlaqələr olduqda istifadə edilir.**

- **Xüsusiyyətləri **dərəcə (degree)** qədər genişləndirir.**
- **Modelə qeyri-xətti sərhədlər öyrənməyə imkan verir.**
- **d böyük → model daha mürəkkəb**
<br><br>
#### 3. **RBF (Radial Basis Function) / Gaussian Kernel**
- **Düstur:**
$$K(x, x') = \exp(-\gamma \|x - x'\|^2)$$

* **$x$** – datasetdə olan **bir obyektin xüsusiyyət vektoru**
* **$x′$** – datasetdə olan **başqa (və ya eyni) obyektin xüsusiyyət vektoru** *(SVM-də çox vaxt **support vector** olur, amma mütləq deyil)*
* **$‖x − x′‖²$** — iki obyektin xüsusiyyət vektorları arasındakı **Evklid məsafəsinin kvadratı**; SVM-də x′ çox vaxt **support vector** olur, amma mütləq deyil.
* **$γ (gamma)$** – **məsafənin təsirini tənzimləyən parametr (böyük olduqca təsir daha lokal olur)**
* **$exp(−γ‖x − x′‖²)$** — **iki obyektin xüsusiyyət vektorları arasındakı Evklid məsafəsinin kvadratını $-\gamma$ ilə çarpıb eksponent funksiyasına tətbiq edən RBF (Radial Basis Function) kernelidir; məsafəsi kiçik olan vektorlar üçün nəticə 1-ə yaxın, uzaq olanlar üçün 0-a yaxın olur və SVM-də qeyri-xətti qərar sərhədləri yaratmaq üçün istifadə olunur.**

<br>

💡 **Niyə modul $‖x - x′‖$?**

* İki nöqtə arasındakı **məsafəni** ölçmək üçün
* Oxşarlıq məsafəyə görə təyin olunur

<br>

💡 **Niyə kvadrat $‖x - x′‖² \ $?**

* **Mənfi işarələri aradan qaldırır (məsafə həmişə müsbət olur)**
* Riyazi olaraq daha rahat və hamardır (diferensiallana bilir)**
* **Böyük məsafələri daha çox “cəzalandırır”**

<br>

💡 **Niyə exp (eksponential)?**

* Oxşarlığı **0 ilə 1 arasında** sıxır
* **Yaxın nöqtələr → $K \approx 1$**
* **Uzaq nöqtələr → $K \approx 0$**
* Yəni **məsafə artdıqca oxşarlıq sürətlə azalır**

<br>

🔑 **Qısa nəticə:**
- **RBF kernel → “nə qədər yaxınsan, o qədər oxşarsan” prinsipi ilə işləyir.**

<br>

* **İzahı:**
> **Bu kernel ən populyar olanıdır. İki nöqtə arasındakı məsafəyə əsaslanır. Nöqtələr bir-birinə yaxın olduqda dəyər 1-ə, uzaq olduqda isə 0-a yaxınlaşır.**
* **Nə zaman istifadə olunur:**
> **Məlumatın paylanması haqqında öncədən məlumat olmadıqda "default" seçimdir. Sonsuz ölçülü fəza yaratmaq gücünə malikdir.**

- **İki nöqtə arasındakı məsafəyə baxır.**
- **Yaxın nöqtələr → böyük dəyər, uzaq nöqtələr → kiçik dəyər.**
- **Ən güclü və çox istifadə olunan kernel**
- **Lokal qərar sərhədləri qurur.**
<br><br>
#### 4. **Sigmoid (Hyperbolic Tangent) Kernel**
- **Düstur:**
$$K(x, x') = \tanh(k_1 \langle x, x' \rangle - k_0)$$

* **$x$** – datasetdə olan **bir obyektin xüsusiyyət vektoru**
* **$x′$** – datasetdə olan **başqa (və ya eyni) obyektin xüsusiyyət vektoru** *(SVM-də çox vaxt **support vector** olur, amma mütləq deyil)*
* **$⟨x, x′⟩$** — **bir obyektin xüsusiyyət vektorunun digər obyektin xüsusiyyət vektoru (və ya support vector-u) ilə skalyar hasilidir (dot product)**.

* **$k₁$** – **skalyar hasilin təsirini gücləndirən əmsal**
* **$k₀$** – **bias (sabit sürüşmə) parametri**
* **$tanh(k₁⟨x, x′⟩ − k₀)$** — **iki obyektin xüsusiyyət vektorlarının skalyar hasilini qeyri-xətti şəkildə (tanh funksiyası ilə) transformasiya edərək onların oxşarlığını ölçən kernel funksiyasıdır; SVM-də mürəkkəb sərhədlər qurmaq üçün istifadə olunur. İki obyektin xüsusiyyət vektorlarının skalyar hasilini (k₁) ilə çarpıb (k₀) ilə azaldaraq hiperbollik tangens funksiyasına tətbiq edən sigmoidal kerneldir; nəticə $-1$ ilə $1$ arasında dəyişir və SVM-də qeyri-xətti sərhədləri yaratmaq üçün istifadə olunur.**


<br>

💡 **tanh niyə istifadə olunur?**

* Dəyərləri **-1 ilə 1 arasında sıxır**
* **Qeyri-xətti transformasiya verir**
* **Neyron şəbəkələrdəki aktivasiya funksiyasına bənzəyir**

<br>

🔑 **Qısa məntiq:**
- Bu kernel → **linear + qeyri-xətti aktivasiya** kimi işləyir.

<br>

* **İzahı:**
> **Süni neyron şəbəkələrində istifadə olunan aktivasiya funksiyasına bənzəyir.**
* **Nə zaman istifadə olunur:**
> **SVM-i bir növ neyron şəbəkəsi kimi işlətmək istədikdə istifadə olunur, lakin hər zaman "müsbət yarı-müəyyən" (positive semi-definite) olmadığı üçün ehtiyatla istifadə edilməlidir.**

- **Mövcud kernel üzərində əlavə transformasiya.**
- **Daha elastik model qurmaq üçün istifadə olunur.**
- **Məsələn: threshold, scaling və s.**
<br><br>
#### **Xülasə Cədvəli:**

| Kernel Növü | Mürəkkəblik | Əsas İstifadə Sahəsi |
| :--- | :--- | :--- |
| **Linear** | Aşağı | Sadə xətti ayırma - Mətn təsnifatı, sürətli hesablamalar |
| **Polynomial** | Orta | Qeyri-xətti (dərəcə ilə)  - Şəkil tanıma |
| **RBF (Gaussian)** | Yüksək | Ümumi təyinatlı, ən güclü kernel - Məsafəyə əsaslanan güclü model |
| **Sigmoid** | Orta | Kernel-i dəyişib daha elastik edir - Neyron şəbəkə bənzətmələri |

> Bu kernellərdən hansının sizin modeliniz üçün daha uyğun olduğunu yoxlamaq üçün **Cross-Validation** metodundan istifadə etməyiniz məsləhətdir. 



#### Sintez Kernel “yaratma qaydalarıdır” → məqsəd SVM və digər kernel metodlarını daha güclü və çevik etməkdir

> Bunlar **kernel (nüvə) funksiyalarını qurma üsullarıdır** — yəni yeni kernel-lər yaratmaq üçün qaydalar:


* **$K(x,x')=\langle x,x' \rangle$** – **iki vektorun skalyar hasilidir və sadə xətti oxşarlığı ölçmək üçün istifadə olunur və linearly separable datalarda linear SVM-də tətbiq edilir.**

* **$K(x,x')=1$** – **sabit kerneldir** və bütün nöqtələri eyni sayaraq **modelə yalnız bias (sabit təsir) əlavə etmək** üçün istifadə olunur və **linear modellərdə** tətbiq oluna bilər.

* **$K(x,x')=K_1(x,x') \cdot K_2(x,x')$** – **iki kernelin hasilidir** və **fərqli oxşarlıq ölçülərini birləşdirib daha mürəkkəb model qurmaq** üçün istifadə olunur və **nonlinearly separable problemlərdə** istifadə olunur. **polynomial kernel və ya kompleks kernel birləşməsi**, fərqli oxşarlıq ölçülərini birləşdirərək **polynomial SVM və ya kernel kompozisiyalarında** istifadə olunur.

* **$K(x,x')=\psi(x)\psi(x')$** – **verilənlərə əvvəl funksiya tətbiq edib sonra kernel hesablayır** və **datanı yeni (daha uyğun) feature space-ə keçirmək** üçün istifadə olunur. **funksional kernel**, verilənləri yeni feature space-ə keçirmək üçün istifadə olunur və **polynomial, RBF və ya neural kernel-lərdə** tətbiq oluna bilər.

* **$K(x,x')=\alpha_1 K_1(x,x')+\alpha_2 K_2(x,x')$** – **kernel-lərin cəmidir** və **bir neçə kernel-i çəkilərlə birləşdirərək daha elastik model yaratmaq** üçün istifadə olunur. **mix kernel SVM-lərində** tətbiq olunur və bəzi hallarda **Süni Neyron şəbəkə motivli kernel kimi** istifadə oluna bilər.


### SVM-in üstünlükləri və mənfi cəhətləri



#### **Üstünlükləri:**
* **Burada yeganə minimum nöqtəsi (yeganə həlli) olan qabarıq kvadratik proqramlaşdırma optimallaşdırma məsələsi həll edilir.**
* **Dəstək vektorlarını, o cümlədən təlim seçimində mövcud ola biləcək kənarlaşmaları (outliers) ayırmağa imkan verir (bəzən SVM-dən kənarlaşmaları tapmaq üçün istifadə olunur).**
* **Neyron şəbəkələri ilə müqayisədə, gizli təbəqənin zəruri neyron sayını **dəstək vektorlarının sayı** kimi müəyyən etməyə imkan verir.**

---

#### **Mənfi cəhətləri:**
* **İxtiyari nüvələr (kernels) üçün optimallaşdırma məsələsinə ümumi yanaşmalar yoxdur**.
* **"Daxili əlamət seçimi" (L1-requlyatorunun analoqu) yoxdur**.
* **Hər bir həll edilən tapşırıq üçün C hiperparametrini seçmək (tənzimləmək) lazımdır.**


### Python - Scikit-learn - `from sklearn import svm` $\omega$ (omega)-nı necƏ tapır?

> `from sklearn import svm`-in **əsas iş prinsipi**:

#### 🔵 **LINEAR KERNEL üçün**

<br>

#### 🔹 **1-ci addım: ω (omega) necə tapılır?**

> sklearn **birbaşa ω tapmır**, əvvəl $λ$ tapır:

<br>

#### 🔹 **λ-lərin tapılması**

<br>

- **Soft-Margin SVM-in primal forması, yəni margin genişliyini maksimum edən və səhv nümunələri cərimələyən optimallaşdırma problemi.**

$$
\begin{cases}
\frac{1}{2}||\omega||^2 + C \sum\limits_{i=1}^{l} \xi_i \rightarrow 
\min\limits_{\omega,b,\xi} \\
M_i(\omega,b) \ge 1-\xi_i \\
\xi_i \ge 0
\end{cases}
$$

<br>

* Metod: **Lagrange multipliers (Lagrange çoxlayıcıları)** və **dual transformasiya** tətbiq olunur, yəni **Primal problemdən Lagrange multipliers vasitəsilə çıxarılan dual problem; primal dəyişənləri ($ω$, $b$, $ξ$) çıxarıldıqdan sonra yalnız λ üzərində optimallaşdırma qalır.**
> Nəticədə **SVM-in Linear Kernel üçün Dual Formu**-dur alınır:

$$
\sum_{i=1}^l \lambda_i - \frac{1}{2} \sum_{i=1}^l \sum_{j=1}^l \lambda_i \lambda_j y_i y_j (x_i \cdot x_j) \rightarrow \max_{\lambda}, \quad 0 \le \lambda_i \le C, \ \sum_{i=1}^l \lambda_i y_i = 0 
$$

**və ya**

$$
\frac{1}{2} \sum_{i=1}^l \sum_{j=1}^l \lambda_i \lambda_j y_i y_j (x_i \cdot x_j) - \sum_{i=1}^l \lambda_i \rightarrow \min_{\lambda}, \quad 0 \le \lambda_i \le C, \ \sum_{i=1}^l \lambda_i y_i = 0 
$$
  
- **Qeyd:** Bu bir **quadratic programming (kvadrat proqramlaşdırma) problemi**-dir, çünki məqsəd funksiyası kvadratik, şərtlər isə xətti.
- Sklearn (və digər kitabxanalar) bunu **iterativ optimallaşdırma** ilə həll edir: məsələn, **SMO (Sequential Minimal Optimization)** kimi metodlardan istifadə olunur.

<br>

#### 🔹 **$\omega$-nin hesablanması**

* λ-lər tapıldıqdan sonra **ağırlıq vektoru** belə hesablanır:

$$
\omega = \sum_{i=1}^l \lambda_i y_i x_i
$$

* Burada:

  * **$x_i$** — support vector-lar (yalnız $λ_i$ > 0 olan nümunələr)
  * **$y_i$** — həmin $x_i$-nin sinfi (+1 və ya -1)
  * **$λ_i$** — əvvəl tapılmış Lagrange çarpanı

* Yəni ω, bütün support vector-ların çəkili cəmi kimi alınır.

<br>

✅ **Qısa xülasə:**

1. λ-lər **quadratic programming** və ya **iterativ optimallaşdırma** ilə tapılır.
2. **Tapılan λ-lər və onların müvafiq $x_i$, $y_i$-ləri ω-nin hesablanmasında istifadə olunur.**
3. **Nəticədə ω, SVM-in linear sərhədini təyin edir.**

---

#### 🔹 **2-ci addım: Loss funksiyası**

$$
\text{Loss}_i = \max(0, \ 1 - y_i(\omega \cdot x_i + b))
$$

- **İzah:**

- Bu **hinge loss** adlanır, SVM üçün standartdır.
- Məqsəd: hər nümunənin düzgün siniflə ayrılması.

  * Əgər $y_i(\omega \cdot x_i + b) \ge 1$, loss = 0 → düzgün ayrılıb.
  * Əks halda, loss > 0 → səhv və ya margin içindədir.

> **Sklearn bunu dual optimallaşdırma yolu ilə həll edir, biz birbaşa loss-u optimize etmirik.**

<br>

#### 🔹 Ümumi optimizasiya funksiyası

$$
\frac{1}{2} |\omega|^2 + C \sum_i \max(0, \ 1 - y_i(\omega \cdot x_i + b))
$$

* Burada:

  - 1.$\frac{1}{2}|\omega|^2$ → **margin-in genişliyini maksimum etmək üçün**, daha böyük margin → daha kiçik ω.
  - 2.$C \sum \text{Loss}_i$ → **səhvlərə cəza**, C nə qədər böyükdürsə, model səhvləri daha çox cəzalandırır.

* **Sklearn praktiki olaraq bunu birbaşa hesablamır**.

  * Onlar **dual SVM optimallaşdırması** ($λ$-ləri tapmaq) ilə bu funksiyanı dolayı yolla həll edirlər.
  * Yəni, hinge loss + margin konsepti **daxili olaraq dual problemə çevrilir**.

<br>

✅ **Qısa xülasə:**

- 1. Loss = $max(0, \ 1 - y_i(ω·x_i + b))$ → margin və səhvləri ölçür.
- 2. Ümumi məqsəd: **margin genişliyini artır, səhvləri azald**.
- 3. Sklearn bunu **dual optimallaşdırma** yolu ilə həll edir, biz birbaşa loss-u optimize etmirik.


---

#### 🔹 **3-cü addım: Predict (təxmin)**

$$
\hat{y} = \text{sign}(\omega \cdot x + b)
$$

- **İzah:**

* **$ω · x + b$** → SVM-in xətti sərhədinə görə nöqtənin “mövqeyi”:

  - Əgər nəticə > 0 → sinif +1
  - Əgər nəticə < 0 → sinif -1

* **sign() funksiyası** sadəcə nəticəni **+1 və -1** kimi sinifə çevirir.

<br>

* **Praktiki olaraq:**

  - 1. SVM $λ$-ləri tapır → ω və b hesablanır.
  - 2. Yeni nümunə x verilir.
  - 3. $ω · x + b$ hesablanır.
  - 4. sign funksiyası ilə sinif təyini edilir.

<br>

* **Qeyd:**

  - Bu linear SVM üçündür.
  - Non-linear kernel-lərdə də proses eynidir, amma $ω · x$ yerinə $\sum_i λ_i y_i K(x_i, x)$ istifadə olunur.

<br>

✅ **Qısa xülasə:**

- 1. $ω$ və $b$ artıq təlim zamanı tapılıb.
- 2. Yeni nümunə üçün **dot product + b** hesablanır.
- 3. sign funksiyası ilə **$+1$ və $-1$ sinif təyini** edilir.


---

#### 🔴 **NON-LINEAR KERNEL üçün (RBF, Poly və s.)**

<br>

#### 🔹 **1-ci addım: Niyə ω yoxdur?**

$$
\phi(x)
$$

* **İzah:**

  * **Non-linear SVM-də data yüksək ölçülü (feature) space-ə **$\phi(x)$** ilə xəritələnir.**
  * **Bu space çox böyük ola bilər → **ω birbaşa hesablamaq mümkün deyil**.**
  * **Yəni linear SVM-də olduğu kimi $ω = Σ λ_i y_i x_i$ hesablamaq çətindir, çünki $x_i$ artıq $\phi(x_i)$ ilə əvəz olunur.**

<br>

#### 🔹 **Kernel istifadə olunur**

$$
K(x, x') = \phi(x) \cdot \phi(x')
$$

* **İzah:**

  - Kernel funksiyası **dot product-u** yüksək ölçülü space-də birbaşa hesablayır, amma $\phi(x)$ vektorunu yaratmadan.
  - Bu sayədə **hesablama və yaddaş sərfi azalır**.

<br>

#### 🔹 **λ necə tapılır?**


- **Soft-Margin SVM-in primal forması, yəni margin genişliyini maksimum edən və səhv nümunələri cərimələyən optimallaşdırma problemi.**

$$
\begin{cases}
\frac{1}{2}||\omega||^2 + C \sum\limits_{i=1}^{l} \xi_i \rightarrow 
\min\limits_{\omega,b,\xi} \\
M_i(\omega,b) \ge 1-\xi_i \\
\xi_i \ge 0
\end{cases}
$$

<br>

* Metod: **Lagrange multipliers (Lagrange çoxlayıcıları)** və **dual transformasiya** tətbiq olunur, yəni **Primal problemdən Lagrange multipliers vasitəsilə çıxarılan dual problem; primal dəyişənləri ($\omega$, $b$, $\xi$) çıxarıldıqdan sonra yalnız $\lambda$ üzərində optimallaşdırma qalır.**
> Nəticədə **SVM-in Non-Linear Kernel üçün Dual Formu** alınır:

$$
\sum_{i=1}^l \lambda_i - \frac{1}{2} \sum_{i=1}^l \sum_{j=1}^l \lambda_i \lambda_j y_i y_j K(x_i, x_j) \rightarrow \max_{\lambda}, \quad 0 \le \lambda_i \le C, \ \sum_{i=1}^l \lambda_i y_i = 0 
$$

**və ya**

$$
\frac{1}{2} \sum_{i=1}^l \sum_{j=1}^l \lambda_i \lambda_j y_i y_j K(x_i, x_j) - \sum_{i=1}^l \lambda_i \rightarrow \min_{\lambda}, \quad 0 \le \lambda_i \le C, \ \sum_{i=1}^l \lambda_i y_i = 0 
$$

* **İzah:**

  - Bu **dual problem**-dir, məqsəd λ-ləri tapmaqdır.
  - $λ$-lər yalnız **support vector-lar üçün sıfırdan fərqli** olur.
  - Sklearn və digər kitabxanalar bunu **SMO (Sequential Minimal Optimization)** və ya digər **iterativ optimallaşdırma** metodları ilə tapır.

<br>

✅ **Qısa xülasə:**

- 1. **Non-linear SVM-də ω birbaşa yoxdur → $\phi(x)$ çox böyükdür**.
- 2. **Kernel $K(x,x')$ ilə ω-nin yerinə dot product hesablanır.**
- 3. **λ-lər iterativ olaraq (SMO ilə) tapılır → support vector-lar müəyyənləşir.**

---


#### 🔹 **2-ci addım: Loss funksiyası (Non-linear / Kernel SVM)**

$$
f(x_i) = \sum_j \lambda_j y_j K(x_j, x_i) + b
$$

* **İzah:**

  > **ω artıq birbaşa yoxdur, ona görə $x_i$ ilə kernel dot product vasitəsilə $f(x_i)$ hesablanır.**
  * Burada:

    - **$x_j$** → **support vector-lar**
    - **$λ_j$** → **SMO ilə tapılmış Lagrange çarpanları**
    - **$y_j$** → **support vector-ların sinfi**
    - **$K(x_j, x_i)$** → **kernel funksiyası (dot product yüksək ölçülü space-də)**

<br>

$$
\text{Loss}_i = \max(0, 1 - y_i f(x_i))
$$

* **İzah:**

  * Hinge loss funksiyası SVM-də səhvləri ölçür.
  * Əgər $y_i f(x_i) \ge 1$ → loss = $0$ (nümunə düzgün margin-dən kənardadır).
  * Əks halda → loss > $0$ (nümunə margin içində və ya səhv ayrılıb).

> **Sklearn bunu dual optimallaşdırma yolu ilə həll edir, biz birbaşa loss-u optimize etmirik.**

<br>

✅ **Qısa xülasə:**

1. **Non-linear SVM-də $f(x_i)$ **support vector-ların $λ$, $y$ və kernel dot product-u** ilə hesablanır.**
2. **Loss = $max(0, 1 - y_i f(x_i))$ → səhvləri və margin içində olan nümunələri ölçür.**
3. **Bu loss, C ilə birləşərək ümumi optimizasiya məqsədini yaradır (sklearn bunu dual problem ilə dolayı həll edir).**

---


#### 🔹 **3-cü addım: Predict (Non-linear / Kernel SVM)**

$$
\hat{y} = \text{sign}\Big(\sum_i \lambda_i y_i K(x_i, x) + b\Big)
$$

- **İzah:**

> Non-linear SVM-də artıq **ω yoxdur**, ona görə prediksiya **support vector-lar və kernel funksiyası** vasitəsilə edilir.

* Addımlar:

  1. **Hər support vector $x_i$ üçün $\lambda_i y_i K(x_i, x)$ hesablanır.**
  2. **Bütün support vector-ların cəmi + bias (b) alınır → $f(x)$.**
  3. **sign() funksiyası** ilə sinif təyini edilir:

     * $f(x) > 0 \Rightarrow \hat{y} = +1$
     * $f(x) < 0 \Rightarrow \hat{y} = -1$

> **Qeyd:** **Kernel sayəsində x nöqtəsinin yüksək ölçülü space-də yerləşməsinə baxmadan linear-like prediksiya etmək mümkün olur.**

<br>

✅ **Qısa xülasə:**

1. $ω$ **yox → support vector-lar + $λ$ + $y$ + kernel istifadə olunur.**
2. **Yeni nümunə üçün $f(x) = (\sum_i \lambda_i y_i K(x_i, x) + b)$.**
3. $sign(f(x))$ → **final sinif prediksiyası.**

#### 🔴 **Ən vacib fərq: Linear vs Non-linear SVM (bunu yadda saxla)**

<br>

|             | Linear                                        | Non-linear                                                        |
| ----------- | --------------------------------------------- | ----------------------------------------------------------------- |
| **ω**       | Var: $\omega = \sum \lambda_i y_i x_i$        | ❌ yoxdur, çünki x → yüksək ölçülü space-ə xəritələnir $(\phi(x))$ |
| **λ**       | Var, support vector-lar üçün sıfırdan fərqli  | Var, support vector-lar üçün sıfırdan fərqli                      |
| **Predict** | $ \hat{y} = \text{sign}(\omega \cdot x + b) $ | $ \hat{y} = \text{sign}(\sum_i \lambda_i y_i K(x_i, x) + b) $     |
| **Kernel**  | Sadə dot product $x_i \cdot x_j$              | Əsas rol oynayır, yüksək ölçülü space-də dot product-u əvəz edir  |

<br>

- **İzah:**

- 1. **Linear SVM→ $ω$ mövcuddur, birbaşa dot product ilə prediksiya edilir.** 
- 2. **Non-linear SVM → ω birbaşa yoxdur, kernel funksiyası vasitəsilə prediksiya hesablanır.** 
- 3. **Hər iki halda $λ$-lər support vector-lar üçün tapılır və prediksiyada istifadə olunur.** 
- 4. **Kernel Non-linear SVM-in əsas “hiyləsi”dir: yüksək ölçülü space-də xətti ayırma kimi işləməyə imkan verir, amma $ω$ birbaşa hesablamaq lazım deyil.**

<br>

✅ **Qısa yaddaş üçün:**

* **Linear → $ω$ var, sadə dot product.**
* **Non-linear → ω yoxdur, kernel + $λ$-lər istifadə olunur.**

#### **🔚 SON YEKUN – sklearn SVM**

<br>

#### 🔹 **Linear SVM**

1. **λ tapılır** → support vector-lar üçün sıfırdan fərqli.
2. **ω hesablanır:**
   $$
   \omega = \sum_i \lambda_i y_i x_i
   $$
3. **Predict:**
   $$
   \hat{y} = \text{sign}(\omega \cdot x + b)
   $$

> Yəni linear SVM-də **ω mövcuddur**, prediksiya üçün birbaşa dot product istifadə olunur.

<br>

#### 🔹 **Non-linear SVM**

- 1. **λ tapılır**, amma **kernel $K(x_i, x_j)$** istifadə olunur.
- 2. **ω yoxdur**, **çünki data yüksək ölçülü space-ə xəritələnir** $(\phi(x))$.
- 3. **Predict:**
   $$
   \hat{y} = \text{sign}\Big(\sum_i \lambda_i y_i K(x_i, x) + b\Big)
   $$

>  **Yəni non-linear SVM-də kernel + $λ$ prediksiya üçün əsasdır, $ω$ birbaşa yoxdur.**

<br>

✅ **Qısa yaddaş üçün:**

* **Linear → $λ$ → $ω$ → predict**
* **Non-linear → $λ$ + kernel → predict, $ω$ yoxdur**
