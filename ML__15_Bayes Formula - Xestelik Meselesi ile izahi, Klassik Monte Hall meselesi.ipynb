1. Uşaqların cinsini (O = oğlan, Q = qız) nəzərə alaq. İki uşağın bütün mümkün cinsi kombinasiyaları:

   1. Oğlan - Oğlan (OO)
   2. Oğlan - Qız (OQ)
   3. Qız - Oğlan (QO)
   4. Qız - Qız (QQ)

2. Biz bilirik ki, **ən azı biri oğlandır**, yəni QQ halını çıxırıq:

   * Qalan kombinasiyalar: OO, OQ, QO

3. İkinci uşağın **qız olma ehtimalını** tapmaq üçün OO, OQ, QO halarında neçə dəfə digər uşaq qızdır:

4. Buna görə, uğurlu hallar: OQ, QO → 2 hal
   Ümumi mümkün hallar: OO, OQ, QO → 3 hal

5. Ehtimal:

$$
P(\text{digər uşaq qızdır}) = \frac{2}{3}
$$

### Klassik Monty Hall məsələsi Bayes ilə sübutu

> .**Üç qapı var. Üçündən birində maşın var. Sən birini seçə bilərsən. Sən 3-cü qapını seçirsən, amma aparıcı 2-ci qapını açır, Soruşur fikrini dəyişib 1-ci qapını seçmək istəyirsən. Məsələ də odur ki aparıcı 2-ci qapını acandan sonra, mən fikrimdə qalsam, yəni 3-cü qapıda etimal nə qədərdir, fikrimi 1-ci qapıya dəyişsəm ehtimal nə qəqədərdir?
Elm sübut edir ki, mən fikrimdə qalsam, yəni 3-cü qapıda etimal 1/3, amma fikrimi 1-ci qapıya dəyişsəm ehtimal 2/3 -dir.**

#### Aparıcı 2-ci qapını açandan sonra:
> **3-ci qapıda seçimi saxlasam Ehtimal = $\frac{1}{3}$**

> **1-ci qapıya seçimi dəyişsəm Ehtimal = $\frac{2}{3}$**


#### **Bayes Teoremi ilə İzah**:
#### **A — maşının yeri**

* **A = 1** → maşın 1-ci qapıdadır
* **A = 2** → maşın 2-ci qapıdadır
* **A = 3** → maşın 3-cü qapıdadır

Maşın təsadüfi qoyulub:

$$
P(A=1)=P(A=2)=P(A=3)=\frac{1}{3}
$$

---

#### **B — aparıcının açdığı qapı**

* **B = 1** → aparıcı 1-ci qapını açdı
* **B = 2** → aparıcı 2-ci qapını açdı
* **B = 3** → aparıcı 3-cü qapını açdı

---

#### 🔴 Şəkildə verilən sual
> Niyə  **1-ci qapıya seçimi dəyişmək sərfəlidir**

> **Aparıcı 2-ci qapını açıbsa (B=2), maşın 1-ci qapıdadırmı?**


Riyazi yazılışı:

$$
P(A=1 \mid B=2)=?
$$

---

#### 🔴 Bayes düsturu (şəkildəki kimi)

$$
P(A=1 \mid B=2)=
\frac{P(A=1)\cdot P(B=2 \mid A=1)}
{P(B=2)}
$$

İndi **düsturdakı HƏR ŞEYİ ayrıca izah edib hesablayırıq** 👇

---

#### 1️⃣ $P(A=1)$ — maşın əvvəlcədən 1-ci qapıda olsun

Bu **prior ehtimaldır**:

$$
P(A=1)=\frac{1}{3}
$$

---

#### 2️⃣ $P(B=2 \mid A=1)$ — maşın 1-ci qapıdadırsa, aparıcı 2-ni açar?

Şəkildə **sən 3-cü qapını seçmisən** (yaşıl ox).

Əgər maşın **1-ci qapıdadırsa**:

* 2-ci qapı → keçi
* 3-cü qapı → sənin seçdiyin, aça bilməz

👉 Aparıcı **məcburdur 2-ci qapını açsın**

$$
P(B=2 \mid A=1)=1
$$

---

#### 3️⃣ $P(B=2)$ — aparıcının 2-ci qapını açma ehtimalı (ümumi)

Bu **ən çətin hissədir**, şəkil də bunu göstərir.

#### Ümumi düstur:

$$
P(B=2)=\sum_{i=1}^{3} P(B=2 \mid A=i)\cdot P(A=i)
$$

İndi **tək-tək** yazırıq 👇

---

#### 🔹 $P(B=2)$ — aparıcının 2-ci qapını açma ehtimalı **A = 1 olarsa**
Şəkildə **sən 3-cü qapını seçmisən** (yaşıl ox).

Əgər maşın **1-ci qapıdadırsa**:

* 2-ci qapı → keçi
* 3-cü qapı → sənin seçdiyin, aça bilməz

👉 Aparıcı **məcburdur 2-ci qapını açsın**

$$
P(B=2 \mid A=1)\cdot P(A=1)=1\cdot\frac{1}{3}=\frac{1}{3}
$$

---

#### 🔹 $P(B=2)$ — aparıcının 2-ci qapını açma ehtimalı **A = 2 olarsa**

* Maşın 2-ci qapıdadır
* Aparıcı **2-ni aça bilməz**
* **Ehtimal = 0**

$$
P(B=2 \mid A=2)\cdot P(A=2) = 0\cdot\frac{1}{3}=0
$$

---

#### 🔹 $P(B=2)$ — aparıcının 2-ci qapını açma ehtimalı **A = 3 olarsa**

* Maşın **3-cü qapıdadır**
* Sən də **3-cü qapını seçmisən**
* Aparıcının qarşısında **iki qapı seçimi qalır: 1 və 2**
* **Ehtimal = $\frac{1}{2}$**
- **Niyə 1/2?**
- Çünki bu halda aparıcı **məcburi**, iki variantdan birini seçir.

👉 Deməli:

$$
P(B=2 \mid A=3)\cdot P(A=3) = \frac{1}{2}\cdot\frac{1}{3}=\frac{1}{6}
$$




---


#### 🔴 İndi Bayes düsturuna qoyuruq

$$
P(A=1 \mid B=2)=
\frac{\frac{1}{3}\cdot 1}{\frac{1}{3}+0+\frac{1}{6}} =\frac{2}{3}
$$


> Yəni: **aparıcı 2-ci qapını açandan sonra, seçimi dəyişmək ehtimalı artırır yəni 1-ci qapıda maşın olma ehtimalı 2/3-dür** 🚗🔥

### Bayes teoremi


> **Tərif:** **Bayes teoremi** — yeni məlumat (dəlil) əldə etdikdən sonra bir hadisənin ehtimalını **yenidən hesablamaq** üsuludur.

> **Tərif:**  **Bayes teoremi** — yeni əldə olunan məlumatlara əsaslanaraq bir hadisənin ehtimalını yeniləməyə imkan verən riyazi qaydadır.

> **Tərif:**  **Bayes teoremi** — **Bir hadisənin (A) yeni bir müşahidə (B) baş verdikdən sonra ehtimalı, müşahidənin həmin hadisə baş verdiyi halda ehtimalı (likelihood) ilə əvvəlki ehtimalın (prior) hasilinin, müşahidənin ümumi ehtimalına (evidence) bölünməsi ilə tapılır.**



#### Riyazi tərifi

$$
\boxed{P(A \mid B) = \frac{P(B \mid A) \ P(A)}{P(B)}}
$$

$$
\boxed{P(B) = P(B \mid A)\cdot P(A) + P(B \mid \neg A)\cdot P(\neg A)}
$$

#### Buradakı anlayışlar:

- **P(A) – A hadisəsinin ilkin ehtimalı (prior) - (müşahidədən əvvəl A-ya inamımız)**
- **P($\neg$ A) – A-nın baş verməməsi ehtimalı - (A-nın əksi, yəni A deyil)**
- **P(B $\mid$ A) – A baş verdiyi halda B-nin olma ehtimalı - (likelihood, müşahidənin A ilə uyğunluğu)**
- **P(B $\mid \neg$ A) – A baş vermədiyi halda B-nin olma ehtimalı - (B-nin alternativ səbəbdən gəlmə ehtimalı)**
- **P(B) – B hadisəsinin ümumi ehtimalı (evidence, normallaşdırıcı) - (B-nin A və ya A olmayan bütün hallarda baş vermə ehtimalı)**
- **P(A $\mid$ B) – B baş verdikdən sonra A-nın ehtimalı - (**posterior**, yenilənmiş inam)**

#### Bir cümləlik tam məna

> **Bayes teoremi deyir ki, bir hadisəyə olan ilkin inamımızı, müşahidənin həmin hadisə ilə və onun alternativi ilə nə qədər uyğun olduğuna baxaraq rasional şəkildə yeniləyirik.**

#### Sadə izahla

> **Əvvəl bildiyimiz ehtimalı, yeni məlumatlara görə yeniləyirik.**

#### Kiçik real nümunə 🧠

Tutaq ki:

* Xəstəlik ehtimalı: **1%**
* Test xəstələrdə **99%** doğru çıxır
* Test sağlam adamlarda **5%** səhv pozitiv verir

Testin pozitiv çıxması **xəstəliyin mütləq olduğunu göstərmir**. Bayes ilə hesablayanda görürük ki real xəstə olma ehtimalı düşündüyümüzdən xeyli azdır.

### Bayes teoremi 🔴 Məsələ həlli ilə izahı

**Bir xəstəlik var.**

**Verilənlər:**

* Cəmiyyətdə xəstəlik ehtimalı: **1%** → $P(X)=0.01$ - **Prior**
* Xəstə adam üçün testin pozitiv çıxması: **99%** → $P(+|X)=0.99$ - **Likelihood**
* Sağlam adam üçün testin səhvən pozitiv çıxması: **5%** → $P(+|S)=0.05$ - **Likelihood**

>❓ **Sual:**: Testin nəticəsi **pozitivdirsə**, insanın **doğrudan xəstə olma ehtimalı neçədir?** - **Posterior**



#### Bayes teoreminin ÜMUMİ DÜSTURU:

$$
\boxed{
P(X|+) = \frac{P(+|X)\cdot P(X)}{P(+)} 
} =
\frac{0.99 \cdot 0.01}
{(0.99 \cdot 0.01) + (0.05 \cdot 0.99)} \approx 0.1667
$$

$$
$$

#### ❗ P(+) = testin pozitiv çıxmasının **ümumi ehtimalıdır**

Bu iki halın cəmidir:

- **Xəstə olub pozitiv çıxanlar**
- **Sağlam olub səhvən pozitiv çıxanlar**



$$
\boxed{
P(+) = P(+|X)\cdot P(X) + P(+|S)\cdot P(S) 
}
$$

Burada:
$$
P(S)=1-P(X) - \ Prior
$$


#### 📌 Düsturun hissələri nədir?

| Hissə  | Adı       | Mənası                              |                                         
| ------ | --------- | ----------------------------------- | 
| $$P(X)$$ | **Prior** | Testdən ƏVVƏL xəstə olma ehtimalı   |    
| $$P(+/X)$$ | Likelihood | Xəstədirsə testin pozitiv çıxma ehtimalı |
| $$P(+)$$ | Evidence  | Testin ümumi pozitiv çıxma ehtimalı |                                         
| $$P(X/+)$$ | **Posterior**| Testdən SONRA xəstə olma ehtimalı |

$$$$

---

#### 🔴 Bayes düsturunda **SURƏT** və **MƏXRƏC** nədir?

Bayes düsturu:
$$
P(X|+) = \frac{P(+|X)\cdot P(X)}{P(+)}
$$



#### 🔵 1️⃣ **SURƏT (HASİL)** nəyi ifadə edir?

#### Surət:

$$
\boxed{P(+|X)\cdot P(X)} = (0.99 \cdot 0.01)
$$

#### Mənası:

> **HƏM xəstə olub  testi pozitiv çıxan insanların payı, Həm də xəstə olan insanların payı**

Yəni bu **iki şərtin birlikdə baş vermə ehtimalıdır**:

* insan **xəstədir**
* test **pozitivdir**

#### Riyazi dildə:

$$
P(+|X)\cdot P(X) = P(X \cap +)
$$

#### Real düşüncə:

> **Cəmiyyətdə olan insanların neçə faizi xəstədir və testdə pozitiv çıxır**

Bu **doğru pozitivlərdir**.

$$$$

#### 🔵 2️⃣ **MƏXRƏC (CƏM)** nəyi ifadə edir?

#### Məxrəc:

$$
\boxed{P(+) = P(+|X)\cdot P(X) + P(+|S)\cdot P(S)} = (0.99 \cdot 0.01) + (0.05 \cdot 0.99)
$$

#### Mənası:

> **Testi pozitiv çıxan BÜTÜN insanların payı**

#### Burada iki fərqli qrup var:

#### 🔹 Xəstə + pozitiv (doğru pozitiv)
- **Cəmiyyətdə olan insanların neçə faizi xəstədir və testdə pozitiv çıxır**

$$
P(+|X)\cdot P(X)
$$

#### 🔹 Sağlam + səhv pozitiv
- **Cəmiyyətin neçə faizi sağlamdır, amma testdə səhvən pozitiv çıxır (yəni false positive)**

$$
P(+|S)\cdot P(S)
$$

#### Yəni məxrəc deyir:

> **Ümumiyyətlə testdə pozitiv çıxanların hamısı kimlərdir?**



#### 🔴 Vacib məqam

> **Məxrəc sadəcə “xəstələr” deyil,
> testi pozitiv çıxan HAMIDIR.**

**Ona görə də cəmdir ❗**



#### 🧠 Bayes-in əsas məntiqi (bir cümlə ilə)

$$
\boxed{
\text{Posterior} =
\frac{\text{Doğru pozitivlər}}
{\text{Bütün pozitivlər}}
}
$$



#### 🔹 Yəni:

* **Surət** → **Doğru səbəbdən pozitiv çıxanlar**
* **Məxrəc** → **Hər hansı səbəbdən pozitiv çıxanlar**



#### 🔚 Çox qısa yadda qalan qayda

> **Bayes soruşur:**
> **Pozitiv çıxanların içində neçəsi doğrudan xəstədir?**

---

#### ✅ YEKUN CAVAB

> **Test pozitiv çıxsa belə, insanın həqiqətən xəstə olma ehtimalı cəmi 👉 16.7%-dir.**



#### 🧠 Bayes düsturunun MƏNTİQİ (ən vacib hissə)

#### 🔹 Prior $P(X)$

> **Xəstəlik nadirdir (1%) → başlanğıcda risk azdır**

#### 🔹 Likelihood $P(+|X)$

> **Test xəstələrdə yaxşı işləyir (99%)**

#### 🔹 Evidence $P(+)$

> **Amma pozitiv nəticələrin çoxu **sağlam adamlardan** gəlir**

#### 🔹 Posterior $P(X|+)$

> **Bu səbəbdən pozitiv test ≠ mütləq xəstəlik**



#### 🧠 Bayes nə deyir?

> **Yeni məlumat (test nəticəsi) əvvəlki biliyimizi (prior) tam silmir,
> sadəcə onu yeniləyir.**



#### 🧠 Vacib nəticə (əsas dərs!)

> Test **çox dəqiq** olsa belə,
> xəstəlik **nadir**dirsə, pozitiv nəticə **mütləq xəstəlik demək deyil**.

Bu məhz **Bayes teoreminin gücüdür**:
ilkin ehtimalı (1%) nəzərə alır.
