## Binar təsnifat (binary classification) problemlərində **hipertəyyar tənliyi**

**Binar təsnifat** (binary classification) problemlərində **hipertəyyar tənliyi**, verilənlər məkanını iki sinifə ayırmaq üçün istifadə olunan riyazi ifadədir. 

Bunu belə təsəvvür etmək olar: verilənləriniz çoxölçülü bir məkan təşkil edir (məsələn, 2D və ya 3D) və siz həmin məkanı iki sinifə (məsələn, "0" və "1") ayırmaq üçün bir düz xətt (2D-də), bir müstəvi (3D-də) və ya ümumiyyətlə bir hipertəyyar (n ölçülü məkan üçün) qurursunuz.

---

### Ümumi tənlik:

Hipertəyyarın tənliyi bu şəkildə yazılır:

$$
w_1x_1 + w_2x_2 + \dots + w_nx_n + b = 0
$$

Burada:
- $x_1, x_2, \dots, x_n$ — xüsusiyyətlər (features) və ya verilənlərinizin koordinatlarıdır,
- $w_1, w_2, \dots, w_n$ — xüsusiyyətlərin çəkiləri (weights),
- $b$ — sərhəd xəttini (hipertəyyarı) yerdəyişdirən əmsaldır (bias),
- $= 0$ — məkanı iki sinifə ayıran sərhəd xəttini ifadə edir.

---

### Binar təsnifat zamanı necə işləyir?
- **Tənliyin nəticəsi $> 0$** olduqda verilənlər bir sinfə aid edilir (məsələn, "1").
- **Tənliyin nəticəsi $< 0$** olduqda verilənlər digər sinfə aid edilir (məsələn, "0").
- **Tənliyin nəticəsi $= 0$** olan nöqtələr hipertəyyarın üzərində yerləşir və iki sinfi ayırır.

---

### Məsələn:
Əgər 2D məkanda təsnifat problemi üzərində işləyirsinizsə, hipertəyyar sadəcə bir düz xətt olur:
$$
w_1x_1 + w_2x_2 + b = 0
$$
Burada xətt, iki sinif arasında sərhəd təşkil edir.

### İstifadə olunduğu modellər:
- **Logistic Regression**
- **Support Vector Machine (SVM)**
- **Neural Networks** (ilk qatlarında)
- **Perceptron**

---

Qısaca, hipertəyyar tənliyi binar təsnifat problemlərində verilənlər arasında sinifləri ayıran qərar sərhədini (decision boundary) təsvir edir.

## Vektorların Skalyar hasil

**Skalyar hasil** (və ya skalyar vurma) iki vektorun riyazi əməliyyatıdır və nəticədə bir **skalyar** (sadə ədədi dəyər) alınır. Bu, vektorların böyüklüyü (uzunluğu) və onların bir-birinə olan bucağı ilə müəyyən edilir.

### Skalyar hasilin formulu:
İki vektor $\vec{a} = (a_1, a_2, \dots, a_n)$ və $\vec{b} = (b_1, b_2, \dots, b_n)$ üçün skalyar hasil belə hesablanır:

$$
\vec{a} \cdot \vec{b} = a_1b_1 + a_2b_2 + \dots + a_nb_n
$$

**Başqa bir ifadə ilə:**
$$
\vec{a} \cdot \vec{b} = |\vec{a}| \cdot |\vec{b}| \cdot \cos \theta
$$
Burada:
- $|\vec{a}|$ və $|\vec{b}|$ — vektorların uzunluqlarıdır,
- $\cos \theta$ — iki vektor arasında olan bucağın kosinusudur.

---

### Skalyar hasilin xüsusiyyətləri:
1. **Skalyar** — nəticə sadə bir ədədi dəyərdir, vektor deyil.
2. Əgər iki vektor bir-birinə **perpendikulyardırsa (90°)**, onların skalyar hasili **0** olur ($\cos 90° = 0$).
3. Əgər iki vektor eyni istiqamətdədirsə ($0°$), skalyar hasil **maksimaldır**.
4. Əgər iki vektor əks istiqamətdədirsə (\$180°$), skalyar hasil mənfi olur.

---

### Geometrik mənası:
Skalyar hasil, bir vektorun digərinə **proyeksiyası** ilə həmin vektorun uzunluğunun hasilini ifadə edir. 

Məsələn,$\vec{a}$ vektorunu $\vec{b}$-yə proyeksiya edirsiniz və bu proyeksiyanın uzunluğunu $|\vec{b}|$-yə vurursunuz.

---

### İstifadə sahələri:
1. **Maşın öyrənməsi** və **neyron şəbəkələrdə**: Xüsusiyyətlərin çəkilər ilə (weights) vurulması skalyar hasilə əsaslanır.
2. **Fizika**: İki qüvvənin bir-birinə təsirinin ölçülməsi.
3. **Kosinus oxşarlığı** (cosine similarity): İki vektorun oxşarlığını ölçmək üçün skalyar hasildən istifadə olunur.

Qısacası, skalyar hasil vektorlar arasındakı əlaqəni ölçmək və onların istiqamətlərini təhlil etmək üçün istifadə olunan bir riyazi anlayışdır.

##  Əgər iki vektor bir-birinə **perpendikulyardırsa (90°)**, onların skalyar hasili **0** olur ($\cos 90° = 0$)

Burada göstərilən düstur, vektorlar arasındakı skalyar hasilə ($(\omega, x)$) əsaslanır və o halda 90 dərəcə bucaq məsələsi ilə bağlıdır. Gəlin bunu ətraflı izah edək.

---

### Skalyar hasil (dot product) nədir?
İki vektor $\vec{\omega}$ və $\vec{x}$-in skalyar hasili belə ifadə olunur:

$$
(\omega, x) = |\omega| \cdot |x| \cdot \cos(\alpha)
$$

Burada:  
- $|\omega|$: $\omega$ vektorunun uzunluğu,  
- $|x|$: $x$ vektorunun uzunluğu,  
- $\alpha$: $\omega$ və $x$ vektorları arasındakı bucaqdır.

---

### 90 dərəcə olduqda $\cos(\alpha) = 0$ niyə olur?
Trigonometrik funksiyalara əsasən, bucaq $90^\circ$ olduqda ($\alpha = 90^\circ$):

$$
\cos(90^\circ) = 0
$$

Bu, trigonometriyanın təməl xüsusiyyətidir. Bu vəziyyətdə, skalyar hasil aşağıdakı kimi olur:

$$
(\omega, x) = |\omega| \cdot |x| \cdot \cos(90^\circ) = |\omega| \cdot |x| \cdot 0 = 0
$$

---

### Fiziki mənası:
Əgər iki vektor bir-birinə **perpendikulyar** (90° bucaq altında) olarsa, onların skalyar hasili həmişə \(0\)-a bərabərdir. Burada verilmiş ifadədə $(\omega, x) = 0$ bu o deməkdir ki, $\omega$ və $x$ vektorları perpendikulyardır.

---

### Qısaca:
1. $\cos(90^\circ) = 0$, çünki bu trigonometriyanın təməl qaydasıdır.
2. $(\omega, x) = 0$olduqda, $\omega$ və $x$) vektorları 90° bucaq altında yerləşir.
3. Bu ifadə, iki vektorun bir-birinə perpendikulyar olduğunu göstərir.

## $
A =
\begin{bmatrix} 
\omega_1 & \omega_2 
\end{bmatrix}_{1 \times 2} \quad
$  $
B =
\begin{bmatrix} 
x_1 \\ 
x_2 
\end{bmatrix}_{2 \times 1}
$ hasili

Sənin verdiyin matrislər belədir:  

$$
A =
\begin{bmatrix} 
\omega_1 & \omega_2 
\end{bmatrix}_{1 \times 2}
$$

$$
B =
\begin{bmatrix} 
x_1 \\ 
x_2 
\end{bmatrix}_{2 \times 1}
$$

---

### **Matris Vurma:**
Matrislərin ölçüləri:

$$
(1 \times 2) \times (2 \times 1) = (1 \times 1)
$$

Yəni nəticə **$ 1 \times 1 $** ölçülü skalyar olacaq.

---

### **Hesablama:**
$$
C = A \times B =
\begin{bmatrix} 
\omega_1 & \omega_2 
\end{bmatrix}
\times
\begin{bmatrix} 
x_1 \\ 
x_2 
\end{bmatrix}
$$

Matris vurma qaydasına görə:

$$
C = (\omega_1 \cdot x_1) + (\omega_2 \cdot x_2)
$$

Nəticə:

$$
C =
\begin{bmatrix} 
\omega_1 x_1 + \omega_2 x_2
\end{bmatrix}_{1 \times 1}
$$

---

Bu kod çıxış olaraq **$ 1 \times 1 $** ölçülü skalyar dəyəri qaytaracaq:
$$
\omega_1 x_1 + \omega_2 x_2
$$

**Bu vurmaya "dot product" (skalyar hasil) deyilir və iki vektor arasındakı daxili hasil kimi tanınır.**

---

```python
import sympy as sp

# Simvolik dəyişənləri təyin edirik
omega_1, omega_2, x_1, x_2 = sp.symbols('omega_1 omega_2 x_1 x_2')

# 1x2 sətir vektoru
A = sp.Matrix([[omega_1, omega_2]])

# 2x1 sütun vektoru
B = sp.Matrix([[x_1], [x_2]])

# Matris vurma
C = A * B

# Nəticəni göstəririk
sp.pprint(C)
```

## $\cos(\alpha)$ müsbət və mənfi halları

$\cos(\alpha)$ funksiyası müsbət və mənfi ola bilər. Bu hal bu funksiyanın bucağa görə necə dəyişdiyini göstərir:

1. **Müsbət olduğu hallar**:
   - **I və IV kuadrantları**: $\cos(\alpha)$ müsbət olur, əgər bucaq $0^\circ \leq \alpha < 90^\circ$ (birinci kuadrant) və $270^\circ < \alpha < 360^\circ$(dördüncü kuadrant) arasında yerləşirsə. Bu zaman bucağın kosinus dəyəri müsbət olur.

2. **Mənfi olduğu hallar**:
   - **II və III kuadrantları**: $\cos(\alpha)$ mənfi olur, əgər bucaq $90^\circ < \alpha < 180^\circ$ (ikinci kuadrant) və $180^\circ < \alpha < 270^\circ$ (üçüncü kuadrant) arasında yerləşirsə. Bu zaman bucağın kosinus dəyəri mənfi olur.

Yəni,$\cos(\alpha)$ funksiyasının müsbət və mənfi olması, bucağın hansı kuadrantda olduğuna bağlıdır.

### cos(𝛼) müsbət və mənfi hallarının Qrafik təsviri

```python
import numpy as np
import matplotlib.pyplot as plt

# Bucaqları təyin edirik (0-dan 2*pi-yə qədər)
angles = np.linspace(0, 2 * np.pi, 1000)

# Cosine dəyərlərini hesablamaq
cos_values = np.cos(angles)

# Qrafik
plt.figure(figsize=(8, 6))
plt.plot(angles, cos_values, label=r'$\cos(\alpha)$', color='b')

# X və Y oxlarını göstəririk
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Cosine funksiyasının müsbət və mənfi hissələrini vurğulamaq
plt.fill_between(angles, cos_values, 0, where=(cos_values >= 0), color='green', alpha=0.3, label='Müsbət Hissə')
plt.fill_between(angles, cos_values, 0, where=(cos_values <= 0), color='red', alpha=0.3, label='Mənfi Hissə')

# X oxunu pi vahidlərində göstərmək üçün
pi_ticks = np.arange(0, 2.1 * np.pi, np.pi / 2)  # 0, π/2, π, 3π/2, 2π
pi_labels = [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']
plt.xticks(pi_ticks, pi_labels)

# Başlıq və etiketlər
plt.xlabel(r'Bucaq $\alpha$ (radian)', fontsize=12)
plt.ylabel('Funksiya Dəyəri', fontsize=12)
plt.legend()

# Qrafiki göstəririk
plt.grid(True)
plt.show()
```
