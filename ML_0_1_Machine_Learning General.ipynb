## Machine Learning nedir?


>**Tərif:** Machine learning – modelin verilənlərdən nümunə və qanunauyğunluqları öyrənərək, insan müdaxiləsi olmadan qərarlar verməsini və proqnozlar etməsini təmin edən süni intellekt sahəsi və statistik hesablama metodudur. Bu yanaşma supervised, unsupervised və reinforcement learning kimi kateqoriyalara bölünür və geniş miqyaslı analitik, proqnozlaşdırıcı və avtomatlaşdırılmış sistemlərdə tətbiq olunur. 🤖📈

---

>**Tərif:** **Machine Learning** (Maşın Öyrənməsi), kompüterlərin verilmiş məlumatlardan öyrənərək müəyyən tapşırıqları yerinə yetirmək üçün proqramlaşdırılmasına əsaslanan bir süni intellekt sahəsidir. Bu prosesdə maşın öyrənmə alqoritmləri, məlumatları analiz edərək və bu məlumatlardan qaydalar çıxararaq özünü inkişaf etdirir. Aşağıda maşın öyrənməsinin əsas aspektləri və tətbiq sahələri təqdim olunur:

### 1. **Maşın Öyrənməsinin Tipləri**
Maşın öyrənməsi əsasən üç tələbat üzrə təsnif edilir:

- **Supervised Learning (Nəzarətli Öyrənmə)**:
  - Məlumat dəstinin etiketli olduğu bir yanaşmadır. Yəni, hər bir giriş məlumatına uyğun bir çıxış dəyəri var. Məqsəd, bu etiketlərə əsaslanaraq yeni, görünməyən məlumatlar üçün proqnozlar verməkdir.
  - Məsələn, müştəri kredit verilib-verilməyəcəyini müəyyənləşdirmək üçün verilənlərdəki müştəri məlumatlarını istifadə etmək.

- **Unsupervised Learning (Nəzarətsiz Öyrənmə)**:
  - Bu yanaşmada məlumatlar etiketlənməyib, yəni çıxış dəyərləri yoxdur. Məqsəd, məlumatlar arasında gizli strukturların və ya əlaqələrin aşkar edilməsidir.
  - Məsələn, müştəriləri oxşar xüsusiyyətlərinə görə klasterlərə ayırmaq.

- **Reinforcement Learning (Mükafatlandırma ilə Öyrənmə)**:
  - Bu yanaşmada agent, mühit ilə qarşılıqlı təsir edir və hər bir hərəkətinin mükafatını (və ya cəzasını) alaraq öyrənir. Məqsəd, ən yüksək mükafatı əldə etmək üçün optimal strategiyaları inkişaf etdirməkdir.
  - Məsələn, bir robotun müəyyən bir mühitdə tapşırıqları yerinə yetirməsi.

### 2. **Maşın Öyrənməsinin Prosesləri**
Maşın öyrənməsi prosesi adətən aşağıdakı mərhələləri əhatə edir:

1. **Məlumat Toplanması**: Problemi anlamaq üçün müvafiq məlumatların toplanması.
2. **Məlumatların Təmizlənməsi**: Məlumatların keyfiyyətini artırmaq üçün təmizlənməsi, eksik dəyərlərin tamamlanması və ya yanlış dəyərlərin düzəldilməsi.
3. **Xüsusiyyətlərin Seçimi**: Modelin performansını artırmaq üçün müvafiq xüsusiyyətlərin seçilməsi.
4. **Modelin Qurulması**: Seçilmiş alqoritmlərlə modelin yaradılması.
5. **Modelin Testi**: Modelin performansının qiymətləndirilməsi üçün test dəstində yoxlanılması.
6. **İcra**: Modelin real dünya tətbiqinə yönəldilməsi və nəticələrin izlənilməsi.

### 3. **Tətbiq Sahələri**
Maşın öyrənməsi müxtəlif sahələrdə geniş tətbiq olunur:

- **Maliyyə**: Kredit riskinin qiymətləndirilməsi, fırıldaqçılıq aşkarlanması.
- **Səhiyyə**: Xəstəliklərin proqnozlaşdırılması, tibbi şəkillərin analizi.
- **Marketinq**: Hədəf auditoriyanın müəyyənləşdirilməsi, müştəri davranışlarının analizi.
- **Robototexnika**: Avtomatlaşdırma, robotların mühitdə fəaliyyət göstərməsi.
- **Tərcümə**: Tərcümə sistemlərində və təbii dil işləmə (NLP) tətbiqlərində.

### Nəticə
Maşın öyrənməsi, verilənlərdən avtomatik öyrənməyi və bu öyrənmələrə əsaslanaraq qərarlar verməyi mümkün edən güclü bir alət və metodologiyadır. Bu sahə, süni intellektin inkişafında və bir çox tətbiq sahəsində inqilabi dəyişikliklərə səbəb olmuşdur.

## Supervised Learning (Nəzarətli Öyrənmə)

>**Tərif:** **Supervised Learning (Nəzarətli Öyrənmə)**, verilənlərdəki **girişlərə (features)** uyğun **çıxışların (labels)** məlum olduğu hallarda, modelin bu giriş-çıxış əlaqəsini öyrənərək yeni verilənlər üçün doğru çıxışları proqnozlaşdırmağa çalışdığı maşın öyrənmə yanaşmasıdır.



Supervised Learning (Nəzarətli Öyrənmə), maşın öyrənmənin əsas kateqoriyalarından biridir. Bu yanaşmada, modelə əvvəlcədən etiketlənmiş məlumatlar təqdim olunur və o, bu məlumatlara əsaslanaraq girişlər ilə çıxışlar arasındakı əlaqəni öyrənir. Məqsəd, modelin yeni və etiketlənməmiş məlumatlar üçün dəqiq proqnozlar verə bilməsini təmin etməkdir.

---

### **Supervised Learning-in Əsas Prinsipi:**
1. **Təlim Məlumatı:** 
   - Təlim dəsti $ X = \{x_1, x_2, ..., x_n\} $girişlərdən (features) və $ Y = \{y_1, y_2, ..., y_n\} $çıxışlardan (labels) ibarətdir.
   - Hər bir nümunə $ (x_i, y_i) $ cütü şəklində təqdim olunur, burada $ x_i $giriş məlumatlarıdır, $ y_i $ isə həmin girişlərin uyğun etiketidir.

2. **Model Öyrənməsi:** 
   - Algoritm $ f(X) \to Y $ funksiyasını öyrənir. Bu funksiya girişləri (features) ilə çıxışlar (labels) arasında əlaqəni təsvir edir.
   - Model yeni məlumatları görsə də, öyrəndiyi bu funksiyadan istifadə edərək proqnozlar verir.

3. **Qiymətləndirmə:**
   - Model test məlumat dəsti ilə sınanır. Test dəsti təlimdə istifadə edilməyən etiketli nümunələrdən ibarətdir.
   - Məqsəd, modelin doğruluğunu və ümumiləşmə qabiliyyətini qiymətləndirməkdir.

---

### **Supervised Learning-in Əsas Tətbiqləri:**

1. **Klassifikasiya (Classification):**
   - Giriş məlumatları müəyyən kateqoriyalara aid edilir.
   - Məsələn:
     - E-poçtların "Spam" və ya "Normal" olaraq təsnif edilməsi.
     - Xəstəlik diaqnozu üçün tibbi məlumatların "Xəstə" və ya "Sağlam" olaraq təsnif edilməsi.
   - Algoritmlər: Logistic Regression, Decision Trees, Random Forest, Support Vector Machines (SVM), Neural Networks.

2. **Reqressiya (Regression):**
   - Giriş məlumatlarına əsasən davamlı (continuous) çıxışlar proqnozlaşdırılır.
   - Məsələn:
     - Ev qiymətlərinin proqnozlaşdırılması.
     - Bir şirkətin gəlirlərinin zaman seriyalarına əsasən proqnozlaşdırılması.
   - Algoritmlər: Linear Regression, Polynomial Regression, Ridge, Lasso Regression.

---

### **Supervised Learning-in Tətbiq Sahələri:**
- **Maliyyə:** Kreditlərin təsdiqi (classification), qiymətlərin təxmin edilməsi (regression).
- **Səhiyyə:** Xəstəliklərin erkən diaqnostikası.
- **Marketinq:** Müştəri davranışlarının təhlili və proqnozlaşdırılması.
- **Avtonom Texnologiyalar:** Obyektlərin aşkarlanması və tanınması (self-driving cars).
- **E-ticarət:** Məhsul tövsiyə sistemləri (recommendation systems).

---

### **Supervised Learning-də Əsas Addımlar:**

1. **Məlumatların Hazırlanması:**
   - Məlumatlar toplanır və təmizlənir (missing values, outliers, normalization, scaling).
   - Təlim və test dəstlərinə bölünür (məsələn, 80% təlim, 20% test).

2. **Xüsusiyyətlərin Seçimi və Ekstraksiyası:**
   - Modelin girişləri üçün ən təsirli xüsusiyyətlər (features) seçilir.
   - Dərin öyrənmə kimi sahələrdə xüsusiyyətlər avtomatik olaraq müəyyən edilir.

3. **Modelin Təlimi:**
   - Seçilmiş alqoritm vasitəsilə model təlim edilir.
   - Təlim prosesində model parametrləri optimallaşdırılır.

4. **Modelin Qiymətləndirilməsi:**
   - Test dəstində dəqiqlik (accuracy), dəqiqlik (precision), çağırış (recall), F1-score kimi metriklərlə model qiymətləndirilir.
   - Performans zəifdirsə, modelin optimallaşdırılması və ya daha yaxşı xüsusiyyətlərin daxil edilməsi tələb olunur.

5. **Modelin İstifadəsi:**
   - Model real dünyada yeni məlumatlara tətbiq olunur.

---

### **Supervised Learning-in Üstünlükləri:**
1. Etiketlənmiş məlumatlardan istifadə etdiyi üçün proqnozların dəqiqliyi yüksəkdir.
2. Müxtəlif sahələrdə tətbiq oluna bilər.
3. Sadə alqoritmlərlə başlayaraq kompleks problemləri həll etmək mümkündür.

### **Supervised Learning-in Çətinlikləri:**
1. **Etiketlənmiş Məlumatın Tələb Olunması:** Keyfiyyətli və böyük həcmli etiketli məlumat toplamaq çətin və baha başa gələ bilər.
2. **Overfitting:** Model yalnız təlim məlumatlarına uyğunlaşa bilər və yeni məlumatlarda zəif performans göstərə bilər.
3. **Vaxt və Hesablama Resursları:** Çox böyük verilənlər üzərində təlim prosesi uzun müddət və yüksək resurs tələb edə bilər.

---

### **Populyar Supervised Learning Algoritmləri:**

1. **Linear Regression:**
   - Sadə reqressiya problemləri üçün istifadə olunur.
2. **Logistic Regression:**
   - İki və ya daha çox sinif arasında təsnifat üçün.
3. **Decision Trees:**
   - Aydın və şərh edilə bilən modellər üçün idealdır.
4. **Random Forest:**
   - Bir neçə decision tree modelindən istifadə edir və dəqiqliyi artırır.
5. **Support Vector Machines (SVM):**
   - Qeyri-xətti məlumatların təsnifi üçün güclü seçimdir.
6. **Neural Networks:**
   - Çox mürəkkəb və böyük məlumat dəstləri üçün güclüdür.

---

### **Nəticə:**
Supervised Learning geniş istifadə olunan bir yanaşmadır və əsas üstünlüyü giriş və çıxış arasındakı əlaqəni öyrənərək yeni məlumatlarda yüksək dəqiqliklə proqnozlar verməsidir. Həm klassifikasiya, həm də reqressiya problemlərində effektivdir. Lakin müvəffəqiyyətli tətbiqi üçün etiketlənmiş məlumatlar tələb olunur və bu, bəzi hallarda çətinlik yarada bilər.

## Regression və Classification arasinda ferq?

**Regression** və **classification**, hər ikisi də **nəzarətli öyrənmə** (supervised learning) metodlarıdır. Lakin onların arasında mühüm fərqlər var, çünki onlar fərqli növ proqnozlar və nəticələr üzərində işləyir. Hər ikisinin necə işlədiyini və əsas fərqlərini nəzərdən keçirək:

### Regression (Regressiya)

>**Tərif:** Regression – bir və ya bir neçə müstəqil dəyişənlərlə asılı dəyişənin arasındakı əlaqəni modelləşdirərək, davamlı bir qiymətin proqnozlaşdırılmasını təmin edən statistik və maşın öyrənməsi metodudur. Bu yanaşma, xətti və qeyri-xətti əlaqələri təhlil edərək, ən uyğun model parametrlərini tapmaq və yeni verilənlər üzərində dəqiq proqnozlar etmək məqsədini güdür.


- **Tərif**: Regressiya, kəmiyyət bir dəyişən üçün proqnoz verməyi hədəfləyir. Bu tip modellər, davamlı (continuous) dəyişənləri proqnozlaşdırır, yəni nəticə rəqəm kimi olur.
- **Misal**: Bir evin qiymətini proqnozlaşdırmaq. Verilənlər evin sahəsi, otaq sayı və digər xüsusiyyətlərə əsaslanır və nəticə evin təxmini qiyməti (rəqəmsal dəyər) olacaq.
- **Məqsəd**: Davamlı bir nəticə proqnozlaşdırmaq (məsələn, bir şeyin qiyməti, temperaturu, çəkisi və s.).
- **Nəticə növü**: **Davamlı (continuous)**, rəqəm şəklində nəticə. Məsələn, 0.0-dan 100.0-a qədər istənilən dəyər ola bilər.
- **Məhsuldarlıq sahələri**:
  - Satışların proqnozlaşdırılması.
  - Bir insanın gəlirini təxmin etmək.
  - Borsada qiymətlərin təxmin edilməsi.

### Classification (Klassifikasiya)

>**Tərif:** Classification – bir və ya bir neçə müstəqil dəyişənlərlə asılı dəyişənin arasındakı əlaqəni modelləşdirərək, verilənləri müəyyən siniflərə ayıran və hər bir nümunənin hansı sinifə aid olduğunu təxmin edən statistik və maşın öyrənməsi metodudur. Bu yanaşma, etiketlənmiş verilənlər üzərində təlim edərək, yeni verilənlər üçün doğru sinif etiketini təyin etməyə yönəlir və məsələn, spamlama, xəstəlik təyini və ya müştəri seqmentasiyası kimi problemlərdə istifadə olunur.


- **Tərif**: Klassifikasiya, fərqli kateqoriyalar və ya siniflər arasında qərar vermək məqsədi daşıyır. Nəticə diskret (diskret) olur, yəni bir neçə əvvəlcədən təyin olunmuş siniflərdən birinə aid olur.
- **Misal**: E-poçtların "spam" və ya "normal" kimi təsnif edilməsi. Verilənlər e-poçt başlığı, göndərən və məzmununa əsaslanır və nəticədə e-poçt müəyyən bir kateqoriyaya (spam və ya normal) daxil edilir.
- **Məqsəd**: Verilən məlumatlara əsaslanaraq müəyyən bir sinfə qərar vermək (diskret, fərqli siniflərə ayırmaq).
- **Nəticə növü**: **Diskret (discrete)**, təyin olunmuş siniflərdən birinə aid edilən nəticə. Məsələn, siniflər 0 və 1 ola bilər, ya da daha çox siniflər (məsələn, 0, 1, 2, 3 və s.) ola bilər.
- **Məhsuldarlıq sahələri**:
  - Müştərinin kredit riskini müəyyənləşdirmək (məsələn, müştəri krediti ödəyə biləcək, ya da ödəyə bilməyəcək).
  - Tibb sahəsində xəstəlik diaqnozu (məsələn, xəstə müəyyən bir xəstəliyə malikdir və ya deyil).
  - İris data setində çiçəklərin növlərini (species) təsnif etmək.

### Əsas fərqlər

| Xüsusiyyət            | **Regression**                         | **Classification**                       |
|-----------------------|----------------------------------------|------------------------------------------|
| **Məqsəd**             | Davamlı bir dəyişəni proqnozlaşdırmaq  | Diskret siniflərə aid etmək              |
| **Nəticə növü**        | Rəqəmsal, davamlı (continuous)         | Diskret, təyin olunmuş siniflər          |
| **Məsələnin tipi**     | Nəticə real ədədi (rəhbərsiz dəyər)    | Nəticə siniflər şəklində (kategoriya)    |
| **Misal**              | Ev qiymətinin təxmin edilməsi          | E-poçtun spam olub-olmaması              |
| **Populyar alqoritmlər**| Linear Regression, Decision Tree Regressor | Logistic Regression, Random Forest Classifier, SVM |

### Regression üçün bəzi alqoritmlər:
- Linear Regression
- Ridge Regression
- Decision Tree Regressor
- Random Forest Regressor

### Classification üçün bəzi alqoritmlər:
- Logistic Regression
- Decision Tree Classifier
- Random Forest Classifier
- Support Vector Machine (SVM)
- K-Nearest Neighbors (KNN)

### Əsas fərq
- **Regression** rəqəmsal nəticə proqnozlaşdırır, məsələn, qiymətlər, ölçülər və digər kəmiyyətlər.
- **Classification** verilənləri müəyyən siniflərə ayırır, məsələn, bir obyektin hansı kateqoriyaya aid olduğunu müəyyən edir.

### Nəticə
Əgər məqsədiniz rəqəmsal dəyər (qiymət, məbləğ) proqnozlaşdırmaqdırsa, **regression** modellərindən istifadə etməlisiniz. Əgər verilənləri müəyyən siniflərə ayırmaq istəyirsinizsə, məsələn, müsbət və ya mənfi nəticələr arasında seçim etmək istəyirsinizsə, **classification** modelləri daha uyğun olacaq.

## Linear Regression nedir?

### **Linear Regression** (Xətti regresiya)

>**Tərif:** Xətti regresiya, ən sadə və ən çox istifadə edilən statistik və maşın öyrənmə metodlarından biridir. **Bu metod, müstəqil dəyişənlər ilə asılı dəyişən arasında xətti əlaqəni modelləşdirir. Əsas məqsəd, bir və ya bir neçə müstəqil dəyişənin asılı dəyişənə təsirini qiymətləndirmək və gələcək üçün proqnozlar verməkdir**.

---

### **Xətti regresiyanın tənliyi**
Xətti regresiya asılı dəyişəni ($ y $) müstəqil dəyişənlərin $(x_1, x_2, \dots, x_n$) xətti kombinasiyası kimi ifadə edir:

$$
y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon
$$

Burada:  
- $y$: Asılı dəyişən (target, proqnoz etmək istədiyimiz dəyər).  
- $x_1, x_2, \dots, x_n$: Müstəqil dəyişənlər (features, girişlər).  
- $\beta_0$: Intercept (başlanğıc nöqtəsi, müstəqil dəyişənlər olmadıqda \(y\)-nin dəyəri).  
- $\beta_1, \beta_2, \dots, \beta_n$: Əmsallar (hər bir müstəqil dəyişənin asılı dəyişənə təsiri).  
- \$\epsilon$: Səhv (error) termini (modellə izah olunmayan təsadüfi səhvlər).

---

### **Xətti regresiyanın növləri**
1. **Sadə xətti regresiya (Simple Linear Regression):**  
   - Bir asılı dəyişən və bir müstəqil dəyişən arasındakı əlaqəni modelləşdirir.  
   - Tənlik:  
     $$
     y = \beta_0 + \beta_1x + \epsilon
     $$

2. **Çoxlu xətti regresiya (Multiple Linear Regression):**  
   - Bir asılı dəyişən və bir neçə müstəqil dəyişən arasındakı əlaqəni modelləşdirir.  
   - Tənlik:  
     $$
     y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n + \epsilon
     $$

---

### **Xətti regresiyanın əsas fərziyyələri**
Modelin düzgün işləməsi üçün aşağıdakı fərziyyələr nəzərə alınmalıdır:

1. **Xətti əlaqə:**  
   - Asılı dəyişən ($y$) ilə müstəqil dəyişənlər ($x_i$) arasında xətti bir əlaqə olmalıdır.

2. **Qalıqların normal paylanması:**  
   - Modeldən qalan səhv dəyərləri ($\epsilon$) normal paylanmalıdır.

3. **Homoskedastiklik:**  
   - Qalıqların varyansı sabit olmalıdır, yəni proqnozlaşdırılmış dəyərlərin səviyyəsindən asılı olaraq qalıqların dispersiyası dəyişməməlidir.

4. **Müstəqil dəyişənlərin multicollinearity olmaması:**  
   - Müstəqil dəyişənlər bir-biri ilə güclü şəkildə əlaqəli olmamalıdır. Əgər multicollinearity varsa, bu, modelin nəticələrinə mənfi təsir göstərə bilər.

5. **Autokorrelyasiyanın olmaması:**  
   - Qalıqlar bir-biri ilə əlaqəli olmamalıdır (xüsusilə zaman seriyasında).

---

### **Xətti regresiyanın üstünlükləri**
1. **Sadə və anlaşılır:** Riyazi olaraq asan başa düşülür və tətbiq olunur.
2. **Hesablama cəhətdən effektiv:** Böyük həcmdə data ilə sürətli nəticələr verir.
3. **Əsaslı interpretasiya:** Hər bir müstəqil dəyişənin asılı dəyişənə təsiri asanlıqla izah edilir.

---

### **Xətti regresiyanın çatışmazlıqları**
1. **Xətti əlaqə tələb edir:** Əgər dəyişənlər arasında xətti əlaqə yoxdursa, model yaxşı işləməz.
2. **Outlier-lərə həssaslıq:** Kənar dəyərlər (\(outliers\)) modelin nəticələrinə ciddi təsir edə bilər.
3. **Multicollinearity problemi:** Müstəqil dəyişənlər bir-biri ilə yüksək korrelyasiya edirsə, əmsallar düzgün qiymətləndirilə bilməz.
4. **Homoskedastiklik tələb edir:** Qalıqların varyansı dəyişirsə, model dəqiqliyi azalır.

---

### **Xətti regresiyanın performans ölçüləri**
Modelin performansını qiymətləndirmək üçün aşağıdakı metriklərdən istifadə olunur:

1. **Mean Absolute Error (MAE):**
   - Model səhvlərinin orta dəyərini göstərir.  
     $$
     MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
     $$

2. **Mean Squared Error (MSE):**
   - Səhvlərin kvadratlarının orta dəyərini ölçür.  
     $$
     MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
     $$

3. **Root Mean Squared Error (RMSE):**
   - MSE-nin kvadrat kökü. Model səhvini daha intuitiv şəkildə göstərir.  
     $$
     RMSE = \sqrt{MSE}
     $$

4. **$R^2$ (Determinasiya əmsalı):**
   - Modelin asılı dəyişənin variasiyasını izah etmə qabiliyyətini göstərir.  
     $$
     R^2 = 1 - \frac{\text{SSR}}{\text{SST}}
     $$
     Burada:
     - SSR: Səhvlərin kvadrat cəmi.
     - SST: Ümumi kvadrat cəm.

## Classification novleri

Klassifikasiya, verilənlərin müəyyən siniflərə və ya kateqoriyalara bölünməsini təmin edən maşın öyrənməsi metodudur. Klassifikasiya problemlərində hədəf dəyişən diskret (diskret dəyərlər ala bilən) olur və müxtəlif klassifikasiya növləri mövcuddur. Klassifikasiya növlərini başa düşmək üçün aşağıdakı əsas kateqoriyaları nəzərdən keçirmək faydalıdır:

### 1. **Binary Classification (İki Sinifli Klassifikasiya)**
   - **Ümumi prinsip**: İki sinif arasında seçim edir.
   - **Misallar**:
      - Xəstə/xəstə olmayan (məsələn, tibbi diaqnoz)
      - Fırıldaq/fırıldaq olmayan (məsələn, bank fırıldaqçılığını aşkar etmək)
   - **Ən çox istifadə olunan metodlar**:
      - **Logistic Regression**: Xətti modelləşdirmə əsaslı və iki sinif arasında ehtimalı təxmin edir.
      - **Support Vector Machines (SVM)**: Siniflər arasında ən yaxşı xətti sərhədi tapmağa çalışır.
      - **Decision Trees**: Məlumatları təkrarən bölərək sinifləri təyin edir.

### 2. **Multi-Class Classification (Çox Sinifli Klassifikasiya)**
   - **Ümumi prinsip**: Üç və ya daha çox sinif arasında seçim edir.
   - **Misallar**:
      - Şəkilləri "it", "pişik", "quş" və s. kimi müxtəlif kateqoriyalara ayırmaq
      - Müştəriləri müxtəlif risk qruplarına ayırmaq
   - **Ən çox istifadə olunan metodlar**:
      - **Softmax Regression**: Logistic regression-ın genişləndirilmiş formasıdır, bir neçə sinif üçün ehtimalları çıxarmağa imkan verir.
      - **Random Forest**: Çox sayda decision tree modelinin birləşməsi ilə yüksək dəqiqlik verir.
      - **Naive Bayes**: Hər bir sinifin ehtimalını təxmin edərək ən uyğun sinfi seçir.
      - **Neural Networks**: Sinir şəbəkələri sinifləndirmə üçün effektiv nəticələr verir, xüsusilə dərin öyrənmə modellərində.

### 3. **Multi-Label Classification (Çox Etiketli Klassifikasiya)**
   - **Ümumi prinsip**: Hər bir nümunəyə birdən çox sinif etiketi verilir.
   - **Misallar**:
      - Şəkildə həm "it", həm "avtomobil", həm də "ağac" varsa, bütün bu etiketləri təyin etmək
      - Mətnlərin birdən çox mövzu ilə əlaqələndirilməsi, məsələn, bir məqalə həm “İqtisadiyyat” həm də “Siyasət” kateqoriyalarında ola bilər.
   - **Ən çox istifadə olunan metodlar**:
      - **Binary Relevance**: Hər bir etiket üçün ayrı binary classifier qurur.
      - **Classifier Chains**: Hər bir sinfi müəyyən etdikdən sonra nəticələri digər siniflərə ötürərək işləyir.
      - **Adapted Algorithms**: Multi-label üçün xüsusi hazırlanmış algoritmlər, məsələn, ML-kNN (Multi-label KNN).

### 4. **Imbalanced Classification (Balanssız Klassifikasiya)**
   - **Ümumi prinsip**: Verilənlərdə siniflərin bərabər sayda olmadığı halları əhatə edir.
   - **Misallar**:
      - Fırıldaqçılıq deteksiyasında "fırıldaq" halları "normal" hallara nisbətən çox az ola bilər.
      - Səhiyyədə nadir xəstəliklərin diaqnozu üçün verilənlərdə xəstələr çox az, sağlamlar isə çox sayda ola bilər.
   - **Həll üsulları**:
      - **Class Weighting**: Modelin balanssızlığı aradan qaldırması üçün sinif çəkilərini tənzimləmək.
      - **Resampling Techniques**: Azlıqda olan sinifləri artırmaq (oversampling) və ya çox olan sinifləri azaltmaq (undersampling).
      - **Specialized Algorithms**: SMOTE (Synthetic Minority Over-sampling Technique) və ADASYN kimi balanssız verilənlər üçün uyğun metodlar.

### 5. **Ordinal Classification (Sıralı Klassifikasiya)**
   - **Ümumi prinsip**: Siniflər arasında müəyyən bir sıralama olduğunda istifadə olunur.
   - **Misallar**:
      - Müştəri məmnuniyyəti sorğuları: “çox narazı”, “narazı”, “neytral”, “razı”, “çox razı”.
      - Kredit reytinqləri: “çox riskli”, “riskli”, “normal”, “yüksək riskli” və s.
   - **Metodlar**:
      - **Ordinal Logistic Regression**: Sıralı klassifikasiya üçün logistic regression-ın genişləndirilmiş variantıdır.
      - **Tree-Based Methods**: Decision tree və random forest metodları modifikasiyalarla sıralı klassifikasiya üçün istifadə edilə bilər.

### 6. **Hierarchical Classification (Hierarşik Klassifikasiya)**
   - **Ümumi prinsip**: Siniflər arasında hierarşik əlaqə olduqda istifadə olunur.
   - **Misallar**:
      - Canlılar aləmi təsnifatı: “Canlılar” → “Heyvanlar” → “Məməlilər” → “İtlər” və s.
   - **Metodlar**:
      - **Top-Down və Bottom-Up Approaches**: Məlumatları yuxarıdan aşağıya və ya əksinə, siniflər arasında ayırmaq.

Bu klassifikasiya növləri tətbiq sahəsinə görə istifadə olunur və modellərin xüsusiyyətlərini nəzərə almaq vacibdir. Modelin düzgün seçilməsi və uyğun parametrlərin təyin edilməsi də nəticələrin dəqiqliyini artırır.

## Logistic Regression nedir?

### **Logistic Regression** (Logistik Regressiya)

>**Tərif:** Logistik Regressiya, təsnifat problemlərini həll etmək üçün geniş istifadə olunan statistik və maşın öyrənmə metodudur. Əsas məqsəd müstəqil dəyişənlər (features) ilə ikili və ya çoxlu sinifli hədəf dəyişəni (target) arasında əlaqəni modelləşdirmək və sinif ehtimallarını proqnozlaşdırmaqdır. Adı "regressiya" olsa da, təsnifat üçün istifadə edilir.

---

### **Logistik Regressiyanın əsas prinsipi**

Logistik Regressiya xətti regresiyanın bir genişləndirilmiş formasıdır. Burada xətti əlaqə **log-odds** (logistik funksiyanın girişi) vasitəsilə proqnozlaşdırılır və ehtimallara çevrilir.

Modelin əsas tənliyi:  
$$
p(y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n)}}
$$

Burada:  
- $p(y=1|X)$: Hədəf dəyişənin \(1\) sinifinə aid olması ehtimalı.  
- $\beta_0$: Intercept (başlanğıc nöqtəsi).  
- $\beta_1, \beta_2, \dots, \beta_n$: Əmsallar (hər bir müstəqil dəyişənin təsiri).  
- $x_1, x_2, \dots, x_n$: Müstəqil dəyişənlər.  
- $e$: Təbii logaritmanın əsas konstantası (təqribən 2.718).  

**Odds** və **Log-Odds**:
- Odds:  
  $$
  Odds = \frac{p}{1-p}
  $$
  Burada \(p\), müsbət hadisənin ehtimalıdır.
- Log-Odds (logit funksiyası):  
  $$
  \text{logit}(p) = \ln\left(\frac{p}{1-p}\right)
  $$

---

### **Logistik Regressiyanın növləri**

1. **Binary Logistic Regression (İkili Logistik Regressiya):**  
   - Hədəf dəyişən yalnız iki sinfə malikdir (məsələn, 0 və 1).  
     Məsələn:  
     - Email spamdır (1) və ya deyil (0).  
     - Müştəri məhsul alacaq (1) və ya almayacaq (0).  

2. **Multinomial Logistic Regression (Çoxlu Sinif Logistik Regressiya):**  
   - Hədəf dəyişən birdən çox sinifə malikdir və siniflər arasında ardıcıllıq yoxdur.  
     Məsələn: Məhsul rəngi (qırmızı, mavi, yaşıl).

3. **Ordinal Logistic Regression:**  
   - Hədəf dəyişən birdən çox sinifə malikdir və siniflər arasında ardıcıllıq vardır.  
     Məsələn: Müştəri məmnuniyyəti (zəif, orta, yüksək).  

---

### **Logistik Regressiyanın fərziyyələri**
1. **Log-odds ilə müstəqil dəyişənlər arasında xətti əlaqə:**  
   - Logistik Regressiyanın xətti quruluşu müstəqil dəyişənlə log-odds arasında əlaqəni ifadə edir.

2. **Multicollinearity olmaması:**  
   - Müstəqil dəyişənlər bir-biri ilə yüksək korrelyasiya etməməlidir.

3. **Hədəf dəyişənin təsnifat problemi olması:**  
   - Logistik Regressiya yalnız təsnifat problemləri üçün işləyir (hədəf dəyişən kəmiyyət deyil, kateqoriyadır).

4. **Qalıqların heteroskedastiklik olmaması:**  
   - Müstəqil dəyişənlər model səhvlərini sabit dispersiya ilə izah etməlidir.

---

### **Logistik Regressiyanın performans ölçüləri**

Logistik Regressiyanın düzgünlüyünü qiymətləndirmək üçün bir neçə metrik istifadə olunur:

1. **Confusion Matrix:**  
   - Doğru və yanlış təsnif edilmiş nümunələri göstərir.  
     - True Positive (TP): Doğru şəkildə 1 olaraq təsnif edilmiş nümunələr.  
     - False Positive (FP): Yanlış şəkildə 1 olaraq təsnif edilmiş nümunələr.  
     - True Negative (TN): Doğru şəkildə 0 olaraq təsnif edilmiş nümunələr.  
     - False Negative (FN): Yanlış şəkildə 0 olaraq təsnif edilmiş nümunələr.

2. **Accuracy (Dəqiqlik):**  
   $$
   Accuracy = \frac{TP + TN}{TP + FP + TN + FN}
   $$

3. **Precision (Dəqiqlik):**  
   $$
   Precision = \frac{TP}{TP + FP}
   $$

4. **Recall (Xatırlama):**  
   $$
   Recall = \frac{TP}{TP + FN}
   $$

5. **F1-Score:**  
   - Precision və Recall arasındakı tarazlığı ölçür.  
     $$
     F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}
     $$

6. **ROC-AUC Score:**  
   - Modelin müxtəlif threshold-lar üçün performansını ölçür.  

---

### **Logistik Regressiyanın üstünlükləri**

1. **Sadə və effektiv:** Qısa zamanda tətbiq edilir və əsaslı nəticələr verir.
2. **Ehtimal proqnozu verir:** Sinif ehtimallarını proqnozlaşdırmaq üçün yaxşıdır.
3. **Təsnifat üçün geniş istifadə olunur:** Bir çox real dünya problemlərində tətbiq olunur.
4. **Çox yönlüdür:** İkili və çoxlu sinif problemləri üçün uyğunlaşa bilir.

---

### **Logistik Regressiyanın çatışmazlıqları**

1. **Xətti əlaqə fərziyyəsi:** Log-odds ilə müstəqil dəyişənlər arasında xətti əlaqə tələb edir.
2. **Outlier-lərə həssasdır:** Kənar dəyərlər əmsalların düzgün qiymətləndirilməsini poza bilər.
3. **Qeyri-xətti problemlər üçün uyğun deyil:** Qeyri-xətti əlaqələri modelləşdirməkdə zəifdir.
4. **Dəyişən seçimi:** Düzgün müstəqil dəyişənlər seçilməzsə, model nəticələri düzgün olmayacaq.

## Log-odds nədir?

Logistic regresiyanın əsas fərziyyələrindən biri olan **"Müstəqil dəyişənlərlə log-odds arasında xətti əlaqə olmalıdır"** ifadəsi, logistic regresiya modelinin düzgün işləməsi üçün modelin əsas prinsipinə bağlıdır. Bunun mənasını izah edək:

---

### Log-odds nədir?

>**Tərif:** Log-odds, bir hadisənin baş vermə ehtimalının baş verməmək ehtimalına nisbətinin loqarifmik transformasiyasıdır və logistika regresiyası kimi modellərdə, ehtimalların daha asan idarə edilə bilməsi üçün istifadə olunur. Bu, odds (qazanma və itirmə nisbəti) funksiyasının loqarifmasını alaraq, ehtimal dəyərlərinin geniş bir intervalda qiymətləndirilməsini təmin edir və modelin nəticələrini daha əlçatan edir.

Logistic regresiya, bir hadisənin (məsələn, $y=1$) ehtimalını təxmin edir. Bu ehtimal ($P(y=1|x)$) ilə əlaqəli log-odds (logarifmik odds) aşağıdakı kimi hesablanır:

$$
\text{Odds} = \frac{P(y=1|x)}{1 - P(y=1|x)}
$$

Log-odds isə bu odds-un təbii logarifmidir:

$$
\text{Log-odds} = \ln\left(\frac{P(y=1|x)}{1 - P(y=1|x)}\right)
$$

Log-odds xətti olaraq logistic regresiyanın tənliyi ilə ifadə olunur:

$$
\ln\left(\frac{P(y=1|x)}{1 - P(y=1|x)}\right) = \beta_0 + \beta_1x_1 + \beta_2x_2 + \dots + \beta_nx_n
$$

---

### Fərziyyənin mənası:
Fərziyyəyə görə, müstəqil dəyişənlər ($x_1, x_2, \dots$) ilə log-odds arasında **xətti münasibət** olmalıdır. Bu o deməkdir ki:

- Müstəqil dəyişənlərin ($x_i$) hər biri log-odds-a **proportional** (mütənasib) təsir göstərir.  
- Dəyişənlər log-odds-u birbaşa artıra və ya azalda bilər, lakin bu dəyişiklik xətti olmalıdır.

Məsələn:
- Bir müstəqil dəyişənin ($x$) $1$ vahid artması, log-odds-da müəyyən bir dəyər ($\beta$) qədər artışa səbəb olur.
- Bu təsir bütün giriş dəyərlərində sabit qalır.

---

### Əsas məqsəd:
Bu fərziyyə, logistic regresiyanın nəticəni düzgün proqnozlaşdırması üçün zəruridir. Əgər xətti münasibət yoxdursa, model qeyri-dəqiq nəticələr verə bilər.

---

### Xətti münasibətin pozulması nə deməkdir?
Əgər müstəqil dəyişənlərlə log-odds arasında xətti münasibət yoxdursa:
- Logistic regresiya düzgün proqnozlar verə bilməz.  
- Model ya underfitting (çatışmayan uyğunlaşma), ya da overfitting (artıq uyğunlaşma) problemi ilə üzləşə bilər.

Məsələn:
- Əgər bir dəyişən log-odds-u eksponent olaraq artırırsa, bu, xətti münasibətə uyğun deyil. Belə bir vəziyyətdə dəyişən üzərində çevrilmələr (məsələn, logaritmik və ya kvadrat kök çevrilməsi) tətbiq edilə bilər.

---

### Bu problemi necə yoxlamaq və həll etmək olar?
1. **Vizual yoxlama:** 
   - Müstəqil dəyişənlə ehtimal ($P(y=1|x)$) arasındakı əlaqəni scatter plot ilə analiz edin. Əgər münasibət xətti deyilsə, bu görünəcək.

2. **Dəyişənləri çevirmək:**
   - Xətti olmayan dəyişənləri logaritm, kvadrat kök və ya polynomial çevrilmələrlə xətti münasibətə uyğunlaşdırmağa çalışın.

3. **İnteraktiv dəyişənlər daxil edin:**
   - Xətti olmayan münasibəti modelləşdirmək üçün dəyişənlərin qarşılıqlı təsirini ($x_1 \cdot x_2$) əlavə edə bilərsiniz.

4. **Alternativ modellər istifadə edin:**
   - Logistic regresiya xətti münasibət tələb etdiyindən, xətti olmayan münasibətləri modelləşdirmək üçün **decision tree**, **random forest** və ya **neural network** kimi metodlara müraciət edə bilərsiniz.

---

## Unsupervised Learning (Nəzarətsiz Öyrənmə)

>**Tərif:** **Unsupervised Learning (Nəzarətsiz Öyrənmə)**, verilənlərdə çıxış etiketləri olmadan **yalnız girişlər** arasındakı **gizli strukturları, nümunələri və əlaqələri** müəyyən etməyə çalışan maşın öyrənmə yanaşmasıdır.


**Unsupervised Learning (Nəzarətsiz Öyrənmə)**, maşın öyrənmənin əsas növlərindən biridir. Bu yanaşmada, model yalnız giriş məlumatlarını istifadə edir və çıxış etiketləri olmadan verilənlər arasında gizli nümunələri, strukturları və əlaqələri öyrənməyə çalışır. Məqsəd məlumatları təsnif etmək, qruplaşdırmaq və ya sıxışdırmaq (dimensionality reduction) kimi vəzifələri yerinə yetirməkdir.

---

### **Unsupervised Learning-in Əsas Prinsipləri**
1. **Etiketlənməmiş Məlumat:**
   - Təlim zamanı yalnız giriş (input) məlumatları təqdim olunur.
   - Çıxış etiketləri yoxdur, çünki məqsəd məlumatdakı gizli əlaqələri tapmaqdır.

2. **Məlumatın Strukturu:**
   - Model məlumatdakı əsas strukturları öyrənir və bu strukturlara əsasən məlumatları qruplaşdırır, azaldır və ya təşkil edir.

3. **Öyrənmə Nəticəsi:**
   - Model məlumatdakı oxşarlıqları və fərqləri əsasən təsnif etmə və ya vizuallaşdırma üçün istifadə olunur.

---

### **Unsupervised Learning-in Əsas Tətbiqləri**

1. **Klasterləşdirmə (Clustering):**
   - Məlumatları oxşarlığına və ya xüsusiyyətlərinə əsasən qruplaşdırmaq.
   - Məsələn:
     - Müştəriləri davranışlarına görə qruplaşdırmaq.
     - Məlumatdan fərqli kateqoriyaları avtomatik müəyyənləşdirmək.
   - Populyar alqoritmlər:
     - K-Means
     - Hierarchical Clustering
     - DBSCAN

2. **Dimensionality Reduction (Ölçü Azaltma):**
   - Çoxsaylı xüsusiyyətləri daha az ölçülərlə təmsil etmək.
   - Bu üsul məlumatı vizuallaşdırmaq və ya hesablama prosesini sürətləndirmək üçün istifadə olunur.
   - Məsələn:
     - Müxtəlif xüsusiyyətlərdən ibarət olan genetik məlumatları analiz etmək.
     - İri ölçülü məlumatları iki və ya üç ölçüdə vizuallaşdırmaq.
   - Populyar alqoritmlər:
     - Principal Component Analysis (PCA)
     - t-SNE
     - UMAP

3. **Assosiasiya Qaydaları (Association Rules):**
   - Məlumatlar arasında əlaqə və qaydaları aşkar etmək.
   - Məsələn:
     - Mağazalarda məhsul tövsiyəsi sistemi qurmaq ("X məhsulunu alanlar Y məhsulunu da alır").
   - Populyar alqoritmlər:
     - Apriori
     - Eclat

4. **Anomaliya Aşkarlanması (Anomaly Detection):**
   - Nadir və ya fərqli nümunələri tapmaq.
   - Məsələn:
     - Kredit kartı fırıldaqlarını aşkar etmək.
     - Sistem məntiqində anormal fəaliyyətləri müəyyənləşdirmək.
   - Alqoritmlər:
     - Isolation Forest
     - One-Class SVM

---

### **Unsupervised Learning-in Üstünlükləri**
1. **Etiketlənmiş Məlumata Ehtiyac Yoxdur:**
   - Etiketli məlumat toplamaq baha və vaxt tələb edən proses olduğu üçün nəzarətsiz öyrənmə daha sərfəlidir.

2. **Yeni Nümunələrin Kəşfi:**
   - Məlumatların insan tərəfindən bilinməyən strukturlarını kəşf etməyə imkan verir.

3. **Çoxşaxəli Tətbiq:**
   - Müxtəlif sahələrdə (biologiya, marketinq, maliyyə) məlumat analizi və proqnozlaşdırma üçün istifadə olunur.

---

### **Unsupervised Learning Alqoritmləri**

#### **Klasterləşdirmə Algoritmləri**
1. **K-Means Clustering:**
   - Məlumatları əvvəlcədən müəyyən edilmiş $ K $ sayda klasterə ayırır.
   - Klasterlərin mərkəzi (centroid) məlumatın xüsusiyyətlərindən asılı olaraq təyin olunur.

2. **Hierarchical Clustering:**
   - Məlumatları iyerarxik strukturda qruplaşdırır.
   - İki növü var: Aşağıdan-yuxarıya (Agglomerative) və Yuxarıdan-aşağıya (Divisive).

3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise):**
   - Yüksək sıxlığa malik bölgələri klaster kimi müəyyən edir və səs-küy məlumatlarını ayırır.

#### **Dimensionality Reduction Algoritmləri**
1. **Principal Component Analysis (PCA):**
   - Məlumatın əsas dəyişkənliyini qorumaq üçün ölçüləri azaldır.
2. **t-SNE (t-Distributed Stochastic Neighbor Embedding):**
   - Ölçü azaldaraq məlumatı iki və ya üç ölçüdə vizuallaşdırmaq üçün istifadə olunur.
3. **UMAP (Uniform Manifold Approximation and Projection):**
   - Daha effektiv və sürətli ölçü azaltma alqoritmidir.

#### **Assosiasiya Qaydaları**
- Apriori və Eclat algoritmləri ilə marketinq və satış sistemlərində məhsul tövsiyəsi sistemi qurulur.

---

### **Unsupervised Learning-in Çətinlikləri**
1. **Məlumatların Şərhi:**
   - Modelin tapdığı strukturların və klasterlərin izah edilməsi çətin ola bilər.
2. **Optimal Klaster Sayının Seçimi:**
   - Klasterlərin sayı əvvəlcədən bilinmədikdə, optimal sayın müəyyən edilməsi çətinlik yaradır.
3. **Overfitting Riski:**
   - Məlumatın yanlış strukturlarını taparaq nəticələrin ümumiləşmə qabiliyyətini azalda bilər.

---

### **Unsupervised Learning-in Real Tətbiqləri**
1. **Müştəri Seqmentasiyası:**
   - Müştəriləri alış-veriş davranışlarına əsasən qruplaşdırmaq.
2. **Tövsiyə Sistemləri:**
   - Məhsul və ya xidmətlər üçün tövsiyələr yaratmaq.
3. **Anomaliyaların Aşkarlanması:**
   - Maliyyə fırıldaqlarını, təhlükəsizlik problemlərini və ya texniki nasazlıqları müəyyənləşdirmək.
4. **Bioloji Tədqiqatlar:**
   - Gen ifadələrinin qruplaşdırılması və ya dərman təsirlərinin təhlili.

---

### **Nəticə**
Unsupervised Learning, məlumatın etiketlənməsi çətin və ya qeyri-mümkün olduğu hallarda effektiv həll yoludur. Klasterləşdirmə, ölçü azaltma və assosiasiya qaydaları kimi metodlarla məlumatların strukturunu tapmaq, vizuallaşdırmaq və proqnozlaşdırmaq üçün geniş tətbiq sahələrinə malikdir. Lakin optimal parametr seçimi və nəticələrin düzgün şərhi müəyyən problemlər yarada bilər.

## Cluster nedir?

 >**Tərif:** **Clustering** – verilənlərdəki strukturları və oxşarlıqları modelləşdirərək, obyektləri heç bir əvvəlcədən təyin olunmuş etiket olmadan təbii qruplara (klasterlərə) ayıran statistik və maşın öyrənməsi metodudur. Bu yanaşma, verilənlər arasındakı nümunələri və qanunauyğunluqları aşkar edərək, müştəri seqmentasiyası, anomaliya aşkarlanması və tövsiyə sistemləri kimi sahələrdə geniş istifadə olunur. 🔍📊

---

**Cluster** (klaster) verilənlərdəki müşahidələrin (data nöqtələrinin) oxşarlıqlarına əsaslanaraq qruplaşdırıldığı bir qrup və ya seqmentdir. **Clustering** isə bu qruplaşdırma prosesidir, hansı ki, verilənləri təbii olaraq bənzər xüsusiyyətlərə malik olan qruplara ayırmaq məqsədi daşıyır. Bu proses **nəzarətsiz öyrənmə** (unsupervised learning) texnikalarından biridir, yəni modelin təlimi üçün etiketli və ya hədəf dəyişəninə ehtiyac yoxdur.

### Klasterləşdirmənin məqsədi nədir?
Clustering-in əsas məqsədi verilənlərdəki təbii qrupları tapmaqdır. Klasterlər daxilindəki müşahidələr bir-birinə oxşar olmalı, fərqli klasterlərdə olan müşahidələr isə bir-birindən fərqlənməlidir. Bu metod, verilənlərdə gizli olan strukturları aşkar etmək və müşahidələri daha kiçik, bənzər qruplara bölmək üçün istifadə olunur.

### Klaster (Cluster) xüsusiyyətləri:
1. **Bənzərlik**: Klaster daxilindəki bütün məlumat nöqtələri bir-birinə mümkün qədər oxşar olmalıdır.
2. **Fərqlilik**: Müxtəlif klasterlərdə olan məlumat nöqtələri mümkün qədər fərqli olmalıdır.
3. **Köhnə etiketlər yoxdur**: Clustering təbii qrupları tapmaq üçün məlumatlardan istifadə edir və əvvəlcədən təyin edilmiş siniflər yoxdur.

### Klasterləşdirmə üçün istifadə olunan əsas alqoritmlər:
1. **K-Means Clustering**: Ən populyar klasterləşdirmə metodlarından biridir. K istifadəçi tərəfindən müəyyən edilir və alqoritm verilənləri "K" ədəd klasterə bölməyə çalışır.
2. **Hierarchical Clustering**: Bu metod verilənləri iyerarxik şəkildə qruplaşdırır, yəni kiçik klasterlər daha böyük klasterlərə qoşularaq ağacvari struktur yaradır.
3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: Bu metod verilənlərin sıxlığına əsasən klasterləri müəyyən edir. Sıx verilənlər klaster kimi qəbul edilir, sıx olmayan nöqtələr isə "şübhəli" nöqtələr hesab olunur.
4. **Mean Shift Clustering**: Verilənlərin sıxlığına əsaslanaraq qruplar tapmağa çalışır və klasterləri müəyyən edir.

### Klasterləşdirmə harada istifadə olunur?
- **Müştəri seqmentasiyası**: Müştəriləri davranışlarına, alış-veriş vərdişlərinə və digər göstəricilərə əsasən qruplara ayırmaq.
- **Marketinq təhlili**: Müxtəlif məhsulların və ya bazar seqmentlərinin təhlil edilməsi.
- **Tibbi tədqiqatlar**: Xəstələri xəstəlik növlərinə və ya oxşar simptomlara görə qruplaşdırmaq.
- **Sosial şəbəkə analizləri**: İstifadəçilərin əlaqələrinə və davranışlarına əsasən sosial qrupların müəyyən edilməsi.
- **Anomaliya aşkarlanması**: Normadan kənar olan verilənləri aşkar etmək (məsələn, fırıldaqçılıq, sistemdəki səhvlər).

### Klasterlər ilə bağlı misal
Təsəvvür edin ki, bir marketinq şirkətinin müştəri bazası var və bu müştəriləri onların satınalma vərdişlərinə əsasən qruplaşdırmaq istəyirsiniz. **Klasterləşdirmə** istifadə edərək, müştəriləri fərqli qruplara ayıra bilərsiniz:
- **Klaster 1**: Daha çox lüks məhsullar alan müştərilər.
- **Klaster 2**: Endirimli məhsulları almağa üstünlük verən müştərilər.
- **Klaster 3**: Tez-tez alış-veriş edən, lakin kiçik məbləğdə alış-veriş edən müştərilər.

Bu qruplar marketinq strategiyasını fərqli qruplar üçün optimallaşdırmaq üçün istifadə edilə bilər.

### Nəticə
Klaster, bənzər verilənlərin bir yerə toplandığı qrupdur. Klasterləşdirmə verilənləri təbii qruplara ayırmaq üçün istifadə olunan nəzarətsiz öyrənmə metodudur. Məqsəd, verilənlər arasında gizli qrupları tapmaq və onları oxşar xüsusiyyətlərə görə qruplaşdırmaqdır.

## Cluster metdolari

Klaster analizində istifadə olunan metodlar, verilənləri qruplara ayırmaq və hər bir qrupa aid oxşar obyektləri təyin etmək üçün müxtəlif yanaşmaları əhatə edir. Əsas klaster metodları bunlardır:

### 1. **K-Means Clustering**
   - **Ümumi prinsip**: Verilənləri müəyyən sayda klasterə (qruplara) ayırır və hər klasterin mərkəzi olan "centroid" hesablanır.
   - **Prosesi**: Verilən hər bir nöqtə ən yaxın centroid-ə təyin olunur, sonra centroid-lər yenidən hesablanır. Bu proses nöqtələr uyğun klasterlərə düşənə qədər təkrarlanır.
   - **Əlavə məlumat**: K-means klasterləri təyin etmək üçün əvvəlcədən klaster sayını (`k`) təyin etməyi tələb edir.

### 2. **Hierarchical Clustering**
   - **Ümumi prinsip**: Verilənləri ardıcıl şəkildə birləşdirərək və ya bölərək klasterlər yaradır.
   - **İki əsas növü**:
      - **Agglomerative (Bottom-Up)**: Hər bir obyekt bir klaster olaraq başlayır, sonra klasterlər tədricən birləşir.
      - **Divisive (Top-Down)**: Verilənlər əvvəlcə bir klasterdə olur, sonra təkrarlayan şəkildə kiçik klasterlərə bölünür.
   - **Dendrogram**: Bütün verilənlərin qruplara necə bölündüyünü göstərən bir diaqram (ağac strukturu) yaradılır.

### 3. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**
   - **Ümumi prinsip**: Verilənləri sıxlıq əsaslı qruplara ayırır.
   - **Prosesi**: Verilən sıxlıq ətrafında toplanmış nöqtələri klaster olaraq təyin edir və "noise" (səs-küy) adlanan ayrılmış nöqtələri qruplaşdırmaz.
   - **Əlavə məlumat**: DBSCAN klaster sayını əvvəlcədən təyin etməyə ehtiyac olmadan sıxlıq əsaslı klasterləri müəyyən etmək üçün effektivdir.

### 4. **Mean-Shift Clustering**
   - **Ümumi prinsip**: Verilən nöqtələrin sıx olduğu bölgələrə doğru nöqtələri tədricən hərəkət etdirir.
   - **Prosesi**: Hər bir nöqtə sıxlığın yüksək olduğu sahəyə doğru iterativ şəkildə hərəkət edir.
   - **Əlavə məlumat**: Mean-Shift, verilənlərdəki sıxlıq dəyişikliklərini analiz edərək klasterlərin təbii sayını tapmağa kömək edir.

### 5. **Gaussian Mixture Models (GMM)**
   - **Ümumi prinsip**: Verilənləri Gaussian (normal) paylanma ilə təmsil olunan alt qruplara ayırır.
   - **Prosesi**: Hər bir klaster Gaussian paylanma olaraq modelləşdirilir və verilənlər bu modellərə əsasən bölünür.
   - **Əlavə məlumat**: GMM, K-Means-in təkmilləşdirilmiş variantı kimi qəbul edilir, çünki GMM klasterlərin formasında daha elastikdir.

### 6. **Spectral Clustering**
   - **Ümumi prinsip**: Verilənləri qraf nəzəriyyəsi əsasında klasterlərə ayırır.
   - **Prosesi**: Verilənlər bir qraf üzərində təmsil edilir və bu qrafın spektral xüsusiyyətlərindən istifadə edilərək klasterlər müəyyən edilir.
   - **Əlavə məlumat**: Bu metod, xüsusilə qeyri-xətti strukturlara malik verilənlər üçün yaxşı nəticələr verir.

### 7. **Agglomerative Clustering**
   - **Ümumi prinsip**: Kiçik klasterləri iterativ olaraq birləşdirərək daha böyük klasterlər yaradır.
   - **Prosesi**: Əvvəlcə hər bir nöqtə öz klasteri kimi qəbul edilir və hər dəfə ən yaxın iki klaster birləşdirilir.
   - **Əlavə məlumat**: Bu metod dendrogram vasitəsilə vizuallaşdırıla bilər və klasterlərin sayını dendrogram əsasında müəyyən etmək mümkündür.

### 8. **Birlikdə İstifadə Edilən Modellər və Klasterlər (Cluster Ensembles)**
   - **Ümumi prinsip**: Müxtəlif klaster metodlarını birlikdə istifadə edərək daha dəqiq klasterlər əldə edir.
   - **Əlavə məlumat**: Müxtəlif klasterlərin nəticələrini birləşdirərək, hər bir metodun çatışmazlıqlarını digərləri ilə kompensasiya edir.

Hər bir metod fərqli verilən strukturlarına və məqsədlərə uyğun gəldiyi üçün, klaster analizində uyğun metodu seçmək üçün verilənlərin xüsusiyyətlərini və tətbiq sahəsini nəzərə almaq vacibdir.

## Association Rule 


>**Tərif:** **Association Rule** – verilənlər toplusunda dəyişənlər arasındakı əlaqələri və birlikdə baş vermə nümunələrini modelləşdirərək, obyektlər arasındakı assosiativ qaydaları müəyyən edən məlumat çıxarma və maşın öyrənməsi metodudur. Bu yanaşma, xüsusilə alış-veriş səbəti təhlili, tövsiyə sistemləri və istifadəçi davranışının analizi kimi sahələrdə tətbiq olunur və "Əgər X baş verərsə, Y-nin baş vermə ehtimalı yüksəkdir" prinsipinə əsaslanır. 🔍📊

---

**Association rule** (assosiasiya qaydaları) verilənlər arasında gizli nümunələri və əlaqələri aşkar etmək üçün istifadə olunan bir **nəzarətsiz öyrənmə** metodudur. Bu metod, xüsusilə pərakəndə satışda və marketinq analizində geniş istifadə olunur. Məsələn, hansı məhsulların tez-tez birlikdə alındığını müəyyən etməyə imkan verir. 

### Association rule nədir?
- **Assosiasiya qaydaları** müxtəlif obyektlər (məhsullar, hadisələr və s.) arasında əlaqələr quraraq onların birlikdə ortaya çıxma ehtimalını aşkar etməyə çalışır. Bu qaydalar əsasən **“Əgər-X Onda-Y”** şəklində yazılır. Yəni, **Əgər** bir hadisə baş veribsə, **Onda** başqa bir hadisənin baş vermə ehtimalı yüksəkdir.
- **Misal**: Bir marketdə tez-tez birlikdə alınan məhsulları təhlil etmək. Məsələn, müştərilər tez-tez **çörək** aldıqda, **pendir** də alırlar. Bu zaman qayda belədir: "Əgər çörək alınıbsa, onda pendir alma ehtimalı yüksəkdir."

### Əsas terminlər
Assosiasiya qaydalarının düzgün başa düşülməsi üçün bəzi əsas terminləri bilmək vacibdir:

1. **Support (Dəstək)**:
   - Bu, müəyyən bir məhsulun və ya məhsul qrupunun ümumi alış-verişdə nə qədər tez-tez baş verdiyini göstərir. Başqa sözlə, "support" həmin qaydanın nə qədər ümumi olduğu haqqında məlumat verir.
   - **Formula**: 
     $
     \text{Support}(X) = \frac{\text{X obyektinin rast gəlmə sayı}}{\text{ümumi alış-veriş sayı}}
     $
   - **Misal**: 100 alış-verişdə 20 dəfə çörək alınıbsa, çörəyin support dəyəri 0.2-dir (yəni, 20%).

2. **Confidence (Əminlik)**:
   - Bu, müəyyən bir məhsulun alışı zamanı digər bir məhsulun alınma ehtimalını göstərir. Yəni, "Əgər X alınıbsa, onda Y də alınır" ehtimalıdır.
   - **Formula**: 
     $
     \text{Confidence}(X \rightarrow Y) = \frac{\text{X və Y birlikdə alınma sayı}}{\text{X alınma sayı}}
     $
   - **Misal**: 100 alış-verişdə 20 dəfə həm çörək, həm də pendir alınıbsa, amma çörək ümumilikdə 50 dəfə alınıbsa, "çörək → pendir" əminlik dəyəri 20/50 = 0.4 olacaq (yəni, 40%).

3. **Lift**:
   - Lift, **X və Y** məhsullarının birlikdə alınmasının təsadüfi olma ehtimalına nisbətən nə qədər daha çox olduğunu göstərir. Başqa sözlə, lift, bir qaydanın önəmli olub-olmamasını göstərir.
   - **Formula**: 
     $
     \text{Lift}(X \rightarrow Y) = \frac{\text{Confidence}(X \rightarrow Y)}{\text{Support}(Y)}
     $
   - **Misal**: Əgər lift dəyəri 1-dən böyükdürsə, deməli, X alındıqda Y alma ehtimalı yüksəkdir. 1-dən kiçikdirsə, o zaman birlikdə alınma ehtimalı aşağıdır.

### Misal
Bir supermarketdə alış-verişlə bağlı assosiasiya qaydalarını yaratmaq üçün belə bir nümunəyə baxaq:

- Müştəri tez-tez **çörək** aldıqda **yağ** da alır. 
- **Support**: 20% (çörək və yağ birlikdə 100 alış-verişin 20-də alınıb).
- **Confidence**: 50% (çörək alındıqda, onun yanında yağ alınması ehtimalı 50%).
- **Lift**: 2 (çörək alındıqda yağ alma ehtimalı, yağın ümumi alınma ehtimalından iki dəfə çoxdur).

Bu qaydanı istifadə edərək market, müştərilər çörək aldıqda yağın yanında endirim təklif edərək satışları artıra bilər.

### Association rule-un istifadə sahələri:
1. **Market Basket Analysis** (Alış-veriş səbəti təhlili): Müştərilərin hansı məhsulları birlikdə almağa meylli olduğunu müəyyən etmək.
2. **Tövsiyə sistemləri**: Amazon və Netflix kimi platformalar istifadəçilərə hansı məhsul və ya filmi tövsiyə edəcəyini müəyyən etmək üçün assosiasiya qaydalarından istifadə edir.
3. **Fırıldaqçılığın aşkarlanması**: Banklar və kredit kartı şirkətləri fırıldaqçı fəaliyyətlərin müəyyən bir nümunəsini aşkar etmək üçün assosiasiya qaydalarından istifadə edir.
4. **Xəstəlik analizi**: Tibb sahəsində müəyyən simptomlar və xəstəliklərin birlikdə ortaya çıxma ehtimalını müəyyən etmək üçün assosiasiya qaydaları istifadə olunur.

### Algoritmlər:
- **Apriori**: Association rule-ları tapmaq üçün ən çox istifadə edilən alqoritmlərdən biridir. Bu metod verilənlərdəki tez-tez rast gəlinən məhsul dəstlərini aşkar edir və onların arasında əlaqələri müəyyən edir.
- **FP-Growth**: Apriori-yə alternativ bir alqoritmdir və daha sürətli hesablamalarla tanınır. Verilənlərin sıx strukturlarını nəzərə alaraq tez-tez rast gəlinən obyekt qruplarını tapır.

### Nəticə
**Association rule learning** verilənlərdə obyektlərin birlikdə yaranma ehtimalını təhlil etmək və onların arasındakı əlaqələri aşkar etmək üçün istifadə olunur. Bu qaydalar marketinq, satış, fırıldaqçılıq aşkarlanması kimi bir çox sahədə biznes qərarlarını optimallaşdırmaq üçün istifadə oluna bilər.

## Supervised ve Unsupervised arasinda ferq?

Supervised (nəzarətli) və unsupervised (nəzarətsiz) öyrənmə, maşın öyrənməsi sahəsində istifadə olunan iki əsas yanaşmadır. Onlar arasında bir neçə mühüm fərq var:

### Supervised Learning (Nəzarətli Öyrənmə)

>**Tərif:** Supervised learning – modelin giriş (input) və çıxış (output) cütləri ilə təlim olunduğu, nümunələrdən qanunauyğunluqları öyrənərək yeni və görünməmiş verilənlər üçün doğru proqnozlar verməsini təmin edən maşın öyrənməsi metodudur. Bu yanaşma təsnifat (classification) və reqressiya (regression) problemlərinin həlli üçün geniş istifadə olunur. 🎯📊

1. **Təyin Edilmiş Məlumat**: Supervised learning-də model, giriş verilənlərinin (x) və onların müvafiq çıxış etiketlərinin (y) olduğu bir dataset ilə öyrədilir. Bu etiketlər, modelin doğru proqnozlar etməsinə kömək edir.

2. **Məqsəd**: Modelin məqsədi, verilən giriş verilənlərinə əsaslanaraq doğru çıxış etiketlərini proqnozlaşdırmaqdır. Məsələn, xəstələrin tibbi məlumatlarından istifadə edərək xəstəliyin olub-olmadığını proqnozlaşdırmaq.

3. **Nümunələr**: Məsələn, sınıflandırma (classification) və regresiya (regression) tapşırıqları.

4. **Tətbiq Sahələri**: E-poçt spam filtrasiya, müştəri davranışı proqnozu, maliyyə analizləri, tibbi diaqnozlar.

5. **Örnek Algoritmlər**: Logistic Regression, Decision Trees, Random Forests, Support Vector Machines (SVM), Neural Networks.

### Unsupervised Learning (Nəzarətsiz Öyrənmə)

>**Tərif:** Unsupervised learning – modelin yalnız giriş (input) verilənləri ilə təlim olunduğu, heç bir etiketlənmiş çıxış (output) məlumatı olmadan, verilənlərdəki gizli nümunələri, strukturları və əlaqələri aşkar etməyə yönələn maşın öyrənməsi metodudur. Bu yanaşma əsasən klasterləşdirmə (clustering) və ölçü azaldılması (dimensionality reduction) kimi problemlərin həllində istifadə olunur. 🔍📊

1. **Təyin Edilməmiş Məlumat**: Unsupervised learning-də model, yalnız giriş verilənləri (x) olan bir dataset ilə öyrədilir, lakin çıxış etiketləri yoxdur. Yəni, model verilənlər arasında əlaqələri tapmağa çalışır.

2. **Məqsəd**: Modelin məqsədi, verilənlər arasında gizli strukturları və ya klasterləri aşkar etməkdir. Məsələn, müştəri seqmentasiyası üçün müştəri məlumatlarını qruplaşdırmaq.

3. **Nümunələr**: Məsələn, klasterləşdirmə (clustering) və əsas komponentlərin analizi (PCA).

4. **Tətbiq Sahələri**: Müştəri seqmentasiyası, məlumatların analizi, anomaliya aşkar etmə, müstəqil öyrənmə.

5. **Örnek Algoritmlər**: K-Means, Hierarchical Clustering, DBSCAN, Principal Component Analysis (PCA), t-SNE.

### Qısaca Fərqlər

- **Etiketlər**: Supervised learning-də etiketli verilənlər var, unsupervised learning-də isə yoxdur.
- **Məqsəd**: Supervised learning-də modelin doğru proqnozlar etməsi üçün öyrədilməsi, unsupervised learning-də isə verilənlərdəki strukturları aşkar etmək.
- **Nümunələr**: Supervised learning, sınıflandırma və regresiya, unsupervised learning, klasterləşdirmə və məlumat analizi.

Bu fərqləri nəzərə alaraq, verilən tapşırığın təbiətinə uyğun öyrənmə metodunu seçmək vacibdir.

## **Classification (Sinifləndirmə)** və **Clustering (Klasterləşdirmə)**  arasında fərq

**Classification (Sinifləndirmə)** və **Clustering (Klasterləşdirmə)** maşın öyrənmədə fərqli yanaşmalardır və əsas fərqləri aşağıdakı kimidir:

### **1. Nəzarət Tipləri:**
- **Classification (Sinifləndirmə):**
  - **Nəzarətli öyrənmə** metodudur.
  - Verilənlərdə girişlərə uyğun olaraq çıxış (etiket) məlumdur.
  - Məqsəd: Girişlərə əsasən düzgün sinif etiketlərini proqnozlaşdırmaq.
  - Məsələn: E-poçtların "Spam" və ya "Spam deyil" olaraq sinifləndirilməsi.

- **Clustering (Klasterləşdirmə):**
  - **Nəzarətsiz öyrənmə** metodudur.
  - Verilənlərdə çıxış etiketləri məlum deyil.
  - Məqsəd: Oxşar xüsusiyyətlərə malik verilənləri qruplara (klasterlərə) ayırmaq.
  - Məsələn: Müştəri seqmentlərinin qruplaşdırılması.

---

### **2. Çıxış Formatı:**
- **Classification:**
  - Model hər bir nümunə üçün müəyyən bir sinif etiketi verir.
  - Çıxış diskret və öncədən təyin edilmiş kateqoriyalardır.

- **Clustering:**
  - Model nümunələri qruplar şəklində təşkil edir.
  - Çıxış qeyri-müəyyən qruplardır və klasterlərin sayı və strukturu əvvəlcədən məlum olmaya bilər.

---

### **3. Tətbiq Sahələri:**
- **Classification:**
  - Tibbi diaqnostika (xəstəliyin proqnozlaşdırılması).
  - Müştəri davranışlarının sinifləndirilməsi.
  - Kredit riskinin qiymətləndirilməsi.

- **Clustering:**
  - Marketinqdə müştəri seqmentasiyası.
  - Anomaliya aşkarlanması.
  - Təbiətdə canlıların taksonomik qruplarının tapılması.

---

### **4. Alqoritmlər:**
- **Classification:**
  - Logistic Regression
  - Decision Trees
  - Random Forest
  - Support Vector Machines (SVM)
  - Neural Networks

- **Clustering:**
  - K-Means
  - DBSCAN
  - Hierarchical Clustering
  - Gaussian Mixture Models (GMM)

---

### **Sadə Nümunə ilə Fərq:**

- **Classification:**
  - Məsələn: Bir şəkil pişikdir ya deyil, bunu müəyyənləşdirmək.
  
- **Clustering:**
  - Məsələn: Şəkilləri növlərinə görə (pişiklər, itlər, quşlar) qruplaşdırmaq.

Classification-də nəticə öncədən məlumdur, Clustering-də isə nəticəni müəyyənləşdirmək öyrənmənin məqsədidir.

## **Reinforcement Learning (Möhkəmləndirici Öyrənmə)**

>**Tərif:** **Reinforcement Learning (Möhkəmləndirici Öyrənmə)**, bir **agentin** **mühitlə qarşılıqlı əlaqə** yaradaraq hərəkətlərinə görə aldığı mükafat və ya cəzaları istifadə edərək optimal strategiyanı öyrəndiyi maşın öyrənmə yanaşmasıdır.Bu öyrənmə metodu, bir agentin müəyyən bir mühitdə qərarlar qəbul edərək **maksimum mükafat** əldə etməyi öyrənməsi üzərində qurulub. **Məqsəd agentin ən yüksək ümumi mükafatı əldə edəcək strategiyanı (policy) öyrənməsidir.**


**Reinforcement Learning (RL)**, maşın öyrənmənin bir alt sahəsidir və qərar qəbul etmə problemlərini həll etmək üçün istifadə olunur. Bu öyrənmə yanaşmasında, agent (öyrənən sistem) bir mühitlə qarşılıqlı əlaqə yaradır və müəyyən hərəkətlərə görə mükafat və ya cəza alaraq öyrənir. Məqsəd agentin ən yüksək ümumi mükafatı əldə edəcək strategiyanı (policy) öyrənməsidir.

---

### **Reinforcement Learning-in Əsas Prinsipləri**

1. **Agent və Mühit:**
   - **Agent:** Öyrənən və qərar qəbul edən sistemdir.
   - **Mühit:** Agentin fəaliyyət göstərdiyi yer və agentin hərəkətlərinə cavab verən sistemdir.

2. **Hərəkət (Action):**
   - Agentin mühitdə edə biləcəyi hərəkətlərdir. Hərəkətlərin dəsti \( A \) ilə təmsil olunur.

3. **Vəziyyət (State):**
   - Agentin mühitdə olduğu anlıq vəziyyətidir. Vəziyyətlər \( S \) ilə ifadə olunur.

4. **Mükafat (Reward):**
   - Agentin bir vəziyyətdən digərinə keçərkən aldığı müsbət və ya mənfi dəyərdir. Məqsəd bu mükafatları maksimal dərəcədə artırmaqdır.

5. **Strategiya (Policy):**
   - Agentin hansı vəziyyətdə hansı hərəkəti etməli olduğunu müəyyənləşdirən qaydalar toplusudur.

6. **Q-Dəyəri (Q-Value):**
   - Vəziyyət-hərəkət cütlərinin gələcək mükafatların gözlənilən dəyərini göstərən funksiyadır.

---

### **Reinforcement Learning-in Əsas Fəaliyyət Mexanizmi**

Reinforcement Learning dövrü belə işləyir:
1. Agent müəyyən bir vəziyyətdə olur (\( S_t \)).
2. Agent mühitdə bir hərəkət (\( A_t \)) edir.
3. Mühit bu hərəkətə cavab olaraq bir mükafat (\( R_t \)) verir və agenti yeni vəziyyətə (\( S_{t+1} \)) keçirir.
4. Agent aldığı mükafata əsaslanaraq strategiyasını yeniləyir.
5. Bu proses dövr edir və agent optimal strategiyanı öyrənir.

---

### **Reinforcement Learning-in Növləri**

1. **Model-Free RL:**
   - Mühitin necə işlədiyini bilmədən agentin öyrənməsinə əsaslanır.
   - Məsələn:
     - **Q-Learning**
     - **SARSA (State-Action-Reward-State-Action)**

2. **Model-Based RL:**
   - Agent mühitin modelini qurmağa çalışır və bu modeli istifadə edərək optimal strategiyanı öyrənir.

3. **On-Policy və Off-Policy RL:**
   - **On-Policy:** Agent öz aktuall strategiyasına əsaslanaraq öyrənir (məsələn, SARSA).
   - **Off-Policy:** Agent başqa bir strategiyadan istifadə edərək öyrənir (məsələn, Q-Learning).

---

### **Əsas Alqoritmlər**

1. **Q-Learning:**
   - Ən çox istifadə edilən model-free RL alqoritmlərindən biridir.
   - Q-funksiyası ilə vəziyyət-hərəkət cütlərinin dəyərini öyrənir:
     \[
     Q(s, a) = Q(s, a) + \alpha \left( R + \gamma \max_{a'} Q(s', a') - Q(s, a) \right)
     \]
     Burada:
     - \( \alpha \): Öyrənmə sürəti (learning rate).
     - \( \gamma \): Gələcək mükafatların diskont faktorudur.

2. **Deep Q-Learning (DQN):**
   - Q-Learning-in dərin neyron şəbəkələrlə birləşdirilmiş formasıdır.
   - Çox böyük vəziyyət fəzalarında tətbiq olunur.

3. **Policy Gradient:**
   - Birbaşa strategiyanı optimallaşdırmağa əsaslanır.
   - Məsələn:
     - REINFORCE alqoritmi
     - Actor-Critic modeli

4. **Actor-Critic:**
   - Strategiya və dəyər funksiyalarını birlikdə öyrənən hibrid modeldir.
   - Actor strategiyanı öyrənir, Critic isə mükafatın nə qədər yaxşı olduğunu qiymətləndirir.

---

### **Reinforcement Learning Tətbiqləri**

1. **Oyun Nəzarəti:**
   - Şahmat, Go, Atari oyunları kimi mürəkkəb strategiyalı oyunlarda istifadə olunur.
   - Məsələn, AlphaGo və AlphaZero.

2. **Robot Texnologiyaları:**
   - Robotların hərəkətini planlaşdırmaq və mühitdə düzgün hərəkət etməsini təmin etmək.

3. **Avtonom Avtomobillər:**
   - Nəqliyyatda sükan idarəsini və təhlükəsiz hərəkəti öyrənmək.

4. **Maliyyə və Ticarət:**
   - Portfel idarəsi və ticarət strategiyalarının optimallaşdırılması.

5. **Tövsiyə Sistemləri:**
   - Müştərilərə fərdiləşdirilmiş tövsiyələrin verilməsi.

6. **Tibbi Məlumatların Analizi:**
   - Dərmanların dozalarını tənzimləmək və müalicə strategiyalarını optimallaşdırmaq.

---

### **Reinforcement Learning-in Üstünlükləri**
1. **Dinamik Mühitlərdə Tətbiq:**
   - Gələcəyə uyğunlaşaraq mühitdə dəyişikliklərə cavab verir.
2. **Uzunmüddətli Mükafatın Maksimallaşdırılması:**
   - Yalnız ani mükafatlara deyil, həm də uzunmüddətli nəticələrə fokuslanır.
3. **Geniş Tətbiq Sahəsi:**
   - Robot texnologiyaları, oyunlar, tibbi sahə və maliyyədə istifadə olunur.

---

### **Reinforcement Learning-in Məhdudiyyətləri**
1. **Çətin və Yavaş Öyrənmə:**
   - Mürəkkəb mühitlərdə agentin öyrənməsi uzun müddət çəkə bilər.
2. **Mükafat Strukturunun Dizaynı:**
   - Əgər mükafat funksiyası düzgün qurulmazsa, agent optimal strategiyanı öyrənməyə bilər.
3. **Hesablama Məsrəfləri:**
   - Çoxsaylı vəziyyət və hərəkət məkanlarında RL alqoritmləri çox vaxt və resurs tələb edir.

---

### **Nəticə**
Reinforcement Learning, mürəkkəb qərar qəbul etmə problemlərində təsirli bir öyrənmə yanaşmasıdır. Agentin mühitlə qarşılıqlı əlaqə yaratdığı bu öyrənmə üsulu həm nəzəri, həm də praktiki baxımdan böyük əhəmiyyət daşıyır. Lakin çətin mühitlərdə performansı artırmaq üçün əlavə resurslar və optimallaşdırma metodları tələb oluna bilər.

## Orta empirik risk(Empirical Risk)

**Orta empirik risk** (ing. "Empirical Risk") maşın öyrənməsində istifadə edilən anlayışdır və bir modelin təlim məlumatları üzərindəki orta səhvini ölçmək üçün istifadə olunur. Bu anlayış əsasən **risk minimallaşdırma** konsepsiyası ilə əlaqədardır.

---

### **Tərif:**
Orta empirik risk bir modelin proqnozlarının real etiketlərdən nə qədər uzaq olduğunu ölçən orta dəyərdir. Riyazi olaraq bu belə ifadə edilir:

$$
R_{\text{emp}}(f) = \frac{1}{n} \sum_{i=1}^{n} L(y_i, f(x_i))
$$

### **Burada:**
- $n$: Təlim nümunələrinin sayı (müşahidələr).
- $x_i$: Giriş məlumatı (xüsusiyyətlər).
- $y_i$: Real nəticə (hədəf dəyəri).
- $f(x_i)$: Modelin proqnozlaşdırdığı dəyər.
- $L(y_i, f(x_i))$: Səhv funksiyası (itki funksiyası), məsələn, Mean Squared Error (MSE), Mean Absolute Error (MAE), və ya Cross-Entropy Loss.

---

### **Sadə İzah:**
Orta empirik risk modelin verilən təlim məlumatları üzərində **orta səhvini** göstərir. Məsələn:
- Əgər bir xətti reqressiya modeliniz varsa və **MSE (Mean Squared Error)** istifadə edirsinizsə:
  $$
  R_{\text{emp}}(f) = \frac{1}{n} \sum_{i=1}^{n} (y_i - f(x_i))^2
  $$
  Bu, modelin nə qədər dəqiq olduğunu ölçür.

---

### **Empirik Risk ilə Əsl Risk Arasındakı Fərq:**
1. **Empirik Risk:** Modelin təlim verilənləri üzərində ölçülən səhvidir.
   - Təlim verilənlərinə əsaslanır.
   - Həmişə müşahidələrdən hesablanır.

2. **Əsl Risk:** Modelin bütün mümkün verilənlər üzərindəki səhvini göstərir.
   - Həqiqi populyasiyaya əsaslanır.
   - Əsl riski bilmək mümkün deyil, çünki bütün verilənlər məlum deyil.

---

### **Nəticə:**
Orta empirik risk modelin təlim verilənlərindəki performansını ölçmək üçün istifadə olunur. Bu riskin minimallaşdırılması təlim prosesinin əsas məqsədlərindən biridir, lakin **overfitting** riskindən qaçmaq üçün diqqətli olmaq lazımdır.
