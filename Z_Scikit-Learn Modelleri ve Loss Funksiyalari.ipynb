### **Scikit-Learn Modelləri və İtki Funksiyaları**

| Model Kateqoriyası | Model Adı (Scikit-learn) | İtki Funksiyası (Loss Function) | Riyazi Düstur / Meyar |
| :--- | :--- | :--- | :--- |
| **Xətti Reqressiya** | `LinearRegression` | Mean Squared Error (MSE) | $$L = \sum (y_i - \hat{y}_i)^2$$ |
| **Xətti Reqressiya** | `Ridge` | MSE + L2 Regularization | $$L = \sum (y_i - \hat{y}_i)^2 + \alpha \sum w_j^2$$ |
| **Xətti Reqressiya** | `Lasso` | MSE + L1 Regularization | $$L = \sum (y_i - \hat{y}_i)^2 + \alpha \sum \|w_j\|$$ |
| **Xətti Reqressiya** | `ElasticNet` | MSE + L1 + L2 | $$L = \text{MSE} + \alpha\rho\|w\|_1 + \frac{\alpha(1-\rho)}{2}\|w\|_2^2$$ |
| **Təsnifat** | `LogisticRegression` | Log Loss (Cross-Entropy) | $$L = -(y \log(p) + (1-y) \log(1-p))$$ |
| **Təsnifat** | `SVC` | Hinge Loss | $$L = \max(0, 1 - y \cdot f(x))$$ |
| **Təsnifat** | `DecisionTreeClassifier` | Gini Impurity və ya Entropy | $$G = 1 - \sum p_i^2$$ |
| **Təsnifat** | `RandomForestClassifier`| Gini / Entropy (Aggregated) | $$\text{Ağacların ortalaması}$$ |
| **Təsnifat** | `KNeighborsClassifier` | Minkovski Məsafəsi (və ya Euclidean) | $$d(x, y) = (\sum \|x_i - y_i\|^p)^{1/p}$$ |
| **Təsnifat** | `GaussianNB` | Log-Likelihood (Qaus paylanması) | $$P(x_i \vert y) = \frac{1}{\sqrt{2\pi\sigma_y^2}} e^{-\frac{(x_i-\mu_y)^2}{2\sigma_y^2}}$$ |
| **Stokastik Gradient**| `SGDClassifier` | Hinge (default) və ya Log Loss | `loss` parametrinə bağlıdır |
| **Reqressiya** | `SVR` | $\epsilon$-insensitive Loss | $$L = \max(0, \|y - \hat{y}\| - \epsilon)$$ |
| **Boosting** | `AdaBoostClassifier` | Exponential Loss | $$L = \exp(-y \cdot f(x))$$ |
| **Boosting** | `GradientBoosting` | Deviance (Log Loss) / MSE | $$-\sum y \log(p) + \dots$$ |
| **Boosting** | `XGBoost` (XGBClassifier) | Log Loss + Regularization | $$L = \sum L(y, p) + \gamma T + \frac{1}{2}\lambda\|w\|^2$$ |
| **Neyron Şəbəkə** | `MLPClassifier` | Cross-Entropy | $$L = -\sum y \log(p)$$ |
| **Neyron Şəbəkə** | `MLPRegressor` | Mean Squared Error (MSE) | $$L = \frac{1}{2} \|y - \hat{y}\|^2$$ |

<br>

#### **Cədvəl üçün qısa izahatlar:**

* **KNN (K-Neighbors):** **Əslində klassik bir "itki funksiyası" ilə optimallaşdırılmır, qonşular arasındakı məsafəyə əsaslanır.**
* **GaussianNB (Naive Bayes):** **İtki funksiyasını minimallaşdırmır, ehtimalın maksimallaşdırılması (Maximum Likelihood Estimation) prinsipi ilə işləyir.**
* **Decision Tree:** **Budama zamanı xətanı azaldır, lakin hər bölmədə Gini və ya Entropy (informasiya qazancı) meyarlarından istifadə edir.**
* **SVC (Support Vector Classification):** **Əgər `probability=True` seçsəniz, model daxildə Plati skalanmasından istifadə edərək Hinge loss nəticələrini ehtimala çevirir.**
