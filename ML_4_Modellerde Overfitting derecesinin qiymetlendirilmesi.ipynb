## Emprik Risk Nədir

> **Tərif:** **Emprik risk**, modelin **təlim datası** üzərindəki performansını ölçən, **faktiki** dəyərlərlə modelin **proqnozları** arasındakı fərqlərin **orta xəta funksiyası** ilə ifadəsidir.

Emprik risk, statistik öyrənmə nəzəriyyəsində modelin "təlim datası" üzərindəki performansını qiymətləndirmək üçün istifadə olunur. Təlim datasında modelin xəta funksiyasını minimallaşdırmaq, yəni modellə faktiki nəticələr arasındakı fərqi azaltmaq əsas məqsəddir. 

### **Emprik Risk Nədir?**
Emprik risk təlim datasında orta xəta kimi ifadə edilə bilər. Bu, hər bir nümunə üçün xəta funksiyası (loss function) hesablanaraq, onların ortalamasını almaqla müəyyən edilir. Daha riyazi desək:

$$
R_{emp} = \frac{1}{n} \sum_{i=1}^n L(y_i, \hat{y}_i)
$$

Burada:
- $n$ — təlim datasındakı nümunələrin sayı,
- $y_i$ — faktiki dəyər,
- $\hat{y}_i$ — model tərəfindən təxmin edilən dəyər,
- $L(y_i, \hat{y}_i)$ — bir nümunə üçün xəta funksiyası.

### **Loss Function (Xəta Funksiyası)**
Xəta funksiyası, modelin verdiyi nəticə ilə faktiki nəticə arasındakı fərqi ölçən bir funksiyadır. Müxtəlif məqsədlərə görə fərqli xəta funksiyalarından istifadə olunur:

1. **Regression üçün:**
   - Mean Squared Error (MSE): 
     $$
     L(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
     $$
   - Mean Absolute Error (MAE): 
     $$
     L(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
     $$

2. **Classification üçün:**
   - Cross-Entropy Loss:
     $$
     L(y, \hat{y}) = - \frac{1}{n} \sum_{i=1}^n \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
     $$

3. **Hinge Loss** (Support Vector Machines üçün):
   $$
   L(y, \hat{y}) = \max(0, 1 - y_i \cdot \hat{y}_i)
   $$

### **Emprik Riskin Əhəmiyyəti**
Emprik risk minimallaşdırılaraq model öyrədilir. Yəni, təlim mərhələsində məqsəd, model parametrlərini $R_{emp}$-i minimallaşdıracaq şəkildə tənzimləməkdir. Lakin, emprik riskin çox kiçik olması modelin overfitting etməsi riskini artırır. Buna görə də, **ümumi risk (generalization risk)** minimal edilməlidir.

### **Ümumi Risk (True Risk) və Emprik Risk Arasındakı Fərq**
- **Emprik Risk:** Təlim datasında modellə faktiki dəyərlər arasındakı fərqə əsaslanır.
- **Ümumi Risk:** Modelin real dünyadakı (yəni test datasında və ya yeni məlumatlarda) performansını əks etdirir.

Ümumiyyətlə, model qurarkən məqsəd emprik risklə ümumi risk arasında balans yaratmaqdır. Bu, müxtəlif texnikalarla (məsələn, regularization, cross-validation) həyata keçirilir.

## Hold-out

>**Tərif:** **Hold-out** metodu, verilənlər toplusunun **təlim** və **test** dəstlərinə bölünərək modelin performansının qiymətləndirildiyi sadə və **təsadüfi validasiya** texnikasıdır.


**Hold-out metodu**, maşın öyrənməsi və statistik modelləşdirmədə **modelin performansını qiymətləndirmək** üçün istifadə edilən ən sadə üsullardan biridir. Bu metod, mövcud verilənlər toplusunu təlim və test (və bəzən də təsdiqləmə) dəstlərinə bölərək işləyir.

---

### **Hold-out metodunun addımları:**

1. **Verilənləri bölmək:**
   - Verilənlər toplusu əvvəlcə **təlim dəsti (training set)** və **test dəsti (test set)** olaraq bölünür. Tipik olaraq:
     - Təlim dəsti (70-80%) — model öyrədilməsi üçün istifadə olunur.
     - Test dəsti (20-30%) — modelin öyrədilməmiş məlumatlar üzərində necə performans göstərdiyini qiymətləndirmək üçün istifadə olunur.

   Əgər təsdiqləmə dəsti də əlavə olunarsa, belə bir bölünmə ola bilər:
   - Təlim dəsti: 60%
   - Təsdiqləmə dəsti (validation set): 20% (hyperparameter tuning üçün)
   - Test dəsti: 20%

2. **Modeli təlimdən keçirmək:**
   - Təlim dəstindən istifadə edərək model qurulur (parametrlər təlim zamanı öyrədilir).

3. **Modeli qiymətləndirmək:**
   - Test dəsti üzərində model tətbiq edilir və performans ölçülür. **Metriklər** olaraq, məsələn, aşağıdakılar istifadə edilə bilər:
     - Klassifikasiya üçün: dəqiqlik (accuracy), F1-score, precision, recall.
     - Regressiya üçün: RMSE, MAE, $R^2$.

---

### **Hold-out metodunun üstünlükləri:**

1. **Sadəlik:** 
   - Verilənlərin bir dəfə bölünməsi və bu bölünməyə əsaslanan qiymətləndirmə çox sadə və sürətlidir.
   
2. **Hesablama cəhətdən ucuzdur:**
   - Verilənlər yalnız bir dəfə bölündüyü üçün daha az hesablama resursları tələb edir.

---

### **Hold-out metodunun çatışmazlıqları:**

1. **Verilənlərin bölünməsinin təsadüfiliyi:**
   - Verilənlərin təsadüfi bölünməsi nəticəsində test və təlim dəstləri arasında qeyri-bərabərlik yarana bilər. Bu, modelin həqiqi performansını düzgün qiymətləndirməyə bilər.

2. **Məlumatın istifadəsinin azlığı:**
   - Test dəstinə ayrılmış məlumatlar təlim üçün istifadə edilmir, bu da modelin öyrənmə potensialını azalda bilər.

3. **Test nəticələrinin dəyişkənliyi:**
   - Verilənlər müxtəlif yollarla bölündükdə, nəticələr fərqli ola bilər (bu problemə "data variance problem" deyilir).

---

### **Hold-out metodunun alternativləri:**

- **K-Fold Cross-Validation:** Verilənlər dəfələrlə müxtəlif hissələrə bölünərək həm təlim, həm də test üçün istifadə olunur. Bu metod daha sabit və etibarlı performans qiymətləndirməsi verir.
- **Stratified Sampling:** Xüsusilə qeyri-balanslı verilənlərdə (məsələn, bir sinif daha çox təmsil olunursa), hər bir sinfi proporsional şəkildə təlim və test dəstlərinə bölməklə nəticələrin daha düzgün olması təmin edilir.

---

### **Hold-out metodunun istifadə sahələri:**
- Kiçik həcmli məlumatlarla işləyərkən sadə bir qiymətləndirmə metodu olaraq istifadə olunur.
- Sürətli model prototipləşdirmədə performansın ilkin qiymətləndirilməsi üçün istifadə olunur.

Əgər böyük həcmli verilənlərdən istifadə edirsinizsə, hold-out üsulu kifayət edə bilər. Ancaq verilənlər azdırsa və ya nəticələrdəki dəyişkənlik əhəmiyyətlidirsə, **cross-validation** üstünlük verilir.

## Leave-One-Out Cross-Validation

>**Tərif:** **Leave-One-Out Cross-Validation (LOOCV)**, verilənlər dəstindəki **hər bir nümunənin** ardıcıl olaraq **test dəsti** kimi istifadə edildiyi və **qalan nümunələrin** modelin **təlimi** üçün ayrıldığı bir qiymətləndirmə metodudur.


**Leave-One-Out Cross-Validation (LOOCV)**, **k-fold cross-validation** metodunun xüsusi bir halıdır. Burada verilənlər dəstindəki hər bir nümunə test dəsti kimi istifadə edilir, qalan bütün nümunələr isə təlim üçün istifadə olunur.

---

### **LOOCV metodu necə işləyir?**

1. Verilənlər dəstində $ n $ nümunə olduğunu düşünək.
2. Hər iterasiyada:
   - Təlim dəsti: $n-1 $ nümunədən ibarət olur.
   - Test dəsti: 1 nümunədən ibarət olur.
3. Proses$ n $-dəfə təkrarlanır və hər dəfə fərqli bir nümunə test üçün seçilir.
4. Sonda bütün iterasiyalar üzrə modelin performansı orta hesabla qiymətləndirilir.

---

### **LOOCV üçün əsas tənlik:**
Modelin ümumi xətası (ortalama xəta funksiyası) belə hesablanır:

$$
R_{LOOCV} = \frac{1}{n} \sum_{i=1}^n L(y_i, \hat{y}_i^{(-i)})
$$

Burada:
- $n $: verilənlər dəstindəki nümunələrin sayı,
- $ y_i$: faktiki dəyər,
- $ \hat{y}_i^{(-i)} $): $ i $-ci nümunə çıxarıldıqdan sonra qurulan modelin verdiyi proqnoz,
- $ L $: xəta funksiyası (məsələn, MSE, MAE və s.).

---

### **LOOCV metodunun üstünlükləri:**

1. **Maksimum məlumat istifadəsi:** Hər iterasiyada verilənlərin böyük hissəsi təlim üçün istifadə edilir ($ n-1 $ nümunə).
2. **Dəqiq qiymətləndirmə:** Verilənlər dəsti az olduqda daha sabit və dəqiq qiymətləndirmə verir.
3. **Overfitting yoxlanışı:** Hər bir nümunənin ayrı-ayrılıqda test edilməsi overfitting riskini azaldır.

---

### **LOOCV metodunun çatışmazlıqları:**

1. **Hesablama xərci yüksəkdir:** Xüsusilə verilənlər çox olduqda ($ n $-dəfələrlə model qurmaq tələb olunur), LOOCV çox vaxt və resurs tələb edə bilər.
2. **Həssaslıq:** Verilənlərdəki fərdi nümunələr performans qiymətləndirməsinə böyük təsir göstərə bilər, xüsusilə outlier-lər mövcuddursa.
3. **Təsadüfi dəyişkənlik:** Hər bir test dəsti yalnız bir nümunədən ibarət olduğundan nəticələr təsadüfi təsirlərə həssas ola bilər.

---

### **LOOCV nə zaman istifadə olunur?**

- Verilənlər dəsti kiçikdirsə və hər bir nümunənin qiymətləndirilməsində maksimum dəqiqlik lazımdırsa.
- Hesablama gücünün yüksək olduğu və ya sadə modellərin istifadə edildiyi hallarda.

---

### **LOOCV-nin alternativləri:**

- **K-Fold Cross-Validation:** Hesablama baxımından daha səmərəlidir, verilənləri $ k $ qrupa bölərək test edir.
- **Stratified K-Fold Cross-Validation:** Xüsusilə qeyri-balanslı verilənlərdə daha balanslı test-təlim dəsti yaradır.

LOOCV əsasən kiçik verilənlərdə yüksək dəqiqlik tələb olunduqda istifadə olunur, lakin böyük verilənlərdə çox vaxt K-Fold daha praktiki hesab edilir.

## K-Fold Cross-Validation

>**Tərif:** K-Fold Cross-Validation**, verilənlər dəstini **k** bərabər hissəyə bölərək hər hissənin ardıcıl olaraq **test** dəsti kimi, qalan hissələrin isə **təlim** üçün istifadə edildiyi model qiymətləndirmə metodudur.

**K-Fold Cross-Validation** üçün ümumi xəta funksiyasının hesablanması aşağıdakı şəkildə verilə bilər:

$$
R_{K-Fold} = \frac{1}{k} \sum_{i=1}^{k} L(y_i, \hat{y}_i^{(i)})
$$

Burada:
- $ k $: fold sayısı (məsələn, $ k=5 $ və ya $ k=10 $),
- $ y_i $: faktiki dəyər,
- $\hat{y}_i^{(i)} $: $ i $ -ci fold-da təlim üçün istifadə olunan modelin test dəstindəki $ i $-ci nümunə üçün proqnozlaşdırılmış dəyəri,
- $ L $: xəta funksiyası (məsələn, MSE, MAE, dəqiqlik və s.).


**K-Fold Cross-Validation**, maşın öyrənməsində modelin performansını qiymətləndirmək üçün geniş istifadə olunan bir texnikadır. Bu metod verilənlər dəstini **k** sayda bərabər hissəyə (**folds**) bölərək modelin müxtəlif hissələrdəki performansını ölçür.

---

### **K-Fold Cross-Validation necə işləyir?**

1. **Verilənləri bölmək:**
   - Verilənlər dəsti təsadüfi şəkildə **k** bərabər hissəyə bölünür.

2. **Təkrarlama:**
   - Hər iterasiyada:
     - Bir fold test dəsti kimi istifadə olunur.
     - Qalan ; k-1 ; fold isə təlim dəsti kimi istifadə olunur.
   - Bu proses **k** dəfə təkrarlanır və hər dəfə fərqli bir fold test dəsti kimi seçilir.

3. **Performansın qiymətləndirilməsi:**
   - Hər iterasiya üçün bir performans ölçüsü (məsələn, dəqiqlik, MSE) hesablanır.
   - Sonda bütün iterasiyaların performans nəticələrinin ortalaması alınır.

---

### **K-Fold Cross-Validation üçün əsas addımlar:**
Misal üçün, ; k=5 ; (5-fold cross-validation):
- Verilənlər 5 hissəyə bölünür.
- 1-ci fold test, qalan 4 fold təlim üçün istifadə edilir.
- 2-ci fold test, qalan 4 fold təlim üçün istifadə edilir.
- Bu proses 5 dəfə təkrarlanır.

---

### **K-Fold Cross-Validation metodunun üstünlükləri:**

1. **Məlumatın effektiv istifadəsi:** Bütün verilənlər həm təlim, həm də test üçün istifadə edilir.
2. **Performans qiymətləndirməsində sabitlik:** Bütün verilənlər dəfələrlə istifadə edildiyi üçün nəticələr daha sabit və dəqiq olur.
3. **Qeyri-təsadüfi nəticələr:** Verilənlərin müxtəlif hissələrinin test kimi istifadə olunması nəticələrin daha ümumi olmasını təmin edir.

---

### **K-Fold Cross-Validation metodunun çatışmazlıqları:**

1. **Hesablama baxımından bahalıdır:** ; k ;-dəfə təlim və test prosesi aparıldığı üçün böyük verilənlərdə hesablama resursları çox tələb olunur.
2. **K-ın seçimi:** ;k ;-ın düzgün seçilməsi nəticələrin keyfiyyətinə təsir göstərə bilər. (Tipik olaraq $ k=5 $ və ya ; $k=10$ ; seçilir.)

---

### **K-Fold Cross-Validation nə zaman istifadə olunur?**

- Verilənlər kifayət qədər böyükdürsə və modelin performansını qiymətləndirmək üçün dəqiq nəticələr tələb olunursa.
- Məlumat balanslıdırsa və hər fold-da məlumatın təmsilçiliyini qorumaq mümkündür.

---

### **K-Fold Cross-Validation metodunun alternativləri:**

- **Stratified K-Fold Cross-Validation:** Qeyri-balanslı verilənlərdə siniflərin proporsiyasını qoruyaraq fold-lar yaratmaq üçün istifadə edilir.
- **Leave-One-Out Cross-Validation (LOOCV):** Hər dəfə yalnız bir nümunə test, qalan bütün nümunələr təlim üçün istifadə edilir.
- **Hold-Out Validation:** Verilənlərin bir hissəsi təlim, qalanı test üçün istifadə olunur (daha sadə, lakin az etibarlıdır).

K-Fold Cross-Validation praktik olaraq maşın öyrənməsi layihələrinin çoxunda istifadə olunur, çünki balanslı və ümumiləşdirilmiş performans ölçüsü verir.

## K-Fold Cross-Validation-a  hansı data verilir və onun nəticəsi hansı nəticə ilə müqayisə olunur?

Əvvəlcə verilənləri (**x** və **y**) train və testə böldükdən sonra**, **K-Fold Cross-Validation** tətbiq etmək üçün **x_train** və **y_train** verilənlərini verirsiniz.

### Addımlar:
1. **Əvvəlcə** **x** və **y** verilənlərini **train** və **test** dəstlərinə bölürsünüz, məsələn, `train_test_split` ilə:
   - **x_train**, **y_train** (təlim dəstləri),
   - **x_test**, **y_test** (test dəstləri).

2. Sonra, **K-Fold Cross-Validation** tətbiq etmək istəyirsinizsə, **x_train** və **y_train** verilənlərini **KFold** və ya **StratifiedKFold** kimi metodlarla bölüb, **train** dəstində təlim edirsiniz.

   **K-Fold Cross-Validation** üçün:

   - **x_train** və **y_train** verilənlərini **k** fold-a bölürsünüz.
   - Hər bir fold-da **k-1** fold təlim üçün **x_train** və **y_train** kimi istifadə edilir, 1 fold isə test üçün **x_test** və **y_test** kimi istifadə olunur.
   
---

### Nəticə:
Əgər **x** və **y** verilənlərini əvvəlcə **train** və **test** dəstlərinə ayırmısınızsa, **K-Fold Cross-Validation** üçün yalnız **x_train** və **y_train** verirsiniz. **x_test** və **y_test** isə yalnız son qiymətləndirmə mərhələsində istifadə olunur.

**Cross-validation**-dan əldə edilən nəticələri **modelin ümumi performansını** qiymətləndirmək üçün istifadə edirsiniz. Bunun üçün, cross-validation-dan əldə olunan ortalama qiymətləndirməni və **test dəsti (x_test, y_test)** ilə modelin performansını müqayisə edirsiniz.

### Müqayisə addımları:

1. **Cross-Validation Nəticələri:**
   - **K-Fold Cross-Validation**-dan sonra hər fold üçün əldə olunan qiymətlər (məsələn, doğruluq, xəta, MSE və s.) toplanır.
   - Sonra bu nəticələrin ortalaması alınır. Bu ortalama, modelin təlim verilənlərindəki **ümumi performansını** göstərir.

2. **Test Dəsti Nəticələri:**
   - **Test dəsti (x_test, y_test)** ilə modelin son qiymətləndirilməsi edilir. Bu nəticə modelin **gerçək dünya verilənlərində** necə işlədiyini göstərir.

### Müqayisə üçün istifadə olunan əsas ölçülər:

- **Doğruluq (Accuracy):** Modelin düzgün təxmin etdiyi nisbət.
- **MSE (Mean Squared Error):** Əgər regressiya modelidirsə, orta kvadrat xətanı istifadə edə bilərsiniz.
- **F1 Score, Precision, Recall:** Sinifləndirmə problemlərində.
- **AUC-ROC:** Sinifləndirmə modellərində performansı qiymətləndirmək üçün.

### Müqayisə etmək:
- **Cross-Validation Nəticələri** (orta qiymətləndirmə) modelin **ümumi təlim performansını** göstərir.
- **Test Dəsti Nəticələri** (son qiymətləndirmə) isə modelin **gerçək dünya verilənlərindəki** performansını göstərir.

İdeal olaraq, **cross-validation** nəticələri ilə **test dəstinin** nəticələri arasında çox böyük fərq olmamalıdır. Əgər fərq böyükdürsə, bu overfitting və ya underfitting olduğunu göstərə bilər.

---

### Nümunə:
Tutaq ki, **cross-validation** nəticəsində doğruluq **0.85** alındı və **test dəsti** ilə doğruluq **0.80** oldu. Bu halda, **cross-validation** nəticələrinin test dəstinə yaxın olması yaxşı modelin göstəricisidir.
