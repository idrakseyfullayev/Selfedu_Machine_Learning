#### 🔍 **1. Linear Regression nədir?**

* **Linear Regression** — **Supervised Learning** metodudur, **Regression (rəqəmsal proqnoz)** üçün istifadə olunur.
* **Məqsədi**: Giriş dəyişənləri (X) əsasında **continuous (kəsilməz) dəyər** tapmaq (məs: qiymət, maaş, temperatur və s.)

> **Tərif:** Linear Regression, müstəqil dəyişənlərlə (features) hədəf dəyişən (target) arasında **xətti əlaqəni modelləşdirən** və bu əlaqə əsasında **ədədi dəyər proqnozlaşdıran** statistik modeldir.

> Logistic-də ehtimal tapırdıq, burada isə **birbaşa dəyər tapırıq**.

<br>

#### 🧮 **Predict funksiyası nədir, necə işləyir?**

#### ⚙️ Məntiq:

**linear kombinasiya** hesablanır:

$$
\hat{y} = \beta_1^T x + \beta_0 = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
$$

* $x$ — input xüsusiyyətlər
* $\beta_1$ — weights
* $\beta_0$ — bias


<br>

#### 🔹 Nəticə necə olur?

* $\hat{y} = 1500$ → məsələn evin qiyməti
* $\hat{y} = 23.5$ → temperatur
* $\hat{y} = 800$ → maaş

> **Output: $(-\infty, +\infty)$ intervalında real ədəddir**

<br>

💡 **Qısa intuisiya:**
> X → xətt çəkilir → həmin xətt üzərində nöqtənin qiyməti tapılır

---

<br>

#### 📉 **Itki funksiyası (Loss Function) nədir?**

Linear Regression **MSE (Mean Squared Error)** istifadə edir:

$$
L(\beta_0, \beta_1) = \frac{1}{n} \sum_{i=1}^N (y_i - \hat{y}_i)^2 = \frac{1}{n} \sum_{i=1}^n \big(y_i - (\beta_0 + \beta_1 x_i)\big)^2
$$

 

<br>

#### 🔹 İzah:

* $y_i$ — real dəyər
* $\hat{y}_i$ — modelin proqnozu

🔹 Məqsəd:
**Proqnoz ilə real dəyər arasındakı fərqi minimum etmək**

<br>

#### 🔹 Niyə kvadrat?

> Böyük səhvləri daha çox cəzalandırır

> Mənfi işarə problemi olmur

<br>

💡 Misal:

* səhv = 1 → $1^2 = 1$
* səhv = 10 → $10^2 = 100$ (daha ciddi cəza)

---

<br>

#### ⚙️ **Itki funksiyası (Loss Function) necə optimallaşdırılır?**

> **Linear Regression modelində xəta funksiyası, yəni Mean Squared Error (MSE), proqnozlaşdırılmış dəyərlər $ \hat{y} $ ilə faktiki dəyərlər  $ y $ arasındakı fərqlərin kvadrat ortalaması kimi müəyyən edilir. Bu funksiyanın törəməsi sıfıra bərabər edilərək, xəta minimal səviyyəyə endirilir və ən uyğun $ \beta $ əmsalları tapılır.**

> **MSEfunksiyasının optimallaşdırılması üçün $\beta_0$ və $\beta_1$-ə görə qismən törəmələri sıfıra bərabər edib həll edək. Bu, xətti regresiya üçün parametrlərin analitik həllini tapmağa imkan verəcək.** 

#### **MSE funksiyası**:

> Bu, **Xətti Reqressiya** modelinin **Ən Kiçik Kvadratlar Üsulu (Ordinary Least Squares - OLS)** ilə skalyar formada həllidir.

$$
L(\beta_0, \beta_1) = \frac{1}{n} \sum_{i=1}^n \big(y_i - (\beta_0 + \beta_1 x_i)\big)^2
$$


#### **$\beta_0$-a görə qismən törəmə**:
Törəmə:
$$
\frac{\partial L}{\partial \beta_0} = -\frac{2}{n} \sum_{i=1}^n (y_i - \beta_0 - \beta_1 x_i)
$$

Bunu sıfıra bərabər edirik:
$$
-\frac{2}{n} \sum_{i=1}^n (y_i - \beta_0 - \beta_1 x_i) = 0
$$
Sadələşdirək:
$$
\sum_{i=1}^n (y_i - \beta_0 - \beta_1 x_i) = 0
$$
$$
\sum_{i=1}^n y_i - n\beta_0 - \beta_1 \sum_{i=1}^n x_i = 0
$$
$$
n\beta_0 = \sum_{i=1}^n y_i - \beta_1 \sum_{i=1}^n x_i
$$
$$
\beta_0 = \frac{\sum_{i=1}^n y_i}{n} - \beta_1 \frac{\sum_{i=1}^n x_i}{n}
$$

<br>

Burada:
- $\frac{\sum_{i=1}^n y_i}{n} =  \bar{y}$ $\rightarrow$ $y$ **dəyişəninin orta qiyməti.**
- $\frac{\sum_{i=1}^n x_i}{n} = \bar{x}$ $\rightarrow$ $x$ **dəyişəninin orta qiyməti,**


<br>

#### **$\beta_1$-ə görə qismən törəmə**:
Törəmə:
$$
\frac{\partial L}{\partial \beta_1} = -\frac{2}{n} \sum_{i=1}^n x_i (y_i - \beta_0 - \beta_1 x_i)
$$

Bunu sıfıra bərabər edirik:
$$
-\frac{2}{n} \sum_{i=1}^n x_i (y_i - \beta_0 - \beta_1 x_i) = 0
$$
Sadələşdirək:
$$
\sum_{i=1}^n x_i (y_i - \beta_0 - \beta_1 x_i) = 0
$$
$$
\sum_{i=1}^n x_i y_i - \beta_0 \sum_{i=1}^n x_i - \beta_1 \sum_{i=1}^n x_i^2 = 0
$$

<br>

#### $\beta_0$-ı ****yuxarıdakı tənlikdə yerinə qoyuruq:****

$$
\beta_0 = \frac{\sum y_i}{n} - \beta_1 \frac{\sum x_i}{n}
$$

yerinə qoyaq:

$$
\sum x_i y_i - \left( \frac{\sum y_i}{n} - \beta_1 \frac{\sum x_i}{n} \right)\sum x_i - \beta_1 \sum x_i^2 = 0
$$

<br>

#### **Açırıq (paylayırıq):**

$$
\sum x_i y_i - \frac{\sum y_i \sum x_i}{n} + \beta_1 \frac{(\sum x_i)^2}{n} - \beta_1 \sum x_i^2 = 0
$$

<br>

####  $\beta_1$ **olanları qrupla:**

$$
\sum x_i y_i - \frac{\sum y_i \sum x_i}{n} = \beta_1 \left( \sum x_i^2 - \frac{(\sum x_i)^2}{n} \right)
$$

<br>

#### $\beta_1$-i **ayır:**

$$
\beta_1 =
\frac{\sum x_i y_i - \frac{\sum x_i \sum y_i}{n}}
{\sum x_i^2 - \frac{(\sum x_i)^2}{n}}
$$

<br>

#### **Bu nədir əslində?**

Bu ifadə **kovariasiya bölünsün varians formasıdır**:

$$
\beta_1 = \frac{\text{Cov}(x,y)}{\text{Var}(x)}
$$

<br>

$$
\beta_1 = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
$$

<br>

#### **$\beta_0$ və $\beta_1$-in final nəticələri**:
1. $\beta_1$:
$$
\beta_1 = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^n (x_i - \bar{x})^2}
$$

2. $\beta_0$:
$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$


### Itki funksiyası (Loss Function) necə optimallaşdırılır? Matris forması

> **Scalar formadakı linear regression MSE törəmələri Matris formasına çevrilə bilər.** 

> **Bu, xüsusilə çox dəyişənli (multiple) linear regression üçün çox faydalıdır.**

<br>

#### **1️⃣ Modelin matris forması**

> **Linear regression modelini matrisdə belə yaza bilərik:**

$$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\epsilon}
$$

* $\mathbf{y}$ — $n \times 1$ çıxış vektoru
* $\mathbf{X}$ — $n \times (p+1)$ giriş matrisi (ilk sütun 1-lərdir bias üçün)

$$
\mathbf{X} =
\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_n
\end{bmatrix}, \quad
\boldsymbol{\beta} =
\begin{bmatrix}
\beta_0 \\
\beta_1
\end{bmatrix}, \quad
\boldsymbol{\epsilon} =
\begin{bmatrix}
\epsilon_1 \\
\epsilon_2 \\
\vdots \\
\epsilon_n
\end{bmatrix}
$$

<br>

#### **2️⃣ MSE funksiyası matrisdə**

<br>

**1️⃣ Adi yazılış (real − predict)**

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big( \mathbf{y} - \hat{\mathbf{y}} \big)^2
$$

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big( \mathbf{y} -\mathbf{X}\boldsymbol{\beta} \big)^2
$$

> Burada
>
> * $\mathbf{y}$ — real dəyərlər
> * $\hat{\mathbf{y}} = \mathbf{X}\boldsymbol{\beta}$ — predict edilmiş dəyərlər (y cap)
> * $\mathbf{y} - \hat{\mathbf{y}}$ sadəcə **real − predict** deməkdir.



**2️⃣ Matris forması açılmış şəkildə**

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big(\mathbf{y} - \mathbf{X}\boldsymbol{\beta}\big)^T \big(\mathbf{y} - \mathbf{X}\boldsymbol{\beta}\big)
$$

<br>

#### **3️⃣ Gradient (törəmə) matrisdə**

MSE-i $\boldsymbol{\beta}$-ə görə törədək:

$$
\frac{\partial L}{\partial \boldsymbol{\beta}} = -\frac{2}{n} \mathbf{X}^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})
$$

> **Scalar formada gördüyümüz $\frac{\partial L}{\partial \beta_0}$ və $\frac{\partial L}{\partial \beta_1}$ matris formasının sadələşdirilmiş versiyasıdır.**

<br>

#### **4️⃣ Törəməni sıfıra bərabərləyərək optimal $\boldsymbol{\beta}$-i tapmaq**

$$
\mathbf{X}^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta}) = 0
$$

$$
\mathbf{X}^T \mathbf{y} - \mathbf{X}^T \mathbf{X} \boldsymbol{\beta} = 0
$$

$$
\mathbf{X}^T \mathbf{X} \boldsymbol{\beta} = \mathbf{X}^T \mathbf{y}
$$

<br>

#### **5️⃣ Nəticə: Matris formasında closed-form düstur (Normal Equations)**

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

> **Bu düstur scalar formada çıxan $\beta_0 = \bar{y} - \beta_1 \bar{x}$, $\beta_1 = \frac{\text{Cov}(x,y)}{\text{Var}(x)}$ nəticələrinin generalizasiyasıdır.**

> **Bir neçə dəyişən olsa belə eyni düstur işləyir, sadəcə $\boldsymbol{\beta}$ və $\mathbf{X}$ daha böyük ölçülərdə olur.** 

<br>

#### 🧠 **Closed-form həll (Normal Equation)**

Linear Regression-un **ən böyük üstünlüyü** 👇

Gradient descent olmadan da həll var:

$w = (X^T X)^{-1} X^T y$

<br>

#### 🔹 **Bu nə deməkdir?**

* Birbaşa optimal $w$ tapılır
* Iterasiya yoxdur
* Gradient descent lazım olmaya bilər

<br>

#### 📊 **Əlaqə xülasəsi**

| Addım | Funksiya              | Məqsəd                 |
| ----- | --------------------- | ---------------------- |
| 1     | $$\hat{y} = w^T x + b$$ | Birbaşa proqnoz        |
| 2     | $$L = (y - \hat{y})^2$$ | Səhvi ölçmək           |
| 3     | Gradient              | Parametrləri yeniləmək |
| 4     | $$(X^TX)^{-1}X^Ty$$     | Birbaşa həll           |

<br>

#### **Linear Regression:**

> **“Ən yaxşı xətti çək ki, nöqtələrə ən yaxın olsun”**

### Itki funksiyası (Loss Function) necə optimallaşdırılır? Matris forması - Daha Geniş

 

> **Scalar formadakı linear regression MSE törəmələri Matris formasına çevrilə bilər. Bu məsələyə Matris (Vektor) formasında baxaq.** 

> **Bu yanaşma daha yığcamdır və çoxölçülü (çoxsaylı asılı olmayan dəyişənlər olan) reqressiya üçün standart üsuldur.**

<br>

#### **Xətti Reqressiyanın Matris Forması**

<br>

- **Əvvəlcə modelimizi matris dilinə çevirək:**
$$
y = X\beta + \epsilon
$$

$$
\mathbf{X} =
\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_n
\end{bmatrix}, \quad
\boldsymbol{\beta} =
\begin{bmatrix}
\beta_0 \\
\beta_1
\end{bmatrix}, \quad
\boldsymbol{\epsilon} =
\begin{bmatrix}
\epsilon_1 \\
\epsilon_2 \\
\vdots \\
\epsilon_n
\end{bmatrix}
$$

- **Burada:**
* $y$: **Asılı dəyişənlər vektoru** ($n \times 1$)
* $X$: **Dizayn matrisi ($n \times (k+1)$). Birinci sütun bias ($\beta_0$) üçün $1$-lərdən ibarətdir.**
* $\beta$: **Tapılmalı olan əmsallar vektoru** ($(k+1) \times 1$)
* $\epsilon$: **Xətalar (qalıqlar) vektoru** ($n \times 1$)

<br>

#### **1. Xəta Funksiyasının (MSE) Tərifi**

<br>

- **Matris formasında Orta Kvadratik Xəta (MSE) qalıqların kvadratları cəmi kimi belə yazılır:**
$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big( \mathbf{y} - \hat{\mathbf{y}} \big)^2
$$

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big( \mathbf{y} -\mathbf{X}\boldsymbol{\beta} \big)^2
$$
$$L(\beta) = \frac{1}{n} (y - X\beta)^T (y - X\beta)$$

- **Mötərizələri açaq (transpoze xassələrindən istifadə edərək):**
$$
L(\beta) = \frac{1}{n} (y^T y - y^T X\beta - \beta^T X^T y + \beta^T X^T X \beta)
$$

- **Burada $y^T X\beta$ və $\beta^T X^T y$ skalyar kəmiyyətlərdir və bir-birinə bərabərdir. Ona görə də:**
$$L(\beta) = \frac{1}{n} (y^T y - 2\beta^T X^T y + \beta^T X^T X \beta)$$

<br>

#### **2. Törəmənin Tapılması və Sıfıra Bərabər Edilməsi**

<br>

- $\beta$ **vektoruna görə qismən törəmə alırıq (Matrix Calculus qaydalarına əsasən):**

1.  $\frac{\partial}{\partial \beta} (y^T y) = 0$
2.  $\frac{\partial}{\partial \beta} (2\beta^T X^T y) = 2X^T y$
3.  $\frac{\partial}{\partial \beta} (\beta^T X^T X \beta) = 2 X^T X \beta$

<br>

- **Beləliklə:**
$$\frac{\partial L}{\partial \beta} = \frac{1}{n} (-2 X^T y + 2 X^T X \beta) = 0$$

- **İndi isə $\beta$-nı tənlikdən ayıraq:**
$$X^T X \beta = X^T y$$

Bu tənlik **Normal Tənlik (Normal Equation)** adlanır.

<br>

#### **3. Final Nəticə**
<br>

- **Əgər $X^T X$ matrisi tərsilənəndirsə (yəni determinat sıfır deyilsə), $\beta$ əmsallarını tapmaq üçün hər iki tərəfi soldan $(X^T X)^{-1}$ matrisinə vururuq:**

$$\beta = (X^T X)^{-1} X^T y$$

<br>

#### **Skalyar və Matris Formasının Müqayisəsi**

| Xüsusiyyət | Skalyar (Sizin yazdığınız) | Matris (Yuxarıdakı) |
| :--- | :--- | :--- |
| **Dəyişən sayı** | Adətən tək $x$ üçün rahatdır. | İstənilən sayda $x$ üçün eynidir. |
| **Hesablama** | $\beta_0$ və $\beta_1$ ayrı hesablanır. | Bütün $\beta$ vektor şəklində bir dəfəyə tapılır. |
| **Gözəlliyi** | Statistik mənanı (kovariasiya) göstərir. | Kompüter alqoritmləri üçün optimaldır. |

> **Qeyd:** **Skalyar formada tapdığımız $\beta_0$ və $\beta_1$ qiymətləri, bu matris həllindəki $\beta$ vektorunun elementlərinə tam olaraq ekvivalentdir.**

####  **3️⃣ İtki (loss) funksiyası**

<br> 

`sklearn.linear_model.LinearRegression` **MSE (Mean Squared Error)** minimallaşdırır:

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
$$

- **lakin matris formasında:**

$$
L(\boldsymbol{\beta}) = \frac{1}{n} (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})^T (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})
$$

> **Amma sklearn bunu “hesablayırmı”?**
- ⚠️ **Vacib nöqtə:**
> `LinearRegression` **loss-u ayrıca hesablayıb iterasiya etmir**.

- **Yəni:**

- ❌ **Gradient descent YOX**
- ❌ **Hər iterasiyada loss hesablanmır**
- ❌ **Loss-u optimallaşdırmaq üçün loop yoxdur**

> **Birbaşa Analitik həll istifadə edir:**

$$
\boldsymbol{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}
$$

> **Bu düstur MSE-ni minimallaşdıran nöqtəni birbaşa verir**.

- **Yəni:**

- **Loss funksiyası var** ✔️
- **Amma onu hesablayıb addım-addım azaltmır** ❌
- **Birbaşa minimum nöqtəni tapır ✔️**

> **İstəsən loss-u özün hesablaya bilərsən**

- **`sklearn` avtomatik göstərmir, amma sən hesablaya bilərsən:**

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print(mse)
```

<br>

#### **1️⃣ Parametrlərin $\beta$ Tapılması: Iterativ Yeniləmə**

> `SGDRegressor` **işə başlayanda çəkiləri (weights $\beta$) və bias-ı ($\beta_0$) təsadüfi və ya sıfır olaraq təyin edir. Hər bir iterasiyada (və ya hər bir data nöqtəsi üçün) gradient hesablanır və parametrlər dərhal yenilənir.**




#### **MSE funksiyası matrisdə**

<br>

- **Adi yazılış (real − predict)**

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big( \mathbf{y} - \hat{\mathbf{y}} \big)^2
$$

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big( \mathbf{y} -\mathbf{X}\boldsymbol{\beta} \big)^2
$$

> Burada
>
> * $\mathbf{y}$ — real dəyərlər
> * $\hat{\mathbf{y}} = \mathbf{X}\boldsymbol{\beta}$ — predict edilmiş dəyərlər (y cap)
> * $\mathbf{y} - \hat{\mathbf{y}}$ sadəcə **real − predict** deməkdir.

<br>

- **Matris forması açılmış şəkildə**

$$
L(\boldsymbol{\beta}) = \frac{1}{n} \big(\mathbf{y} - \mathbf{X}\boldsymbol{\beta}\big)^T \big(\mathbf{y} - \mathbf{X}\boldsymbol{\beta}\big)
$$

<br>

#### **Gradient (törəmə) matrisdə**

- **MSE-i $\boldsymbol{\beta}$-ə görə törədək:**

$$
\frac{\partial L}{\partial \boldsymbol{\beta}} = -\frac{2}{n} \mathbf{X}^T (\mathbf{y} - \mathbf{X} \boldsymbol{\beta})
$$

$$
\frac{\partial L}{\partial \boldsymbol{\beta}} = \frac{2}{n} X^T (X\beta - y)
$$

<br>

#### **Gradientin Yekun forması - $\beta_0$ də əlavə olunur (Matris Forması)**

<br>

- **Tutaq ki, bizim $n$ sayda nümunəmiz və $m$ sayda xüsusiyyətimiz (features) var. Matris formasında gradient belə hesablanır:**

$$\nabla_{\beta} L = \frac{2}{n} X^T (X\beta + \beta_0 - y)$$

Burada:
* $X$: $n \times m$ ölçülü giriş matrisi.
* $y$: $n \times 1$ ölçülü hədəf vektoru.
* $(X\beta + \beta_0 - y)$: Modelin proqnozu ilə real dəyər arasındakı **səhv (error)** vektoru.

<br>

#### **Əmsalların Yenilənməsi**

<br>

- **Hər addımda $\beta$ (çəkilər) və $\beta_0$ (bias) aşağıdakı düsturla yenilənir:**

$$\beta^{(t+1)} = \beta^{(t)} - \alpha \cdot \nabla_{\beta} L$$
$$\beta_0^{(t+1)} = \beta_0^{(t)} - \alpha \cdot \frac{2}{n} \sum (X\beta + \beta_0 - y)$$

> $\mathbf{X}^T$ **yalnız $\beta$ üçün gradient-də var, çünki $\beta$ vektordur və matris ilə vurulmalıdır; $\beta_0$ isə skalar olduğu üçün sadəcə elementləri toplamaq kifayətdir, ona görə $\mathbf{X}^T$ yoxdur. İfadəni sadələşdirəndə $\mathbf{X}^T$ yalnız $\beta$ üçün qalır, $\beta_0$-a görə isə toplama kifayət edir, ona görə $\mathbf{X}^T$ “çıxır”.**

> **Qeyd:** `SGDRegressor` **adı üstündə "Stochastic" olduğu üçün, o, $n$-i adətən 1 götürür (hər sətir üçün yeniləmə edir) və ya kiçik "batch"-lərlə işləyir. Bu, hesablama sürətini artırır.**

---

<br>

#### **2️⃣ İtki Funksiyasının Hesablanması (Loss Tracking)**

 - **Hər iterasiyadan sonra modelin nə dərəcədə yaxşı öyrəndiyini yoxlamaq üçün İtki Funksiyası (MSE) hesablanır. Sklearn bunu numerik stabillik üçün matris formasında belə icra edir:**
 
- **Sadə MSE (loss tracking üçün):**

$$J(\beta) = \frac{1}{n} \| X\beta + \beta_0 - y \|^2_2$$

- **Burada $\| \cdot \|^2_2$ ifadəsi L2 normasını (kvadratlar cəmini) bildirir.**

<br>

- **Əgər modelə Regularization (L2 - Ridge) əlavə olunubsa (Sklearn-də susmaya görə `penalty='l2'`), düstur belə şəkil alır:**

- **L2 Regularization əlavə olunanda:**
$$
J(\beta) = \frac{1}{n} \| X\beta + \beta_0 - y \|_2^2 + \lambda \|\beta\|_2^2
$$

> **Yəni həm sadə MSE, həm də Ridge versiyası matris/vetor əməliyyatı ilə eyni formada hesablanır.**

> **Bu mərhələdə model yoxlayır: "İtki əvvəlki addıma nisbətən azaldımı?". Əgər azalma `tol` (tolerance) parametrindən kiçikdirsə, öyrənmə dayandırılır.**

---

<br>

#### **3️⃣ Proqnoz (Predict) Mexanizmi**

<br>

- **Realda `SGDRegressor` matris əməliyyatı ilə işləyir, yəni proqnozu bütün nümunələr üçün bir dəfə vektor / matris hasilatı ilə hesablayır. Əmsallar ($\beta$ və $\beta_0$) tapıldıqdan sonra proqnoz vermək ən sadə mərhələdir. Bu, birbaşa matris hasilidir:**

$$
\hat{y} = X_{new} \cdot \beta + \beta_0
$$

- **İzahı real prosesdə:**

1. $X_{new}$ — **bütün yeni nümunələrin xüsusiyyət matrisi.**
2. $\beta$ — **öyrənilmiş çəkilər vektoru.**
3. $X_{new} \cdot \beta$ — **bütün nümunələr üçün linear kombinasiya (matris hasilatı).**
4. $+ \beta_0$ — **bias hər sətirə əlavə olunur (broadcasting ilə).**
5. **Nəticə** $\hat{y}$ — **bütün nümunələr üçün vektor olaraq qaytarılır, skalyar tək-tək nümunələr üçün deyil, bütün batch üzrə.**

- **Qısaca:** **SGDRegressor realda bütün nümunələri eyni anda matris hasilatı ilə proqnozlayır**, **tək-tək skalyar hesablama etmir.**

<br>

#### **Xülasə: SGD-nin OLS-dən Fərqi**
* **OLS:** $(X^T X)^{-1} X^T y$ **düsturu ilə birbaşa bütün matrisi həll edir (Yaddaş tələb edir).**
* **SGD:** **Hər iterasiyada gradienti tapır, $\beta$-nı bir az "tərpədir" (azaldır) və bunu itki minimum olana qədər təkrar edir (Böyük datalarda effektivdir).**

#### **Gradient Descent** prosesi necə işləyir və **Dayanma şərtləri** (Stopping criteria)

#### 🔹 **1️⃣ Başlanğıc (Initialization)**

- Model **ω (weights)** və **b (bias)**-ı **təsadüfi** seçir:

$$
\omega^{(0)}, \ b^{(0)}
$$

<br>

#### 🔹 **2️⃣ Proqnoz hesablanır**

$$
\hat{y} = X\omega + b
$$

<br>

#### 🔹 **3️⃣ İtki (Loss) hesablanır**

- **Adətən MSE:**

$$
L = \frac{1}{n} \sum (y - \hat{y})^2
$$

<br>

#### 🔹 **4️⃣ Gradient hesablanır**

- **Bu ən vacib hissədir 👇**

$$
\nabla_{\omega} L = \frac{2}{n} X^T (X\omega + b - y)
$$

$$
\nabla_{b} L = \frac{2}{n} \sum (X\omega + b - y)
$$

<br>

#### 🔹 **5️⃣ Parametrlər yenilənir**

$$
\omega := \omega - \alpha \cdot \nabla_{\omega} L
$$

$$
b := b - \alpha \cdot \nabla_{b} L
$$

<br>

#### 🔁 **6️⃣ Bu proses təkrarlanır**

- **Vacib sual məhz buradır 👇**

> **“itki sıfır olana qədər davam edir?”**

> ❌ **Xeyr, mütləq 0 olmur və olmaq da lazım deyil**

<br>

#### 🔹 **Dayanma şərtləri (Stopping criteria)**

> **Gradient Descent bunlardan biri ödənəndə dayanır**:

#### ✔️ **1. Maksimum iterasiya sayı**

```python
max_iter = 1000
```

<br>

#### ✔️ **2. İtki artıq çox az dəyişir (convergence)**

$$
|L^{(t)} - L^{(t-1)}| < \epsilon
$$

- **yəni:**

> **artıq model öyrənmir → dayan**

<br>

#### ✔️ **3. Gradient çox kiçik olur**

$$
\nabla L \approx 0
$$

- **yəni:**

> **minimum nöqtəyə çatmışıq**

<br>

#### **🔴 Vacib məqam**

* **Loss = 0 demək olar ki, real datada mümkün deyil**
* **Çünki:**

  * **səs (noise) var**
  * **model sadədir**
  * **data tam linear deyil**

<br>

#### 🔹 **Qısa məntiq (sənin dediyin dillə)**

- ✔️ Başlayır random ω ilə
- ✔️ Loss hesablanır
- ✔️ Böyükdür → gradient ilə düzəldilir
- ✔️ Yenə hesablanır
- ✔️ Yenə düzəldilir

- **🔁 Bu proses davam edir**

**⛔ Amma:**

- **“loss 0 olsun” deyə yox**
- **“artıq yaxşılaşma yoxdur” deyə dayanır**

<br>

#### **🔥 Vizual düşünmə**

- **Gradient descent belədir:**
  - **Dağın başındasan ⛰️**
  - **şağı düşmək istəyirsən**
  - **Hər addımda ən dik istiqamətə gedirsən**

<br>

- **Sonda:**
  - **ən aşağı nöqtəyə yaxınlaşırsan**
  - **amma tam “0 nöqtə” olmaya bilər**


### Gradient Descent HEÇ VAXT istiqaməti dəyişmir.
#### **itki artdısa istiqaməti dəyişib, yəni “+ gradient etmir”**

#### 🔹 **Əsas qayda**

$$
\omega := \omega - \alpha \cdot \nabla L
$$

> **Gradient ($\nabla L$) → itkini artıran istiqaməti göstərir**

> **Biz isə Mənfi gradient ($-\nabla L$) → itkini azaldan istiqamətə gedirik**

- 👉 **Bizim məqsəd:** **minimum tapmaq (minimize etmək)**
- 👉 **Ona görə həmişə** **çıxırıq (−)**

<br>

#### 🔥 Vacib insight



✔️ **Bəli, qısa cavab:**

👉 **Gradientin işarəsi (sign) dəyişir**, sən yox.

* Sən həmişə:
  $$
  \omega := \omega - \alpha \cdot \nabla L
  $$

* Amma:
  $$
  \nabla L ; (+) \rightarrow (-) \text{ dəyişə bilər}
  $$

👉 Buna görə update bəzən **+ kimi görünür**

---

✅ Yekun:

> **İstiqaməti gradient özü dəyişir, sən həmişə “− gradient” edirsən.**



Loss artanda Gradient artıq özü istiqaməti “deyir”:

* Əgər sağa getsən loss artır → gradient səni sola aparacaq
* Əgər sola getsən loss artır → gradient səni sağa aparacaq

Yəni:

> sən ayrıca “+ ya da - edim?” deyə qərar vermirsən

---

## 🔹 Bəs niyə bəzən loss artır?

Bu çox vacibdir 👇

### Səbəb 1️⃣: Learning rate çox böyükdür

$$
\alpha \text{ çox böyükdür}
$$

Nəticə:

* minimumu “ötürsən” (overshoot edirsən)
* loss artıb-azalıb “zıplayır”

---

### Səbəb 2️⃣: SGD-də noise var

`SGDRegressor`-də:

* hər dəfə bütün data yox, **bir hissə (batch)** istifadə olunur
* buna görə:

  * loss bəzən artır
  * amma ümumi trend azalır

---

## 🔹 Sualına konkret cavab

> “itki artdı deyə + gradient edirik?”

❌ Xeyr

✔️ Həmişə eyni formula:

$$
\omega := \omega - \alpha \cdot \nabla L
$$

Sadəcə:

* gradient özü istiqaməti dəyişir
* sən yox

---

#### 🔹 Sadə intuisiya

Təsəvvür et:

* gradient = “yuxarı istiqamət”
* sən = “aşağı düşmək istəyən”

👉 Həmişə:

* gradient göstərən istiqamətin ƏKSİNƏ gedirsən

<br>

#### **🔥 Bonus (çox vacib real məqam)**

- Əgər həqiqətən:

  - **loss tez-tez artır** → **learning rate azalt**
  - **loss çox yavaş azalır** → **learning rate artır**







<br>

#### 🔴 Bəs “+ gradient” nə vaxt olur?

Bu artıq başqa şeydir 👇

#### ✔️ Gradient Ascent

$$
\omega := \omega + \alpha \cdot \nabla L
$$

Bu istifadə olunur:

* **loss-u maksimum etmək** istəyəndə
* məsələn:

  * likelihood **maximize** edəndə
  * bəzi RL (reinforcement learning) alqoritmlərində

<br>

#### 🔥 Sənin confusion-un səbəbi

Sən belə düşünürsən:

> “loss artdı → istiqaməti dəyişim (+ edim)”

Amma realda:

* istiqaməti **sən dəyişmirsən**
* gradient artıq dəyişir

<br>

#### 🔹 Kiçik nümunə

Tutaq ki:

* əvvəl gradient = +5 → sən − edirsən → sola gedirsən
* sonra nöqtəni keçdin → gradient = −3 olur

İndi update:

$$
\omega := \omega - \alpha \cdot (-3) = \omega + 3\alpha
$$

👉 Bax burada **riyazi olaraq + oldu**, amma:

* sən qaydanı dəyişmədin ❗
* gradient özü işarəni dəyişdi

<br>

#### ✅ Nəticə

* Gradient Descent:

  * həmişə → **− gradient**
* Gradient Ascent:

  * həmişə → **+ gradient**
* Amma:

  * gradient özü dəyişdiyi üçün bəzən update “+ kimi” görünür



Əla, indi tam nüvəyə gəldik 👇

## 🔹 Necə?

Gradient bu düsturdur:

[
\nabla L = \frac{dL}{d\omega}
]

👉 Yəni:

* **funksiyanın meyli (slope)**

---

## 🔥 İşarə nə deməkdir?

### ✔️ 1. Əgər:

[
\nabla L > 0
]

* funksiya **yuxarı artır**
* minimum sola tərəfdədir
  👉 sən:
  [
  \omega := \omega - \alpha \cdot (+)
  ]
  → **sola gedirsən**

---

### ✔️ 2. Əgər:

[
\nabla L < 0
]

* funksiya **aşağı düşür**
* minimum sağdadır
  👉 sən:
  [
  \omega := \omega - \alpha \cdot (-)
  ]
  → **sağa gedirsən**

---

## 🔹 Bəs gradient niyə dəyişir?

Çünki:

[
\nabla L = f(\omega)
]

👉 yəni gradient **ω-dan asılıdır**

]

Gradient:

]

### Nöqtələr:

* ω = 0 → ∇L = 0
* ω = -2 → ∇L = -4

👉 Görürsən?

* ω dəyişdikcə → gradientin işarəsi dəyişir

* ω dəyişir

* → istiqamət dəyişir

* buna görə **gradientin işarəsi dəyişir**
