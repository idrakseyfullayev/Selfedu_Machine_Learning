## Probability Məsələlər



**Məsələ 1.**
Təsadüfi eksperimentdə simmetrik pul iki dəfə atılır. Tapın ehtimalı ki, **təkcə bir dəfə yazı (orel) düşsün**.

**İzah:**

* Simmetrik pulun hər atışında `orel` və `yazı` ehtimalları bərabərdir: $P(\text{orel}) = 0.5$ və $P(\text{yazı}) = 0.5$.
* İki dəfə atış üçün nümunələr: (O, O), (O, Y), (Y, O), (Y, Y).
* Təkcə bir dəfə O olması nümunələr: (O, Y) və (Y, O) → 2 hadisə.
* Ehtimal: $P = \frac{2}{4} = 0.5$.

---

**Məsələ 2.**
Simmetrik pul üç dəfə atılır. Tapın ehtimalı ki, **yazı (reshka) dəqiq olaraq 2 dəfə düşsün**.

**İzah:**

* Üç dəfə atış nümunələri: 2 yazı və 1 orel üçün kombinasiyalar: (Y, Y, O), (Y, O, Y), (O, Y, Y) → 3 hadisə.
* Ümumilikdə mümkün nəticələr: $2^3 = 8$.
* Ehtimal: $P = \frac{3}{8} = 0.375$.

---

**Məsələ 3.**
Üç ədəd oyun zarı eyni anda atılır. Tapın ehtimalı ki, **1, 2 və 3 ədədləri çıxsın, sıra önəmsizdir**.

**İzah:**

* Mövcud sıra önəmsizdir, yəni (1,2,3), (1,3,2), (2,1,3) və s. hamısı eyni hadisə kimi sayılır.
* Üç fərqli ədədi üç zar üzərində yerləşdirmək üçün permutasiya sayı: $3! = 6$.
* Ümumilikdə mümkün nəticələr: $6^3 = 216$ (hər zar 1-6 arası).
* Ehtimal: $P = \frac{6}{6^3?}$ – dəqiq hesab etmək üçün:
  Sıra önəmsiz olduğuna görə 1,2,3 çıxan nümunə yalnız bir hadisə kimi sayılır. Hər zarın fərqli olma ehtimalı:

  $$
  P = \frac{3!}{6^3} = \frac{6}{216} = \frac{1}{36} \approx 0.0278
  $$

---

**Məsələ 4.**
Simmetrik pul dörd dəfə atılır. Tapın ehtimalı ki, **dəqiq olaraq 3 dəfə orel düşsün**.

**İzah:**

* Dörd atış üçün kombinasiya: 3 orel və 1 yazı.
* Kombinasiya sayı: $C_4^3 = 4$ (hansı atış yazı olacaq).
* Ümumi nəticə sayı: $2^4 = 16$.
* Ehtimal: $P = \frac{4}{16} = 0.25$.

---

**Məsələ 5.**
Oyun zarı iki dəfə atılır. Tapın ehtimalı ki, **birinci atışda 3-dən böyük, ikinci atışda 3-dən kiçik çıxsın**.

**İzah:**

* Birinci zar: 3-dən böyük → 4,5,6 → 3 mümkün nəticə.
* İkinci zar: 3-dən kiçik → 1,2 → 2 mümkün nəticə.
* Ümumi nəticələr: $6 \times 6 = 36$.
* Xüsusi hadisələr sayı: $3 \times 2 = 6$.
* Ehtimal: $P = \frac{6}{36} = \frac{1}{6} \approx 0.1667$.



## Probability Məsələlər geniş izahlı



### **Məsələ 1.**
Təsadüfi eksperimentdə simmetrik pul iki dəfə atılır. Tapın ehtimalı ki, **təkcə bir dəfə yazı (orel) düşsün**.

**İzah:**

* Simmetrik pulun hər atışında `orel` və `yazı` ehtimalları bərabərdir: $P(\text{orel}) = 0.5$ və $P(\text{yazı}) = 0.5$.
* İki dəfə atış üçün nümunələr: (O, O), (O, Y), (Y, O), (Y, Y).
* Təkcə bir dəfə O olması nümunələr: (O, Y) və (Y, O) → 2 hadisə.
* Ehtimal: $P = \frac{2}{4} = 0.5$.

**Kombinasiya istifadə etmədən (sadə məntiq):**
İki dəfə atışda təkcə 1 dəfə orel düşməsini düşünək:

* Hansı atışın orel düşəcəyini seçmək lazımdır → 2 variant var.
* Qalan atış yazı olmalıdır → ehtimalı $0.5$.
* Orel atılan atışın ehtimalı → $0.5$.
* Hər variantın ehtimalı: $(0.5)^2 = 0.25$.
* Ümumi ehtimal: $2 \cdot 0.25 = 0.5$. ✅

**Binomial formul:**

$$
P(X=1) = \binom{2}{1} (0.5)^1 (0.5)^{2-1} = 2 \cdot 0.5 \cdot 0.5 = 0.5
$$

---

### **Məsələ 2.**

Simmetrik pul üç dəfə atılır. Tapın ehtimalı ki, **yazı (reshka) dəqiq olaraq 2 dəfə düşsün**.

**İzah:**

* Ümumilikdə mümkün nəticələr: $2^3 = 8$.
* Ehtimal: $P = \frac{3}{8} = 0.375$.

**Kombinasiya istifadə etmədən (sadə məntiq):**
Üç dəfə atışda təkcə 2 dəfə yazı düşməsini düşünək:

* Hansı atışın orel düşəcəyini seçmək lazımdır → 3 variant var.
* Qalan 2 atış yazı olmalıdır → ehtimalı $(0.5)^2$.
* Orel atılan atışın ehtimalı → $0.5$.
* Hər variantın ehtimalı: $(0.5)^3 = 0.125$.
* Ümumi ehtimal: $3 \cdot 0.125 = 0.375$. ✅

**Binomial formul:**

$$
P(X=2) = \binom{3}{2} (0.5)^2 (0.5)^1 = 3 \cdot 0.25 \cdot 0.5 = 0.375
$$

Üç ədəd oyun zarı eyni anda atılır. Tapın ehtimalı ki, **1, 2 və 3 ədədləri çıxsın, sıra önəmsizdir**.

## Müstəqil və  Asılı hadisələr



### **1️⃣ Müstəqil vs. Asılı hadisələr**

* Bu təsnifat **hadisələrin bir-birinə təsirinə görə** aparılır.
* **Müstəqil hadisələr:** Bir hadisənin baş verməsi **digər hadisəyə təsir etmir**.
* **Asılı hadisələr:** Bir hadisənin baş verməsi **digər hadisənin ehtimalına təsir edir**.

**Məsələlər:**

* İki zər atmaq → **müstəqil**
* Torbadan top çıxarmaq (geri qoymadan) → **asılı**

**Əsas əsası:** **hadisələrin baş vermə ehtimalları bir-birinə bağlıdırmı yoxsa yox?**

---

### **2️⃣ Uyğun (uyuşan) vs. Uyğun gəlməyən (uyuşmayan) hadisələr**

* Bu təsnifat **hadisələrin nəticələrinin bir-birinə uyğunluğuna görə** aparılır.
* **Uyğun hadisələr:** İki hadisə eyni nəticəyə malikdir və bir-birinə təsir etmədən baş verə bilər (müstəqil ola bilər).
* **Uyğun gəlməyən (uyuşmayan) hadisələr:** İki hadisə bir-birinin baş verməsinə mane olur, eyni vaxtda baş verə bilməz.

**Məsələlər:**

* Bir zər atıldığında “2 gəlmək” və “3 gəlmək” → uyğun gəlməyən, çünki eyni vaxtda hər ikisi olmaz
* Eyni zər atıldığında “cəmi 2 və ya 3 gəlməsi” → uyğun

**Əsas əsası:** **hadisələrin nəticələri üst-üstə düşürmü yoxsa biri digərini istisna edirmi?**

---

### **Xülasə**

| Təsnifat          | Əsas əsası                                     | Misal                                       |
| ----------------- | ---------------------------------------------- | ------------------------------------------- |
| Müstəqil / Asılı  | Hadisələrin baş verməsi bir-birinə təsir edir? | İki zər atmaq / torbadan top çıxarmaq       |
| Uyğun / Uyuşmayan | Hadisələrin nəticələri üst-üstə düşürmü?       | “2 gəlmək” və “3 gəlmək” / eyni zər üstündə |



## Uyuşan, Uyuşmayan, Cüt-cüt uyuşmayan hadisələr, Eyni ehtimallı hadisələr


### 🔹 1. **Uyuşan hadisələr (Kəsişən hadisələr)**  
İki hadisə **eyni anda baş verə bilərsə**, bu hadisələr **uyuşan** sayılır.  
**Riyazi işarəsi:**  
- A və B hadisələri üçün:  
  $$ A \cap B \neq \emptyset $$  
  (Yəni, A ilə B-nin kəsişməsi boş deyil.)

**Ehtimalı:**  
$$
P(A \cap B)
$$

**Nümunə:**  
A: Zar atıldıqda cüt ədəd gəlsin. → {2, 4, 6}  
B: Zar atıldıqda 3-dən böyük ədəd gəlsin. → {4, 5, 6}  
A ∩ B = {4, 6} → Uyuşan hadisələr.

---

### 🔹 2. **Uyuşmayan hadisələr (Disjoint / bir-birini istisna edən hadisələr)**  
İki hadisə **eyni anda baş verə bilməzsə**, onlar **uyușmayan** hadisələrdir.  
**Riyazi işarəsi:**  
- $A \cap B = \emptyset$

**Ehtimalı:**  
$$
P(A \cup B) = P(A) + P(B)
$$

**Nümunə:**  
A: Zar atıldıqda 2 gəlsin. → {2}  
B: Zar atıldıqda 5 gəlsin. → {5}  
A ∩ B = ∅ → Uyuşmayan hadisələr.

---

### 🔹 3. **Cüt-cüt uyuşmayan hadisələr (Pairwise disjoint events)**  
Bir neçə hadisə varsa və **hər bir iki hadisə arasında kəsişmə yoxdursa**, yəni **bütün cütlər arasında uyuşmayanlıq varsa**, bunlar **cüt-cüt uyuşmayan** adlanır.

**Riyazi işarəsi:**  
Hadisələr $A₁, A₂, ..., Aₙ$ üçün:
$$
\forall i \neq j,\quad A_i \cap A_j = \emptyset
$$

**Ehtimalı (toplam ehtimal):**  
$$
P\left(\bigcup_{i=1}^{n} A_i\right) = \sum_{i=1}^{n} P(A_i)
$$

**Nümunə:**  
$A₁$: Zar atıldıqda 1 gəlsin → {1}  
$A₂$: Zar atıldıqda 3 gəlsin → {3}  
$A₃$: Zar atıldıqda 6 gəlsin → {6}  
Bunlar cüt-cüt uyuşmayan hadisələrdir.

---

### 🔹 4. **Eyni ehtimallı hadisələr (Bərabər ehtimallı hadisələr)**  
Əgər bütün nəticələrin ehtimalı eynidirsə, onlar **eyni ehtimallı hadisələr** adlanır.

**Riyazi işarəsi:**  
Əgər hadisə A, bərabər ehtimallı $n$ nəticədən $m$-ni əhatə edirsə:
$$
P(A) = \frac{m}{n}
$$

**Nümunə:**  
Zar atıldıqda hər bir nəticənin (1, 2, 3, 4, 5, 6) ehtimalı eynidir:  
$$
P(1) = P(2) = \cdots = P(6) = \frac{1}{6}
$$

---

### 🔹 Vizual olaraq ümumi cədvəl:

| Hadisə tipi                 | Şərt                                      | İşarəsi                        | Ehtimal düsturu                          |
|----------------------------|--------------------------------------------|--------------------------------|------------------------------------------|
| Uyuşan                     | A və B birlikdə baş verə bilər             | $$ A \cap B \neq \emptyset $$   | $$ P(A \cap B) $$                          |
| Uyuşmayan                  | A və B eyni anda baş verə bilməz           | $$ A \cap B = \emptyset $$      | $$ P(A \cup B) = P(A) + P(B) $$            |
| Cüt-cüt uyuşmayan          | Hər iki hadisə cüt-cüt uyuşmur             | $$ A_i \cap A_j = \emptyset $$  | $$ P(\cup A_i) = \sum P(A_i) $$            |
| Eyni ehtimallı             | Bütün nəticələrin ehtimalı eynidir         | -                              | $$ P(A) = \frac{m}{n} $$                  |
