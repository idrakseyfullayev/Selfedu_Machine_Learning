## Məsəslə 1



**Məsələ.** Otaq iki eyni lampası olan bir işıqla işıqlandırılır. Bir lampanın bir il ərzində yanmaması ehtimalı **0,1**-dir. Eyni zamanda hər iki lampanın bir il ərzində yanmaması ehtimalı **0,05**-dir. Tapın ki, bir il ərzində ən azı bir lampa yanmasa ehtimalı.


````python
# Ehtimallar
P_A = 0.1       # Bir lampanın yanmaması ehtimalı
P_B = 0.1       # Digər lampanın yanmaması ehtimalı
P_A_and_B = 0.05  # Hər iki lampanın yanmaması ehtimalı

# Ən azı bir lampanın yanmaması
P_at_least_one = P_A + P_B - P_A_and_B
print("Ən azı bir lampanın yanmaması ehtimalı:", P_at_least_one)
````

## ✅ **Məsələ 1**

**Tərcümə:**
Kanselariya mağazasında 120 qələm var: 15-i qırmızı, 22-si yaşıl, 27-si bənövşəyi. Qalanları mavi və qara rəngdədir, onlar da bərabər paylanıb. Alisa təsadüfən bir qələm seçir. Onun mavi və ya yaşıl qələm götürmə ehtimalını tapın.

**Həll:**

* Ümumi qələm sayı: 120
* Qırmızı = 15, Yaşıl = 22, Bənövşəyi = 27
* Qalan = 120 − (15+22+27) = 56
* Deməli, mavi = 28, qara = 28
* Mavi və yaşıl qələm sayı = 28 + 22 = 50
* Ehtimal:

$$
P = \frac{50}{120} = \frac{5}{12}
$$

**Python ilə:**

```python
total = 120
red, green, violet = 15, 22, 27
rest = total - (red + green + violet)
blue = black = rest // 2
prob = (blue + green) / total
prob
```

## ✅ **Məsələ 3**

**Tərcümə:**
Geometriya imtahanında şagirdə təsadüfi sual düşür.

* "Romb" mövzusu üzrə ehtimal = 0.1
* "Çevrə yazılmış çoxbucaq" mövzusu üzrə ehtimal = 0.15
* İkisinə birdən aid olan ehtimal = 0.05
  Tapın: bu iki mövzudan hər hansı birinə aid sual düşmə ehtimalı.

**Həll:**
Burada kəsişmə var → düstur:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

$$
P = 0.1 + 0.15 - 0.05 = 0.2
$$

**Python ilə:**

```python
p_rombus, p_circle, p_both = 0.1, 0.15, 0.05
prob = p_rombus + p_circle - p_both
prob
```

## **İki hadisənin cəmi ehtimalı $P(A + B)$**



#### 1️⃣ Uyuşmayan (müstəqil olmayan, eyni vaxtda baş verməyən) hadisələr üçün:

Əgər $A$ və $B$ hadisələri **eyni vaxtda baş verə bilmirsə** (yəni bir-birini istisna edirlər), onda ehtimal belə hesablanır:

$$
P(A + B) = P(A) + P(B)
$$

**İzah:**
Məsələn, bir zər atılırsa, həm 2, həm də 5 gəlmək eyni anda mümkün deyil. Bu hadisələr bir-birinə uyğun gəlmir (müstəqil deyil). Ona görə də sadəcə ehtimalları toplamaq kifayətdir.

**Hesablama nümunəsi:**
Bir zər atılır. \$A\$: 2 gəlmə ehtimalı, \$B\$: 5 gəlmə ehtimalı.
$P(A) = \frac{1}{6}, \quad P(B) = \frac{1}{6}$
Çünki hadisələr eyni anda baş vermir:
$P(A + B) = P(A) + P(B) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}$

---

#### 2️⃣ Uyuşan (hər iki hadisə eyni vaxtda baş verə bilər) hadisələr üçün:

Əgər $A$ və $B$ hadisələri **eyni anda baş verə bilirsə**, onda ehtimal belə hesablanır:

$$
P(A + B) = P(A) + P(B) - P(AB)
$$

Burada $P(AB)$ — $A$ və $B$-nin **hər ikisinin eyni anda baş vermə ehtimalı**dır.

**İzah:**
Əgər hər iki hadisə eyni anda baş verə bilirsə, sadəcə $P(A) + P(B)$ etsək, $AB$ ehtimalını iki dəfə saymış olarıq. Ona görə onu bir dəfə çıxırıq.

**Məsələn:**

* Bir sinifdə oğlanların ehtimalı $P(A) = 0.6$
* Qırmızı köynək geyənlərin ehtimalı $P(B) = 0.3$
* Həm oğlan, həm qırmızı köynək geyənlərin ehtimalı $P(AB) = 0.2$

Onda oğlan və ya qırmızı köynək ehtimalı belə olacaq:

$$
P(A + B) = 0.6 + 0.3 - 0.2 = 0.7
$$
