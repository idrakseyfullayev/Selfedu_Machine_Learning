## Probability Məsələlər

### **Məsələ:**

Otaq 3 lampalı bir işıqla işıqlandırılır. Bir lampanın bir il ərzində xarab olma ehtimalı $P(\text{yanmaz}) = 0.3$. Tapın ehtimalı ki, il ərzində ən azı bir lampa işləsin (yəni yanmasın).

---

## **1️⃣ Ədəbiyyat (adi ehtimal) üsulu ilə**

Ən azı bir lampa işləsin ehtimalı:

$$
P(\text{ən azı bir lampa işləsin}) = 1 - P(\text{heç bir lampa işləməsin})
$$

Heç bir lampa işləməsin ehtimalı isə:

$$
P(\text{hamsı yanmaz}) = P(\text{1-ci yanmaz}) \cdot P(\text{2-ci yanmaz}) \cdot P(\text{3-cü yanmaz})
$$

Lampalar müstəqil olduğuna görə:

$$
P(\text{hamsı yanmaz}) = 0.3 \cdot 0.3 \cdot 0.3 = 0.027
$$

Deməli:

$$
P(\text{ən azı bir işləyir}) = 1 - 0.027 = 0.973
$$

✅ Cavab: **0.973**

---

## **2️⃣ Python ilə**

```python
# Lampaların yanmama ehtimalı
p_burnout = 0.3  # yanmama ehtimalı
n_lamps = 3

# Heç bir lampa işləməsin ehtimalı
p_none_work = p_burnout ** n_lamps

# Ən azı bir lampa işləsin ehtimalı
p_at_least_one = 1 - p_none_work

print("Ən azı bir lampa işləyəcək ehtimal:", p_at_least_one)
```

**Nəticə:**

```
Ən azı bir lampa işləyəcək ehtimal: 0.973
```

### **Məsələ 1**

**Tərcüməsi:**
Mağazada 3 satıcı var. Hər biri müştəri ilə məşğul olur ehtimalı 0,2-dir. Tapın ehtimalı ki, təsadüfi bir anda bütün 3 satıcı eyni anda məşğul olsun (tutaq ki, satıcılar bir-birindən müstəqildir).

**Adi həll:**
Hər satıcının müştəri ilə məşğul olma ehtimalı: $P(A) = 0.2$

3 satıcının eyni anda məşğul olma ehtimalı müstəqil hadisələr üçün:

$$
P(\text{hamısı məşğul}) = 0.2 \times 0.2 \times 0.2 = 0.008
$$

**Python ilə həll:**

```python
p = 0.2
prob_all_busy = p**3
prob_all_busy
```

**Nəticə:** `0.008`

### **Məsələ 2**

**Tərcüməsi:**
Əgər qrossmeyster A ağ fiqurlarla oynasa, o, qrossmeyster B-ni 0,6 ehtimalı ilə udur. Əgər A qara fiqurlarla oynasa, A B-ni 0,4 ehtimalı ilə udur. Hər iki oyunçu iki oyun oynayır və ikinci oyunda rənglərini dəyişirlər. Tapın ehtimalı ki, A hər iki oyunu udacaq.

**Adi həll:**

* Birinci oyunda A ağ oynayır: $P_1 = 0.6$
* İkinci oyunda A qara oynayır: $P_2 = 0.4$

Hər iki oyunu udma ehtimalı (müstəqil hadisələr üçün):

$$
P(\text{A udur hər iki oyun}) = 0.6 \times 0.4 = 0.24
$$

**Python ilə həll:**

```python
p_first_game = 0.6
p_second_game = 0.4
prob_win_both = p_first_game * p_second_game
prob_win_both
```

**Nəticə:** `0.24`

### **Məsələ 3**

**Tərcüməsi:**
Bir ərazidə müşahidə edilmişdir:

1. İyun səhəri günəşli olarsa, o gün yağışın ehtimalı 0,1-dir.
2. İyun səhəri buludlu olarsa, o gün yağış ehtimalı 0,4-dür.
3. İyun səhəri buludlu olma ehtimalı 0,3-dür.

Tapın ehtimalı ki, təsadüfi seçilmiş bir iyun günü yağış yağmayacaq.

**Adi həll:**

* Səhər günəşli olma ehtimalı: $P(\text{sun}) = 1 - 0.3 = 0.7$
* Səhər buludlu olma ehtimalı: $P(\text{cloud}) = 0.3$
* Yağış olmama ehtimalı:

$$
P(\text{no rain}) = P(\text{sun}) \cdot (1 - 0.1) + P(\text{cloud}) \cdot (1 - 0.4)
$$

$$
P(\text{no rain}) = 0.7 \cdot 0.9 + 0.3 \cdot 0.6 = 0.63 + 0.18 = 0.81
$$

**Python ilə həll:**

```python
p_sun = 0.7
p_cloud = 0.3
p_rain_given_sun = 0.1
p_rain_given_cloud = 0.4

prob_no_rain = p_sun*(1 - p_rain_given_sun) + p_cloud*(1 - p_rain_given_cloud)
prob_no_rain
```

**Nəticə:** `0.81`

### **Məsələ 4**

**Tərcüməsi:**
Kovboy Con divardakı milçəyə atəş açır. Əgər o, atəş açarsa, tapançanın atəşlənmə ehtimalı 0,7-dir. Əgər Con atəş açmaq üçün qüsursuz olmayan tapança istifadə edirsə, milçəyə dəymə ehtimalı 0,3-dür. Masada 10 tapança var, onlardan yalnız 2-si atəşlənmişdir. Con divarda milçəni görür, təsadüfi bir tapança götürür və atəş açır. Tapın ehtimalı ki, Con miss etsin.

**Adi həll:**

* Təsadüfi tapança atəşlənmiş ehtimalı: $2/10 = 0.2$
* Təsadüfi tapança atəşlənməyəcək ehtimalı: $0.8$

Ehtimal ki, Con milçəyə dəyməyəcək:

1. Tapança atəşlənməsə (0.8): mütləq miss = 1
2. Tapança atəşlənmişsə (0.2): miss ehtimalı = $1 - 0.7 \cdot 0.3 = 1 - 0.21 = 0.79$

Tam ehtimal:

$$
P(\text{miss}) = 0.8 \cdot 1 + 0.2 \cdot 0.79 = 0.8 + 0.158 = 0.958
$$

**Python ilə həll:**

```python
p_loaded = 2/10
p_not_loaded = 1 - p_loaded
p_hit_if_loaded = 0.7 * 0.3  # Con atəş açdı və milçəyə dəydi ehtimalı

prob_miss = p_not_loaded*1 + p_loaded*(1 - p_hit_if_loaded)
prob_miss
```

**Nəticə:** `0.958`



## Müstəqil və Asılı (müstəqil olmayan) hadisələr



## 🔹 1. Müstəqil hadisələr

İki hadisə **bir-birinə təsir etmirsə**, onlar müstəqildir.
Yəni bir hadisənin baş verib-verməməsi digərinin ehtimalını dəyişmir.

👉 Riyazi tərif:

$$
P(A \cap B) = P(A) \cdot P(B)
$$

**Misal:**

* Bir zər atılır və bir sikkə atılır.

  * \$A\$: Zərdə 3 düşsün → \$P(A)=\tfrac{1}{6}\$
  * \$B\$: Sikkədə yazı düşsün → \$P(B)=\tfrac{1}{2}\$
    Burada \$A\$ və \$B\$ müstəqildir, çünki birinin nəticəsi digərinə təsir etmir:

  $$
  P(A \cap B) = \tfrac{1}{6} \cdot \tfrac{1}{2} = \tfrac{1}{12}
  $$

---

## 🔹 2. Asılı (müstəqil olmayan) hadisələr

Əgər bir hadisənin baş verməsi digərinin ehtimalını dəyişirsə, hadisələr **asılıdır**.

👉 Riyazi tərif:

$$
P(A \cap B) = P(A) \cdot P(B|A)
$$

burada \$P(B|A)\$ – \$A\$ baş verəndə \$B\$ hadisəsinin şərti ehtimalıdır.

**Misal:**
52 kartdan ibarət göyərtədən iki kart çıxarılır.

* \$A\$: Birinci kart qırmızı olsun.
* \$B\$: İkinci kart qırmızı olsun.

Əgər birinci kart geri qoyulmazsa, onda ikinci kartın ehtimalı birincidən asılı olur.

$$
P(A) = \tfrac{26}{52} = \tfrac{1}{2}, \quad
P(B|A) = \tfrac{25}{51}
$$

$$
P(A \cap B) = P(A) \cdot P(B|A) = \tfrac{1}{2} \cdot \tfrac{25}{51} = \tfrac{25}{102}
$$

## 🔑 Qısa müqayisə

| Hadisə tipi  | Formula                          | Məsələyə təsiri              |                                            |
| ------------ | -------------------------------- | ---------------------------- | ------------------------------------------ |
| **Müstəqil** | $$P(A \cap B) = P(A)\cdot P(B)$$ | Bir hadisə digərini dəyişmir |                                            |
| **Asılı**    | $$P(A \cap B) = P(A)\cdot P(B | A)$$ | Bir hadisə digərinin ehtimalına təsir edir |
