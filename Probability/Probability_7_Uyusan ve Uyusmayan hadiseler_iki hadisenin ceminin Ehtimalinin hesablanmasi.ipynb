## Uyuşan və Uyuşmayan hadisələr, iki hadisənin cəminin Ehtimalının hesablanması.

```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Selfedu\Probability\Pictures\Probability_7_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 2):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 100:
        plt.figure(figsize=(10, 4))
    else:     
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
    plt.title('Uyuşan və Uyuşmayan hadisələr, iki hadisənin cəminin Ehtimalının hesablanması', size='12')    
    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show() 
```

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

$P = P(A) \cdot P(B) + P(B) \cdot P(A)$

✅ **Əhəmiyyətli fərq**:

**Bu formula “birinci zər 2 gəlir və ya ikinci zər 3 gəlir”** ehtimalını verir.


$$
\boxed{P(A \cup B) = P(A) + P(B) - P(A \cap B)}
$$

* Burada $A$ və $B$ iki hadisəni göstərir.
* $A \cup B$ = “A və ya B baş verir”.
* $P(A \cap B)$ = “A və B eyni anda baş verir”.



**Bu formula isə birində 2, digərində 3 gəlməsi**

$$
\boxed{P = P(A) \cdot P(B) + P(B) \cdot P(A)}
$$

Burada:

* $A$ = “zər 2 gəlir”
* $B$ = “zər 3 gəlir”

Bu, “birində 2, digərində 3 gəlməsi” ehtimalını verir.
