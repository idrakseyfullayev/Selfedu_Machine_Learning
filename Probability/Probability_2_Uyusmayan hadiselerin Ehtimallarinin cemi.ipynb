## Probability Məsələlər



### **Məsələ 1**

**Rusca:**
На экзамене 60 билетов, Андрей не выучил 3 из них. Найдите вероятность того, что ему попадется выученный билет.

**Azərbaycan dilində:**
İmtahanda 60 biletdən 3-nü Andrey öyrənməyib. Tapın ehtimal ki, ona öyrəndiyi bilet düşsün.

**Həll:**

* Öyrənilməyən biletlərin sayı: 3
* Öyrənilmiş biletlərin sayı: $60 - 3 = 57$
* Ehtimal:

$$
P = \frac{\text{öyrənilmiş biletlərin sayı}}{\text{ümumi biletlərin sayı}} = \frac{57}{60} = 0.95
$$

**Cavab:** 0.95 (yəni 95%)

---

### **Məsələ 2**

**Rusca:**
На тарелке 16 пирожков: 8 с мясом, 3 с яблоками и 5 с луком. Настя наугад выбирает один пирожок. Найдите вероятность того, что он окажется с мясом.

**Azərbaycan dilində:**
Tabaqda 16 pirojki var: 8 ətl, 3 alma və 5 soğanlı. Nastya təsadüfi olaraq birini götürür. Tapın ehtimal ki, o, ətli pirojki götürsün.

**Həll:**

* Ümumi pirojk sayı: 16
* Ətlilərin sayı: 8
* Ehtimal:

$$
P = \frac{8}{16} = 0.5
$$

**Cavab:** 0.5 (yəni 50%)

---

### **Məsələ 3**

**Rusca:**
Игральную кость (кубик) бросили один раз. Какова вероятность того, что выпало очков не меньше, чем 3?

**Azərbaycan dilində:**
Təsadüfi bir zar atılır. Tapın ehtimal ki, 3 və ya daha çox nöqtə düşsün.

**Həll:**

* Zarın mümkün nəticələri: 1, 2, 3, 4, 5, 6
* 3 və daha böyük: 3, 4, 5, 6 → 4 nəticə
* Ehtimal:

$$
P = \frac{4}{6} = \frac{2}{3} \approx 0.667
$$

**Cavab:** \~0.667 (yəni 66.7%)

---

### **Məsələ 4**

**Rusca:**
Игральный кубик бросили один раз. Какова вероятность того, что выпало очков больше, чем 4?

**Azərbaycan dilində:**
Təsadüfi bir zar atılır. Tapın ehtimal ki, 4-dən çox nöqtə düşsün.

**Həll:**

* 4-dən böyük nöqtələr: 5, 6 → 2 nəticə
* Ehtimal:

$$
P = \frac{2}{6} = \frac{1}{3} \approx 0.333
$$

**Cavab:** \~0.333 (yəni 33.3%)

---

### **Məsələ 5**

**Rusca:**
Вероятность того, что новая шариковая ручка пишет плохо (или не пишет), равна 0.1. Покупатель в магазине выбирает одну ручку. Найдите вероятность того, что эта ручка пишет хорошо.

**Azərbaycan dilində:**
Yeni qələm pis yazır (və ya yazmır) ehtimalı 0.1-dir. Mağazada müştəri bir qələm seçir. Tapın ehtimal ki, bu qələm yaxşı yazsın.

**Həll:**

* Pis yazma ehtimalı: 0.1
* Yaxşı yazma ehtimalı (tamamlayıcı hadisə):

$$
P_{\text{yaxşı}} = 1 - 0.1 = 0.9
$$

**Cavab:** 0.9 (yəni 90%)




## Uyuşan, Uyuşmayan, Cüt-cüt uyuşmayan hadisələr, Eyni ehtimallı hadisələr


### 🔹 1. **Uyuşan hadisələr (Kəsişən hadisələr)**  
İki hadisə **eyni anda baş verə bilərsə**, bu hadisələr **uyuşan** sayılır.  
**Riyazi işarəsi:**  
- A və B hadisələri üçün:  
  $$ A \cap B \neq \emptyset $$  
  (Yəni, A ilə B-nin kəsişməsi boş deyil.)

**Ehtimalı:**  
$$
P(A \cap B)
$$

**Nümunə:**  
A: Zar atıldıqda cüt ədəd gəlsin. → {2, 4, 6}  
B: Zar atıldıqda 3-dən böyük ədəd gəlsin. → {4, 5, 6}  
A ∩ B = {4, 6} → Uyuşan hadisələr.

---

### 🔹 2. **Uyuşmayan hadisələr (Disjoint / bir-birini istisna edən hadisələr)**  
İki hadisə **eyni anda baş verə bilməzsə**, onlar **uyușmayan** hadisələrdir.  
**Riyazi işarəsi:**  
- $A \cap B = \emptyset$

**Ehtimalı:**  
$$
P(A \cup B) = P(A) + P(B)
$$

**Nümunə:**  
A: Zar atıldıqda 2 gəlsin. → {2}  
B: Zar atıldıqda 5 gəlsin. → {5}  
A ∩ B = ∅ → Uyuşmayan hadisələr.

---

### 🔹 3. **Cüt-cüt uyuşmayan hadisələr (Pairwise disjoint events)**  
Bir neçə hadisə varsa və **hər bir iki hadisə arasında kəsişmə yoxdursa**, yəni **bütün cütlər arasında uyuşmayanlıq varsa**, bunlar **cüt-cüt uyuşmayan** adlanır.

**Riyazi işarəsi:**  
Hadisələr $A₁, A₂, ..., Aₙ$ üçün:
$$
\forall i \neq j,\quad A_i \cap A_j = \emptyset
$$

**Ehtimalı (toplam ehtimal):**  
$$
P\left(\bigcup_{i=1}^{n} A_i\right) = \sum_{i=1}^{n} P(A_i)
$$

**Nümunə:**  
$A₁$: Zar atıldıqda 1 gəlsin → {1}  
$A₂$: Zar atıldıqda 3 gəlsin → {3}  
$A₃$: Zar atıldıqda 6 gəlsin → {6}  
Bunlar cüt-cüt uyuşmayan hadisələrdir.

---

### 🔹 4. **Eyni ehtimallı hadisələr (Bərabər ehtimallı hadisələr)**  
Əgər bütün nəticələrin ehtimalı eynidirsə, onlar **eyni ehtimallı hadisələr** adlanır.

**Riyazi işarəsi:**  
Əgər hadisə A, bərabər ehtimallı $n$ nəticədən $m$-ni əhatə edirsə:
$$
P(A) = \frac{m}{n}
$$

**Nümunə:**  
Zar atıldıqda hər bir nəticənin (1, 2, 3, 4, 5, 6) ehtimalı eynidir:  
$$
P(1) = P(2) = \cdots = P(6) = \frac{1}{6}
$$

---

### 🔹 Vizual olaraq ümumi cədvəl:

| Hadisə tipi                 | Şərt                                      | İşarəsi                        | Ehtimal düsturu                          |
|----------------------------|--------------------------------------------|--------------------------------|------------------------------------------|
| Uyuşan                     | A və B birlikdə baş verə bilər             | $$ A \cap B \neq \emptyset $$   | $$ P(A \cap B) $$                          |
| Uyuşmayan                  | A və B eyni anda baş verə bilməz           | $$ A \cap B = \emptyset $$      | $$ P(A \cup B) = P(A) + P(B) $$            |
| Cüt-cüt uyuşmayan          | Hər iki hadisə cüt-cüt uyuşmur             | $$ A_i \cap A_j = \emptyset $$  | $$ P(\cup A_i) = \sum P(A_i) $$            |
| Eyni ehtimallı             | Bütün nəticələrin ehtimalı eynidir         | -                              | $$ P(A) = \frac{m}{n} $$                  |



## Ehtimal məntiqi



### **1️⃣ Bir zər atıldıqda 2 və ya 3 gəlmək ehtimali**

* Bir zər atıldığında nəticələr: 1, 2, 3, 4, 5, 6 → cəmi **6 nəticə**.
* İstədiyimiz nəticələr: 2 **və ya** 3 → cəmi 2 nəticə.

Bu hadisələr **üst-üstə düşmür** (eyni vaxtda həm 2, həm 3 olmaz), ona görə **toplama qaydasını** istifadə edirik:

$$
P(2 \text{ və ya } 3) = P(2) + P(3) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}
$$

✅ Cavab: $\frac{1}{3}$

---

### **2️⃣ İki zər atıldıqda hər ikisində 2 gəlmək ehtimali**

* İki zər müstəqildir, yəni bir zərin nəticəsi digərinə təsir etmir.
* Hər zərdə **2 gəlmək ehtimali** = $\frac{1}{6}$

İki zərdə də 2 gəlməsi üçün **vurma qaydasını** istifadə edirik:

$$
P(\text{birinci 2 və ikinci 2}) = \frac{1}{6} \times \frac{1}{6} = \frac{1}{36}
$$

✅ Cavab: $\frac{1}{36}$

---

### **3️⃣ İki zər atıldıqda birincidə 2 və ikincidə 3 gəlmək ehtimali**

* Birinci zərdə **2 gəlməsi ehtimali** = $\frac{1}{6}$
* İkinci zərdə **3 gəlməsi ehtimali** = $\frac{1}{6}$

Hadisələr müstəqil olduğundan vurma qaydasını tətbiq edirik:

$$
P(\text{birinci 2, ikinci 3}) = \frac{1}{6} \times \frac{1}{6} = \frac{1}{36}
$$

✅ Cavab: $\frac{1}{36}$

---

### **4️⃣ İki zər atıldıqda birində 2 və digərində 3 gəlmək ehtimali** (sıralama fərqi nəzərə alınmır)

* İki mümkün hal var: (2,3) və (3,2)
* Hər bir halın ehtimalı: $\frac{1}{36}$ (bax 3-cü hal kimi)

Toplama qaydasını tətbiq edirik:

$$
P(\text{birində 2 və birində 3}) = \frac{1}{36} + \frac{1}{36} = \frac{2}{36} = \frac{1}{18}
$$

✅ Cavab: $\frac{1}{18}$

---

## Ehtimal məntiqi



### 1️⃣ Bir zər atmaq prosesi iki dəfə baş verib, 2 və ya 3 gəlmək ehtimalı

* Bir zər üçün 2 və ya 3 gəlmə ehtimalı: $1/3$ (əvvəlki hesablamadan)
* İki dəfə bu prosesi etdikdə: hər iki zər müstəqildir, ehtimal “hər ikisi 2 və ya 3 gəlmə”:

$$
P(\text{hər iki zər 2 və ya 3}) = \frac{1}{3} \times \frac{1}{3} = \frac{1}{9} \approx 0.111
$$

---

### 2️⃣ İki zər atmaq prosesi iki dəfə baş verib, hər ikisində 2 gəlmək ehtimalı

* Bir prosesi iki zər atmaq: ehtimal $1/36$ (əvvəlki hesablamadan)
* İki dəfə bu prosesi etdikdə (yəni iki dəstə iki zər):

$$
P(\text{hər iki prosesdə hər ikisi 2}) = \frac{1}{36} \times \frac{1}{36} = \frac{1}{1296} \approx 0.0007716
$$

---

### 3️⃣ İki zər atmaq prosesi iki dəfə baş verib, birinci 2 və ikinci 3 gəlmək ehtimalı

* Bir proses üçün ehtimal: $1/36$
* İki dəfə baş verəndə:

$$
P(\text{hər iki prosesdə birinci 2 və ikinci 3}) = \frac{1}{36} \times \frac{1}{36} = \frac{1}{1296} \approx 0.0007716
$$

---

### 4️⃣ İki zər atmaq prosesi iki dəfə baş verib, birində 2 və digərində 3 gəlmək ehtimalı

* Bir proses üçün ehtimal: $1/18$
* İki dəfə baş verəndə:

$$
P(\text{hər iki prosesdə birində 2 və birində 3}) = \frac{1}{18} \times \frac{1}{18} = \frac{1}{324} \approx 0.003086
$$

---
