## Probability Məsələ



**Rusca mətn:**
*Задача. Бросается игральный кубик, имеющий шесть абсолютно симметричных граней. Определите вероятность того, что при бросании этого кубика выпадет число 3.*

---

### Azərbaycan dilinə tərcümə:

**Məsələ.** Altı tamamilə simmetrik üzü olan zər atılır. Zəri atarkən 3 rəqəminin düşmə ehtimalını müəyyən edin.

---

### Riyazi həll:

Altıüzlü zərdə hər üzü eyni ehtimalla düşür.

* Ümumi nəticələrin sayı: $6$
* Tələb olunan nəticələrin sayı: $1$ (yalnız “3”)

$$
P = \frac{1}{6}
$$



```python
# Ümumi nəticələrin sayı
total_outcomes = 6

# 3 düşməsi üçün uğurlu nəticələrin sayı
favorable_outcomes = 1

# Ehtimalın hesablanması
probability = favorable_outcomes / total_outcomes
print("3 düşmə ehtimalı:", probability)
```
