## Məhsulun bazara girmək ehtimalı analizi: Z score və Qaus ehtimal sıxlığını (PDF) 



## ✅ **Bazara girmək ehtimalı analizi**

### **Verilənlər**:
- **Gelir1**: 1900  
- **Gelir2**: 3000  
- **Orta bazar gəliri (mean, μ)**: 2200  
- **Standart sapma (σ)**: 500  

---

## 🔹 **1) Z-score ilə hesabla və faizini tap**

### **Z-score düsturu**:
$$
z = \frac{X - \mu}{\sigma}
$$

---

### **a) Gelir = 1900 üçün**:
$$
z = \frac{1900 - 2200}{500} = -0.6
$$

Z-score cədvəlindən:  
$$
P(Z < -0.6) ≈ 0.2743 = 27.43\%
$$

> 📌 **Interpretasiya**:  
Bu məhsulun **ortalama bazar gəlirindən daha aşağı gəlir gətirmə ehtimalı 27.43%**-dir.  
Bu o deməkdir ki, bu gəlir bazarda çox da nadir deyil – **təxminən hər 4 məhsuldan 1-i** bu səviyyədə və ya daha aşağı gəlir gətirir.

---

### **b) Gelir = 3000 üçün**:
$$
z = \frac{3000 - 2200}{500} = 1.6
$$

Z-score cədvəlindən:
$$
P(Z > 1.6) = 1 - P(Z < 1.6) = 1 - 0.9452 = 0.0548 = 5.48\%
$$

> 📌 **Interpretasiya**:  
Bu məhsulun **ortalama gəlirdən daha yüksək gəlir gətirmə ehtimalı yalnız 5.48%**-dir.  
Bu gəlir **çox yüksək və bazarda nadir hallarda rast gəlinən gəlir səviyyəsidir** – **100 məhsuldan yalnız 5-6-sı** bu səviyyəyə çata bilir.

---

## 🔹 **2) Qaus ehtimal sıxlığını (PDF) tap və faizlə ifadə et**

Normal paylanmada ehtimal sıxlığı funksiyası (PDF):
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }
$$

---

### **a) Gelir = 1900 üçün**:
$$
z = -0.6 \Rightarrow f(1900) ≈ 0.000798 \Rightarrow 0.0798\%
$$

> 📌 **Interpretasiya**:  
Bu dəyər bizə göstərir ki, **1900 gəliri olan məhsul bazar daxilində təxminən 0.08% sıxlıqla rast gəlinir**.  
Yəni, bu gəlir **nə çox nadirdir, nə də çox sıxdır**, bazarın **orta aşağı hissəsində yerləşir**.

---

### **b) Gelir = 3000 üçün**:
$$
z = 1.6 \Rightarrow f(3000) ≈ 0.000216 \Rightarrow 0.0216\%
$$

> 📌 **Interpretasiya**:  
Bu məhsulun gəliri **bazar daxilində yalnız 0.0216% sıxlıqla rast gəlinir**, yəni **çox nadir gəlir səviyyəsidir**.  
Bazarın əsas hissəsi bu qədər yüksək gəliri görmür.

---

## 🔚 **Nəticə**:

- **1900** gəliri olan məhsul bazarda **normaldan bir qədər aşağı gəlirli** məhsul sayılır – çox da nadir deyil.
- **3000** gəliri olan məhsul isə **bazar üçün yüksək performanslı və nadir məhsuldur** – həm Z-score, həm də ehtimal sıxlığı ilə bu təsdiqlənir.

---

Hazırsansa, bu nəticələri **normal əyri** üzərində vizual formada da göstərə bilərəm – maraqlıdırsa, sadəcə “bəli” de 😎

##  İnterpretasiya



Çox gözəl sualdır! Gəlin **1900 AZN** və **3000 AZN** gəlirli məhsulları müqayisə edək, həm **risk**, həm də **potensial** baxımından:

---

### 🔍 **Z-score baxımından müqayisə**:
- **1900 AZN**: $Z = -0.6$ → bazar ortalamasından *bir qədər aşağıdır*, amma çox da uzaq deyil.
- **3000 AZN**: $Z = 1.6$ → bazar ortalamasından *çox yuxarıdır*, amma **çox nadirdir**.

---

### 🎯 **Ehtimal baxımından**:
- **1900 gəlirin baş vermə ehtimalı**: **27.43%** → kifayət qədər real ehtimaldır.
- **3000 gəlirin baş vermə ehtimalı**: **5.48%** → çox aşağı ehtimaldır.

---

### 📊 **Qaus ehtimal sıxlığı (PDF) baxımından**:
- **1900 gəlirin sıxlığı**: 0.0798% → bu gəlir səviyyəsi **bazarın "normal" aşağı hissəsində yerləşir**.
- **3000 gəlirin sıxlığı**: 0.0216% → **bazarın çox nadir yüksək hissəsidir**.

---

### 🧠 **Nəticə (Hansı daha yaxşıdır?)**

| Mehsul | Gəlir | Ehtimal | Risk | Tövsiyə |
|--------|--------|-----------|------|----------|
| A | 1900 AZN | 27.43% | Aşağı | **Daha stabil və real** nəticə verir |
| B | 3000 AZN | 5.48% | Yüksək | **Yüksək qazanc, amma risklidir** |

---

### ✅ **Yekun qərar**:
- Əgər **riskdən qorxursansa və sabit gəlir istəyirsənsə**, **1900** daha real və məqsədəuyğundur.
- Əgər **riskə getməyə hazırsansa və böyük gəlir üçün şans axtarırsansa**, **3000** daha cəlbedicidir, amma ehtimalı aşağıdır.

---

İstəyirsənsə, mən **expected value (gözlənilən gəlir)** ilə bu iki variantı müqayisə edə bilərəm – bu daha dəqiq iqtisadi qərar verməyə kömək edər. Deyim onu da hesablayım?

---

Bu şəkildə istədiyiniz formatda təqdim etdim. Əgər əlavə dəyişiklər etmək istəsəniz, mənə bildirə bilərsiniz.

## Qrafiklər

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Verilənlər
mu = 2200  # orta gəlir
sigma = 500  # standart sapma
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Gəlir nöqtələri
income_1900 = 1900
income_3000 = 3000

# Z-score-lar
z_1900 = (income_1900 - mu) / sigma
z_3000 = (income_3000 - mu) / sigma

# Plot
plt.figure(figsize=(8, 5))  # Qrafik ölçüsünü kiçiltmək üçün dəyişiklik
plt.plot(x, y, label='Normal Paylanma (μ=2200, σ=500)', color='darkblue')

# Şaquli xətlər
plt.axvline(income_1900, color='orange', linestyle='--', label=f'1900 AZN (Z={z_1900:.2f})')
plt.axvline(income_3000, color='green', linestyle='--', label=f'3000 AZN (Z={z_3000:.2f})')
plt.axvline(mu, color='red', linestyle='-', label=f'Orta Gəlir (μ={mu})')  # Orta gəlir xətti

# Annotasiyalar
plt.text(income_1900, stats.norm.pdf(income_1900, mu, sigma)+0.00001, '1900 AZN', color='orange', ha='center')
plt.text(income_3000, stats.norm.pdf(income_3000, mu, sigma)+0.00001, '3000 AZN', color='green', ha='center')
plt.text(mu, stats.norm.pdf(mu, mu, sigma)+0.00001, f'Orta Gəlir ({mu})', color='red', ha='center')

# Ayarlar
plt.title('Gəlir üzrə Normal Paylanma və Məhsul Mövqeləri')
plt.xlabel('Gəlir (AZN)')
plt.ylabel('Ehtimal sıxlığı')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Yuxarıdakı qrafikdə:

- **Mavi əyri** – bazar gəlirlərinin **normal paylanmasını** göstərir (ortalama 2200, $\sigma=500$).
- **Narıncı xətt (1900 AZN)** – ortalamadan bir qədər aşağıda yerləşir, bazarda **qismən rast gəlinən** gəlirdir.
- **Yaşıl xətt (3000 AZN)** – ortalamadan çox uzaqda, **çox nadir** və yüksək gəlirli məhsuldur.
- **Qırmızı xətt (Ortalama 2200 AZN)** – bazarın **orta gəlir** nöqtəsini göstərir.

Bu vizual sadəcə sözlə deyil, **rəqəmlərlə də hiss etdiyin fərqi** göstərir. İstəsən bu əraziyə uyğun ehtimalları (altında qalan sahə) da rəngləyib göstərə bilərəm.
