#### Likelihood,  MLE (Maximum Likelihood), Prior,  Posterior (Bayes), Bayes-də MLE haradadır? - $\theta$ parameter baxımından



#### 1️⃣ Likelihood

$$
P(Y \mid \theta)
$$

* Data sabitdir, parametr dəyişəndir.
> **👉 “θ doğru olsaydı, bu data nə qədər mümkündür?”**

---

#### 2️⃣ MLE (Maximum Likelihood)

$$
\hat{\theta}_{\text{MLE}} = \arg\max_{\theta} P(Y \mid \theta)
$$

> **👉 Likelihood-u maksimum edən parametr.**

* Yəni:
* **Sadəcə data ilə işləyir, prior yoxdur.**

---

#### 3️⃣ Prior

$$
P(\theta)
$$

> **👉 Datadan əvvəl parametr haqqında inanc.**

---

#### 4️⃣ Posterior (Bayes)

$$
P(\theta \mid Y) = \frac{P(Y \mid \theta) \ P(\theta)}{P(Y)}
$$

və ya qısa forma:

$$
Posterior \propto Likelihood \times Prior
$$

> **👉 Data + əvvəlki inanc birlikdə.**

---

#### 5️⃣ Bayes-də MLE haradadır?

Əgər prior sabitdirsə (uniform):

$$
MAP = MLE
$$

MAP:

$$
\hat{\theta}_{\text{MAP}} = \arg\max_{\theta} P(Y \mid \theta) \ P(\theta)
$$

---

#### Qısa müqayisə

* **MLE** → yalnız likelihood
* **Bayes** → likelihood + prior
* **Posterior** → yenilənmiş paylanma

---

$$
$$

#### Nümunə

> **Tutaq ki: 10 atış → 7 Heads, 3 Tails, və prior $ \theta \sim Beta(1,1) $ (uniform prior).**

$$
$$

#### 1️⃣ Likelihood (Binomial)

$$
P(Y \mid \theta) = \theta^7 (1-\theta)^3
$$

$$
$$

#### 2️⃣ MLE

$$
\hat{\theta}_{MLE} = \frac{7}{10} = 0.7
$$

(Yəni likelihood-u maksimum edən ehtimal 0.7-dir)

$$
$$

#### 3️⃣ Bayes (Prior + Posterior)

Prior:

$$
\theta \sim Beta(\alpha=1, \beta=1)
$$

Posterior:

$$
\theta \mid Y \sim Beta(\alpha+7, \beta+3) = Beta(8,4)
$$

* Posterior mean:

$$
E[\theta \mid Y] = \frac{8}{8+4} = \frac{8}{12} \approx 0.667
$$

> Yeni inanc = Köhnə prior + Data



#### Likelihood,  MLE (Maximum Likelihood), Prior,  Posterior (Bayes), Bayes-də MLE haradadır? - Bayes Classification baxımından

**Bayes Classification** üçün $P(Y \mid X)$ formasında yazaq. Burada hədəf dəyişənimiz $Y$, xüsusiyyətlərimiz isə $X$ olacaq:

$$
$$

#### 1️⃣ Likelihood (X verilmiş Y)

$$
P(X \mid Y)
$$

* Data sabitdir, klass $Y$ dəyişəndir.
> **👉 “Əgər klass $Y$ doğru olsaydı, bu xüsusiyyətlər $X$ nə qədər mümkündür?”**

---

#### 2️⃣ MLE (Maximum Likelihood Estimation)

$$
\hat{P}(X \mid Y)_{\text{MLE}} = \text{likelihood-u maksimum edən Class}
$$

> **👉 Likelihood-u maksimum edən Class.**
* Yəni: **yalnız data ilə işləyir, prior yoxdur.**

---

#### 3️⃣ Prior (klass ehtimalı)

$$
P(Y)
$$

> **👉 Class $Y$ haqqında əvvəlki inanc.**

---

#### 4️⃣ Posterior (Bayes Theorem)

$$
P(Y \mid X) = \frac{P(X \mid Y) \ P(Y)}{P(X)}
$$

və ya qısa formada:

$$
Posterior \propto Likelihood \times Prior
$$

> **👉 $X$ məlumatı + əvvəlki inanc birlikdə.**

---

#### 5️⃣ Bayes-də MLE haradadır?

Əgər prior sabitdirsə (hər klass eyni ehtimallı):

$$
MAP = MLE
$$

MAP:

$$
\hat{Y}_{\text{MAP}} = \arg\max_{Y} P(X \mid Y) \, P(Y)
$$

> **Sabit prior → sadəcə likelihood-u maksimum edən Class seçilir (yəni MLE).**

---

#### Qısa müqayisə

* **MLE** → yalnız $P(X \mid Y)$
* **Bayes** → $P(X \mid Y) \times P(Y)$
* **Posterior** → yenilənmiş $P(Y \mid X)$

---

$$
$$

#### Nümunə

> **Tutaq ki, 10 nümunə var: 7-si klass A, 3-ü klass B, prior hər ikisi üçün 0.5-dir.**

$$
$$

#### 1️⃣ Likelihood (X verilmiş Y)

$$
P(X \mid Y=A) = \frac{7!}{7!0!} p^7 (1-p)^0 = 0.7^7 \cdot 0.3^3 \approx 0.0082,
$$

$$
P(X \mid Y=B)  = \frac{3!}{3!0!} q^3 (1-q)^0 = 0.3^7 \cdot 0.7^3 \approx 0.00018
$$


* 👉 “Əgər klass A doğru olsaydı, bu 7 müşahidə nə qədər mümkündür?”

$$
$$

#### 2️⃣ MLE

$$
\hat{p} {MLE} = \frac{7}{10} = 0.7, \quad \hat{q} {MLE} = \frac{3}{10} = 0.3
$$

* Yəni likelihood-u maksimum edən klass ehtimalları.

$$
$$

#### 3️⃣ Bayes (Prior + Posterior)

Prior:

$$
P(Y=A) = 0.5, \quad P(Y=B) = 0.5
$$

Posterior:

$$
P(Y=A \mid X) \propto P(X \mid Y=A) \cdot P(Y=A) = 0.0082 \cdot 0.5 = 0.0041
$$

$$
P(Y=B \mid X) \propto P(X \mid Y=B) \cdot P(Y=B) = 0.00018 \cdot 0.5 = 0.00009
$$

Normalizasiya:

$$
P(Y=A \mid X) = \frac{0.0041}{0.0041+0.00009} \approx 0.9785
$$

$$
P(Y=B \mid X) = \frac{0.00009}{0.0041+0.00009} \approx 0.0215
$$

> Yeni inanc = Köhnə prior × Data


✅ Nəticə: 7/10 müşahidəyə görə **klass A-nın posterior ehtimalı ~97.85%**, klass B-nin isə ~2.15%.
