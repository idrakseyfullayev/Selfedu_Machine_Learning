### Accuracy 

> **Accuracy (dəqiqlik)** — **modelin düzgün təsnif etdiyi nümunələrin bütün nümunələrə nisbətidir. Yəni, modelin nə qədər “doğru” işlədiyini göstərir.**

> **Accuracy (dəqiqlik) maşın öyrənməsi və statistika sahəsində ən sadə performans ölçülərindən biridir. O, modelin düzgün proqnoz verdiyi halların ümumi hallara nisbətini göstərir.**

<br>

**Tərifi:**

$$
\text{Accuracy} = \frac{\text{Düzgün təsnif olunan nümunələrin sayı}}{\text{Ümumi nümunələrin sayı}}
$$

<br>

**Confuison Matrix**


|                   | **Model Positive**           | **Model Negative**            |
| ----------------- | ---------------------------- | ----------------------------- |
| **Real Positive** | <div style="text-align:center">**TP**                       | <div style="text-align:center">**FN** <br> **Type II Error** |
| **Real Negative** |  <div style="text-align:center">**FP** <br> **Type I Error** | <div style="text-align:center">**TN**                        |
    
<br>
    
**Simvolik düsturla:**

$$
\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

Burada:

* $TP$ = True Positive (doğru müsbət)
* $TN$ = True Negative (doğru mənfi)
* $FP$ = False Positive (səhv müsbət)
* $FN$ = False Negative (səhv mənfi)


**Misal:**
100 nümunədən 90-ını düzgün proqnoz etsək, accuracy belə hesablanır:
$$
\text{Accuracy} = \frac{90}{100} = 0.9 = 90%
$$


Qısa desək: **Accuracy = Düzgünlərin faizi**.

### **Precision**

> **Precision (Dəqiqlik)** — **Modelini Müsbət olaraq proqnozlaşdırdığı halların nə qədərinin həqiqətən Müsbət olduğunu göstərən bir metrikdir.**

> **Sadə dillə desək: "Model bir şeyi 'müsbət' (məsələn, spam) adlandırırsa, buna nə dərəcədə güvənə bilərik?" sualına cavab verir.**

> **Modelin 1 dediklərinin həqiqətən də neçəsi 1-dir.**

#### **Riyazi Düsturu**
- **Precision, Modelin doğru müsbət proqnozlarının ($TP$), Modelin etdiyi bütün Müsbət proqnozlara ($TP + FP$) nisbətidir:**

$$Precision = \frac{TP}{TP + FP}$$

- **TP (True Positive):** **Model "müsbətdir" dedi və düz çıxdı.**
- **FP (False Positive):** **Model "müsbətdir" dedi, amma əslində mənfi idi (Yalançı həyəcan).**

<br>

#### **Sizin Şəkildəki Nümunə Üzərindən İzah**
- **Əvvəlki şəkildəki rəqəmləri xatırlayaq:**
    * $TP = 90$ **(90 normal mesaj düzgün tapılıb)**
    * $FP = 3$ **(3 spam mesaj səhvən "normal" kimi qeyd olunub)**

    - **Burada əgər biz "normal" mesajları tapmaq istəyiriksə:**
$$
Precision = \frac{90}{90 + 3} = \frac{90}{93} \approx 0.967 \text{ (və ya 96.7%)}
$$

> **Bu o deməkdir ki, model bir mesajın "normal" olduğunu deyirsə, 96.7% hallarda haqlıdır.**

<br>

#### **Precision Nə Zaman Kritikdir?**
- **Precision-un yüksək olması False Positive (səhv müsbət) xətalarının baha başa gəldiyi yerlərdə vacibdir:**

    * **Kredit Riski: Əgər Model krediti ödəyə bilməyən bir müştəriyə səhvən "ödəyə bilər" ($FP$) deyib təsdiq verərsə, bu bank üçün birbaşa maliyyə itkisidir və burada Precision-un yüksək olması şərtdir.**.
    * **Məhkəmə Qərarları:** **Günahsız bir insanın səhvən cinayətkar elan edilməsi ($FP$) yolverilməzdir.**

#### **Precision vs Recall ilə fərqi**
- **Precision-u çox vaxt Recall (Həssaslıq) ilə qarışdırırlar:**
    - **Precision:** **"Tapdığım müsbətlərin neçəsi doğrudur?" (Səhv etməməyə fokuslanır).**
    - **Recall:** **"Bütün müsbətlərin neçəsini tapa bildim?" (Heç nəyi qaçırmamağa fokuslanır).**

> **Bu iki metrik arasında çox vaxt balans ($Trade-off$) olur; birini artıranda digəri azala bilir.**

<br>

#### **📌 Qısa izah:**

> **Model “bu spamdır” deyirsə** → **neçə faizində haqlıdır?**

<br>

#### **📌 Qısa yekun:**

> **Precision yüksəkdirsə → model “pozitiv” deyəndə çox vaxt düz deyir.**

### Recall

> **Recall** **(Həssaslıq və ya Tamlıq) — Modelin  Realda bütün mövcud Müsbət hallardan nə qədərini tapa bildiyini göstərən metrikdir.**

> **Əgər Precision "tapdıqlarımızın nə qədəri doğrudur?" sualına cavab verirdisə, Recall "tapmalı olduqlarımızın neçəsini qaçırmadıq?" sualına cavab verir.**

> **Realda 1 olanların Model tərəfindən nə qədəri həqiqətən 1 olaraq təsbit edilmişdir.**

<br>

#### **Riyazi Düsturu**
> **Recall, Modelin doğru müsbət proqnozlarının ($TP$), bütün həqiqi müsbət hallara ($TP + FN$) nisbətidir:**

$$Recall = \frac{TP}{TP + FN}$$

* **TP (True Positive):** **Model düzgün şəkildə "müsbətdir" dedi.**
* **FN (False Negative):** **Əslində müsbət idi, amma model onu səhvən "mənfi" kimi qeyd etdi (Qaçırılmış hal).**

<br>

#### **Sizin Şəkildəki Nümunə Üzərindən İzah**
- **Gəlin yenidən şəkildəki rəqəmlərə baxaq (burada +1 sinfi normal mesajlardır):**
    * $TP = 90$ **(90 normal mesaj tapılıb)**
    * $FN = 10$ **(10 normal mesaj "spam" sanılaraq qaçırılıb)**

    - **Bu modelin normal mesajları tapma "həssaslığı" (Recall):**
$$Recall = \frac{90}{90 + 10} = \frac{90}{100} = 0.90 \text{ (və ya 90%)} $$

> **Bu o deməkdir ki, sistem bütün normal mesajların **90%-ni** istifadəçiyə çatdıra bilir, 10%-ni isə səhvən bloklayır.**

<br>

#### **Recall Nə Zaman Kritikdir?**
- **Recall-un yüksək olması **False Negative** (qaçırılmış hal) xətalarının çox təhlükəli olduğu sahələrdə vacibdir:**

    * **Xərçəng Diaqnozu:** **Əgər bir xəstədə xərçəng varsa, onu "sağlam" ($FN$) deyib evə göndərmək həyatı təhlükə yaradır. Burada Recall-un 100%-ə yaxın olması şərtdir.**
    * **Bank Fırıldaqçılığı (Fraud Detection):** **Bir oğurluq əməliyyatını qaçırmaq bank üçün itkidir.**
    * **Minaların Tapılması:** **Ərazidəki minalardan birini belə qaçırmaq (FN) faciəyə səbəb olar.**

<br>

#### Precision və Recall arasındakı fərq (Xülasə)

| **Metrik** | **Sualı** | **Fokuslandığı Səhv** |
| :--- | :--- | :--- |
| **Precision** | **"Müsbət" dediklərin nə dərəcədə etibarlıdır? | **FP** (Yalançı həyəcanı azaltmaq)** |
| **Recall** | **Həqiqi müsbətlərin neçəsini "tutmusan"?** | **FN** **(Heç nəyi qaçırmamaq)** |

<br>

> **Çox vaxt birini artıranda (məsələn, daha çox adamı xəstəxanaya yatıranda ki, Recall artsın), digəri azalır (çoxlu sağlam adamı da səhvən yatırmış oluruq, Precision düşür)**.

<br>

#### **📌 Qısa izah:**

> **Model bütün “bu spamdır” mesajları neçə faizində düzgün tapır?**

<br>

#### **📌 Qısa yekun:**

> **Recall yüksəkdirsə → model əksər həqiqi pozitivləri (“spam” mesajları) düzgün aşkarlayır.**

$$
\text{Precision}_1 = \frac{TP}{TP + FP}
$$

#### **Qısa izah:**

* **TP (True Positive)** → **1 olub düzgün 1 deyilib**
* **FP (False Positive)** → **0 olub səhvən 1 deyilib**

> **Yəni Model “1” dediklərinin nə qədəri həqiqətən 1-dir.**

<br>

#### **Precision (0 sinfi üçün):**
$$
\text{Precision}_0 = \frac{TN}{TN + FN}
$$

#### **Qısa izah:**

* **TN (True Negative)** → **0 olanları düzgün 0 demək**
* **FN (False Negative)** → **1 olub səhvən 0 demək**

> **Yəni Model “0” dediklərinin nə qədəri həqiqətən 0-dır.**


### **Recall - 1 və 0 sinfi üçün düstur** 

#### **Recall (1 sinfi üçün):** 

$$
\text{Recall}_1 = \frac{TP}{TP + FN}
$$

#### **Qısa izah:**

* **TP (True Positive)** → **1 olub düzgün 1 deyilib**
* **FN (False Negative)** → **1 olub səhvən 0 deyilib**

> **Yəni Real 1-lərin  Model tərəfindən nə qədəri düzgün tapılıb.**

<br>

#### **Recall (0 sinfi üçün):** 

$$
\text{Recall}_0 = \frac{TN}{TN + FP}
$$

#### **Qısa izah:**

* **TN (True Negative)** → **0 olub düzgün 0 deyilib**
* **FP (False Positive)** → **0 olub səhvən 1 deyilib**

> **Yəni Real 0-ların Model tərəfindən nə qədəri düzgün tapılıb.**



### **F-score - Harmonic Mean**

> **F-score** (və ya **F1-score**) — **Precision və Recall metrikləri arasında tarazlıq yaradan və modelin ümumi effektivliyini tək bir rəqəmlə ifadə edən göstəricidir.**

> **Əgər bir modelin Precision-u çox yüksək, Recall-u isə çox aşağıdırsa (və ya əksinə), təkbaşına bu metriklər bizi yanıltmış olar. F1-score hər iki göstəricini nəzərə alaraq bizə daha obyektiv bir nəticə verir.**

#### **Riyazi Düsturu**
- **F1-score Precision və Recall-un harmonik ortasıdır (Harmonic Mean)**

$$F1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall}$$

- **Niyə adi orta yox, harmonik orta? Çünki harmonik orta, göstəricilərdən biri çox aşağı olduqda (məsələn, 0-a yaxın), ümumi nəticəni də kəskin şəkildə aşağı salır. Bu da bizi "tənbəl" və ya qərəzli modellər qurmaqdan qoruyur.**

<br>

#### **Sizin Nümunə Üzərindən Hesablama**
- **Əvvəlki cavablarda sizin şəkildəki rəqəmlər üçün bunları hesablamışdıq:**
* **Precision:** $0.967$
* **Recall:** $0.90$

- **İndi F1-score-u hesablayaq:**
$$F1 = 2 \cdot \frac{0.967 \cdot 0.90}{0.967 + 0.90} = 2 \cdot \frac{0.8703}{1.867} \approx 0.932$$

- **Yəni bu modelin ümumi performansı 93.2%-dir.**

<br>

#### **F-score Nə Zaman İstifadə Olunur?**
- **F1-score xüsusilə balanssız verilənlər bazasında (imbalanced datasets) çox vacibdir.**

    - **Məsələn, 1000 nəfərdən yalnız 10-u xəstədirsə:**
        * **Model hamıya "sağlamdır" desə, Accuracy (Dəqiqlik) 99% olacaq. Amma bu model heç bir xəstəni tapmadığı üçün əslində yararsızdır.**
        * **Belə halda F1-score hesablansa, Recall 0 olduğu üçün nəticə 0 çıxacaq. Bu da modelin uğursuz olduğunu dərhal üzə çıxaracaq.**

<br>

#### **Ümumi F-beta Ölçüsü**
- **Bəzən bizə Recall, Precision-dan daha vacib olur (məsələn, tibbdə) və ya əksinə. Bu zaman $F_{\beta}$ düsturundan istifadə edilir:**
    * **F1-score:** **Precision və Recall eyni dərəcədə vacibdir ($\beta = 1$).**
    * **F2-score:** **Recall-a (həssaslığa) daha çox önəm verir.**
    * **F0.5-score:** **Precision-a (dəqiqliyə) daha çox önəm verir.**

<br>

#### **$F_{\beta}$ ölçüsünün müxtəlif variantlarını və onların tətbiq sahələri**
#### **Ümumi $F_{\beta}$ Düsturu**
- **Bütün bu ölçülər aşağıdakı ümumi düsturdan törəyir:**

$$F_{\beta} = (1 + \beta^2) \cdot \frac{Precision \cdot Recall}{(\beta^2 \cdot Precision) + Recall}$$

> **Burada $\beta$ parametri Recall-un Precision-dan neçə dəfə daha vacib olduğunu müəyyən edir.**

<br>

#### **$F$-Ölçülərinin Müqayisəsi və Tətbiqi**

| Metrik | Düstur | Nəyə önəm verir? | Harada istifadə olunur? |
| :--- | :--- | :--- | :--- |
| **F1-Score** | $$2 \cdot \frac{P \cdot R}{P + R}$$ | **Balans: Hər ikisi eyni dərəcədə vacibdir ($\beta=1$).** | **Siniflərin paylanması qeyri-bərabər olan ümumi tapşırıqlarda.** |
| **F0.5-Score** | $$1.25 \cdot \frac{P \cdot R}{(0.25 \cdot P) + R}$$ | **Precision: Səhv həyəcan (FP) verməmək daha vacibdir.** | **Spam filtrləri, reklam hədəfləmə, rəqəmsal marketinq.** |
| **F2-Score** | $$5 \cdot \frac{P \cdot R}{(4 \cdot P) + R}$$ | **Recall: Heç bir halı qaçırmamaq (FN) daha vacibdir.** | **Tibbi diaqnostika, təhlükəsizlik sistemləri, mina axtarışı.** |

<br>

#### **Qısa Nümunələr və Məntiq**

<br>

#### **F0.5 — Precision-un üstünlüyü (Dəqiqlik önəmlidir)**
* **Ssenari:** **Bir istifadəçiyə maraqlı ola biləcək bahalı bir məhsulun reklamını göstərirsiniz.**
* **Məntiq:** **Əgər siz "səhvən" (False Positive) reklamı marağı olmayan adama göstərsəniz, pulunuz boşa gedəcək. Burada Recall-un aşağı olması (bəzi maraqlı insanları qaçırmaq) o qədər də dəhşətli deyil, amma Precision-un yüksək olması (reklamın hədəfə dəyməsi) kritikdir.**

#### **F1 — Balanslı yanaşma**
* **Ssenari:** **Müştərilərin bankdan gedib-getməyəcəyini (Churn prediction) təxmin edirsiniz.**
* **Məntiq:** **Həm gedən müştəriləri qaçırmamaq (Recall), həm də qalacaq müştərilərə boş yerə endirim təklif edib gəliri azaltmamaq (Precision) eyni dərəcədə vacibdir.**

#### **F2 — Recall-un üstünlüyü (Həssaslıq önəmlidir)**
* **Ssenari:** **Hava limanında rentgen cihazı ilə baqajda təhlükəli maddə axtarılır.**
* **Məntiq:** **Burada bir təhlükəni qaçırmaq (False Negative) faciəvi ola bilər. Cihaz bəzən səhvən "təhlükə var" deyib (False Positive) çantanı əlavə yoxlamaya göndərsə belə (aşağı Precision), əsas odur ki, heç bir real təhlükəli maddə keçməsin (yüksək Recall).**


#### `sklearn.metrics` kitabxanasından istifadə edərək **F-beta score**-un hesablanması
- `from sklearn.metrics import fbeta_score`

- **`sklearn.metrics` kitabxanasından istifadə edərək F-beta score-un hesablanmasına dair sadə və aydın bir nümunəni**

#### **Python-da F-Beta Score Hesablanması**

- **F-beta score həm Precision (dəqiqlik), həm də Recall (yada salma) göstəricilərini birləşdirir. Burada $\beta$ parametri hansı tərəfə daha çox üstünlük verəcəyinizi təyin edir:**
* **$\beta = 1$**: **F1-score (hər ikisi bərabərdir).**
* **$\beta < 1$**: **Precision-a daha çox önəm verilir.**
* **$\beta > 1$**: **Recall-a daha çox önəm verilir.**


```python
from sklearn.metrics import fbeta_score

# Həqiqi nəticələr (y_true) və modelin təxminləri (y_pred)
y_true = [0, 1, 1, 0, 1, 1]
y_pred = [0, 0, 1, 0, 0, 1]

# 1. F2-score hesablayaq (Recall Precision-dan 2 dəfə daha vacibdir)
f2_score = fbeta_score(y_true, y_pred, beta=2.0)

# 2. F0.5-score hesablayaq (Precision Recall-dan 2 dəfə daha vacibdir)
f05_score = fbeta_score(y_true, y_pred, beta=0.5)

print(f"F2 Score (Recall öncəlikli): {f2_score:.2f}")
print(f"F0.5 Score (Precision öncəlikli): {f05_score:.2f}")
```

#### **Diqqət etməli olduğunuz məqamlar:**

* **Multiclass problemlər:** **Əgər verilənlərinizdə ikidən çox sinif varsa (məsələn: alma, armud, heyva), `average` parametrini qeyd etməlisiniz:**
    * `average='macro'`: **Bütün siniflərin nəticələrinin ortalamasını alır.**
    * `average='weighted'`: **Siniflərin sayına görə çəkili ortalama hesablayır.**
* **Beta dəyəri:** $\beta$ **parametri həmişə müsbət olmalıdır.**

```python
from sklearn.metrics import fbeta_score

# Həqiqi nəticələr (y_true) və modelin təxminləri (y_pred)
y_true = [0, 1, 1, 0, 1, 1]
y_pred = [0, 0, 1, 0, 0, 1]

# 1. Bütün siniflərin nəticələrinin ortalamasını alır
f2_score = fbeta_score(y_true, y_pred, beta=2.0, average='macro')

# 2. Siniflərin sayına görə çəkili ortalama hesablayır.
f05_score = fbeta_score(y_true, y_pred, beta=0.5, average='weighted')

print(f"F2 Score (Recall öncəlikli): {f2_score:.2f}")
print(f"F0.5 Score (Precision öncəlikli): {f05_score:.2f}")
```
