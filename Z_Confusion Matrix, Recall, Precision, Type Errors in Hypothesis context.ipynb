### Machine Learning-də Confusion Matrix



#### 1️⃣ Confusion Matrix cədvəli

| **Real** \ Model | **Model = 1** | **Model = 0** |
| ------------ | --------- | --------- |
| **Real = 1**     | $$TP$$        | $$FN$$ **Type II Error**        |
| **Real = 0**     | $$FP$$ **Type I Error**         | $$TN$$        |

* **TP (True Positive)** → Real = 1, Model = 1
* **FN (False Negative)** → Real = 1, Model = 0
* **FP (False Positive)** → Real = 0, Model = 1
* **TN (True Negative)** → Real = 0, Model = 0

Bu cədvəl **hər bir proqnozun doğruluğunu** göstərir.

---


#### 2️⃣ Formullar

**Precision (Dəqiqlik)** → Modelin `1` proqnozlarından neçə faizi doğru olub:

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

**Recall (Çağırış / True Positive Rate)** → Real `1`lərin neçə faizi model tərəfindən düzgün tapılıb:

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

---

#### 3️⃣ Misal

Əvvəlki nümunədən:

```python
y_real = [1, 0, 1, 1, 0, 0, 1]
y_pred = [1, 0, 1, 0, 0, 1, 1]
```

Confusion Matrix:

| **Real** \ **Model** | **Model = 1** | **Model = 0** |
| ------------ | --------- | --------- |
| **Real = 1**     | $$3 (TP)$$    | $$1 (FN)$$    |
| **Real = 0**     | $$1 (FP)$$    | $$2 (TN)$$    |

**Precision**:

$$
\text{Precision} = \frac{TP}{TP + FP} = \frac{3}{3 + 1} = 0.75
$$

**Recall**:

$$
\text{Recall} = \frac{TP}{TP + FN} = \frac{3}{3 + 1} = 0.75
$$

---

- Yəni **Precision** deyir ki, model `1` dediyi vaxtın 75%-i doğrudur.
- **Recall** deyir ki, əsl `1`lərin 75%-i model tərəfindən tapılıb.


### Type I və Type II Error (səhvləri) hipotez testləri kontekstində izah



#### 1️⃣ Terminlər

**Null Hipotez (H₀)**: Adətən “heç bir effekt yoxdur” / “model doğru deyil”
**Alternativ Hipotez (H₁)**: “Effekt var” / “model doğru işləyir”

* **Type I Error (α)** → **False Positive**

  * H₀ doğru olduğu halda onu rədd etmək
  * “Yanlış olaraq fərziyyəni rədd etdik”
* **Type II Error (β)** → **False Negative**

  * H₀ səhv olduğu halda onu qəbul etmək
  * “Yanlış olaraq fərziyyəni qəbul etdik”

---

#### 2️⃣ Confusion Matrixlə əlaqə

| **Real** / **H₀ vəziyyəti** | **H₀ qəbul edilir**        | **H₀ rədd edilir**        |
| ------------------- | ---------------------- | --------------------- |
| **H₀ doğru**            | $$TN$$                     | **Type I Error** $$(FP)$$ |
| **H₀ səhv**             | **Type II Error** $$(FN)$$ | $$TP$$                   |

* **Type I (α)** = H₀ doğru olduğu halda rədd etdiyimiz hallar
* **Type II (β)** = H₀ səhv olduğu halda qəbul etdiyimiz hallar

---

#### 3️⃣ Misal ilə

Tutaq ki, xəstəlik test edirik:

* H₀: Xəstə deyil
* H₁: Xəstədir

| **Real** / **(Xəstəlik)** | **Test nəticəsi = Xəstə** | **Test nəticəsi = Sağlam** |
| ---------------               | ------------------------- | ----------------------     |
| **Sağlam**                    | $$FP$$ $\qquad$ **Type I Error**    | $$TN$$                     |
| **Xəstə**                     | $$TP$$                    | $$FN$$ $\qquad$  **Type II Error**  |

* **Type I Error** → Xəstə deyil, amma test deyir “xəstəsən” (yalancı alarm)
* **Type II Error** → Xəstədir, amma test deyir “sağlamsan” (qaçırılmış diaqnoz)
