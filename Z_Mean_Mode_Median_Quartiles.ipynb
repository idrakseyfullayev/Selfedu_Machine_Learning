## Ədədi Orta (Arithmetic Mean) vs Çəkili Orta (Weighted Mean)

### 1. **Ədədi Orta (Arithmetic Mean):**  
Ədədi orta bir sıra ədədi dəyərlərin sadəcə toplamının həmin dəyərlərin sayına bölünməsi ilə hesablanır. Bu, verilənlərdəki mərkəz meylini göstərən əsas statistik ölçülərdən biridir.

**Düstur:**  
$$
\text{Arithmetic Mean} = \frac{\sum_{i=1}^n x_i}{n}
$$ 

**Burada:**  
- $x_i$ - verilənlərin hər bir dəyəri,  
- $n$ - dəyərlərin sayı,  
- $\sum$ - toplama operatoru.

**Məsələn:**  
Verilənlər: $5, 10, 15$  
$$
\text{Arithmetic Mean} = \frac{5 + 10 + 15}{3} = \frac{30}{3} = 10
$$  

**Xüsusiyyətlər:**  
- Ədədi orta simmetrik paylanmada verilənlərin mərkəzində yerləşir.  
- Hər bir dəyər eyni çəkiyə malikdir.

---

### 2. **Çəkili Orta (Weighted Mean):**  
Çəkili orta, hər bir dəyərin fərqli çəkisi olduğu vəziyyətlər üçün istifadə olunur. Yəni hər dəyərin təsiri (çəki) fərqli ola bilər. Bu, daha əhəmiyyətli dəyərlərə daha çox təsir etməyə imkan verir.

**Düstur:**  
$$
\text{Weighted Mean} = \frac{\sum_{i=1}^n w_i x_i}{\sum_{i=1}^n w_i}
$$  

**Burada:**  
- $x_i$ - verilənlərin hər bir dəyəri,  
- $w_i$ - həmin dəyərlə əlaqəli çəki,  
- $\sum$ - toplama operatoru.

**Məsələn:**  
Verilənlər: $5, 10, 15$, çəkilər: $1, 2, 3$  
$$
\text{Weighted Mean} = \frac{(1 \cdot 5) + (2 \cdot 10) + (3 \cdot 15)}{1 + 2 + 3} = \frac{5 + 20 + 45}{6} = \frac{70}{6} \approx 11.67
$$  

**Xüsusiyyətlər:**  
- Çəkilər fərqli əhəmiyyət səviyyələrini ifadə edir.  
- Çəkilər düzgün normallaşdırılmalıdır (cəmi sıfır olmamalıdır).

---

### **Fərqlər:**  
| Xüsusiyyət             | Ədədi Orta                      | Çəkili Orta                         |
|-------------------------|----------------------------------|--------------------------------------|
| **Çəki paylanması**     | Hər bir dəyər eyni çəkiyə malikdir. | Dəyərlərə fərqli çəkilər verilir.   |
| **Düstur**              | $\frac{\sum x_i}{n}$         | $\frac{\sum w_i x_i}{\sum w_i}$   |
| **İstifadə sahəsi**     | Sadə paylanmalar.              | Çəkilər əhəmiyyətli olduqda.        |

---

### **Hər İki Orta Həyatımızda:**  
- Ədədi orta: Orta ballar, gəlirlər, xərclər kimi sadə hesablamalarda istifadə edilir.  
- Çəkili orta: İmtahan nəticələrinin hesablanmasında, maliyyə analitikalarında, portfel idarəçiliyində istifadə edilir.  

---

### 3. **Ehtimal ilə Orta (Probability Mean):** 


### 🔹 **Verilən Data:**

```text
Data = [5, 10, 10, 15, 15, 15, 20, 25, 30, 30]
```

Bu siyahıda:

| Dəyər $$(x\_i)$$ | Təkrarlanma sayı | Ehtimal $$P(x\_i) = \frac{\text{say}}{10}$$ |
| ---------------- | ---------------- | ------------------------------------------- |
| 5                | 1                | 1 / 10 = 0.1                                |
| 10               | 2                | 2 / 10 = 0.2                                |
| 15               | 3                | 3 / 10 = 0.3                                |
| 20               | 1                | 1 / 10 = 0.1                                |
| 25               | 1                | 1 / 10 = 0.1                                |
| 30               | 2                | 2 / 10 = 0.2                                |

---

### 🔹 **Ehtimalların Cəmi:**

$$
0.1 + 0.2 + 0.3 + 0.1 + 0.1 + 0.2 = 1.0 \quad ✅
$$

---

### 🔹 **Gözlənilən Dəyər (Expected Value):**

$$
E(X) = \sum x_i \cdot P(x_i)
$$

Hesablayırıq:

$$
E(X) = (5 \cdot 0.1) + (10 \cdot 0.2) + (15 \cdot 0.3) + (20 \cdot 0.1) + (25 \cdot 0.1) + (30 \cdot 0.2)
$$

$$
= 0.5 + 2 + 4.5 + 2 + 2.5 + 6 = 17.5
$$

---

### ✅ **Nəticə:**

* **Ehtimalları**: dəyərlərin neçə dəfə təkrarlandığına görə tapdıq
* **Orta (gözlənilən dəyər)**: 17.5 çıxdı



## Median (Orta qiymət) vs Moda (Ən çox rast gəlinən dəyər)

### **Median (Orta qiymət)**

**Tərifi:**  
Median verilənlərin sıralanmış halda ortasında yerləşən qiymətdir. O, verilənlər toplusunun mərkəzini müəyyən edir və **dəyərlərin bərabər yarısını ayırır**. Median xüsusilə ekstremal dəyərlərə (outlierlərə) həssas olmadığı üçün statistikanın yayılma ilə bağlı təhlillərində geniş istifadə edilir.

---

**Düsturları:**

1. **Tək sayda verilənlər üçün:**
   $$
   \text{Median} = x_{(n+1)/2}
   $$

2. **Cüt sayda verilənlər üçün:**
   $$
   \text{Median} = \frac{x_{n/2} + x_{(n/2)+1}}{2}
   $$

Burada:
- $ n $: Verilənlərin sayı,
- $ x $: Sıralanmış verilənlər.

---

**Xüsusiyyətləri:**  
1. Dəyərlər sıralanmış olmalıdır.  
2. Outlierlərdən təsirlənmir (orta hesabdan fərqli olaraq).  
3. Verilənlərin mərkəzini müəyyən edir.  

---

**Nümunə:**  
Verilənlər: $ 3, 7, 2, 9, 5 $  
1. **Sıralanmış formada:** $ 2, 3, 5, 7, 9 $  
2. **Tək sayda dəyərlər: Median = 5** (ortadakı dəyər).  

Verilənlər: $ 8, 4, 6, 2 $  
1. **Sıralanmış formada:** $ 2, 4, 6, 8 $  
2. **Cüt sayda dəyərlər: Median = $\frac{4+6}{2} = 5$**.

---

### **Moda (Ən çox rast gəlinən dəyər)**

**Tərifi:**  
Moda verilənlər toplusunda **ən çox rast gəlinən qiyməti** göstərir. Moda həm kəmiyyət, həm də keyfiyyət dəyişənləri üçün istifadə edilə bilər. Verilənlərdə bir neçə moda ola bilər və bu halda verilənlər **bimodal** və ya **multimodal** adlanır.

---

**Xüsusiyyətləri:**  
1. Moda həm say, həm də kateqorik verilənlər üçün uyğundur.  
2. Verilənlərdə modanın olmaması mümkündür (hər dəyər yalnız bir dəfə rast gəlinirsə).  
3. Çox modanın olması mümkündür.  

---

**Düstur:**  
- Verilənlərdə **ən çox təkrarlanan dəyər modadır**.
- Dağılım üçün sıxlıq təhlilində, modanı qrafikdə **pik nöqtə** kimi görmək mümkündür.

---

**Nümunə:**  
Verilənlər: $ 4, 6, 6, 8, 10, 6, 8, 8 $  
1. **Moda = 6** və **Moda = 8** (Bimodal verilənlər).  

Verilənlər: $ 1, 2, 3, 4, 5 $  
1. Moda yoxdur (hər bir dəyər yalnız bir dəfə rast gəlinir).

---

### **Median və Modanın Müqayisəsi**

| Xüsusiyyət      | Median                | Moda                   |
|------------------|-----------------------|------------------------|
| **Nəyi göstərir?** | Ortadakı qiymət        | Ən çox rast gəlinən qiymət |
| **Tətbiq sahəsi**  | Kəmiyyət dəyişənləri   | Kəmiyyət və keyfiyyət dəyişənləri |
| **Outlierlərə təsir** | Təsir etmir           | Təsir etmir            |
| **Bir neçə dəyər**  | Yalnız bir median olur | Çox moda ola bilər      |

---

### **Tətbiqlər:**
1. **Median:**
   - Gəlir, qiymət, zaman analitikası üçün uyğundur.
   - Mərkəzi meyli göstərir.
   
2. **Moda:**
   - Ən çox tələb olunan məhsulu müəyyənləşdirmək üçün istifadə edilir.
   - Ən çox rast gəlinən kateqoriyanı (rəng, marka və s.) təhlil etmək üçün tətbiq olunur.

## Q1 və Q3 hesablanma qaydası 

Q1 (Birinci kvartil) və Q3 (Üçüncü kvartil) verilənlərdəki paylanmanın 25% və 75% nöqtələrini ifadə edir. Onları tapmaq üçün aşağıdakı addımları izləyin:

### Addımlar:

1. **Verilənləri sırala:**
   Verilənləri kiçikdən böyüyə doğru sıralayın.

2. **Medianı tap:**
   Median verilənlərin tam ortasındakı nöqtədir. Verilənlərin sayı cüt olduqda, ortadakı iki ədədin orta qiymətini tapın.

3. **Q1 (Birinci kvartil):**
   - Medianın **solunda** qalan verilənləri götürün.
   - Bu yarının medianını hesablayın. Bu Q1-dir.

4. **Q3 (Üçüncü kvartil):**
   - Medianın **sağında** qalan verilənləri götürün.
   - Bu yarının medianını hesablayın. Bu Q3-dir.

---

### Məsələn:

Verilmiş data:  
`1, 3, 5, 7, 9, 11, 13, 15, 17`

1. Median (ortadakı dəyər): **9**  
2. Sol hissə (Medianın solu): `1, 3, 5, 7`  
   - Bu verilənlərin medianı (Q1): `(3+5)/2 = 4`  
3. Sağ hissə (Medianın sağı): `11, 13, 15, 17`  
   - Bu verilənlərin medianı (Q3): `(13+15)/2 = 14`  

Nəticə:  
Q1 = 4, Q3 = 14

--- 

Əgər sualınız daha riyazi və ya spesifikdirsə, əlavə izah verə bilərəm.

## Q1 və Q3  hesablanması üçün düstur 

Q1 (birinci kvartil) və Q3 (üçüncü kvartil) hesablanması üçün düstur konkret verilənlər siyahısının mövqelərinə əsaslanır. Düsturlar verilənlər ardıcıllığı sıralandıqdan sonra tətbiq olunur.

---

### Düsturlar:

**1. Q1 (Birinci kvartil):**  
$ Q1 = \text{Data at } \left(\frac{n+1}{4}\right)\text{-th position} $

**2. Q3 (Üçüncü kvartil):**  
$ Q3 = \text{Data at } \left(\frac{3(n+1)}{4}\right)\text{-th position} $

Burada:  
- $ n $: verilənlərin ümumi sayı  
- Mövqe tam ədədlə düşmədikdə, interpolasiya (aralıq qiymətlərin tapılması) istifadə olunur.

---

### Nümunə:

**Verilən data:**  
$ 2, 4, 6, 8, 10, 12, 14, 16 $ (sıralanmışdır, $ n = 8 $)

**1. Q1 hesablanması:**  
$ \frac{n+1}{4} = \frac{8+1}{4} = 2.25 \  $ 

2-ci və 3-cü elementlər arasında interpolasiya:  
$ Q1 = 4 + 0.25 \times (6 - 4) = 4 + 0.5 = 4.5 $

**2. Q3 hesablanması:**  
$ \frac{3(n+1)}{4} = \frac{3 \times (8+1)}{4} = 6.75 $  

6-cı və 7-ci elementlər arasında interpolasiya:  
$ Q3 = 12 + 0.75 \times (14 - 12) = 12 + 1.5 = 13.5 $

**Nəticə:**  
Q1 = 4.5, Q3 = 13.5

---

Əgər əlavə nümunə və ya spesifik verilənlər üçün izah istəyirsinizsə, qeyd edin!
