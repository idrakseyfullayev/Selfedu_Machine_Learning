```python
import matplotlib.pyplot as plt

# Məlumatlar
categories = ['A', 'B', 'C', 'D']
values = [70, 85, 60, 90]
target = 75  # Hədəf xətti

# Bar chart
bars = plt.bar(categories, values, color=['red' if v < target else 'green' for v in values])

# Hədəf xətti
plt.axhline(y=target, color='blue', linestyle='--', label=f'Hədəf: {target}')
plt.text(len(categories)-2, target + 1, f'{target}', color='blue', ha='right')

# Dəyərləri barların üzərinə yaz
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, str(height),
             ha='center', va='bottom')

# Əlavə
plt.title("Performans və Hədəf")
plt.ylabel("Dəyər")
plt.legend()
plt.show()
```
