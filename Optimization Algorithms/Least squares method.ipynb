## Standart Qaus (Normal) sapmasƒ±

Standart Qaus (Normal) sapmasƒ± (standart sapma) statistikada dispersiyanƒ±n kvadrat k√∂k√º kimi m√º…ôyy…ôn edilir v…ô veril…ônl…ôrin orta qiym…ôtd…ôn n…ô q…ôd…ôr yayƒ±ndƒ±ƒüƒ±nƒ± √∂l√ß√ºr. Normal paylanma √º√ß√ºn standart sapma a≈üaƒüƒ±dakƒ± d√ºsturla hesablanƒ±r:

$$
\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}
$$

Burada:
- $ \sigma $ ‚Äî standart sapmadƒ±r,
- $ x_i $ ‚Äî veril…ônl…ôrin i-ci elementi,
- $ \mu $ ‚Äî orta qiym…ôt ($\mu = \frac{\sum x_i}{N}$),
- $ N $ ‚Äî √ºmumi m√º≈üahid…ôl…ôrin sayƒ±.

∆èg…ôr n√ºmun…ô √ºz…ôrind…ô hesablanƒ±rsa, o zaman d√ºstur d…ôyi≈üir v…ô $ N $ …ôv…ôzin…ô $ N-1 $ istifad…ô olunur:

$$
s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N-1}}
$$

Bu zaman $ s $ n√ºmun…ô standart sapmasƒ± adlanƒ±r.

Standart sapma b√∂y√ºk olduqda, veril…ônl…ôr orta qiym…ôtd…ôn daha √ßox yayƒ±lmƒ±≈ü olur, ki√ßik olduqda is…ô veril…ônl…ôr orta qiym…ôt …ôtrafƒ±nda c…ôml…ônmi≈ü olur.

S…ôn normal paylanmada standart sapmanƒ±n t…ôsirini v…ô t…ôtbiq sah…ôl…ôrini daha …ôtraflƒ± √∂yr…ônm…ôk ist…ôyirs…ôn?

## Standart Qaus (Normal) paylanmasƒ±nƒ±n sƒ±xlƒ±q funksiyasƒ± (Probability Density Function, PDF)

Standart Qaus (Normal) paylanmasƒ±nƒ±n sƒ±xlƒ±q funksiyasƒ± **(Probability Density Function, PDF)** bel…ô verilir:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Burada:  
- $ x $ ‚Äî t…ôsad√ºfi d…ôyi≈ü…ôn,  
- $ \mu $ ‚Äî orta qiym…ôt,  
- $ \sigma $ ‚Äî standart sapma,  
- $ e $ ‚Äî t…ôqrib…ôn 2.718 olan Eyler sabiti,  
- $ \pi $ ‚Äî 3.1416 olan Pi sabiti.  

### **X√ºsusiyy…ôtl…ôri:**
1. **Simmetrikdir**: Orta qiym…ôt …ôtrafƒ±nda simmetrik olur.
2. **Bell-≈ü…ôkilli (z…ông formalƒ±) qrafik…ô malikdir**.
3. **T…ôk maksimum n√∂qt…ôsi var**: Bu n√∂qt…ô $ x = \mu $ -da yerl…ô≈üir.
4. **Standart sapma daƒüƒ±lƒ±mƒ±n enini t…ôyin edir**:
   - $ \mu \pm \sigma $ aralƒ±ƒüƒ±nda t…ôxmin…ôn **68%**,
   - $ \mu \pm 2\sigma $ aralƒ±ƒüƒ±nda **95%**,
   - $ \mu \pm 3\sigma $ aralƒ±ƒüƒ±nda **99.7%** m√º≈üahid…ôl…ôr olur (**Empirik qayda**).

∆èg…ôr **standart normal paylanmadan** danƒ±≈üƒ±rƒ±qsa, y…ôni $ \mu = 0 $ v…ô $ \sigma = 1 $, o zaman sƒ±xlƒ±q funksiyasƒ± sad…ôl…ô≈üir:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

Bu paylanmanƒ±n qrafiki **Z-≈ükalasƒ±** il…ô √∂l√ß√ºl…ôn standart normal paylanmanƒ± g√∂st…ôrir.

S…ôn hansƒ±sa konkret t…ôtbiq √º√ß√ºn bu funksiyanƒ±n istifad…ôsini √∂yr…ônm…ôk ist…ôyirs…ôn?

### **Standart Qaus Sapmasƒ± n…ôdir?**  

**Standart Qaus Sapmasƒ±** (**Gaussian Standard Deviation**) statistikada v…ô siqnal/g√∂r√ºnt√º emalƒ±nda **normal paylanmanƒ±n yayƒ±lma d…ôr…ôc…ôsini** √∂l√ß…ôn parametrdir. Ad…ôt…ôn **$ \sigma $** (sigma) il…ô i≈üar…ô olunur v…ô **standart sapma** (standard deviation) kimi tanƒ±nƒ±r.

## **1. Standart Sapmanƒ±n Riyazi ƒ∞zahƒ±**  

Normal (Qaus) paylanmanƒ±n √ºmumi t…ônliyi bel…ôdir:  

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Burada:  
- **$ \mu $** ‚Äî ortalama (mean), paylanmanƒ±n m…ôrk…ôzi.  
- **$ \sigma $** ‚Äî standart sapma, yayƒ±lmanƒ±n √∂l√ß√ºs√º.  
- **$ x $** ‚Äî verilmi≈ü d…ôy…ôr.  
- **$ e $** ‚Äî t…ôbii logaritmanƒ±n …ôsasƒ± ($ \approx 2.718 $).  

**$ \sigma $** **b√∂y√ºd√ºkc…ô paylanma daha geni≈ül…ônir v…ô k…ônarlara yayƒ±lƒ±r**, **$ \sigma $** **ki√ßildikc…ô paylanma daralƒ±r v…ô m…ôrk…ôzd…ô c…ôml…ônir.**

---

## **2. Standart Sapmanƒ±n G√∂rs…ôll…ô≈üdirilm…ôsi**  

- **Ki√ßik $ \sigma $** ‚Üí Paylanma dar olur, m…ôlumatlar orta …ôtrafƒ±nda toplanƒ±r.  
- **B√∂y√ºk $ \sigma $** ‚Üí Paylanma geni≈ü olur, m…ôlumatlar daha √ßox yayƒ±lƒ±r.  

üìä **Normal paylanmanƒ±n f…ôrqli standart sapmalara g√∂r…ô d…ôyi≈üm…ôsi**:

- **$ \sigma = 0.5 $** (√ßox dar, k…ôskin pik)  
- **$ \sigma = 1 $** (normal yayƒ±lma)  
- **$ \sigma = 2 $** (geni≈ü yayƒ±lma)  



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5, 5, 1000)
mu = 0

plt.plot(x, norm.pdf(x, mu, 0.5), label=r'$\sigma=0.5$')
plt.plot(x, norm.pdf(x, mu, 1), label=r'$\sigma=1$')
plt.plot(x, norm.pdf(x, mu, 2), label=r'$\sigma=2$')

plt.legend()
plt.title("Normal Paylanmanƒ±n F…ôrqli Standart Sapmalarla G√∂rs…ôll…ô≈üdirilm…ôsi")
plt.xlabel("x")
plt.ylabel("Ehtimallƒ±q sƒ±xlƒ±ƒüƒ±")
plt.grid()
plt.show()
```

Bu qrafikd…ô $ \sigma $ artdƒ±qca paylanmanƒ±n geni≈ül…ôndiyini g√∂r√ºr√ºk.

## **4. Qaus Filtrind…ô Standart Sapma (Sigma)**
G√∂r√ºnt√º emalƒ±nda Gaussian filter t…ôtbiq edil…ôrk…ôn, **$ \sigma $** n…ô q…ôd…ôr b√∂y√ºk olsa, g√∂r√ºnt√º daha bulanƒ±q olacaq.

**OpenCV il…ô Gaussian blur n√ºmun…ôsi:**


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ≈û…ôkili oxumaq
image = cv2.imread(r"C:\Users\dell\Desktop\Selfedu\Optimization Algorithms\Car_image_for_Qaus.jpg")

# OpenCV ≈ü…ôkilini BGR-d…ôn RGB-…ô √ßevirm…ôk (√ß√ºnki OpenCV BGR, Matplotlib is…ô RGB formatƒ±nƒ± istifad…ô edir)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Ki√ßik sigma (az bulanƒ±qlƒ±q)
blurred_1 = cv2.GaussianBlur(image, (5,5), 1)
blurred_1_rgb = cv2.cvtColor(blurred_1, cv2.COLOR_BGR2RGB)

# Ki√ßik sigma (az bulanƒ±qlƒ±q)
blurred_2 = cv2.GaussianBlur(image, (5,5), 2)
blurred_2_rgb = cv2.cvtColor(blurred_1, cv2.COLOR_BGR2RGB)

# B√∂y√ºk sigma (√ßox bulanƒ±qlƒ±q)
blurred_3 = cv2.GaussianBlur(image, (5,5), 3)
blurred_3_rgb = cv2.cvtColor(blurred_2, cv2.COLOR_BGR2RGB)

# ≈û…ôkill…ôri Jupyter-d…ô g√∂st…ôrm…ôk
plt.figure(figsize=(10, 7))

# Orijinal ≈ü…ôkil
plt.subplot(1, 4, 1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

# Sigma = 1 il…ô bulanƒ±qlanmƒ±≈ü ≈ü…ôkil
plt.subplot(1, 4, 2)
plt.imshow(blurred_1_rgb)
plt.title("Sigma = 1")
plt.axis("off")

# Sigma = 2 il…ô bulanƒ±qlanmƒ±≈ü ≈ü…ôkil
plt.subplot(1, 4, 3)
plt.imshow(blurred_2_rgb)
plt.title("Sigma = 2")
plt.axis("off")

# Sigma = 3 il…ô bulanƒ±qlanmƒ±≈ü ≈ü…ôkil
plt.subplot(1, 4, 4)
plt.imshow(blurred_3_rgb)
plt.title("Sigma = 3")
plt.axis("off")

# G√∂st…ôrm…ôk
plt.show()
```

## **5. N…ôtic…ô**  
- **Standart Qaus Sapmasƒ± ($ \sigma $) paylanmanƒ±n yayƒ±lmasƒ±nƒ± t…ôyin edir.**  
- **Ki√ßik $ \sigma $ ‚Üí Dar paylanma, az yayƒ±lma.**  
- **B√∂y√ºk $ \sigma $ ‚Üí Geni≈ü paylanma, √ßox yayƒ±lma.**  
- **Machine Learning, g√∂r√ºnt√º emalƒ± v…ô statistik analizl…ôrd…ô geni≈ü istifad…ô olunur.**  

## $$L = \prod_{i=1}^{N} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_i - f(x_i))^2}{2\sigma^2}}$$

S…ônin verdiyin ifad…ôd…ô **$\prod$** i≈üar…ôsi ehtimallarƒ±n hasili kimi g√∂st…ôrilib. **B…ôs niy…ô bu ehtimallar bir-birin…ô vurulur?**  

Bu, **statistikada m√º≈üahid…ôl…ôr m√ºst…ôqil olduƒüu halda ehtimalƒ±n hesablanma qaydasƒ±dƒ±r**. G…ôlin, addƒ±m-addƒ±m izah edim.  

---

### **1. Bayes n…ôz…ôriyy…ôsi v…ô m√ºst…ôqillik anlayƒ±≈üƒ±**  
Tutaq ki, **$N$ m√º≈üahid…ô var** v…ô h…ôr bir m√º≈üahid…ô **$y_i$** verilmi≈ü **$x_i$** √º√ß√ºn **≈ü…ôrti ehtimalla** ifad…ô olunur:  

$$
\omega(y_i | f(x_i))
$$

∆èg…ôr m√º≈üahid…ôl…ôr **bir-birind…ôn asƒ±lƒ± deyils…ô (y…ôni statistik olaraq m√ºst…ôqildirs…ô)**, o zaman **√ºmumi ehtimalƒ±** hesablamaq √º√ß√ºn onlarƒ±n **hasili** g√∂t√ºr√ºl√ºr:  

$$
P(y_1, y_2, ..., y_N | f(x_1), f(x_2), ..., f(x_N)) = \prod_{i=1}^{N} P(y_i | f(x_i))
$$

Bu **m√ºst…ôqil hadis…ôl…ôrin birg…ô ehtimalƒ±nƒ±n** hesablanma qaydasƒ±dƒ±r!  

∆èg…ôr **hadis…ôl…ôr m√ºst…ôqil olmasa**, onda hasili yox, daha m√ºr…ôkk…ôb bir ≈ü…ôrti ehtimal d√ºsturu istifad…ô olunardƒ±.

---

### **2. Normal paylanmƒ±≈ü m√º≈üahid…ôl…ôr √º√ß√ºn ehtimal sƒ±xlƒ±ƒüƒ± funksiyasƒ±**
∆èg…ôr h…ôr bir **$y_i$** **normal paylanmƒ±≈üdƒ±r**, y…ôni:  

$$
y_i \sim \mathcal{N}(f(x_i), \sigma^2)
$$

o zaman ehtimal sƒ±xlƒ±ƒüƒ± funksiyasƒ± bel…ôdir:

$$
\omega(y_i | f(x_i)) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_i - f(x_i))^2}{2\sigma^2}}
$$

ƒ∞ndi **b√ºt√ºn m√º≈üahid…ôl…ôr m√ºst…ôqil olduƒüuna g√∂r…ô**, √ºmumi ehtimal sƒ±xlƒ±ƒüƒ± funksiyasƒ± (likelihood function) onlarƒ±n **hasili** olacaq:  

$$
L = \prod_{i=1}^{N} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_i - f(x_i))^2}{2\sigma^2}}
$$

---

### **3. Hasili niy…ô istifad…ô edirik?**
- ∆èg…ôr m√º≈üahid…ôl…ôr **bir-birind…ôn m√ºst…ôqildirs…ô**, **√ºmumi ehtimalƒ±** tapmaq √º√ß√ºn onlarƒ±n hasili g√∂t√ºr√ºl√ºr.
- **Normal paylanmada** h…ôr bir m√º≈üahid…ônin ehtimal sƒ±xlƒ±ƒüƒ± funksiyasƒ± eyni olduƒüu √º√ß√ºn **b√ºt√ºn m√º≈üahid…ôl…ôr √º√ß√ºn √ºmumi likelihood hasili** istifad…ô edilir.
- Bu metod **MLE (Maximum Likelihood Estimation - Maksimum Ehtimallƒ±q Qiym…ôtl…ôndirm…ôsi)** √º√ß√ºn …ôsasdƒ±r, √ß√ºnki **log-likelihood** g√∂t√ºr…ônd…ô hasili toplama √ßevirm…ôk olur v…ô hesablama daha rahat gedir.

---

### **N…ôtic…ô**
Ehtimallarƒ±n bir-birin…ô vurulmasƒ±nƒ±n s…ôb…ôbi **m√º≈üahid…ôl…ôrin statistik olaraq m√ºst…ôqil olmasƒ±dƒ±r**. **Statistikada m√ºst…ôqil hadis…ôl…ôrin birg…ô ehtimalƒ± onlarƒ±n hasili il…ô hesablanƒ±r**, buna g√∂r…ô d…ô **$\prod$** i≈üar…ôsi istifad…ô edilir.

## Variasiya (variance) v…ô standart sapma (standard deviation) *∆èsas F…ôrql…ôr

Variasiya (variance) v…ô standart sapma (standard deviation) statistikada dispersiyanƒ±n √∂l√ß√ºl…ôri kimi istifad…ô olunur. Aralarƒ±ndakƒ± …ôsas f…ôrq ondan ibar…ôtdir ki, standart sapma variansƒ±n kvadrat k√∂k√º kimi hesablanƒ±r.

### **1. Variasiya (Variance)**
- Variasiya, m…ôlumat d…ôst…ôsind…ôki d…ôy…ôrl…ôrin orta qiym…ôtd…ôn n…ô q…ôd…ôr uzaq olduƒüunu √∂l√ß…ôn statistik g√∂st…ôricidir.
- Formula:
  $$
  \sigma^2 = \frac{\sum (x_i - \mu)^2}{N}
  $$
  Burada:
  - $ x_i $ ‚Äî veril…ônl…ôrin f…ôrdi d…ôy…ôrl…ôri,
  - $ \mu $ ‚Äî orta qiym…ôt (mean),
  - $ N $ ‚Äî m√º≈üahid…ôl…ôrin √ºmumi sayƒ±.

- Variasiya kvadrat vahidl…ôrl…ô ifad…ô olunduƒüuna g√∂r…ô, onu birba≈üa interpretasiya etm…ôk √ß…ôtindir.

---

### **2. Standart Sapma (Standard Deviation)**
- Standart sapma, variansƒ±n kvadrat k√∂k√ºd√ºr v…ô m…ôlumatlarƒ±n yayƒ±lmasƒ±nƒ± √∂l√ßm…ôk √º√ß√ºn istifad…ô olunur.
- Formula:
  $$
  \sigma = \sqrt{\sigma^2}
  $$
  - Variansdan f…ôrqli olaraq, standart sapma ilkin veril…ônl…ôrl…ô eyni vahidd…ô √∂l√ß√ºl√ºr v…ô interpretasiya etm…ôk daha asandƒ±r.

---

### **∆èsas F…ôrql…ôr**
| X√ºsusiyy…ôt  | Variasiya (Variance) | Standart Sapma (Standard Deviation) |
|-------------|---------------------|-----------------------------------|
| Formula    | $ \sigma^2 = \frac{\sum (x_i - \mu)^2}{N} $ | $ \sigma = \sqrt{\sigma^2} $ |
| Vahid      | Kvadrat vahidl…ôrd…ô (m…ôs…ôl…ôn, $ cm^2 $, $ kg^2 $) | ∆èsas vahidd…ô (m…ôs…ôl…ôn, $ cm $, $ kg $) |
| Yorumlama  | Yayƒ±lmanƒ± kvadrat ≈ü…ôkild…ô g√∂st…ôrir, amma interpretasiya √ß…ôtindir | Yayƒ±lmanƒ± daha aydƒ±n g√∂st…ôrir v…ô ≈ü…ôrh etm…ôk asandƒ±r |
| M√ºnasiblik | R…ôq…ômsal hesablamalar √º√ß√ºn …ôlveri≈ülidir | Statistik analizl…ôrd…ô daha √ßox istifad…ô olunur |

Qƒ±saca des…ôk, **varians yayƒ±lmanƒ±n kvadrat √∂l√ß√ºs√ºd√ºr, standart sapma is…ô daha real interpretasiya √º√ß√ºn onun kvadrat k√∂k√ºd√ºr.**

## Ehtimal N…ôz…ôriyy…ôsind…ô Ba≈ülanƒüƒ±c Anlar v…ô Riyazi G√∂zl…ôm…ô

### **Ehtimal N…ôz…ôriyy…ôsind…ô Ba≈ülanƒüƒ±c Anlar v…ô Riyazi G√∂zl…ôm…ô**

Ehtimal n…ôz…ôriyy…ôsind…ô **ba≈ülanƒüƒ±c anlar (momentl…ôr)** t…ôsad√ºfi d…ôyi≈ü…ônin paylanmasƒ±nƒ± xarakteriz…ô ed…ôn m√ºh√ºm √∂l√ß√ºl…ôrdir. Burada **birinci ba≈ülanƒüƒ±c an**, **birinci qarƒ±≈üƒ±q ba≈ülanƒüƒ±c an**, **ikinci ba≈ülanƒüƒ±c an** v…ô **riyazi g√∂zl…ôm…ô** anlayƒ±≈ülarƒ±nƒ± izah ed…ôc…ôy…ôm.  

---

## **1. Birinci Ba≈ülanƒüƒ±c An (First Raw Moment) v…ô Riyazi G√∂zl…ôm…ô**
**Birinci ba≈ülanƒüƒ±c an**, ba≈üqa s√∂zl…ô **riyazi g√∂zl…ôm…ô** **(expected value, $\mathbb{E}[X]$)** t…ôsad√ºfi d…ôyi≈ü…ônin orta d…ôy…ôrini g√∂st…ôrir.  

**Diskret t…ôsad√ºfi d…ôyi≈ü…ôn √º√ß√ºn:**  
$$
M_1 = \mathbb{E}[X] = \sum_{i} x_i P(x_i)
$$

**K…ôsilm…ôz t…ôsad√ºfi d…ôyi≈ü…ôn √º√ß√ºn:**  
$$
M_1 = \mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) dx
$$

Empirik hesablama √º√ß√ºn:  
$$
M_1^* = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

üí° **Misal:**  
Dey…ôk ki, bir t…ôsad√ºfi d…ôyi≈ü…ôn $X$ d…ôy…ôrl…ôri 2, 3, 5, 7, 9-dur v…ô onlarƒ±n ehtimallarƒ± 0.1, 0.2, 0.3, 0.2, 0.2-dir. O zaman:

$$
M_1 = \mathbb{E}[X] = (2 \times 0.1) + (3 \times 0.2) + (5 \times 0.3) + (7 \times 0.2) + (9 \times 0.2) = 5.4
$$

Bu o dem…ôkdir ki, t…ôsad√ºfi d…ôyi≈ü…ônin g√∂zl…ônil…ôn orta qiym…ôti **5.4**-d√ºr.

---

## **2. ƒ∞kinci Ba≈ülanƒüƒ±c An (Second Raw Moment)**
ƒ∞kinci ba≈ülanƒüƒ±c an t…ôsad√ºfi d…ôyi≈ü…ônin **kvadratƒ±nƒ±n g√∂zl…ônil…ôn d…ôy…ôridir**.  
Bu **dispersiya hesablanmasƒ±nda istifad…ô olunur**.

$$
M_2 = \mathbb{E}[X^2] = \sum_{i} x_i^2 P(x_i)
$$

Empirik hesablama:  
$$
M_2^* = \frac{1}{N} \sum_{i=1}^{N} x_i^2
$$

**Bu Dispersiyanƒ± hesablanmasƒ±nda istifad…ô olunur:**
$$
Var(X) = M_2 - (\mathbb{E}[X])^2
$$

üí° **Misal:**  
∆èvv…ôlki **$X$ d…ôy…ôrl…ôri √º√ß√ºn** ikinci ba≈ülanƒüƒ±c an bel…ô olacaq:

$$
M_2 = (2^2 \times 0.1) + (3^2 \times 0.2) + (5^2 \times 0.3) + (7^2 \times 0.2) + (9^2 \times 0.2)
$$

$$
M_2 = (4 \times 0.1) + (9 \times 0.2) + (25 \times 0.3) + (49 \times 0.2) + (81 \times 0.2) = 34.2
$$

**Dispersiya:**
$$
Var(X) = 34.2 - (5.4)^2 = 34.2 - 29.16 = 5.04
$$

## **3. Birinci Qarƒ±≈üƒ±q Ba≈ülanƒüƒ±c An (First Mixed Moment)**
**Qarƒ±≈üƒ±q ba≈ülanƒüƒ±c anlar** bir ne√ß…ô t…ôsad√ºfi d…ôyi≈ü…ôn arasƒ±nda asƒ±lƒ±lƒ±qlarƒ± √∂yr…ônm…ôk √º√ß√ºn istifad…ô olunur.  
M…ôs…ôl…ôn, iki d…ôyi≈ü…ôn $X$ v…ô $Y$ √º√ß√ºn **birinci qarƒ±≈üƒ±q ba≈ülanƒüƒ±c an** bu d√ºsturla verilir:

$$
M_{xy} = \mathbb{E}[X Y] = \sum_{i} x_i y_i P(x_i, y_i)
$$

Empirik hesablama √º√ß√ºn:  
$$
M_{xy}^* = \frac{1}{N} \sum_{i=1}^{N} x_i y_i
$$

**Bu kovariasiya hesablanmasƒ±nda istifad…ô olunur:** 

$$
Cov(X, Y) = \mathbb{E}[X Y] - \mathbb{E}[X] \mathbb{E}[Y]
$$

üí° **Misal:**  
∆èg…ôr $X$ v…ô $Y$ d…ôyi≈ü…ônl…ôri √º√ß√ºn verilmi≈ü d…ôy…ôrl…ôr v…ô ehtimallar varsa, biz $M_{xy}$-ni hesablaya bil…ôrik. Bu g√∂st…ôrici $X$ v…ô $Y$ arasƒ±nda …ôlaq…ônin olub-olmadƒ±ƒüƒ±nƒ± g√∂st…ôrir.

```python
# Jupyter Notebook-da interaktiv qrafikl…ôrin g√∂st…ôrilm…ôsi √º√ß√ºn %matplotlib notebook komandasƒ±nƒ± istifad…ô edirik
# %matplotlib notebook

# NumPy kitabxanasƒ±nƒ± idxal edir, riyazi …ôm…ôliyyatlar √º√ß√ºn istifad…ô olunur
import numpy as np  
# Matplotlib kitabxanasƒ±nƒ± idxal edir, qrafikl…ôr √º√ß√ºn istifad…ô olunur
import matplotlib.pyplot as plt  

# Eksperimentl…ôrin sayƒ±
N = 100     
# M√º≈üahid…ô edil…ôn d…ôy…ôrl…ôrin standart sapmasƒ±
sigma = 3   
# Teorik k parametrinin qiym…ôti
k = 0.5     
# Teorik b parametrinin qiym…ôti
b = 2       

# x d…ôy…ôrl…ôrini 0-dan N-…ô q…ôd…ôr aralƒ±qda yaratmaq
x = np.array(range(N))  
# f d…ôy…ôrl…ôri, h…ôr bir x √º√ß√ºn k*x + b funksiyasƒ±nƒ±n d…ôy…ôrl…ôri
f = np.array([k*z+b for z in range(N)])  
# y d…ôy…ôrl…ôri f d…ôy…ôrl…ôrin…ô t…ôsad√ºfi normal paylanma …ôlav…ô edil…ôr…ôk yaradƒ±lƒ±r
y = f + np.random.normal(0, sigma, N)  


# Teorik funksiyanƒ± √ß…ôkirik
plt.plot(f)  
# x v…ô y d…ôy…ôrl…ôrini qƒ±rmƒ±zƒ± n√∂qt…ôl…ôr ≈ü…ôklind…ô t…ôsvir etm…ôk
plt.scatter(x, y, s=2, c='red')  
plt.show() 
```

## Python dilind…ô izah

```python
# Jupyter Notebook-da interaktiv qrafikl…ôrin g√∂st…ôrilm…ôsi √º√ß√ºn %matplotlib notebook komandasƒ±nƒ± istifad…ô edirik
# %matplotlib notebook

# NumPy kitabxanasƒ±nƒ± idxal edir, riyazi …ôm…ôliyyatlar √º√ß√ºn istifad…ô olunur
import numpy as np  
# Matplotlib kitabxanasƒ±nƒ± idxal edir, qrafikl…ôr √º√ß√ºn istifad…ô olunur
import matplotlib.pyplot as plt  

# Eksperimentl…ôrin sayƒ±
N = 100     
# M√º≈üahid…ô edil…ôn d…ôy…ôrl…ôrin standart sapmasƒ±
sigma = 3   
# Teorik k parametrinin qiym…ôti
k = 0.5    
# Teorik b parametrinin qiym…ôti
b = 2       

# x d…ôy…ôrl…ôrini 0-dan N-…ô q…ôd…ôr aralƒ±qda yaratmaq
x = np.array(range(N))  
# f d…ôy…ôrl…ôri, h…ôr bir x √º√ß√ºn k*x + b funksiyasƒ±nƒ±n d…ôy…ôrl…ôri
f = np.array([k*z+b for z in range(N)])  
# y d…ôy…ôrl…ôri f d…ôy…ôrl…ôrin…ô t…ôsad√ºfi normal paylanma …ôlav…ô edil…ôr…ôk yaradƒ±lƒ±r
y = f + np.random.normal(0, sigma, N)  

# Koefisiyentl…ôri hesablamaq √º√ß√ºn a≈üaƒüƒ±dakƒ± …ôm…ôliyyatlar
mx = x.sum()/N  # x d…ôy…ôrl…ôrinin ortalamasƒ±nƒ± hesablamaq
my = y.sum()/N  # y d…ôy…ôrl…ôrinin ortalamasƒ±nƒ± hesablamaq
a2 = np.dot(x.T, x)/N  # x d…ôy…ôrl…ôrinin kvadratlarƒ±nƒ±n ortalamasƒ±nƒ± hesablamaq
a11 = np.dot(x.T, y)/N  # x v…ô y d…ôy…ôrl…ôrinin skalyar hasilini hesablamaq

# Regresiya x…ôttinin …ômsalƒ±nƒ± (k) hesablamaq
kk = (a11 - mx*my)/(a2 - mx**2)  
# Regresiya x…ôttinin interceptini (b) hesablamaq
bb = my - kk*mx  
# Regresiya x…ôttinin y t…ôxminl…ôrini yaratmaq
ff = np.array([kk*z+bb for z in range(N)])  

# x v…ô y d…ôy…ôrl…ôrini qƒ±rmƒ±zƒ± n√∂qt…ôl…ôr ≈ü…ôklind…ô t…ôsvir etm…ôk
plt.scatter(x, y, s=2, c='red')  
# Teorik funksiyanƒ± √ß…ôkirik
plt.plot(f)  
# Hesablanmƒ±≈ü regresiya x…ôttini qƒ±rmƒ±zƒ± r…ôngd…ô √ß…ôkirik
plt.plot(ff, c='red')  
# Qrafikin √ºz…ôrind…ô grid x…ôtl…ôri g√∂st…ôrm…ôk
plt.grid(True)  
# Qrafiki ekranda g√∂st…ôrm…ôk
plt.show()

print(f"Teorik k={k}, Proqnoz: k={kk}")
print(f"Teorik b={b}, Proqnoz: b={bb}")
```
