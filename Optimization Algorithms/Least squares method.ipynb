## Standart Qaus (Normal) sapması

Standart Qaus (Normal) sapması (standart sapma) statistikada dispersiyanın kvadrat kökü kimi müəyyən edilir və verilənlərin orta qiymətdən nə qədər yayındığını ölçür. Normal paylanma üçün standart sapma aşağıdakı düsturla hesablanır:

$$
\sigma = \sqrt{\frac{\sum (x_i - \mu)^2}{N}}
$$

Burada:
- $ \sigma $ — standart sapmadır,
- $ x_i $ — verilənlərin i-ci elementi,
- $ \mu $ — orta qiymət ($\mu = \frac{\sum x_i}{N}$),
- $ N $ — ümumi müşahidələrin sayı.

Əgər nümunə üzərində hesablanırsa, o zaman düstur dəyişir və $ N $ əvəzinə $ N-1 $ istifadə olunur:

$$
s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{N-1}}
$$

Bu zaman $ s $ nümunə standart sapması adlanır.

Standart sapma böyük olduqda, verilənlər orta qiymətdən daha çox yayılmış olur, kiçik olduqda isə verilənlər orta qiymət ətrafında cəmlənmiş olur.

Sən normal paylanmada standart sapmanın təsirini və tətbiq sahələrini daha ətraflı öyrənmək istəyirsən?

## Standart Qaus (Normal) paylanmasının sıxlıq funksiyası (Probability Density Function, PDF)

Standart Qaus (Normal) paylanmasının sıxlıq funksiyası **(Probability Density Function, PDF)** belə verilir:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Burada:  
- $ x $ — təsadüfi dəyişən,  
- $ \mu $ — orta qiymət,  
- $ \sigma $ — standart sapma,  
- $ e $ — təqribən 2.718 olan Eyler sabiti,  
- $ \pi $ — 3.1416 olan Pi sabiti.  

### **Xüsusiyyətləri:**
1. **Simmetrikdir**: Orta qiymət ətrafında simmetrik olur.
2. **Bell-şəkilli (zəng formalı) qrafikə malikdir**.
3. **Tək maksimum nöqtəsi var**: Bu nöqtə $ x = \mu $ -da yerləşir.
4. **Standart sapma dağılımın enini təyin edir**:
   - $ \mu \pm \sigma $ aralığında təxminən **68%**,
   - $ \mu \pm 2\sigma $ aralığında **95%**,
   - $ \mu \pm 3\sigma $ aralığında **99.7%** müşahidələr olur (**Empirik qayda**).

Əgər **standart normal paylanmadan** danışırıqsa, yəni $ \mu = 0 $ və $ \sigma = 1 $, o zaman sıxlıq funksiyası sadələşir:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

Bu paylanmanın qrafiki **Z-şkalası** ilə ölçülən standart normal paylanmanı göstərir.

Sən hansısa konkret tətbiq üçün bu funksiyanın istifadəsini öyrənmək istəyirsən?

## Standart Qaus (Normal) paylanmasının sıxlıq funksiyası (PDF) Görsəlləşdirilməsi

### **Standart Qaus Sapması nədir?**  

**Standart Qaus Sapması** (**Gaussian Standard Deviation**) statistikada və siqnal/görüntü emalında **normal paylanmanın yayılma dərəcəsini** ölçən parametrdir. Adətən **$ \sigma $** (sigma) ilə işarə olunur və **standart sapma** (standard deviation) kimi tanınır.

---

## **1. Standart Sapmanın Riyazi İzahı**  

Normal (Qaus) paylanmanın ümumi tənliyi belədir:  

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Burada:  
- **$ \mu $** — ortalama (mean), paylanmanın mərkəzi.  
- **$ \sigma $** — standart sapma, yayılmanın ölçüsü.  
- **$ x $** — verilmiş dəyər.  
- **$ e $** — təbii logaritmanın əsası ($ \approx 2.718 $).  

**$ \sigma $** **böyüdükcə paylanma daha genişlənir və kənarlara yayılır**, **$ \sigma $** **kiçildikcə paylanma daralır və mərkəzdə cəmlənir.**

---

## **2. Standart Sapmanın Görsəlləşdirilməsi**  

- **Kiçik $ \sigma $** → Paylanma dar olur, məlumatlar orta ətrafında toplanır.  
- **Böyük $ \sigma $** → Paylanma geniş olur, məlumatlar daha çox yayılır.  

📊 **Normal paylanmanın fərqli standart sapmalara görə dəyişməsi**:

- **$ \sigma = 0.5 $** (çox dar, kəskin pik)  
- **$ \sigma = 1 $** (normal yayılma)  
- **$ \sigma = 2 $** (geniş yayılma)  



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-5, 5, 1000)
mu = 0

plt.plot(x, norm.pdf(x, mu, 0.5), label=r'$\sigma=0.5$')
plt.plot(x, norm.pdf(x, mu, 1), label=r'$\sigma=1$')
plt.plot(x, norm.pdf(x, mu, 2), label=r'$\sigma=2$')

plt.legend()
plt.title("Normal Paylanmanın Fərqli Standart Sapmalarla Görsəlləşdirilməsi")
plt.xlabel("x")
plt.ylabel("Ehtimallıq sıxlığı")
plt.grid()
plt.show()
```

Bu qrafikdə $ \sigma $ artdıqca paylanmanın genişləndiyini görürük.

## $$L = \prod_{i=1}^{N} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_i - f(x_i))^2}{2\sigma^2}}$$

Sənin verdiyin ifadədə **$\prod$** işarəsi ehtimalların hasili kimi göstərilib. **Bəs niyə bu ehtimallar bir-birinə vurulur?**  

Bu, **statistikada müşahidələr müstəqil olduğu halda ehtimalın hesablanma qaydasıdır**. Gəlin, addım-addım izah edim.  

---

### **1. Bayes nəzəriyyəsi və müstəqillik anlayışı**  
Tutaq ki, **$N$ müşahidə var** və hər bir müşahidə **$y_i$** verilmiş **$x_i$** üçün **şərti ehtimalla** ifadə olunur:  

$$
\omega(y_i | f(x_i))
$$

Əgər müşahidələr **bir-birindən asılı deyilsə (yəni statistik olaraq müstəqildirsə)**, o zaman **ümumi ehtimalı** hesablamaq üçün onların **hasili** götürülür:  

$$
P(y_1, y_2, ..., y_N | f(x_1), f(x_2), ..., f(x_N)) = \prod_{i=1}^{N} P(y_i | f(x_i))
$$

Bu **müstəqil hadisələrin birgə ehtimalının** hesablanma qaydasıdır!  

Əgər **hadisələr müstəqil olmasa**, onda hasili yox, daha mürəkkəb bir şərti ehtimal düsturu istifadə olunardı.

---

### **2. Normal paylanmış müşahidələr üçün ehtimal sıxlığı funksiyası**
Əgər hər bir **$y_i$** **normal paylanmışdır**, yəni:  

$$
y_i \sim \mathcal{N}(f(x_i), \sigma^2)
$$

o zaman ehtimal sıxlığı funksiyası belədir:

$$
\omega(y_i | f(x_i)) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_i - f(x_i))^2}{2\sigma^2}}
$$

İndi **bütün müşahidələr müstəqil olduğuna görə**, ümumi ehtimal sıxlığı funksiyası (likelihood function) onların **hasili** olacaq:  

$$
L = \prod_{i=1}^{N} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y_i - f(x_i))^2}{2\sigma^2}}
$$

---

### **3. Hasili niyə istifadə edirik?**
- Əgər müşahidələr **bir-birindən müstəqildirsə**, **ümumi ehtimalı** tapmaq üçün onların hasili götürülür.
- **Normal paylanmada** hər bir müşahidənin ehtimal sıxlığı funksiyası eyni olduğu üçün **bütün müşahidələr üçün ümumi likelihood hasili** istifadə edilir.
- Bu metod **MLE (Maximum Likelihood Estimation - Maksimum Ehtimallıq Qiymətləndirməsi)** üçün əsasdır, çünki **log-likelihood** götürəndə hasili toplama çevirmək olur və hesablama daha rahat gedir.

---

### **Nəticə**
Ehtimalların bir-birinə vurulmasının səbəbi **müşahidələrin statistik olaraq müstəqil olmasıdır**. **Statistikada müstəqil hadisələrin birgə ehtimalı onların hasili ilə hesablanır**, buna görə də **$\prod$** işarəsi istifadə edilir.

## Variasiya (variance) və standart sapma (standard deviation) *Əsas Fərqlər

Variasiya (variance) və standart sapma (standard deviation) statistikada dispersiyanın ölçüləri kimi istifadə olunur. Aralarındakı əsas fərq ondan ibarətdir ki, standart sapma variansın kvadrat kökü kimi hesablanır.

### **1. Variasiya (Variance)**
- Variasiya, məlumat dəstəsindəki dəyərlərin orta qiymətdən nə qədər uzaq olduğunu ölçən statistik göstəricidir.
- Formula:
  $$
  \sigma^2 = \frac{\sum (x_i - \mu)^2}{N}
  $$
  Burada:
  - $ x_i $ — verilənlərin fərdi dəyərləri,
  - $ \mu $ — orta qiymət (mean),
  - $ N $ — müşahidələrin ümumi sayı.

- Variasiya kvadrat vahidlərlə ifadə olunduğuna görə, onu birbaşa interpretasiya etmək çətindir.

---

### **2. Standart Sapma (Standard Deviation)**
- Standart sapma, variansın kvadrat köküdür və məlumatların yayılmasını ölçmək üçün istifadə olunur.
- Formula:
  $$
  \sigma = \sqrt{\sigma^2}
  $$
  - Variansdan fərqli olaraq, standart sapma ilkin verilənlərlə eyni vahiddə ölçülür və interpretasiya etmək daha asandır.

---

### **Əsas Fərqlər**
| Xüsusiyyət  | Variasiya (Variance) | Standart Sapma (Standard Deviation) |
|-------------|---------------------|-----------------------------------|
| Formula    | $ \sigma^2 = \frac{\sum (x_i - \mu)^2}{N} $ | $ \sigma = \sqrt{\sigma^2} $ |
| Vahid      | Kvadrat vahidlərdə (məsələn, $ cm^2 $, $ kg^2 $) | Əsas vahiddə (məsələn, $ cm $, $ kg $) |
| Yorumlama  | Yayılmanı kvadrat şəkildə göstərir, amma interpretasiya çətindir | Yayılmanı daha aydın göstərir və şərh etmək asandır |
| Münasiblik | Rəqəmsal hesablamalar üçün əlverişlidir | Statistik analizlərdə daha çox istifadə olunur |

Qısaca desək, **varians yayılmanın kvadrat ölçüsüdür, standart sapma isə daha real interpretasiya üçün onun kvadrat köküdür.**

## Ehtimal Nəzəriyyəsində Başlanğıc Anlar və Riyazi Gözləmə

### **Ehtimal Nəzəriyyəsində Başlanğıc Anlar və Riyazi Gözləmə**

Ehtimal nəzəriyyəsində **başlanğıc anlar (momentlər)** təsadüfi dəyişənin paylanmasını xarakterizə edən mühüm ölçülərdir. Burada **birinci başlanğıc an**, **birinci qarışıq başlanğıc an**, **ikinci başlanğıc an** və **riyazi gözləmə** anlayışlarını izah edəcəyəm.  

---

## **1. Birinci Başlanğıc An (First Raw Moment) və Riyazi Gözləmə**
**Birinci başlanğıc an**, başqa sözlə **riyazi gözləmə** **(expected value, $\mathbb{E}[X]$)** təsadüfi dəyişənin orta dəyərini göstərir.  

**Diskret təsadüfi dəyişən üçün:**  
$$
M_1 = \mathbb{E}[X] = \sum_{i} x_i P(x_i)
$$

**Kəsilməz təsadüfi dəyişən üçün:**  
$$
M_1 = \mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) dx
$$

Empirik hesablama üçün:  
$$
M_1^* = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

💡 **Misal:**  
Deyək ki, bir təsadüfi dəyişən $X$ dəyərləri 2, 3, 5, 7, 9-dur və onların ehtimalları 0.1, 0.2, 0.3, 0.2, 0.2-dir. O zaman:

$$
M_1 = \mathbb{E}[X] = (2 \times 0.1) + (3 \times 0.2) + (5 \times 0.3) + (7 \times 0.2) + (9 \times 0.2) = 5.4
$$

Bu o deməkdir ki, təsadüfi dəyişənin gözlənilən orta qiyməti **5.4**-dür.

---

## **2. İkinci Başlanğıc An (Second Raw Moment)**
İkinci başlanğıc an təsadüfi dəyişənin **kvadratının gözlənilən dəyəridir**.  
Bu **dispersiya hesablanmasında istifadə olunur**.

$$
M_2 = \mathbb{E}[X^2] = \sum_{i} x_i^2 P(x_i)
$$

Empirik hesablama:  
$$
M_2^* = \frac{1}{N} \sum_{i=1}^{N} x_i^2
$$

**Bu Dispersiyanı hesablanmasında istifadə olunur:**
$$
Var(X) = M_2 - (\mathbb{E}[X])^2
$$

💡 **Misal:**  
Əvvəlki **$X$ dəyərləri üçün** ikinci başlanğıc an belə olacaq:

$$
M_2 = (2^2 \times 0.1) + (3^2 \times 0.2) + (5^2 \times 0.3) + (7^2 \times 0.2) + (9^2 \times 0.2)
$$

$$
M_2 = (4 \times 0.1) + (9 \times 0.2) + (25 \times 0.3) + (49 \times 0.2) + (81 \times 0.2) = 34.2
$$

**Dispersiya:**
$$
Var(X) = 34.2 - (5.4)^2 = 34.2 - 29.16 = 5.04
$$

---

## **3. Birinci Qarışıq Başlanğıc An (First Mixed Moment)**
**Qarışıq başlanğıc anlar** bir neçə təsadüfi dəyişən arasında asılılıqları öyrənmək üçün istifadə olunur.  
Məsələn, iki dəyişən $X$ və $Y$ üçün **birinci qarışıq başlanğıc an** bu düsturla verilir:

$$
M_{xy} = \mathbb{E}[X Y] = \sum_{i} x_i y_i P(x_i, y_i)
$$

Empirik hesablama üçün:  
$$
M_{xy}^* = \frac{1}{N} \sum_{i=1}^{N} x_i y_i
$$

**Bu kovariasiya hesablanmasında istifadə olunur:** 

$$
Cov(X, Y) = \mathbb{E}[X Y] - \mathbb{E}[X] \mathbb{E}[Y]
$$

💡 **Misal:**  
Əgər $X$ və $Y$ dəyişənləri üçün verilmiş dəyərlər və ehtimallar varsa, biz $M_{xy}$-ni hesablaya bilərik. Bu göstərici $X$ və $Y$ arasında əlaqənin olub-olmadığını göstərir.

---

## **Nəticə**
| An | Riyazi Tərif | Mənası |
|----|--------------|--------|
| **Birinci başlanğıc an ($M_1$)** | $ \mathbb{E}[X] $ | Orta qiymət |
| **İkinci başlanğıc an ($M_2$)** | $ \mathbb{E}[X^2] $ | Kvadrat ortalama |
| **Birinci qarışıq başlanğıc an ($M_{xy}$)** | $ \mathbb{E}[XY] $ | İki dəyişən arasındakı orta əlaqə |

Bu anlayışlar ehtimal nəzəriyyəsində statistik xarakteristikaların təhlili üçün əsasdır.

```python
# Jupyter Notebook-da interaktiv qrafiklərin göstərilməsi üçün %matplotlib notebook komandasını istifadə edirik
# %matplotlib notebook

# NumPy kitabxanasını idxal edir, riyazi əməliyyatlar üçün istifadə olunur
import numpy as np  
# Matplotlib kitabxanasını idxal edir, qrafiklər üçün istifadə olunur
import matplotlib.pyplot as plt  

# Eksperimentlərin sayı
N = 100     
# Müşahidə edilən dəyərlərin standart sapması
sigma = 3   
# Teorik k parametrinin qiyməti
k = 0.5     
# Teorik b parametrinin qiyməti
b = 2       

# x dəyərlərini 0-dan N-ə qədər aralıqda yaratmaq
x = np.array(range(N))  
# f dəyərləri, hər bir x üçün k*x + b funksiyasının dəyərləri
f = np.array([k*z+b for z in range(N)])  
# y dəyərləri f dəyərlərinə təsadüfi normal paylanma əlavə edilərək yaradılır
y = f + np.random.normal(0, sigma, N)  


# Teorik funksiyanı çəkirik
plt.plot(f)  
# x və y dəyərlərini qırmızı nöqtələr şəklində təsvir etmək
plt.scatter(x, y, s=2, c='red')  
plt.show() 
```

## Python dilində izah

```python
# Jupyter Notebook-da interaktiv qrafiklərin göstərilməsi üçün %matplotlib notebook komandasını istifadə edirik
# %matplotlib notebook

# NumPy kitabxanasını idxal edir, riyazi əməliyyatlar üçün istifadə olunur
import numpy as np  
# Matplotlib kitabxanasını idxal edir, qrafiklər üçün istifadə olunur
import matplotlib.pyplot as plt  

# Eksperimentlərin sayı
N = 100     
# Müşahidə edilən dəyərlərin standart sapması
sigma = 3   
# Teorik k parametrinin qiyməti
k = 0.5    
# Teorik b parametrinin qiyməti
b = 2       

# x dəyərlərini 0-dan N-ə qədər aralıqda yaratmaq
x = np.array(range(N))  
# f dəyərləri, hər bir x üçün k*x + b funksiyasının dəyərləri
f = np.array([k*z+b for z in range(N)])  
# y dəyərləri f dəyərlərinə təsadüfi normal paylanma əlavə edilərək yaradılır
y = f + np.random.normal(0, sigma, N)  

# Koefisiyentləri hesablamaq üçün aşağıdakı əməliyyatlar
mx = x.sum()/N  # x dəyərlərinin ortalamasını hesablamaq
my = y.sum()/N  # y dəyərlərinin ortalamasını hesablamaq
a2 = np.dot(x.T, x)/N  # x dəyərlərinin kvadratlarının ortalamasını hesablamaq
a11 = np.dot(x.T, y)/N  # x və y dəyərlərinin skalyar hasilini hesablamaq

# Regresiya xəttinin əmsalını (k) hesablamaq
kk = (a11 - mx*my)/(a2 - mx**2)  
# Regresiya xəttinin interceptini (b) hesablamaq
bb = my - kk*mx  
# Regresiya xəttinin y təxminlərini yaratmaq
ff = np.array([kk*z+bb for z in range(N)])  

# x və y dəyərlərini qırmızı nöqtələr şəklində təsvir etmək
plt.scatter(x, y, s=2, c='red')  
# Teorik funksiyanı çəkirik
plt.plot(f)  
# Hesablanmış regresiya xəttini qırmızı rəngdə çəkirik
plt.plot(ff, c='red')  
# Qrafikin üzərində grid xətləri göstərmək
plt.grid(True)  
# Qrafiki ekranda göstərmək
plt.show()

print(f"Teorik k={k}, Proqnoz: k={kk}")
print(f"Teorik b={b}, Proqnoz: b={bb}")
```
