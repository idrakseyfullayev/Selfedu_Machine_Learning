## Gradient descent alqoritmasında gradientin normallaşdırılması

Gradient descent alqoritmasında gradientin normallaşdırılması, ümumiyyətlə, iki əsas səbəbdən edilir:

### 1. **Öyrənmə Sürətinin Stabilizasiyası (Learning Rate Control)**
Gradient descent alqoritmasında, hər iterasiya üçün parameterlərin dəyişməsi gradientin dəyəri ilə ölçülür. Lakin gradientin ölçüsü (norması) çox böyük və ya çox kiçik ola bilər. Bu, öyrənmə sürətini tənzimləməyin çətin olmasına səbəb olur və nəticədə modelin konvergensiya sürətini azalda bilər.

- **Böyük gradientlər**: Əgər gradient çox böyükdürsə, parametrlər çox sürətlə dəyişir və bu, optimum həlli tapmağı çətinləşdirə bilər (overshooting problemi). 
- **Kiçik gradientlər**: Əgər gradient çox kiçikdirsə, çox kiçik addımlar atılacaq və bu da modelin çox yavaş öyrənməsinə səbəb olar.

Gradienti normallaşdırmaqla, bu dəyişkənliyi və fərqləri ortadan qaldıraraq, daha sabit və effektiv öyrənmə sürəti təmin edilə bilər.

### 2. **Çeşidli Əmsalların Uyğunlaşdırılması (Feature Scaling)**
Əgər modelin təlimi çoxsaylı xüsusiyyətlərdən (features) ibarətdirsə və bu xüsusiyyətlərin ölçüləri bir-birindən çox fərqlənirsə (məsələn, biri 1000-ə qədər qiymətə sahib olsa, digəri 0 ilə 1 arasında ola bilər), gradientin ölçüsü də bu xüsusiyyətlərə əsasən dəyişə bilər. Bu, öyrənmə prosesini çətinləşdirə bilər.

- **Çox fərqli miqyaslar**: Əgər xüsusiyyətlər fərqli miqyaslarda (məsələn, biri 1000, digəri isə 0 ilə 1 arasında) olsa, bəzən bəzi xüsusiyyətlər (yüksək dəyərə sahib olanlar) daha çox təsir edəcək, nəticədə modelin öyrənmə prosesi qeyri-bərabər olacaq.
  
Gradientin normallaşdırılması, xüsusiyyətlər arasındakı miqyas fərqlərini aradan qaldırmaqla, bütün xüsusiyyətlərə eyni dərəcədə təsir etməyə kömək edir. Bu, alqoritmanın daha sürətli və düzgün konvergensiyasını təmin edir.

### Nəticə olaraq:
Gradientin normallaşdırılması öyrənmə sürətini daha sabit hala gətirir və xüsusiyyətlər arasında miqyas fərqlərini azaldır, nəticədə daha sürətli və səmərəli öyrənmə təmin edir.

## Sign funksiyası

**Gradient Descent** alqoritmasında **sign funksiyası** adətən funksiyanın gradientini (yəni, törəməsini) istifadə etməklə addım ölçüsünü müəyyən etməkdə istifadə olunur. **sign** funksiyası, funksiyanın tərs işarəsini (yəni, mənfi və ya müsbət olmasını) təyin edir.

### **Sign Funksiyasının Tərifi:**

$$
\text{sign}(x) =
\begin{cases}
1, & \text{əgər } x > 0 \\
0, & \text{əgər } x = 0 \\
-1, & \text{əgər } x < 0
\end{cases}
$$

Bu, **x**-in işarəsini təyin edən bir funksiyadır:
- Əgər $ x $ müsbətdirsə, $\text{sign}(x) = 1$,
- Əgər $ x $ mənfidirsə, $\text{sign}(x) = -1$,
- Əgər $ x $ sıfıra bərabərdirsə, $\text{sign}(x) = 0$.

### **Gradient Descent-də Sign Funksiyasının İstifadəsi:**

Gradient Descent metodunda ümumiyyətlə, funksiyanı minimuma endirmək üçün **gradient** (törəmə) istifadə olunur. **Gradient**, funksiyanın nöqtəsindəki dəyişiklik sürətini və istiqamətini təyin edir. Bu istiqamətə görə, addım atılır və minimizasiyanı sürətləndirir.

Ancaq bəzi hallarda, funksiyanın törəməsi (gradienti) sadəcə mənfi və ya müsbət qiymət daşıyan qiymətlərdən ibarət ola bilər. **sign** funksiyası, **gradient descent** metodunda təkcə **gradientin işarəsini** təyin etməyə kömək edir, yəni **dəyişiklik istiqamətini**.

Misal üçün, **sign funksiyasının tətbiqi**:

- Əgər **df(x)** müsbət dəyərə malikdirsə, bu, funksiyanın **artdığını** göstərir və addım ölçüsü müsbət olacaq.
- Əgər **df(x)** mənfi dəyərə malikdirsə, bu, funksiyanın **azaldığını** göstərir və addım ölçüsü mənfi olacaq.
- Əgər **df(x)** sıfırdırsa, heç bir dəyişiklik olmayacaq.

### **Kodda Sign Funksiyasının Tətbiqi:**

Baxaq, aşağıdakı **Gradient Descent** alqoritmasında **sign funksiyasının** tətbiqinə:

```python
xx = xx - lmd * np.sign(df(xx))  # Cari iterasiyada addımın ölçüsünü dəyişdiririk
```

Bu, sadəcə **df(x)** (yəni, funksiyanın törəməsi) işarəsini istifadə edərək **xx** qiymətini yeniləyir. Burada:

- **$ lmd $**: Addım ölçüsüdür (learning rate).
- **$ np.sign(df(xx)) $**: Törəmənin işarəsini alır (təsəvvür edin ki, **df(xx)** funksiyasının dəyəri müsbət və ya mənfi olacaq).

Əgər **df(xx)** müsbət olarsa, **sign(df(xx)) = 1** olacaq və **xx**-i azaltmaq üçün **xx = xx - lmd** olacaq. Əgər **df(xx)** mənfi olarsa, **sign(df(xx)) = -1** olacaq və **xx**-i artırmaq üçün **xx = xx + lmd** olacaq.

### **Nə Üçün Sign Funksiyası İstifadə Edilir?**

- **Sürət** və **stabilizasiya**: Bu yanaşma, optimallaşdırmanı daha sürətli hala gətirə bilər, çünki adətən qismən sabit addım ölçüsü ilə çalışır. Sadəcə **dirençli** addımlar istifadə edərək daha dəqiq nəticələr əldə edilir.
  
- **Qlobal Minimumun Axtarılması**: Əgər bir çox yerli minimumlar varsa, sign funksiyasının istifadə edilməsi təkrarlanan, daha stabil addım ölçüsünü və daha yaxşı tərs işarə ilə addım atılmasını təmin edir.

### **Misal:**

Əgər funksiyanız $ f(x) = \sin(x) + 0.5x $ kimi bir funksiya olsa, **sign(df(x))** istifadə edərək bu funksiyanı optimallaşdırmaq istəsəniz:

- $ df(x) = \cos(x) + 0.5 $
- Törəmənin işarəsini dəyişdirərək (müsbət və ya mənfi) addımı müəyyən edirsiniz.

Əgər $ \cos(x) + 0.5 > 0 $, bu zaman nöqtənin yuxarı hissəsində oluruq, və buna görə funksiyanı aşağı salmaq üçün **xx**-i kiçiltmək lazımdır.
  
Əks halda, $ \cos(x) + 0.5 < 0 $, bu zaman funksiyanı yuxarı qaldırmaq üçün **xx**-i artırmaq lazımdır.

Bu cür yanaşmalar həm **Stochastic Gradient Descent** (SGD) kimi yanaşmalarda da istifadə edilə bilər.

## $\frac{1}{\min(i + 1, mn)}$ ifadəsi  funksiyanı optimallaşdırmaq

$$\frac{1}{\min(i + 1, mn)}$$ ifadəsi, müəyyən bir funksiyanı optimallaşdırmaq və ya təkrarlanan bir hesablamada addım ölçüsünü tənzimləmək üçün istifadə olunur. İndi bu ifadənin hissələrinə nəzər salaq:

### İfadənin Hissələri:
1. **$i + 1$**: Bu, iterasiyanın cari addımını göstərir. $i$ ümumiyyətlə dövrün cari mərhələsini təmsil edir. $i + 1$ isə növbəti mərhələnin sayını göstərir. Məsələn, birinci iterasiya üçün $i = 0$, ikinci iterasiya üçün $i = 1$ və s.
   
2. **$mn$**: Bu, ümumiyyətlə, müəyyən edilmiş bir sərhəd və ya maksimum dəyəri təmsil edir. Məsələn, əgər təkrarlamalar çox böyükdürsə, addım ölçüsünü kiçiltmək üçün bir limit kimi çıxış edir.

3. **$\min(i + 1, mn)$**: Bu, ən kiçik dəyəri seçir. Yəni, $i + 1$ ilə $mn$ arasından hansının daha kiçik olduğunu müəyyən edir. Bu, təkrarlanan prosesin sayı ilə əlaqəli olaraq addım ölçüsünün tənzimlənməsi üçün vacibdir.

4. **$\frac{1}{\min(i + 1, mn)}$**: İndi bu ifadənin tam mənası başa düşülür. Əgər $i + 1$ kiçikdirsə, yəni başlanğıcda çox sürətli bir şəkildə irəliləyirsinizsə, addım ölçüsü böyük olur. Amma iterasiyalar artdıqca, adətən optimallaşdırma prosesinin daha dəqiq olması gözlənildiyi üçün, addım ölçüsünü azaldır.

### Funksiyada Yeri:
Bu tip ifadələr, xüsusən optimallaşdırma metodlarında, məsələn, **Gradient Descent** və **Stochastic Gradient Descent**-də (SGD) istifadə olunur. İterasiya sayının artması ilə addım ölçüsünü kiçiltmək, metodu daha sabit və dəqiq hala gətirir. Bu da modelin daha yaxşı konvergensiyasını təmin edir.

### Nümunə:
Məsələn, əgər $i = 5$ və $mn = 10$ olduqda:

$$
\frac{1}{\min(5 + 1, 10)} = \frac{1}{\min(6, 10)} = \frac{1}{6}
$$

Əgər $i = 15$ və $mn = 10$ olduqda:

$$
\frac{1}{\min(15 + 1, 10)} = \frac{1}{\min(16, 10)} = \frac{1}{10}
$$

Beləliklə, iterasiya sayı artdıqca, addım ölçüsü tənzimlənir və optimallaşdırma daha dəqiq olur.

## Python dilində  izah

````python
# Jupyter Notebook-da interaktiv qrafiklərin göstərilməsi üçün %matplotlib notebook komandasını istifadə edirik
%matplotlib notebook  

# Lazımi kitabxanaları daxil edirik
import time                # Vaxt gecikmələrini idarə etmək üçün
import numpy as np         # Numpy kitabxanası, riyazi əməliyyatlar üçün
import matplotlib.pyplot as plt  # Qrafikləri vizuallaşdırmaq üçün matplotlib

# Funksiyanı təyin edirik: f(x) = x^2 - 5x + 5
def f(x):
    return x*x - 5*x + 5

# Törəməni təyin edirik: df(x) = 2x - 5
def df(x):
    return 2*x - 5

# Iterasiya sayı (bu halda 20 dəfə təkrarlayacağıq)
N = 20 

# Başlanğıc nöqtəsini təyin edirik, burada x = 0 olaraq başlayacaq
xx = 0    

# Yanaşma addımını təyin edirik, öyrənmə sürəti (0.9)
lmd = 0.9 

# Qrafik üçün x koordinatlarını təyin edirik (0-dan 5-ə qədər, 0.1 interval ilə)
x_plt = np.arange(0, 5.0, 0.1)

# f(x)-i hər x üçün hesablamaq və nəticələri f_plt listinə saxlamaq
f_plt = [f(x) for x in x_plt]

# Jupyter Notebook-da interaktiv rejimi aktivləşdiririk
plt.ion()  

# Yeni bir şəkil (pəncərə) və oxlar yaradırıq
fig, ax = plt.subplots()    

# Qrafikdə şəbəkə xətləri göstərmək üçün grid əlavə edirik
ax.grid(True)   

# Funksiyanı qrafiklə göstəririk
ax.plot(x_plt, f_plt)                  

# Başlanğıc nöqtəsini qırmızı rəngdə göstəririk
point = ax.scatter(xx, f(xx), c='red')  

# Gradient descent (və ya başqa adı ilə ən kiçik kvadratlar metodu) ilə iterasiyanı başladırıq
for i in range(N):
    # Cari nöqtədən təxmin etməyi yeniləyirik: xx = xx - yanaşma addımı * törəmə
    xx = xx - lmd*df(xx)    

    # Yeni nöqtənin mövqeyini (xx, f(xx)) ilə qrafikdə göstəririk
    point.set_offsets([xx, f(xx)])  

    # Qrafiki yenidən çəkmək, hər dəfə yeniləmələri göstərmək üçün:
    fig.canvas.draw()    # Qrafiki yeniləyirik
    fig.canvas.flush_events()  # Göstərilən dəyişiklikləri dərhal əks etdiririk
    time.sleep(0.02)     # Hər iterasiyadan sonra 20 millisecond gözləyirik

# Qrafikdə interaktiv rejimi dayandırırıq ki, qrafiklər sabit qalsın
plt.ioff()   

# Nəticəni çap edirik (son iterasiya nəticəsi)
print(xx)

# Son nöqtəni mavi rəngdə göstəririk
ax.scatter(xx, f(xx), c='blue')

# Qrafiki göstəririk (interaktiv rejim bitdikdən sonra sabit şəkildə)
plt.show()
````

## Python dilində izah: Sign və  $ \frac{1}{\min(i + 1, mn)}$ funksiyası ilə

````python
# Jupyter Notebook-da interaktiv qrafiklərin göstərilməsi üçün %matplotlib notebook komandasını istifadə edirik
%matplotlib notebook  

# Lazımi kitabxanaları daxil edirik
import time                # Vaxt gecikmələrini idarə etmək üçün
import numpy as np         # Numpy kitabxanası, riyazi əməliyyatlar üçün
import matplotlib.pyplot as plt  # Qrafikləri vizuallaşdırmaq üçün matplotlib

# Funksiyanı təyin edirik: f(x) = x^2 - 5x + 5
def f(x):
    return x*x - 5*x + 5

# Törəməni təyin edirik: df(x) = 2x - 5
def df(x):
    return 2*x - 5

# Iterasiya sayı (bu halda 20 dəfə təkrarlayacağıq)
N = 20 

# Başlanğıc nöqtəsini təyin edirik, burada x = 0 olaraq başlayacaq
xx = 0    

# Yanaşma addımını təyin edirik, öyrənmə sürəti (0.9)
# lmd = 0.3

# Qrafik üçün x koordinatlarını təyin edirik (0-dan 5-ə qədər, 0.1 interval ilə)
x_plt = np.arange(0, 5.0, 0.1)

# f(x)-i hər x üçün hesablamaq və nəticələri f_plt listinə saxlamaq
f_plt = [f(x) for x in x_plt]

# Jupyter Notebook-da interaktiv rejimi aktivləşdiririk
plt.ion()  

# Yeni bir şəkil (pəncərə) və oxlar yaradırıq
fig, ax = plt.subplots()    

# Qrafikdə şəbəkə xətləri göstərmək üçün grid əlavə edirik
ax.grid(True)   

# Funksiyanı qrafiklə göstəririk
ax.plot(x_plt, f_plt)                  

# Başlanğıc nöqtəsini qırmızı rəngdə göstəririk
point = ax.scatter(xx, f(xx), c='red')  


mn = 100
# Gradient descent (və ya başqa adı ilə ən kiçik kvadratlar metodu) ilə iterasiyanı başladırıq
for i in range(N):
    
    
    # Addım ölçüsünü (learning rate) iterasiya ilə kiçiltmək üçün bu kod istifadə olunur
    # `lmd` hər iterasiyada daha kiçik dəyərlərə enir ki, model son mərhələlərdə daha incə şəkildə optimallaşsın.
    # `i + 1` iterasiyanın nömrəsi (başlanğıcda 1, sonra 2, 3 və s.), bu dəyər artır və nəticədə addım ölçüsü kiçilir.
    # `mn` isə maksimum addım ölçüsüdür (məsələn, 100), yəni `i+1` hər iterasiya ilə böyüyərkən, ən yüksək dəyər 
#                                                                                                     `mn` olacaq.
    # `min(i+1, mn)` isə iki dəyəri müqayisə edir və ən kiçiyini seçir. Beləliklə:
    # - İlk iterasiyada `i + 1` kiçik olur və addım ölçüsü daha böyük olur.
    # - Sonrakı iterasiyalarda isə addım ölçüsü kiçilir və ən son iterasiyada `lmd` artıq `mn` ilə limitlənir.

    lmd = 1/min(i+1, mn)
    
    # Cari nöqtədən gradientin işarəsi ilə hərəkət edirik (gradientin işarəsi - "addım ölçüsü")
    xx = xx - lmd*np.sign(df(xx))    

    # Yeni nöqtənin mövqeyini (xx, f(xx)) ilə qrafikdə göstəririk
    point.set_offsets([xx, f(xx)])  

    # Qrafiki yenidən çəkmək, hər dəfə yeniləmələri göstərmək üçün:
    fig.canvas.draw()    # Qrafiki yeniləyirik
    fig.canvas.flush_events()  # Göstərilən dəyişiklikləri dərhal əks etdiririk
    time.sleep(0.02)     # Hər iterasiyadan sonra 20 millisecond gözləyirik

# Qrafikdə interaktiv rejimi dayandırırıq ki, qrafiklər sabit qalsın
plt.ioff()   

# Nəticəni çap edirik (son iterasiya nəticəsi)
print(xx)

# Son nöqtəni mavi rəngdə göstəririk
ax.scatter(xx, f(xx), c='blue')

# Qrafiki göstəririk (interaktiv rejim bitdikdən sonra sabit şəkildə)
plt.show()
````

## Python dilində izah: Qlobal və Lokal minumum Sign və  $ \frac{1}{\min(i + 1, mn)}$ funksiyası ilə 

````python
# Gradient descent metodu ilə funksiya minimumunu tapmaq üçün istifadə olunan addımlar.

# Cari nöqtədə funksiyanı qiymətləndirən funksiya tərifi
def f(x):
    return np.sin(x) + 0.5*x  # Funksiya f(x) = sin(x) + 0.5x

# Cari nöqtədə funksiyanın törəməsini hesablayan funksiya tərifi
def df(x):
    return np.cos(x) + 0.5  # Funksiyanın törəməsi: f'(x) = cos(x) + 0.5

N = 20     # İterasiya sayı (addımların sayı)
xx = 2.5   # Başlanğıc nöqtəsi
# lmd = 0.9  # Addım ölçüsü (learning rate)

# qrafik üçün istifadə edilən x koordinatlarının aralığı
x_plt = np.arange(-5.0, 5.0, 0.1)
# Funksiya f(x)-in hər bir x nöqtəsindəki qiymətini hesablamaq
f_plt = [f(x) for x in x_plt]

# Jupyter Notebook-da qrafikləri interaktiv şəkildə göstərmək üçün aktivləşdiririk
plt.ion()   # Interaktiv rejimi aktivləşdiririk
fig, ax = plt.subplots()    # Yeni bir qrafik pəncərəsi və oxlar yaradılır
ax.grid(True)   # Qrafikdə grid (şəbəkə) göstərilsin

# qrafikdə funksiyanı çəkirik
ax.plot(x_plt, f_plt)                   
# Başlanğıc nöqtəni qırmızı ilə göstəririk
point = ax.scatter(xx, f(xx), c='red')  

# Gradient descent iterasiyaları (adımlar)
for i in range(N):
    # Addım ölçüsünü təyin edirik, hər iterasiyada yavaş-yavaş kiçildirik
    lmd = 1 / min(i + 1, 100)  # Addım ölçüsünün iterasiya ilə azaldılması

    # Cari nöqtədən gradientin işarəsi ilə hərəkət edirik (gradientin işarəsi - "addım ölçüsü")
    xx = xx - lmd * np.sign(df(xx))  # İterasiya üzrə argumenti dəyişirik

    # Yenicə təyin edilmiş nöqtəni qrafik üzərində yeniləyirik
    point.set_offsets([xx, f(xx)])  # Nöqtənin yeni yerini təyin edirik

    # Qrafiki yeniləyirik və animasiya üçün qısa gecikmə əlavə edirik
    fig.canvas.draw()  # Qrafik üzərindəki dəyişiklikləri yeniləyirik
    fig.canvas.flush_events()  # Yeniləməni tətbiq edirik
    time.sleep(0.02)  # 20 ms gözləyirik

# Animasiyanın sonunda nöqtəni mavi ilə göstəririk
plt.ioff()   # Interaktiv rejimi söndürürük
print(xx)    # Hesablanmış minimum nöqtəsini ekrana çap edirik
# Minimum nöqtəsini mavi ilə göstəririk
ax.scatter(xx, f(xx), c='blue')

# Qrafiki son olaraq göstəririk
plt.show()
````
