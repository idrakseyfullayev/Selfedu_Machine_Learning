### Moment-l…ôr anlayƒ±≈üƒ±



#### üîπ Moment n…ôdir?

Moment ‚Äî t…ôsad√ºfi d…ôyi≈ü…ônin paylanmasƒ±nƒ±n **riyazi x√ºsusiyy…ôtl…ôrini** (orta, yayƒ±lma v…ô s.) √∂l√ßm…ôk √º√ß√ºn istifad…ô olunur.

Momentl…ôr **2 c√ºr olur**:

1. **Ba≈ülanƒüƒ±ca g√∂r…ô momentl…ôr (Raw moments)**
2. **M…ôrk…ôz…ô g√∂r…ô momentl…ôr (Central moments)**

---

#### 1Ô∏è‚É£ Ba≈ülanƒüƒ±c Momentl…ôr (Ba≈ülanƒüƒ±ca g√∂r…ô)

#### √úmumi d√ºstur:

$$
\mu'_k = E[X^k]
$$

Burada:

* (k) ‚Äî momentin d…ôr…ôc…ôsi
* (E) ‚Äî riyazi g√∂zl…ôm…ô

---

#### ‚úÖ Birinci Ba≈ülanƒüƒ±c Moment

$$
\mu'_1 = E[X]
$$

üîπ Bu **riyazi g√∂zl…ôm…ôdir (orta qiym…ôt)**

---

#### ‚úÖ ƒ∞kinci Ba≈ülanƒüƒ±c Moment

$$
\mu'_2 = E[X^2]
$$

üîπ Orta deyil
üîπ Dispersiya hesablamaq √º√ß√ºn istifad…ô olunur

---

#### 2Ô∏è‚É£ M…ôrk…ôzi Momentl…ôr (Orta …ôtrafƒ±nda)

#### √úmumi d√ºstur:

$$
\mu_k = E[(X - \mu)^k]
$$

Burada:

* $\mu = E[X]$

---

#### ‚úÖ Birinci M…ôrk…ôzi Moment

$$
\mu_1 = E[X - \mu]
$$

üìå **N…ôtic…ô h…ômi≈ü…ô 0-dƒ±r**
$$
\mu_1 = 0
$$

üîπ √á√ºnki orta …ôtrafƒ±nda m√ºsb…ôt v…ô m…ônfi f…ôrql…ôr bir-birini l…ôƒüv edir

---

#### ‚úÖ ƒ∞kinci M…ôrk…ôzi Moment

$$
\mu_2 = E[(X - \mu)^2]
$$

üìå Bu **dispersiyadƒ±r**
$$
\mu_2 = \sigma^2
$$

üîπ Paylanmanƒ±n n…ô q…ôd…ôr yayƒ±ldƒ±ƒüƒ±nƒ± g√∂st…ôrir

---

#### üîÑ ∆èlaq…ô (√áox vacib!)

Dispersiya ba≈ülanƒüƒ±c momentl…ôrl…ô bel…ô yazƒ±lƒ±r:
$$
\text{Var}(X) = E[X^2] - (E[X])^2
$$

Y…ôni:
$$
\sigma^2 = \mu'_2 - (\mu'_1)^2
$$

---

#### üìä Qƒ±sa x√ºlas…ô c…ôdv…ôli

| Moment n√∂v√º | Adƒ±  | D√ºstur         | M…ônasƒ±     |
| ----------- | ---- | -------------- | ---------- |
| Ba≈ülanƒüƒ±c   | 1-ci | $$E[X]$$        | Orta       |
| Ba≈ülanƒüƒ±c   | 2-ci | $$E[X^2]$$       | K√∂m…ôk√ßi    |
| M…ôrk…ôzi     | 1-ci | $$E[X-\mu]$$     | 0          |
| M…ôrk…ôzi     | 2-ci | $$E[(X-\mu)^2]$$ | Dispersiya |


### EHTƒ∞MAL N∆èZ∆èRƒ∞YY∆èSƒ∞ND∆è MOMENTL∆èR

Tutaq ki,
(X) ‚Äî ehtimal m…ôkanƒ±nda verilmi≈ü **t…ôsad√ºfi d…ôyi≈ü…ôndir**
v…ô onun ehtimal paylanmasƒ± m…ôlumdur.

Momentl…ôr **riyazi g√∂zl…ôm…ô operatoru** √ºz…ôrind…ôn t…ôyin olunur.

---

#### üîπ 1. Ba≈ülanƒüƒ±ca g√∂r…ô momentl…ôr (Raw moments)

#### √úmumi t…ôrif:

$$
\boxed{\mu'_k = \mathbb{E}[X^k]}
$$

---

#### üî∏ Diskret halda:

∆èg…ôr (X) diskret t…ôsad√ºfi d…ôyi≈ü…ôndirs…ô:
$$
\mu'_k = \sum_{i} x_i^k , P(X = x_i)
$$

---

#### üî∏ Fasil…ôsiz halda:

∆èg…ôr (X) fasil…ôsizdirs…ô v…ô sƒ±xlƒ±q funksiyasƒ± (f(x))-dirs…ô:
$$
\mu'_k = \int_{-\infty}^{\infty} x^k f(x),dx
$$

---

#### ‚úÖ Birinci ba≈ülanƒüƒ±c moment

$$
\boxed{\mu'_1 = \mathbb{E}[X]}
$$

‚û° **Riyazi g√∂zl…ôm…ô**

---

#### ‚úÖ ƒ∞kinci ba≈ülanƒüƒ±c moment

$$
\boxed{\mu'_2 = \mathbb{E}[X^2]}
$$

- ‚û° Birba≈üa fiziki m…ônasƒ± yoxdur
- ‚û° Dispersiya √º√ß√ºn …ôsas rol oynayƒ±r

---

#### üîπ 2. M…ôrk…ôz…ô g√∂r…ô momentl…ôr (Central moments)

#### √úmumi t…ôrif:

$$
\boxed{\mu_k = \mathbb{E}[(X - \mathbb{E}[X])^k]}
$$

---

#### üî∏ Diskret halda:

$$
\mu_k = \sum_i (x_i - \mu)^k P(X=x_i)
$$

---

#### üî∏ Fasil…ôsiz halda:

$$
\mu_k = \int_{-\infty}^{\infty} (x - \mu)^k f(x),dx
$$

Burada:
$$
\mu = \mathbb{E}[X]
$$

---

#### ‚úÖ Birinci m…ôrk…ôzi moment

$$
\mu_1 = \mathbb{E}[X - \mu] = 0
$$

üìå **H…ôr zaman 0 olur**
(Bu artƒ±q teoremdir)

---

#### ‚úÖ ƒ∞kinci m…ôrk…ôzi moment

$$
\boxed{\mu_2 = \mathbb{E}[(X-\mu)^2] = \mathrm{Var}(X)}
$$

‚û° **Dispersiya**

---

#### üîÅ ∆èsas …ôlaq…ô (teorem)

$$
\boxed{\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2}
$$

Y…ôni:
$$
\mu_2 = \mu'_2 - (\mu'_1)^2
$$

---

#### üìê Riyazi baxƒ±mdan vacib ≈ü…ôrt

Moment m√∂vcuddursa:
$$
\mathbb{E}[|X|^k] < \infty
$$

∆èks halda moment **m√∂vcud olmaya bil…ôr**
(m…ôs…ôl…ôn: Cauchy paylanmasƒ±)

---

#### üß† Qƒ±sa n…ôtic…ô (imtahan c√ºml…ôsi kimi)

> T…ôsad√ºfi d…ôyi≈ü…ônin $k$-cƒ± ba≈ülanƒüƒ±c momenti $\mu'_k = \mathbb{E}[X^k]$, $k$-cƒ± m…ôrk…ôzi momenti is…ô $\mu_k = \mathbb{E}[(X-\mathbb{E}[X])^k]$ ≈ü…ôklind…ô t…ôyin olunur.




### 3-c√º v…ô 4-c√º M∆èRK∆èZƒ∞ MOMENTL∆èR



#### ‚ú≥Ô∏è 3-c√º v…ô 4-c√º M∆èRK∆èZƒ∞ MOMENTL∆èR

Tutaq ki,
- $X$ ‚Äî t…ôsad√ºfi d…ôyi≈ü…ôn,
- $\mu = \mathbb{E}[X]$,
- $\sigma^2 = \mathrm{Var}(X)$.

---

#### 1Ô∏è‚É£ √ú√ß√ºnc√º m…ôrk…ôzi moment ‚Äî **Asimmetriya**

#### üîπ T…ôrif:

$$
\boxed{\mu_3 = \mathbb{E}[(X-\mu)^3]}
$$

---

#### üîπ Standartla≈üdƒ±rƒ±lmƒ±≈ü forma (asimmetriya …ômsalƒ±):

$$
\boxed{\gamma_1 = \frac{\mu_3}{\sigma^3}}
$$

---

#### üîπ ≈û…ôrh:

* $\gamma_1 > 0$ ‚Üí **saƒüa …ôyilmi≈ü** (right-skewed)
* $\gamma_1 < 0$ ‚Üí **sola …ôyilmi≈ü** (left-skewed)
* $\gamma_1 = 0$ ‚Üí **simmetrik paylanma**

üìå **Normal paylanmada**:
$$
\mu_3 = 0,\quad \gamma_1 = 0
$$

---

#### 2Ô∏è‚É£ D√∂rd√ºnc√º m…ôrk…ôzi moment ‚Äî **Kurtosis (sivrilik)**

#### üîπ T…ôrif:

$$
\boxed{\mu_4 = \mathbb{E}[(X-\mu)^4]}
$$

---

#### üîπ Standartla≈üdƒ±rƒ±lmƒ±≈ü forma:

$$
\boxed{\gamma_2 = \frac{\mu_4}{\sigma^4}}
$$

---

#### üîπ Artƒ±q kurtosis (excess kurtosis):

$$
\boxed{\beta_2 = \gamma_2 - 3}
$$

---

#### üîπ ≈û…ôrh:

| Kurtosis n√∂v√º | ≈û…ôrh                               |
| ------------- | ---------------------------------- |
| $$\beta_2 > 0$$ | Sivri, aƒüƒ±r quyruqlu (leptokurtic) |
| $$\beta_2 = 0$$ | Normal paylanma                    |
| $$\beta_2 < 0$$ | Yastƒ± (platykurtic)                |

üìå **Normal paylanma √º√ß√ºn**:
$$
\gamma_2 = 3,\quad \beta_2 = 0
$$

---

#### üìò KONKRET N√úMUN∆è (Dƒ∞SKRET)

Tutaq ki, (X) a≈üaƒüƒ±dakƒ± qiym…ôtl…ôri alƒ±r:

| $$x$$      | 1   | 2   | 3   |
| -------- | --- | --- | --- |
| $$P(X=x)$$ | 0.2 | 0.5 | 0.3 |

---

#### 1Ô∏è‚É£ Riyazi g√∂zl…ôm…ô:

$$
\mu = 1(0.2)+2(0.5)+3(0.3)=2.1
$$

---

#### 2Ô∏è‚É£ Dispersiya:

$$
\mu_2 = \sum (x-\mu)^2P(X=x)=0.49
$$
$$
\sigma=\sqrt{0.49}=0.7
$$

---

#### 3Ô∏è‚É£ √ú√ß√ºnc√º m…ôrk…ôzi moment:

$$
\mu_3=\sum (x-\mu)^3P(X=x)
$$
$$
=(-1.1)^3(0.2)+(-0.1)^3(0.5)+(0.9)^3(0.3)=0.084
$$

---

#### 4Ô∏è‚É£ Asimmetriya:

$$
\gamma_1=\frac{0.084}{0.7^3}\approx0.24
$$

üìå **Z…ôif saƒüa …ôyilm…ô**

---

#### 5Ô∏è‚É£ D√∂rd√ºnc√º m…ôrk…ôzi moment:

$$
\mu_4=\sum (x-\mu)^4P(X=x)=0.377
$$

---

#### 6Ô∏è‚É£ Kurtosis:

$$
\gamma_2=\frac{0.377}{0.7^4}\approx1.57
$$
$$
\beta_2=1.57-3=-1.43
$$

üìå **Yastƒ± paylanma**

---

#### üìó FASIL∆èSƒ∞Z N√úMUN∆è (QISA)

#### Normal paylanma:

$$
X\sim N(\mu,\sigma^2)
$$

$$
\mu_3=0,\quad \mu_4=3\sigma^4
$$

$$
\gamma_1=0,\quad \gamma_2=3
$$

---

#### üß† ƒ∞MTAHAN √ú√á√úN QISA YADDA≈û

* 3-c√º moment ‚Üí **asimmetriya**
* 4-c√º moment ‚Üí **kurtosis**
* Standartla≈üdƒ±rma vacibdir:
  $$
  \frac{\mu_3}{\sigma^3},\quad \frac{\mu_4}{\sigma^4}
  $$




### EHTƒ∞MAL N∆èZ∆èRƒ∞YY∆èSƒ∞ND∆è 3-C√ú V∆è 4-C√ú M∆èRK∆èZƒ∞ MOMENTL∆èR




Tutaq ki,
$$
(\Omega,\mathcal{F},\mathbb{P})
$$
‚Äî ehtimal m…ôkanƒ±,
$$
X:\Omega\to\mathbb{R}
$$
‚Äî t…ôsad√ºfi d…ôyi≈ü…ôn olsun.

F…ôrz ed…ôk ki,
$$
\mathbb{E}[|X|^4]<\infty
$$
(y…ôni 4-c√º moment m√∂vcuddur).

---

#### üîπ T∆èRƒ∞F (M…ôrk…ôzi moment)

T…ôsad√ºfi d…ôyi≈ü…ônin (k)-cƒ± m…ôrk…ôzi momenti
$$
\boxed{
\mu_k=\mathbb{E}\big[(X-\mathbb{E}[X])^k\big]
}
$$
kimi t…ôyin olunur.

---

#### 1Ô∏è‚É£ √ú√á√úNC√ú M∆èRK∆èZƒ∞ MOMENT (Asimmetriya)

#### üîπ T…ôrif:

$$
\boxed{
\mu_3=\mathbb{E}\big[(X-\mu)^3\big]
}
\quad\text{burada } \mu=\mathbb{E}[X]
$$

---

#### üîπ Asimmetriya …ômsalƒ±:

$$
\boxed{
\gamma_1=\frac{\mu_3}{\sigma^3}
}
\quad\text{burada } \sigma^2=\mu_2
$$

---

#### üîπ Teorem (Simmetriya):

∆èg…ôr $X$ d…ôyi≈ü…ôni $\mu$ …ôtrafƒ±nda simmetrik paylanmaya malikdirs…ô, onda
$$
\mu_3=0
$$

üìå ∆èksi √ºmumi halda doƒüru deyil.

---

#### üîπ ≈û…ôrh:

* $\gamma_1>0$ ‚Üí saƒüa meyilli paylanma
* $\gamma_1<0$ ‚Üí sola meyilli paylanma
* $\gamma_1=0$ ‚Üí simmetrik

---

### 2Ô∏è‚É£ D√ñRD√úNC√ú M∆èRK∆èZƒ∞ MOMENT (Kurtosis)

#### üîπ T…ôrif:

$$
\boxed{
\mu_4=\mathbb{E}\big[(X-\mu)^4\big]
}
$$

---

#### üîπ Kurtosis …ômsalƒ±:

$$
\boxed{
\gamma_2=\frac{\mu_4}{\sigma^4}
}
$$

---

#### üîπ Artƒ±q kurtosis:

$$
\boxed{
\beta_2=\gamma_2-3
}
$$

---

#### üîπ Teorem (Normal paylanma):

∆èg…ôr
$$
X\sim\mathcal{N}(\mu,\sigma^2)
$$
olarsa, onda
$$
\mu_3=0,\qquad
\mu_4=3\sigma^4
$$

v…ô n…ôtic…ôd…ô
$$
\gamma_1=0,\qquad
\gamma_2=3
$$

---

#### üîπ Dƒ∞SKRET HAL (r…ôsmi)

∆èg…ôr (X) diskret t…ôsad√ºfi d…ôyi≈ü…ôndirs…ô:
$$
\mu_k=\sum_i (x_i-\mu)^k\mathbb{P}(X=x_i)
$$

---

#### üîπ FASƒ∞L∆èSƒ∞Z HAL (r…ôsmi)

∆èg…ôr (X) fasil…ôsizdirs…ô v…ô sƒ±xlƒ±q funksiyasƒ± (f_X(x))-dirs…ô:
$$
\mu_k=\int_{-\infty}^{\infty}(x-\mu)^k f_X(x),dx
$$

---

#### üìå VACƒ∞B QEYD (m√∂vcudluq)

M…ôrk…ôzi momentin m√∂vcudluƒüu √º√ß√ºn:
$$
\mathbb{E}[|X-\mu|^k]<\infty
$$

M…ôs…ôl…ôn:

* **Cauchy paylanmasƒ±** √º√ß√ºn he√ß bir m…ôrk…ôzi moment m√∂vcud deyil.

---

#### üß† ƒ∞MTAHAN √ú√á√úN FORMAL N∆èTƒ∞C∆è

> T…ôsad√ºfi d…ôyi≈ü…ônin √º√ß√ºnc√º m…ôrk…ôzi momenti paylanmanƒ±n asimmetriyasƒ±nƒ±, d√∂rd√ºnc√º m…ôrk…ôzi momenti is…ô paylanmanƒ±n sivriliyini xarakteriz…ô edir.  Normal paylanma √º√ß√ºn √º√ß√ºnc√º m…ôrk…ôzi moment sƒ±fƒ±r,  d√∂rd√ºnc√º m…ôrk…ôzi moment is…ô (3\sigma^4)-…ô b…ôrab…ôrdir.
