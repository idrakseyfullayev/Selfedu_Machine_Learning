### Histogram, Statistik Sıraların Hamarlanması (Smoothing)

```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Selfedu\Probability_Main\Pictures\Probability_16_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 11):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 9:
        plt.figure(figsize=(8.8, 4))      
    else:     
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü    
        
    if i < 5:
        plt.title('Histogram', size='18')
    elif i < 6:
        plt.title('Normal Paylanmada Riyazi Gözləmə və Dispersiya', size='15')
    elif i < 7:
        plt.title('Statistik Sıralarn Uyğunlaşdırılması və ya Hamarlanması', size='15')
    elif i < 8:
        plt.title("""Ən Optimal Teoretik Curve-ni tapmaq
                     i-ci mövqedəki Teoretik paylanmanın dəyəri tapmaq
                     Variance düstrunda yerinə qoymaq - minimumlaşdırmaq""", size='15')
    elif i < 9:
        plt.title('Max Error → Min', size='18') 
    elif i < 10:
        plt.title("""Riyazi Gözləmə və Dispersiyanın Optimal Dəyərlərini tapmaq qaydası
          Ən kiçik Kvadratlar Metodu - Least Squares Method """, size='12')        
        
        
    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show()
```

### Histogram qurarkən bin (interval) sayını seçmək

#### 1️⃣ **Sturges düsturu**

#### **Sturges düsturu (10-luq loqarifm ilə)**

$$
k = 1 + 3.322 \log_{10}(n)
$$

#### **Sturges düsturu (2-lik loqarifm ilə)**

$$
k = 1 + \log_2(n)
$$

Kiçik və orta həcmli datalar üçün uyğundur.

$$
k = 1 + \log_2(n)
$$

* (k) — bin sayı
* (n) — müşahidələrin sayı

📌 Məsələn:
(n = 1000)

$$
k = 1 + \log_2(1000) \approx 11
$$

---

#### 2️⃣ **Rice düsturu**

Sturges-dən bir az daha çox bin verir.

$$
k = 2 \cdot n^{1/3}
$$

📌 (n = 1000)

$$
k = 2 \cdot 10 = 20
$$

---

#### 3️⃣ **Scott düsturu** (bin eni üzrə)

Datanın dispersiyasını nəzərə alır.

$$
h = \frac{3.5 \cdot \sigma}{n^{1/3}}
$$

* (h) — bin eni
* (\sigma) — standart sapma

Sonra:
$$
k = \frac{\max(x) - \min(x)}{h}
$$

---

#### 4️⃣ **Freedman–Diaconis düsturu** (ən tövsiyə edilən)

Outlier-lərə daha davamlıdır.

$$
h = \frac{2 \cdot IQR}{n^{1/3}}
$$

* (IQR = Q_3 - Q_1)

$$
k = \frac{\max(x) - \min(x)}{h}
$$
