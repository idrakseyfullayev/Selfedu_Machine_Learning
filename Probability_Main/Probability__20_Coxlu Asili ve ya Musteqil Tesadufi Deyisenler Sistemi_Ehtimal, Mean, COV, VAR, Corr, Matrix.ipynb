### Bir neçə Təsadüfi Dəyişən Sistemi

```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Selfedu\Probability_Main\Pictures\Probability_20_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 8):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 6:
        plt.figure(figsize=(8.8, 4))       
    else:     
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü    
        
    if i < 1:
        plt.title("""Bir neçə Təsadüfi Dəyişən Sistemi""", size='18')
    elif i < 2:
        plt.title("""Çox ölçülüdən Bir ölçülüyə keçid Ehtimal Sıxlığı Düsturu""", size='15')
    elif i < 3:
        plt.title("""Çox ölçülü Ehtimal Paylanması Sıxlığı Düsturları - 
Asılı və Asılı olmayan Təsadufi Dəyişənlər üçün""", size='15')
    elif i < 4:
        plt.title("""Bir neçə Təsadüfi Dəyişən Sisteminin Riyazi Gözləməsi və Dispersiyası""", size='13')
    elif i < 5:
        plt.title("""Bir neçə Təsadüfi Dəyişən Sisteminin Covariance-ı""", size='13')
    elif i < 6:
        plt.title("""Bir neçə Təsadüfi Dəyişən Sisteminin Covariance və Correlation Matrix-ı""", size='13')
    elif i < 7:
        plt.title("""Bir neçə Asılı olmayan Təsadüfi Dəyişən Sisteminin Matrix-ı""", size='13')        

    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show()
```

### двумерной системы случайных величин к многомерной случайной

#### Sözbəsöz tərcümə:

* **двумерной** → ikiölçülü
* **системы случайных величин** → təsadüfi dəyişənlər sistemi
* **к многомерной** → çoxölçülü (ümumi halına)
* **случайной** → təsadüfi

#### Azərbaycan dilində mənası:

👉 **İki təsadüfi dəyişəndən ibarət sistemdən çoxsaylı (n ədəd) təsadüfi dəyişənlər sisteminə keçid**

---

#### 💡 “2 ölçülü” nə deməkdir?

Bu o deməkdir ki, **cəmi 2 təsadüfi dəyişən var**:

$$
(X, Y)
$$

* birgə paylanma
* birgə sıxlıq funksiyası
* kovariasiya, korelyasiya və s.

Buna **ikiölçülü təsadüfi vektor** deyilir.

---

#### 💡 “Çoxölçülü” nə deməkdir?

Bu isə **2-dən çox təsadüfi dəyişən** deməkdir:

$$
(X_1, X_2, X_3, \dots, X_n)
$$

Məsələn:

* $X_1$ – yaş
* $X_2$ – maaş
* $X_3$ – iş təcrübəsi

Bu artıq **n-ölçülü təsadüfi vektor**dur.

---

#### 📐 Fərqi sadə misalla:

| **Hal**               | **Ölçü**      |
| ----------------- | --------- |
| Tək dəyişən $$X$$   | **1 ölçülü**  |
| $$(X, Y)$$          | **2 ölçülü**  |
| $$(X, Y, Z)$$       | **3 ölçülü**  |
| $$(X_1,\dots,X_n)$$ | **çoxölçülü** |

---

#### 🧠 Niyə “2 ölçülüdən çoxölçüyə keçid” deyilir?

Çünki:

* 2 ölçülü hal **ən sadə birgə paylanma nümunəsidir**
* Çoxölçülü hal isə onun **ümumiləşdirilmiş formasıdır**
* Formullar eyni məntiqdədir, sadəcə:

  * iki inteqral → n inteqral
  * kovariasiya → kovariasiya matrisi

---

#### Qısa yekun:

> - **2 ölçülü sistem** = 2 təsadüfi dəyişən
> - **Çoxölçülü sistem** = 2-dən çox təsadüfi dəyişən
> - İfadə isə **ümumi hala keçidi** bildirir.
