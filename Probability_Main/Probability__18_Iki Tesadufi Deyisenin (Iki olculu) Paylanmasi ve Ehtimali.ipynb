#### $\omega_X(x) = \int_{-\infty}^{+\infty} \omega(x,y) \ dy$ birgə sıxlıqdan(joint density) marjinal sıxlıq(marginal density) necə alınır

#### Düstur nə deyir?

Şəkildəki ifadə budur:

$$
\omega_X(x) = \int_{-\infty}^{+\infty} \omega(x,y) \ dy
$$

📌 Mənası:

* $\omega(x,y)$ — **X və Y-nin birgə sıxlıq funksiyasıdır**
* $\omega_X(x)$ — **yalnız X üçün sıxlıqdır**
* Y-ni **inteqrasiya etməklə “yox edirik”**, X qalır

Yəni:
👉 *“X-in ehtimalını tapmaq üçün Y-nin bütün mümkün dəyərlərini nəzərə alırıq.”*

---

#### 🔹 Sade misal (ən başa düşülən)

Tutaq ki:

* X və Y **müstəqildir**
* Hər ikisi **[0,1] intervalında uniform paylanıb**

Birgə sıxlıq:
$$
\omega(x,y) =
\begin{cases}
1, & 0 \le x \le 1, \ 0 \le y \le 1 \\
0, & \text{əks halda}
\end{cases}
$$

---

#### İndi X-in marjinal sıxlığını tapaq

$$
\omega_X(x) = \int_{-\infty}^{+\infty} \omega(x,y) \ dy
$$

Amma $\omega(x,y)$ **yalnız $0 \le y \le 1$** üçün $1$-dir, qalan yerdə $0$:

$$
\omega_X(x) = \int_0^1 1 \ dy
$$

$$
\omega_X(x) = 1
$$

və bu yalnız **$0 \le x \le 1$** üçün doğrudur.

📌 Nəticə:
$$
\omega_X(x)=
\begin{cases}
1, & 0 \le x \le 1 \\
0, & \text{əks halda}
\end{cases}
$$

👉 Yəni X də **uniform(0,1)** imiş.

---

#### 🔹 İntuitiv (sözlə) izah

Təsəvvür et:

* Düzbucaqlı sahə var (0–1, 0–1)
* Bu sahənin içində hər nöqtə **eyni ehtimallıdır**
* Sən **sabit bir x** götürüb:
* onun üzərində **şaquli xətt çəkirsən**
* həmin xətt boyunca bütün y-ləri toplayırsan
- Bu “toplama” — **inteqrasiya**dır.

---

#### 🔑 Niyə belə edilir?

Çünki:

* X-in ehtimalı **Y-nin hansı dəyəri almasının fərqi olmadan** tapılmalıdır
* Ona görə Y üzrə **hamısını cəmləyirik (inteqrasiya edirik)**


#### $\omega_X(x) = \int_{-\infty}^{+\infty} \omega(x,y),dy$
- ❌ **Bu düstur sırf riyazi “qayda” kimi inteqraldan çıxmır**
- ✅ **Bu ehtimalın tərifindən (axiomlarından) gəlir**

#### 1️⃣ Törəmədə nə olur?

İki dəyişənli funksiya üçün:
$$
f(x,y)
$$

**Qismən törəmə**:
$$
\frac{\partial f}{\partial x}
$$

📌 Nə edirik?

* **y-ni sabit qəbul edirik**
* yalnız **x üzrə dəyişməni ölçürük**

👉 Məna: *“y dəyişməsəydi, x dəyişəndə funksiya necə dəyişərdi?”*

---

#### 2️⃣ İnteqralda analoji anlayış varmı?

**Bəli var.** Buna deyilir:

> **qismən inteqral / marjinalizasiya**

Amma:

* törəmədə → **sabit saxlayırıq**
* inteqralda → **digər dəyişəni TAMAMİLƏ YOX EDİRİK**

---

#### 3️⃣ Qismən inteqral necə yazılır?

İki dəyişənli funksiya üçün:

$$
\int f(x,y) \ dy
$$

📌 Bu nə deməkdir?

* **x sabitdir**
* y üzrə **inteqrasiya aparılır**
* nəticə artıq **yalnız x-dən asılı funksiyadır**

Bu, törəmədəki:
$$
\frac{\partial f(x,y)}{\partial x}
$$
ilə **struktur baxımından eynidir**, sadəcə əməliyyat fərqlidir.

---

#### 4️⃣ Sənin şəkildən olan düstur haradan gəlir?

Şəkildəki düstur:

$$
\omega_X(x) = \int_{-\infty}^{+\infty} \omega(x,y) \ dy
$$

📌 Cavab:
- ❌ **Bu düstur sırf riyazi “qayda” kimi inteqraldan çıxmır**
- ✅ **Bu ehtimalın tərifindən (axiomlarından) gəlir**

---

#### 5️⃣ Məntiqi çıxarılış (niyə məhz belə?)

Təsadüfi kəmiyyətlər üçün:

* $\omega(x,y) \ dx \ dy$ — *$(X,Y)$-nin bu kiçik sahəyə düşmə ehtimalıdır*

X-in **x ətrafında** olma ehtimalı:
$$
P(x \le X \le x+dx)
$$

Bu baş verə bilər **istənilən y üçün**:

$$
P(x \le X \le x+dx) = \int_{-\infty}^{+\infty}
\omega(x,y) \ dy \cdot dx
$$

Buradan:
$$
\omega_X(x) = \int_{-\infty}^{+\infty}
\omega(x,y) \ dy
$$

📌 Yəni:
👉 **Bu düstur ehtimalın cəmlənməsi prinsipindən çıxır**, yoxsa inteqralın texniki xüsusiyyətindən yox.

---

#### 6️⃣ Törəmə vs inteqral — əsas fərq

| **Əməliyyat**       | **Digər dəyişənlə nə olur**  | **Məna**            |
| --------------- | ------------------------ | --------------- |
| **Qismən törəmə**   | **Sabit saxlanır**          | **Lokal dəyişmə**  |
| **Qismən inteqral** | **Tam “yığılır”**            | **Ümumi ehtimal**   |
| **Marjinal sıxlıq** | **Digər dəyişən yox edilir** | **Paylanma alınır** |

---

#### 7️⃣ Qısa yekun

- ✔ Bəli, inteqralda **tam analoji anlayış var**
- ✔ Amma məqsəd fərqlidir
- ✔ Şəkildəki düstur:

* **inteqralın qaydası deyil**
* **ehtimal nəzəriyyəsinin tərifidir**
* adı: **marjinalizasiya**


#### $F(x,y)=P(X\le x,;Y\le y)$ ikiölçülü paylanma funksiyasının əsas sərhəd (limit) xassələridir

#### 1️⃣ Tərifi xatırlayaq

$$
F(x,y)=P(X\le x, \ Y\le y)
$$

Yəni: **X x-dən kiçik olsun VƏ Y y-dən kiçik olsun** ehtimalı.

---

#### 2️⃣ Niyə bu limitlər belədir?

---

#### 🔹 $F(x,-\infty)=0$

$$
F(x,-\infty)=P(X\le x, \ Y\le -\infty)
$$

📌 Amma:

* təsadüfi kəmiyyət $-\infty$ **ola bilməz**
* $Y\le -\infty$ → **imkansız hadisə**

❗ Ehtimal = **0**

---

#### 🔹 $F(-\infty,y)=0$

$$
P(X\le -\infty, \ Y\le y)
$$

Eyni məntiq:

- * $X\le -\infty$ mümkün deyil
-  👉 ehtimal **0**

---

#### 🔹 $F(-\infty,+\infty)=0$

$$
P(X\le -\infty, \ Y\le +\infty)
$$

- * Birinci şərt mümkünsüzdür
-  👉 ehtimal **0**

---

#### 3️⃣ Sonsuzluq +∞ tərəfdə

---

#### 🔹 $F(x,+\infty)=F_1(x)$

$$
F(x,+\infty)=P(X\le x, \ Y\le +\infty)
$$

📌 Amma:

* $Y\le +\infty$ **həmişə doğrudur**

Ona görə hadisə sadələşir:
$$
P(X\le x)
$$

Bu da **X-in birölçülü paylanma funksiyasıdır**:
$$
F_1(x)
$$

---

#### 🔹 $F(+\infty,y)=F_2(y)$

Tam analoji:
$$
P(X\le +\infty, \ Y\le y)=P(Y\le y)=F_2(y)
$$

---

#### 4️⃣ Ən son və ən vacib

#### 🔹 $F(+\infty,+\infty)=1$

$$
P(X\le +\infty, \ Y\le +\infty)
$$

📌 Hər iki şərt **həmişə ödənir**
→ bütün mümkün hallar

❗ Ehtimal = **1**

---

#### 5️⃣ Qısa cədvəl xülasəsi

| **İfadə**                | **Məna**           | **Nəticə**   |
| -------------------- | -------------- | -------- |
| $$F(x,-\infty)$$       | **Y mümkün deyil** | $$0$$        |
| $$F(-\infty,y)$$       | **X mümkün deyil** | $$0$$        |
| $$F(x,+\infty)$$       | **yalnız X qalır** | $$F_1(x)$$ |
| $$F(+\infty,y)$$       | **yalnız Y qalır** | $$F_2(y)$$ |
| $$F(+\infty,+\infty)$$ | **bütün hallar**   | $$1$$       |

---

#### 6️⃣ İntuitiv şəkil ilə düşün

Koordinat müstəvisində:

* $(-\infty,-\infty)$ → boş sahə
* $(x,+\infty)$ → şaquli zolaq
* $(+\infty,+\infty)$ → bütün müstəvi
