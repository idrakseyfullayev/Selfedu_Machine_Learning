### Chebyshev bərabərsizliyi Tərif:

> **Tərif:** **Təsadüfi dəyişənin öz riyazi gözləməsindən uzaq düşməsinin $a$ ədədindən böyük olma ehtimalı, onun dispersiyasının $a^2$-ə nisbətindən böyük ola bilməz.**

$$
P\bigl(|X-\mathbb{E}X| \ge a\bigr) \ \le \ \frac{\mathrm{Var}(X)}{a^{2}},
\qquad a>0
$$


Daha da sadə desək:

> **Dəyişənin orta qiymətdən çox uzaqlaşma ehtimalı məhduddur.**

---

**Çebışev (Chebyshev) bərabərsizliyi** belə deyilir:

> **Tərif:** **İstənilən təsadüfi dəyişən üçün (sonlu riyazi gözləmə və dispersiyası varsa), onun qiymətlərinin riyazi gözləmədən $k$ standart sapmadan çox kənara çıxma ehtimalı ən çox $ \frac{1}{k^2} $-dir.**

$$
P\bigl(|X-\mathbb{E}X| \ge k\bigr) \ \le \ \frac{\mathrm{Var}(X)}{k^{2}},
\qquad a>0
$$

Qısa və cümlə formasında:

> **Təsadüfi dəyişən böyük ehtimalla riyazi gözləməyə yaxın qiymətlər alır.**


###  Çebışev Bərabərsizliyinin Tam İzahı (Ardıcıl və Ətraflı) və Misal



#### **1. Başlanğıc Məlumat: Diskret Təsadüfi Kəmiyyət**

**Cədvəl:**
- | $ x_i $ | $ x_1 $ | $ x_2 $ | ... | $ x_n $ | 
- | $ p_i $ | $ p_1 $ | $ p_2 $ | ... | $ p_n$) |

**Burada:**
- $ x_i $: Təsadüfi kəmiyyətin mümkün qiymətləri (məsələn: zər atanda 1, 2, 3, 4, 5, 6)
- $ p_i $: Hər bir qiymətin ehtimalı
  - Şərtlər: $ p_i \ge 0 $ və $ \sum_{i=1}^n p_i = 1 $
- $ n $: Mümkün qiymətlərin sayı

---

#### **2. Riyazi Gözləmə (Orta Qiymət) - $ m_x $**

**Düstur:**
$$
m_x = E[X] = \sum_{i=1}^n x_i p_i
$$

**Nə deməkdir?**
- Bu, təsadüfi kəmiyyətin "orta gözlənilən qiyməti"dir. Hər qiyməti onun ehtimalına vurub cəmləyirik.

**Misal:** Zər atma (hər üzün ehtimalı 1/6):
$$
m_x = 1\cdot\frac{1}{6} + 2\cdot\frac{1}{6} + \cdots + 6\cdot\frac{1}{6} = 3.5
$$

---

#### **3. Dispersiya (Varians) - $ D_X $**

**Düstur:**
$$
D_X = \sum_{i=1}^n (x_i - m_x)^2 p_i
$$

**Nə deməkdir?**
- $(x_i - m_x)^2$: Hər qiymətin ortadan nə qədər uzaq olduğunun kvadratı
- Bu kvadrat sapmaları ehtimallarına vurub cəmləyirik
- Nəticə: Dəyişkənliyin (yayılmanın) ölçüsü

**Xüsusiyyətləri:**
- $ D_X \ge 0 $ (həmişə mənfi olmayan)
- $ D_X = 0 $ o zaman ki, bütün $ x_i $ eynidir (heç dəyişkənlik yoxdur)

---

#### **4. Çebışev Bərabərsizliyinin Məqsədi**

**Problemi:** Orta qiymətdən "çox uzaq" düşmə ehtimalını tapmaq istəyirik.

**Riyazi ifadə:**
$$
P(|X - m_x| \geq a)
$$
Burada $ a > 0 $ verilmiş bir ədəddir.

**Çebışev bərabərsizliyi:**
$$
P(|X - m_x| \geq a) \leq \frac{D_X}{a^2}
$$

**Mənası:**
- "Dəyişkənlik ($ D_X $) kiçikdirsə və ya uzaqlıq ($ a $) böyükdürsə, belə böyük sapmalar nadir baş verir."

---

### **5. İsbatın Tam Addımları**

#### **Addım 1: Çoxluqları Təyin Edək**

**AB çoxluğu** = Kiçik sapmalar çoxluğu:
$$
AB = \{ j : |x_j - m_x| < a \}
$$
Yəni AB-də olan indekslər üçün sapma \( a \)-dan kiçikdir.

**Böyük sapmalar çoxluğu** = AB-də olmayanlar:
$$
j \notin AB \quad \text{yəni} \quad |x_j - m_x| \geq a
$$

**Qeyd:** Şəkildə "j daxil deyil AB" ifadəsi bu ikinci çoxluğu bildirir.

---

#### **Addım 2: Ehtimalı Böyük Sapmalar Cəmi kimi Yazmaq**

**Üçüncü sətirdəki düstur:**
$$
P(|X - m_x| \geq a) = \sum_{j=1}^n p_j \quad \text{[j daxil deyil AB]}
$$

**Düzgün yazılış:**
Bu cəm yalnız $ j \notin AB $ üçündür, yəni:
$$
P(|X - m_x| \geq a) = \sum_{j \notin AB} p_j
$$

**Səbəbi:**
- Ehtimal tərifinə görə, hadisənin baş vermə ehtimalı onu təşkil edən elementar hadisələrin ehtimallarının cəmidir.
- $ |X - m_x| \geq a $ hadisəsi: $ \ |x_j - m_x| \geq a $ olan bütün $ j $-ləri əhatə edir
- Hər belə $ j $ üçün ehtimal $ p_j $-dir
- Onların cəmi tam olaraq bu hadisənin ehtimalıdır

---

#### **Addım 3: Dispersiyanı Parçalamaq**

**Beşinci sətirdəki düstur (dispersiya tərifi):**
$$
D_X = \sum_{i=1}^n (x_i - m_x)^2 p_i
$$

Bu cəmi iki hissəyə ayıraq:
1. Kiçik sapmalar ($ i \in AB $)
2. Böyük sapmalar ($ i \notin AB $)

$$
D_X = \underbrace{\sum_{i \in AB} (x_i - m_x)^2 p_i}_{\text{kiçik sapmalar}} + \underbrace{\sum_{i \notin AB} (x_i - m_x)^2 p_i}_{\text{böyük sapmalar}}
$$

---

#### **Addım 4: Altıncı Sətirdəki Əhəmiyyətli Bərabərsizlik**

**Altıncı sətir:**
$$
D_X \geq \sum_{j=1}^n |x_i - m_x|^2 p_j \quad \text{[j daxil deyil AB]}
$$

**Düzgün yazılış:**
$$
D_X \geq \sum_{j \notin AB} (x_j - m_x)^2 p_j
$$

**Niyə doğrudur?**
1. $D_X $ = kiçik sapmalar cəmi + böyük sapmalar cəmi
2. Kiçik sapmalar cəmi $ \ge 0 $ (çünki kvadratlar mənfi olmur)
3. Onda $ D_X $ ≥ böyük sapmalar cəmi (çünki qeyri-mənfi ədəd əlavə edirik)

**Riyazi izah:**
$$
D_X = A + B \quad \text{burada } A \ge 0, B \ge 0
$$
Onda $D_X \ge B \ $ (çünki \$A \ge 0 $ olduğu üçün $ A+B \ge B $)

---

#### **Addım 5: Dördüncü Sətir - Əsas Məhdudlaşdırma**

**Dördüncü sətir:**
$$
D_X \geq \sum_{j=1}^n a^2 p_j = a^2 \cdot \sum_{j=1}^n p_j
$$

**Düzgün yazılış ( $ \ j $  daxil deyil AB nəzərə alınmaqla):**
$$
D_X \geq \sum_{j \notin AB} a^2 p_j
$$

**Niyə doğrudur?**
1. Addım 4-də: $ D_X \geq \sum_{j \notin AB} (x_j - m_x)^2 p_j $
2. $ j \notin AB $ olduğu üçün: $ |x_j - m_x| \geq a $
3. Kvadrat alsaq: $(x_j - m_x)^2 \geq a^2 $
4. Onda: $ (x_j - m_x)^2 p_j \geq a^2 p_j $ (hər bir toplanan üçün)

**Yəni:**
$$
\sum_{j \notin AB} (x_j - m_x)^2 p_j \geq \sum_{j \notin AB} a^2 p_j
$$
$$
D_X \geq \sum_{j \notin AB} a^2 p_j
$$

---

#### **Addım 6: Son Addım - Çebışev Bərabərsizliyinin Alınması**

**Xatırlayaq ki:**
$$
\sum_{j \notin AB} p_j = P(|X - m_x| \geq a) \quad \text{(Addım 2-dən)}
$$

Onda:
$$
D_X \geq a^2 \cdot \sum_{j \notin AB} p_j = a^2 \cdot P(|X - m_x| \geq a)
$$

**Hər tərəfi $ a^2 $-ə bölək:**
$$
\frac{D_X}{a^2} \geq P(|X - m_x| \geq a)
$$

**Və ya eyni şey:**
$$
P(|X - m_x| \geq a) \leq \frac{D_X}{a^2}
$$

**Bu da Çebışev bərabərsizliyidir!**

---

#### **6. Bütün İsbatın Xülasəsi**

1. **Böyük sapmaları təyin et:** $ |x_j - m_x| \geq a $ olan $ j $-ləri seç
2. **Ehtimalı cəm kimi yaz:** $ P(\text{böyük sapma}) = \sum_{j \notin AB} p_j $
3. **Dispersiyanı parçala:** $ D_X = \text{kiçik} + \text{böyük} $
4. **Dispersiyanı aşağıdan məhdudlaşdır:** $ D_X \geq \text{böyük sapmalar cəmi} $
5. **Böyük sapmaları aşağıdan məhdudlaşdır:** Hər böyük sapma $\geq a $, kvadratı $ \geq a^2 $
6. **Birləşdir:** $ D_X \geq a^2 \cdot P(\text{böyük sapma}) $
7. **Yekunlaşdır:** $ P \leq D_X / a^2 $

---

#### **7. Praktiki Məsələ: Niyə Əhəmiyyətlidir?**

**Misal:** Bir sinifdə imtahan nəticələrinin ortalaması 70, dispersiyası 100 olsun.

**Sual:** Tələbənin qiyməti ortalamadan 30 bal uzaq olma ehtimalı nədir?

**Çebışev ilə:**
$$
P(|X - 70| \geq 30) \leq \frac{100}{30^2} = \frac{100}{900} \approx 0.111
$$

**Nəticə:** Ortalamadan 30 bal uzaq düşmə ehtimalı 11.1%-dən çox deyil.

---

#### **8. Ümumi Qayda**

> Çebışev bərabərsizliyi **hər hansı paylanma üçün doğrudur** (normal, birmənalı, simmetrik, asimmetrik - fərqi yoxdur). O, yalnız orta qiymət və dispersiyadan istifadə edir.

> **Əsas məntiq:** Dəyişkənlik kiçikdirsə, böyük kənarlaşmalar nadirdir. Dəyişkənlik böyükdürsə, böyük kənarlaşmalar daha mümkündür, amma yenə də onların üst həddi var: $ D_X / a^2 $.


#### Bir Təsadüfi Dəyişənin ortalamadan uzaqlaşmasının 3 Stdev-dən böyük olma ehtimalı,  yəni outlier olma ehitmalı  maksimum 1/9 dir

#### **Düzgün ifadə:**

**Çebışev bərabərsizliyinə görə:**
$$
P(|X - \mu| \geq 3\sigma) \leq \frac{1}{3^2} = \frac{1}{9} \approx 0.111
$$

**Bu deməkdir ki:**
##### Hər hansı bir təsadüfi dəyişən (paylanmasından asılı olmayaraq) üçün:
- Ortalamadan 3 standart sapma və ya daha çox uzaq düşmə ehtimalı
- **Maksimum 1/9 (≈ 11.1%)** ola bilər

---

#### **Əhəmiyyəti:**

1. **Universal hədd:** Bu bərabərsizlik bütün paylanmalar üçün doğrudur
   - Normal paylanma, vahid paylanma, eksponensial, hər hansı biri

2. **Normal paylanma ilə müqayisə:**
   - Normalda: $ P(|X-\mu| \geq 3\sigma) \approx 0.0027 $ (0.27%)
   - Çebışev: $ \leq 11.1\% $
   - Çebışev daha "təhlükəsiz" (gevşek) hədd verir

3. **Praktik məna:** Əgər bizə paylanmanın forması məlum deyilsə, amma ortalama və standart sapma məlumdursa, outlier-ların faizini **zəmanətli şəkildə** məhdudlaşdıra bilərik

---

#### **Misal:**
Tutaq ki, bir məktəbdə şagirdlərin boyunun:
- Ortalaması: 170 sm
- Standart sapması: 10 sm

O zaman:
$$
P(\text{boy} \leq 140 \text{ sm VƏ ya } \text{boy} \geq 200 \text{ sm}) \leq \frac{1}{9}
$$
(Çünki 170 ± 3×10 = 170 ± 30 = [140, 200])

Yəni: 140 sm-dən qısa və ya 200 sm-dən uzun olma ehtimalı **ən çox 11.1%**-dir.

---

#### **Qeyd:**
Bu "maksimum" hədddır. Əslində real paylanmalarda bu ehtimal adətən **1/9-dan çox kiçik** olur (normalda 0.27%, vahid paylanmada 0%, vs.). Çebışev bərabərsizliyi **ən pis halda** nə ola biləcəyini deyir.

**Yəni cavab bəli, tam düzgündür:** 1/9 ≈ 11.1% maksimum hədddır.

#### Praktik məna: Əgər bizə paylanmanın forması məlum deyilsə, amma ortalama və standart sapma məlumdursa, outlier-ların faizini zəmanətli şəkildə məhdudlaşdıra bilərik

### **Çebışev Bərabərsizliyi ilə "Outlier"ları Zəmanətli Şəkildə Məhdudlaşdırmaq**

#### **1. "Outlier"ın Tərifi:**

Praktikada "outlier" (kənar qiymət) adətən:
- **Ortalamadan 2-3 standart sapma uzaq** olan qiymətlər
- Statistikada tez-tez $ \mu \pm 2\sigma $ və ya $ \mu \pm 3\sigma $ hədləri istifadə olunur

---

#### **2. Məsələnin Formalaşdırılması:**

**Bizə məlumdur:**
- $ \mu $: ortalama (məsələn, 100)
- $ \sigma $: standart sapma (məsələn, 15)
- **Bilmirik:** Paylanmanın forması (normal, əyri, çarpıq, çox modlu, vs.)

**Sual:** Ən çox neçə faiz data nöqtəsi $\mu \pm 3\sigma $ xaricində ola bilər?

---

#### **3. Çebışev Bərabərsizliyinin Tətbiqi:**

#### **3.1. 3 sigma üçün:**
$$
P(|X - \mu| \geq 3\sigma) \leq \frac{1}{3^2} = \frac{1}{9} \approx 0.111
$$

**Bu o deməkdir:**
- **Məlumatların ən çox 11.1%-i** ortalamadan 3 sigma uzaq ola bilər
- **Ən az 88.9%-i** $ \mu \pm 3\sigma $ intervalında olmalıdır

---

#### **3.2. 2 sigma üçün:**
$$
P(|X - \mu| \geq 2\sigma) \leq \frac{1}{2^2} = \frac{1}{4} = 0.25
$$

**Bu o deməkdir:**
- **Məlumatların ən çox 25%-i** ortalamadan 2 sigma uzaq ola bilər
- **Ən az 75%-i** $ \mu \pm 2\sigma $ intervalında olmalıdır

---

#### **4. Praktik Nümunə:**

**Məsələ:** Bir şirkətin işçilərinin maaşları:
- Ortalama maaş: 5000 AZN
- Standart sapma: 1000 AZN

Paylanmanın forması məlum deyil (normal olmaya da bilər).

---

#### **4.1. "Yüksək outlier" tərifi:**
Maaşı $ 5000 + 3\times1000 = 8000 $ AZN-dən yuxarı olanlar

#### **4.2. Çebışevə görə:**
$$
P(\text{maaş} \geq 8000) \leq P(|\text{maaş}-5000| \geq 3000) \leq \frac{1}{9} \approx 11.1\%
$$

**Yəni:** 8000 AZN-dən yuxarı maaş alan işçilərin nisbəti **ən çox 11.1%** ola bilər.

---

#### **4.3. "Aşağı outlier" üçün də eyni:**
Maaşı $ 5000 - 3\times1000 = 2000 $ AZN-dən aşağı olanlar da **ən çox 11.1%** ola bilər.

---

#### **4.4. Ümumi nəticə:**
- Maaşı ya 2000-dən aşağı, ya da 8000-dən yuxarı olanların **cəmi** ≤ 11.1%
- Maaşı 2000-8000 AZN arasında olanlar ≥ 88.9%

---

#### **5. Niyə Bu "Zəmanətli"dir?**

1. **Heç bir fərziyyə yoxdur:** Paylanmanın normal olmasına ehtiyac yoxdur
2. **Ən pis hal:** Bu hədd **ən pis halda** baş verə biləcək maksimum faizdir
3. **Real həyatda:** Əksər paylanmalarda bu faiz **daha kiçik** olur
   - Normal paylanmada: 3 sigma xarici ≈ 0.27%
   - Vahid paylanmada: 0%
   - Lakin bizim heç bir fərziyyəmiz olmadığı üçün, ən təhlükəsiz həddi (11.1%) qəbul edirik

---

#### **6. İdarəetmədə Tətbiqi:**

**Məsələn, keyfiyyət nəzarəti:**
- Məhsulun çəkisi: orta 100q, standart sapma 5q
- Çebışevə görə: 100 ± 3×5 = [85q, 115q] aralığında olmayan məhsullar ≤ 11.1%
- **Zəmanət:** Partiyanın ən az 88.9%-i bu çəki aralığındadır

**Yaxud risk idarəetməsi:**
- Portfel gəliri: orta 8%, standart sapma 4%
- 8% ± 3×4% = [-4%, 20%] aralığı xaricində qalma ehtimalı ≤ 11.1%
- **Zəmanət:** Mənfəətin mənfiyə düşməsi (-4% aşağı) ehtimalı 11.1%-dən çox deyil

---

#### **7. Üstünlükləri və Məhdudiyyətləri:**

#### **Üstünlükləri:**
1. **Universal:** Hər paylanma üçün işləyir
2. **Sadə:** Yalnız ortalama və standart sapma lazımdır
3. **Mühafizəkar:** Təhlükəsiz (pessimist) qiymətləndirmə verir

#### **Məhdudiyyətləri:**
1. **Gevşək hədd:** Realda çox daha sıx hədlər ola bilər
2. **Simmetrik:** Hər iki tərəfi eyni qaydada məhdudlaşdırır (asimetrik paylanmalarda optimal deyil)
3. **Kvadrat asılılıq:** $ 1/k^2 $ → böyük $ k $ üçün faydalı, kiçik $ k $ üçün az faydalı

---

#### **Xülasə:**

**Çebışev bərabərsizliyi bizə deyir:**
> "Paylanmanın nə olduğunu bilməsəniz də, ortalama və standart sapmanı bilməklə, outlier-ların faizini zəmanətli şəkildə üst hədd kimi təyin edə bilərsiniz."

**Düstur:**  
Əgər outlier = $ \mu \pm k\sigma $ xarici, onda:
$$
\text{Outlier faizi} \leq \frac{1}{k^2}
$$

Bu, **paylanma haqqında minimal məlumatla maksimum zəmanət** verməyin riyazi yoludur.

Bu halda:

* sample üzrə baxanda **12% görünə bilər**
* amma bu, **Çebışevin pozulması deyil**

### Chebyshev bərabərsizliyi - Böyük Saylar Qanununda (Law of Large Numbers, LLN) (1)

### 1️⃣ Riyazi ifadə

Şəkildə verilən ifadə:

$$
P \Big ( | Y - m _ { Y } | \geq \varepsilon \Big ) \leq \frac { D _ { y } } { \varepsilon ^ { 2 } } = \frac { \sigma _ { x } ^ { 2 } } { n \varepsilon ^ { 2 } }
$$

$$
\begin{align*} 
P\Big(|Y-m_{Y}|<\varepsilon\Big)>1-\frac{\sigma_{x}^{2}}{n\varepsilon^{2}}
\end{align*}
$$

$$
\begin{align*}
P\left(\left|\frac{1}{n}\sum_{i=1}^n x_i - m_X\right| < \varepsilon\right) > 1- \frac{\sigma_x^2}{n\varepsilon^2} \qquad\qquad\frac{\sigma_x^2}{n\varepsilon} = \delta
\end{align*}
$$

$$
P\Bigg(\Big|\frac{1}{n}\sum_{i=1}^{n} X_i - m_X \Big| < \varepsilon \Bigg) > 1 - \delta
$$

və ya sadələşdirilmiş formada:

$$
P\Big( |\bar{X}_n - m_X| < \varepsilon \Big) > 1 - \delta
$$

**Burada:**

* $X_1, X_2, \dots, X_n$ – müstəqil və eyni paylanmaya malik təsadüfi dəyişənlər.
* $m_X = E[X_i]$ – hər $X_i$-nin gözlənilən dəyəri (ortalaması).
* $ \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i $ – nümunə ortalaması.
* $\varepsilon > 0$ – kiçik bir müsbət ədəd, “tolerans” kimi düşünsək olar.
* $\delta > 0$ – kiçik ehtimal dəyəri (səhv ehtimalı).


> **Tərif**: Nümunə sayı $n$ kifayət qədər böyük olduqda, müstəqil və eyni paylanmaya malik təsadüfi dəyişənlərin nümunə ortalamasının $\bar X_n$, riyazi gözləmədən $m_X$ istənilən qədər kiçik $\varepsilon$ məsafəsində qalması ehtimalı 1-ə (yəni $1-\delta$-ya) istənilən qədər yaxın olur.


---

#### 2️⃣ Məntiqi izah

Bu ifadə deyir ki:

> Nümunə ortalaması $\bar{X}_n$ çox sayda müşahidə (yəni $n$ böyük olduqda) **həqiqi orta dəyərə $m_X$ yaxın olacaq**, və bunun ehtimalı $1 - \delta$-dən böyükdür.

Başqa sözlə, **çox sayda təcrübə aparsaq, orta nəticə həqiqi ortalama ətrafında “yığılır”**.

* $|\bar{X}_n - m_X| < \varepsilon$ → nümunə ortalaması həqiqi ortalamadan $\varepsilon$ qədər fərqlənir.
* $1 - \delta$ → bu yaxınlığın baş vermə ehtimalı yüksəkdir.

Bu, **Böyük Saylar Qanununun** klassik tərifidir: “nümunə ortalaması, nümunə sayı artdıqca əsl orta dəyərə yaxınlaşır.”

---

#### 3️⃣ Sadə misal

Tutaq ki, **sikkə tullayırıq**:

* Hər tullanış $X_i$ – 1 (baş) və 0 (qüvvədə olmayan baş) dəyərlərini ala bilər.
* Başın gözlənilən ehtimalı $m_X = 0.5$.

Nümunə ortalaması:

$$
\bar{X}_n = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

* Əgər 10 dəfə tullasaq ($n = 10$), $\bar{X}_{10}$ 0.2–0.8 arasında ola bilər.
* Amma 1000 dəfə tullasaq ($n = 1000$), $\bar{X}_{1000}$ artıq 0.49–0.51 aralığında yığılacaq.

Yəni **daha çox sınaq → nümunə ortalaması həqiqi ehtimala yaxınlaşır**.




### Chebyshev bərabərsizliyi - Böyük Saylar Qanununda (Law of Large Numbers, LLN) (2)

#### Burada **1-in rolu və mənası**ni izah edək:

---

Şəkildəki ifadə:

$$
P\Big(\big|\bar{X}_n - m_X\big| < \varepsilon\Big) > 1 - \delta
$$

və əvvəlki formada:

$$
P\Big(\big|\frac{1}{n}\sum_{i=1}^{n} X_i - m_X\big| < \varepsilon\Big) > 1 - \frac{\sigma_X^2}{n\varepsilon^2}
$$

---

#### 1️⃣ “1” nəyi ifadə edir?

* **1** burada **maksimum ehtimal**ı göstərir.
* Dispersiya ilə bağlı yuxarı sərhəd ($\frac{\sigma^2}{n \varepsilon^2}$) çıxarılanda, ehtimal **1 – kiçik bir dəyər** olur.
* Yəni:

$$
P(\text{ortalamanın } m_X \text{ ətrafında olması}) > 1 - \delta
$$

* Buradakı 1 → **100% ehtimal** kimi başlanğıc nöqtəsidir. Sonra “səhv ehtimalı” ($\delta = \frac{\sigma^2}{n \varepsilon^2}$) çıxılır.

Sadə dillə: “ortalamamız həqiqi ortalama ətrafında yığılacaq” deyirik, **demək olar ki, tam ehtimalla (1-ə yaxın)**.

---

#### 2️⃣ “Yaxın olmaq” nə deməkdir?

* $|\bar{X}_n - m_X| < \varepsilon$ → nümunə ortalaması **$m_X$-dən $\varepsilon$ qədər uzaqlıqda**dır.

* Misal: $\varepsilon = 0.01$ və $m_X = 0.5$ → demək olur ki, nümunə ortalaması 0.49–0.51 arasında olacaq.

* Yəni “yaxın olmaq” **arzu olunan dəqiqlik (tolerans) ilə həqiqi ortalama ətrafında olma** deməkdir.

---

#### 3️⃣ Məntiqi izah

* 1 → **maksimum ehtimal** (tam dəqiqlik).
* 1 – δ → **realistik ehtimal** ki, ortalama $\varepsilon$ qədər yaxın olacaq.
* Nümunə sayı $n$ artdıqca δ kiçilir → ehtimal 1-ə yaxınlaşır.
