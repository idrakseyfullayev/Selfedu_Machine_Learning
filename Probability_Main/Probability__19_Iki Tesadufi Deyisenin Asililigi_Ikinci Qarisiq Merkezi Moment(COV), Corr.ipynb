### İki Təsadüfi Dəyişənin İkinci Qarışıq Mərkəzi Momenti

#### Düstur

$$
\mu_{1,1} = M{(X - m_X)(Y - m_Y)}
= \int_{-\infty}^{+\infty}!!\int_{-\infty}^{+\infty}
(x - m_X)(y - m_Y) \ \omega(x,y) \ dx \ dy
$$

Rusca qeyd olunan ad:
**“второй смешанный центральный момент”**
(yəni *ikinci qarışıq mərkəzi moment*).

---

#### Simvolların mənası

* **$X, Y$** – iki təsadüfi dəyişən
* **$m_X = E[X]$** – $X$-in riyazi gözləməsi (orta qiyməti)
* **$m_Y = E[Y]$** – $Y$-in riyazi gözləməsi
* **$M{\cdot} = E[\cdot]$** – riyazi gözləmə operatoru
* **$\omega(x,y)$** – $X$ və$(Y$-in **birgə sıxlıq funksiyası** (joint pdf)

---

#### Riyazi məna

$$
(X - m_X)(Y - m_Y)
$$

ifadəsi göstərir ki:

* $X$ öz ortasından nə qədər yayınır,
* $Y$ öz ortasından nə qədər yayınır,
* və bu yayınmalar **birlikdə necə dəyişir**.

Bu ifadənin riyazi gözləməsi isə onların **birgə dəyişmə meylini** ölçür.

---

#### Statistik mənası (çox vacib)

$$
\mu_{1,1} = \operatorname{Cov}(X,Y)
$$

Yəni bu düstur **kovariasiyadır**.

#### Şərh:

* **$\mu_{1,1} > 0$** → $X$ artanda,  $Y$ də artmağa meyllidir
* **$\mu_{1,1} < 0$** →$X$ artanda, $Y$ azalır
* **$\mu_{1,1} = 0$** → xətti asılılıq yoxdur (amma bu, müstəqillik demək deyil!)

---

#### Niyə “ikinci” momentdir?

Çünki:

* hasildə qüvvətlərin cəmi:
  $$
  1 + 1 = 2
  $$
* yəni ikinci tərtibli momentdir
* amma **qarışıqdır**, çünki həm $X$, həm də $Y$ iştirak edir

---

#### Qısa xülasə

* Bu düstur **kovariasiyanı** verir
* $X$ və $Y$-in ortalamadan yayınmalarının birlikdə davranışını ölçür
* Korelyasiyanın əsası məhz budur:
  $$
  \rho_{X,Y} = \frac{\operatorname{Cov}(X,Y)}{\sigma_X \sigma_Y}
  $$



### Əgər İki Təsdufi dəyişən asılı deyilsə:
$\mu_{1,1} = \operatorname{Cov}(X,Y) = 0$
### Amma
$\mu_{1,1} = \operatorname{Cov}(X,Y) = 0:$
- > **xətti asılılıq yoxdur (amma bu, müstəqillik demək deyil)**

#### 1️⃣ Xətti asılılıq nədir?

Xətti asılılıq o deməkdir ki, dəyişənlər təxminən belə bağlıdır:

$$
Y = aX + b
$$

məsələn:

* $X$ artdıqca $Y$ də **düz xətt üzrə** artır
* və ya $X$ artdıqca $Y$ **düz xətt üzrə** azalır

👉 **Kovariasiya yalnız bu tip əlaqəni ölçür.**

---

#### 2️⃣ Cov(X,Y) = 0 nə deyir?

$$
\operatorname{Cov}(X,Y) = E[(X-m_X)(Y-m_Y)] = 0
$$

Bu o deməkdir ki:

* $X$-in ortadan **yuxarı** olduğu hallarda,
* $(Y$-in də müntəzəm olaraq yuxarı və ya aşağı olması **müşahidə olunmur**

Yəni **düzxətli trend yoxdur**.

📌 Amma bu, **heç bir əlaqə yoxdur** demək deyil!

---

#### 3️⃣ NİYƏ müstəqillik demək deyil?

#### 🔴 Çox vacib nümunə

Tutaq ki:

$$
X \sim U(-1,1), \quad Y = X^2
$$

#### Nə baş verir?

* $X$ və $Y$ **açıq-aşkar asılıdır**
* çünki $Y$, $X$-dən **tam asılıdır**

Amma:

$$
\operatorname{Cov}(X,Y) = 0
$$

#### NİYƏ?

* $X>0$ olduqda $Y>0$
* $X<0$ olduqda də $Y>0$
* müsbət və mənfi tərəflər **bir-birini kompensasiya edir**

➡️ nəticədə kovariasiya sıfır çıxır

---

#### 4️⃣ Deməli nəticə nədir?

| **Şərt**                  | **Nə deməkdir**                                |
| --------------------- | ------------------------------------------ |
| **Cov(X,Y) ≠ 0**          | **Xətti asılılıq var**                         |
|**Cov(X,Y) = 0**          | **Xətti asılılıq yoxdur**                      |
| **Müstəqildir**           | **HƏM xətti, HƏM qeyri-xətti asılılıq yoxdur** |
| **Cov = 0 amma asılıdır** | **Qeyri-xətti asılılıq var**                   |

---

#### 5️⃣ Yalnız hansı halda bərabərlik var?

$$
X \perp Y ;\Rightarrow; \operatorname{Cov}(X,Y)=0
$$

amma

$$
\operatorname{Cov}(X,Y)=0 ;\nRightarrow; X \perp Y
$$

📌 Yəni:

> * müstəqillik → kovariasiya 0
> * kovariasiya 0 → müstəqillik **olmaya bilər**

---

#### 6️⃣ Bir cümləlik yadda saxla 🧠

- > **Kovariasiya yalnız düzxətli əlaqəni görür.
- > Əlaqə əyri (qeyri-xətti)dirsə, onu “görmür”.**

### Əgər $X$ və $Y$ birgə normal (Gaussian) paylanmaya malikdirsə və
#### $\mu_{1,1} = \operatorname{Cov}(X,Y) = 0$ dırsa
#### onda:
> $\Longrightarrow \ X \ \text {və} \ Y \ \text{müstəqildir}$  
### İzah:

#### 1️⃣ Düzgün ifadə (TEOREM)

> **Əgər $X$ və $Y$ birgə normal (Gaussian) paylanmaya malikdirsə**, onda

$$
\operatorname{Cov}(X,Y)=0 ;\Longrightarrow \ X \ \text {və} \ Y \ \text{müstəqildir}
$$

📌 **Açar söz:** **birgə normal paylanma** (*jointly normal*).

---

#### 2️⃣ Ən çox edilən səhv ❌

❌ “$X$ normaldır və $Y$ normaldır” → **bəs etmir**

✔️ “$(X,Y)$ birlikdə 2-ölçülü normal paylanmaya malikdir” → **şərtdir**

---

#### 3️⃣ Birgə normal paylanma nə deməkdir?

#### Tərif (qısa):

$(X,Y)$ **birgə normal paylanıb**, əgər onların **birgə sıxlıq funksiyası** belədirsə:

$$
f_{X,Y}(x,y)=\frac{1}{2\pi\sigma_X\sigma_Y\sqrt{1-\rho^2}}
\exp!\left(
-\frac{1}{2(1-\rho^2)}
\Big[
\frac{(x-\mu_X)^2}{\sigma_X^2}
-2\rho\frac{(x-\mu_X)(y-\mu_Y)}{\sigma_X\sigma_Y}
+\frac{(y-\mu_Y)^2}{\sigma_Y^2}
\Big]
\right)
$$

Burada:
$$
\rho = \frac{\operatorname{Cov}(X,Y)}{\sigma_X\sigma_Y}
$$

---

#### 4️⃣ Niyə burada Cov=0 ⇒ müstəqillik olur?

Əgər:
$$
\operatorname{Cov}(X,Y)=0 \quad\Rightarrow\quad \rho=0
$$

onda birgə sıxlıq:

$$
f_{X,Y}(x,y)= f_X(x) \ f_Y(y)
$$

şəklində **ayrılır**.

📌 Bu isə **müstəqilliyin tərifidir**.

---

#### 5️⃣ Niyə bu yalnız normal paylanmada doğrudur?

Çünki:

* Normal paylanma **tamamilə** yalnız

  * orta
  * dispersiya
  * kovariasiya
    ilə təyin olunur

Başqa paylanmalarda:

* qeyri-xətti asılılıqlar ola bilər
* Cov=0 olsa da, asılılıq qalır

#### Klassik əks-nümunə:

$$
X \sim U(-1,1), \quad Y=X^2
$$

* Cov(X,Y)=0
* amma **tam asılıdır**
* və **normal DEYİL**

---

#### 6️⃣ Ayrı-ayrı normal paylanma niyə yetərli deyil?

Tutaq ki:
$$
X \sim N(0,1), \quad Y \sim N(0,1)
$$

amma:
$$
Y =
\begin{cases}
X, & \text{ehtimal } \frac{1}{2} \\
- X, & \text{ehtimal } \frac{1}{2}
\end{cases}
$$

* Hər ikisi **ayrıca normaldır**

* Cov(X,Y)=0

* Amma **müstəqil deyillər**

📌 Çünki birgə paylanma normal deyil.

---

#### 7️⃣ Qısa cədvəl (imtahan üçün ideal)

| **Şərt**                     | **Nəticə                |
| ------------------------ | --------------------- |
| **Cov(X,Y)=0**               | **Xətti asılılıq yoxdur** |
| **Cov=0 + birgə normal**     | **X və Y müstəqildir ✅**  |
| **Ayrı-ayrı normal + Cov=0** | **Yetərli DEYİL ❌**       |
| **Müstəqillik**              | **Cov=0 (həmişə)**        |

---

#### 8️⃣ Bir cümləlik qızıl qayda 🧠

> **“Zero covariance implies independence ONLY for jointly Gaussian variables.”**

> **“Kovariasiya sıfırdırsa, bu yalnız birgə Qaus (normal) paylanmış dəyişənlər üçün müstəqillik deməkdir.”**

İstəsən, bir az daha açıq formada da demək olar:

> **“İki təsadüfi dəyişənin kovariasiyası sıfırdırsa, onlar yalnız o halda müstəqil sayılır ki, bu dəyişənlər birgə normal (Qaus) paylanmaya malik olsunlar.”**


 ### **linear regression**-un ən vacib məqamlarından biridir.

#### 1️⃣ Model nəyi öyrənib?

Deyək ki, sən modeli belə data ilə öyrətmisən:

$$
x \in [1, 100]
$$

Model bu aralıqda **x ilə y arasındakı xətti əlaqəni** öyrənib:

$$
\hat{y} = \beta_0 + \beta_1 x
$$

Yəni:

* yalnız **1–100 aralığında olan davranışı** görüb,
* bu aralıqdan kənarda nə baş verdiyini **görməyib**.

---

#### 2️⃣ $x = 120$ verəndə nə olur?

Model **hesablaya bilir**:

$$
\hat{y}(120) = \beta_0 + \beta_1 \cdot 120
$$

📌 **Texniki problem yoxdur** — riyazi olaraq işləyir.

Amma bu artıq:

$$
\boxed{\text{Ekstrapolyasiya}}
$$

---

#### 3️⃣ İnterpolasiya vs Ekstrapolyasiya

| **Növ**                 | **$x$ haradadır** | **Etibarlılıq** |
| ------------------- | ------------- | ----------- |
| **İnterpolasiya**   | **1–100 arası**   | **✅ yüksək**    |
| **Ekstrapolyasiya** | **<1 və ya >100** | **⚠️ riskli**   |

---

#### 4️⃣ Niyə risklidir?

Çünki real həyatda:

* əlaqə **xətti olmaya bilər**,
* 100-dən sonra **doyma**, **sərhəd**, **qırılma** ola bilər,
* model bunu **görməyib**.

📌 Model deyir:

> “Mən 1–100 aralığında gördüyümü davam etdirirəm.”

Amma reallıq fərqli ola bilər.

---

#### 5️⃣ Praktik qayda (müsahibə cavabı ⭐)

> **Linear regression modeli yalnız təlim datasının əhatə etdiyi (x) aralığında etibarlıdır.
> Aralıqdan kənar proqnozlar ekstrapolyasiyadır və zəmanət verilmir.**

---

#### 6️⃣ Nə vaxt $x=120$ məntiqli ola bilər?

✔ Əgər:

* fiziki qanunla xətti əlaqə **sübut olunubsa**
  (məsələn, Ohm qanunu, Hooke qanunu)
* və ya əlavə data ilə **100-dən yuxarı** da model yoxlanılıbsa

❌ Əks halda:

* biznes,
* iqtisadiyyat,
* sosial data üçün **təhlükəlidir**

---

#### 7️⃣ Qısa xülasə

* ❓ $x=120$ olar? → **Bəli**
* 🔍 Etibarlıdır? → **Yox, zəmanətli deyil**
* 📌 Adı nədir? → **Ekstrapolyasiya**
