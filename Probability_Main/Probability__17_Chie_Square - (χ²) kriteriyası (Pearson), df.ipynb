### Chi-square $χ²$

#### 1️⃣ Ümumi Chi-square düsturu

$$
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
$$

#### Burada:

* $O_i$ — **müşahidə olunan** (real) qiymət
* $E_i$ — **gözlənilən** (teorik) qiymət
* $\sum$ — bütün kateqoriyalar üzrə cəmləmə

📌 Yəni: *real nəticə ilə nəzəri nəticə arasındakı fərqin kvadratını, gözlənilən qiymətə bölüb hamısını toplayırıq.*

---

#### 2️⃣ Uyğunluq testi (Goodness of Fit)

Paylanmanın nəzəri paylanmaya uyğun olub-olmadığını yoxlayır.

#### Gözlənilən qiymət:

$$
E_i = n \cdot p_i
$$

* $n$ — ümumi müşahidə sayı
* $p_i$ — nəzəri ehtimal

---

#### 3️⃣ Asılılıq testi (Contingency Table)

İki kateqorik dəyişənin **asılı olub-olmadığını** yoxlayır.

#### Gözlənilən tezlik:

$$
E_{ij} = \frac{(\text{Sətir cəmi}) \cdot (\text{Sütun cəmi})}{\text{Ümumi cəm}}
$$

Sonra yenə əsas düstur:
$$
\chi^2 = \sum \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
$$

---

#### 4️⃣ Azadlıq dərəcəsi (df)

* **Uyğunluq testi:**
  $$
  df = k - 1
  $$
  (k — kateqoriya sayı)

* **Asılılıq testi:**
  $$
  df = (r-1)(c-1)
  $$
  (r — sətir sayı, c — sütun sayı)

---

#### 5️⃣ Nəticənin mənası

* χ² **kiçikdirsə** → nəzəri modelə uyğundur
* χ² **böyükdürsə** → ciddi fərq var

📌 Kritik dəyər cədvəldən (χ² table) götürülür və müqayisə edilir.



### Chi-square düsturunun tətbiqi

#### 📊 MÜŞAHİDƏ CƏDVƏLİ (Asılılıq testi)

Sorğu nəticəsi:

|          | Kişi | Qadın | Cəm |
| -------- | ---- | ----- | --- |
| **Bəli** | $$30$$   | $$20$$    | $$50$$  |
| **Xeyr** | $$10$$   | $$40$$   | $$50$$  |
| **Cəm**  | $$40$$   | $$60$$   | $$100$$ |

---

#### 📌 1️⃣ Gözlənilən dəyərlərin hesablanması

Düstur:
$$
E_{ij} = \frac{(\text{Sətir cəmi}) \cdot (\text{Sütun cəmi})}{\text{Ümumi cəm}}
$$

#### Hesablamalar:

|          | Kişi $E$                         | Qadın $E$                        |
| -------- | -------------------------------- | -------------------------------- |
| **Bəli** | $$ \frac{50 \cdot 40}{100} = 20 $$ | $$ \frac{50 \cdot 60}{100} = 30 $$ |
| **Xeyr** | $$ \frac{50 \cdot 40}{100} = 20 $$ | $$\frac{50 \cdot 60}{100} = 30 $$ |

---

#### 📊 GÖZLƏNİLƏN DƏYƏRLƏR CƏDVƏLİ

|          | Kişi | Qadın |
| -------- | ---- | ----- |
| **Bəli** | $$20$$   | $$30$$    |
| **Xeyr** | $$20$$   | $$30$$    |

---

#### 🧮 2️⃣ Chi-square hesablanması

Düstur:
$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

#### Hesab:

| Hüceyrə    | O  | E  | $(O-E)^2/E$ |
| ---------- | -- | -- | ----------- |
| **Bəli–Kişi**  | $$30$$ | $$20$$ | $$5$$          |
| **Bəli–Qadın** | $$20$$ | $$30$$ | $$3.33$$        |

| **Xeyr–Qadın** | $$40$$ | $$30$$ | $$3.33$$       |

$$
\chi^2 = 5 + 3.33 + 5 + 3.33 = 16.66
$$

---

#### 📐 3️⃣ Azadlıq dərəcəsi

$$
df = (r-1)(c-1) = (2-1)(2-1) = 1
$$

---

#### ✅ 4️⃣ Qərar

📌 χ² cədvəlinə görə

* **df = 1**
* **kritik dəyər = 3.84**

$$
16.66 > 3.84
$$

#### 🔴 Nəticə:

👉 **Kişi/Qadın ilə cavab (Bəli/Xeyr) arasında asılılıq VAR**

### Pearson kriteriyasının əsas ideyası

#### İdeyanın mənası belədir:

1. **Histogram və ehtimal:**

   * Hər bir interval (və ya “razryad”) üçün gözlənilən say var.
   * Əgər bu intervala düşəcək eksperiment nəticələrinin sayı çox azdırsa, deməli həmin intervalın “çəki”si (yəni təsir gücü) də kiçik olacaq.
   * Əksinə, ehtimalı böyük olan interval üçün çəkini böyük götürürük.

2. **Chi-Square kriteriyası:**

   * Pirson $χ²$ testində uyğunluq kriteriyası hesablanır:
     $$
     \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
     $$
     Burada:

     * $O_i$ — müşahidə olunan say (observed count)
     * $E_i$ — gözlənilən say (expected count)
     * $(O_i - E_i)^2 / E_i$ — hər intervalın “çəkiyə” çevrilmiş fərqi
   * Yəni, **az ehtimallı interval üçün $E_i$ kiçikdir**, buna görə də bu intervalın fərqi $(O_i - E_i)^2 / E_i$ ümumi $χ²$-ə daha az təsir edir.

> **Qısaca:** Az ehtimallı nəticələr testdə daha az “çəki” alır, çox ehtimallı nəticələr isə daha çox töhfə verir. Bu da Pirsonun kriteriyasının əsas ideyasıdır.


### Ronald A. Fisher - Degrees of freedom

1. **Сislo razryadov ($k$)** — histogramda və ya kateqoriyalarda olan interval / razryadların sayı.

   * Məsələn, əgər siz $5$ müxtəlif kateqoriya üzrə müşahidə aparırsınızsa, onda $k = 5$.

2. **Сislo zavisimix nablyudeniy ($s$)** — müəyyən parametrin (məsələn, ehtimal parametri) təxmin edilməsi üçün sərbəst istifadə olunan dərəcələrin sayı.

   * Məsələn, əgər siz ortalama və ya paylanma parametrini müşahidələrdən çıxarırsınızsa, s = 1, 2 və s.

3. **Azadlıq dərəcəsi ($r$)** — $χ²$ testində istifadə olunur və belə hesablanır:
   $$
   r = k - s
   $$

* $k$ — kateqoriyaların sayı
* $s$ — təxmin edilən parametr sayı - (Orta, Dispersiya və s.)
* Nəticədə $r$ — $χ²$ paylanmasının **azadlıq dərəcəsi**

**Qısaca:**

* Daha çox razryad varsa → $χ²$ testində daha çox azadlıq dərəcəsi
* Daha çox parametr təxmin etsən (Orta, Dispersiya və s.) → azadlıq dərəcəsi azalır


#### $χ²$  nəticəsi ilə Curve aproksimasiyasının Nəzəri paylanmaya uyğunluq interpretasiyası belədir(1)

#### Şərt:

$$
\chi^2_{\text{hesab}} < \chi^2_{\text{kritik}}
$$

---

#### 📌 Mənası:

* **Sıfır hipotezini (H₀) rədd ETMİRİK**
* Müşahidə olunan paylanma **nəzəri (gözlənilən) paylanma ilə uyğundur**
* Fərqlər **təsadüfi dalğalanma ilə izah oluna bilər**

---

#### Sadə dildə:

> “Məlumatlarımız nəzəri paylanmaya zidd deyil.”

və ya

> “Uyğunsuzluq tapılmadı.”

---

#### Statistik dildə düzgün ifadə:

* “Seçilmiş əhəmiyyətlilik səviyyəsində ($\alpha$) H₀ hipotezini rədd etmək üçün kifayət qədər sübut yoxdur.”

---

#### Vacib qeyd ⚠️

Bu **H₀ doğrudur** demək deyil.
Sadəcə:

* “H₀-nın səhv olduğunu göstərən dəlil tapmadıq.”

---

### Kiçik xülasə:

| Şərt                                | Nəticə          |
| ----------------------------------- | --------------- |
| $$\chi^2 < \chi^2_{\text{kritik}}$$   | H₀ rədd edilmir |
| $$\chi^2 \ge \chi^2_{\text{kritik}}$$ | H₀ rədd edilir  |


#### $χ²$  nəticəsi ilə Curve aproksimasiyasının Nəzəri paylanmaya uyğunluq interpretasiyası belədir(2)

#### Şərt:

$$
\chi^2_{\text{hesab}} < \chi^2_{\text{kritik}}
$$

---
#### Sizin dediyiniz:

> **“Curve aproksimasiyası nəzəri paylanmanı izah edir”**

Bu **tam düzgün cümlə deyil**, amma **ideya baxımından yaxındır**.

---

#### Düzgün statistik məna nədir?

χ² uyğunluq testində belə deyirik:

> **Empirik paylanma (histogram / curve aproksimasiyası)**  **nəzəri paylanma ilə ziddiyyət təşkil etmir**

yəni:

> **Məlumatların qurduğu əyri (curve)**  **nəzəri paylanmaya kifayət qədər yaxındır**

---

#### Daha düzgün ifadə variantları ✅

Aşağıdakılardan istənilənini istifadə edə bilərsiniz:

* **“Empirik paylanma nəzəri paylanma ilə uyğundur.”**
* **“Seçilmiş əhəmiyyətlilik səviyyəsində empirik əyri nəzəri paylanmadan statistik olaraq fərqlənmir.”**
* **“Curve aproksimasiyası nəzəri paylanmanı inkar etmir.”**

---

#### Niyə “izah edir” sözü risklidir? ⚠️

“İzah edir” desək, sanki:

* nəzəri paylanma **empirik məlumatdan çıxarılıb**
* və ya **səbəb-nəticə əlaqəsi var**

amma χ² testində biz **sadəcə uyğunluq yoxlayırıq**, izah etmirik.

---

#### Bir cümləlik imtahan cavabı (ideal variant):

> **χ² testinin nəticəsinə görə empirik paylanma nəzəri paylanma ilə uyğundur və onu rədd etmək üçün əsas yoxdur.**



#### $χ²$  nəticəsi ilə Curve aproksimasiyasının Nəzəri paylanmaya uyğunluq interpretasiyası belədir(3)

Əgər **$χ²$ böyük olarsa**, yəni

$$
\chi^2_{\text{hesab}} > \chi^2_{\text{kritik}}
$$

onda **interpretasiya tam əksinədir**:

---

#### 📌 Nəticə:

* **Sıfır hipotezi (H₀) RƏDD EDİLİR**
* Empirik paylanma **nəzəri paylanma ilə uyğun DEYİL**
* Fərqlər artıq **təsadüfi hesab edilə bilməz**

---

#### Sadə dildə:

> **Məlumatların qurduğu əyri (curve) nəzəri paylanmadan əhəmiyyətli dərəcədə fərqlənir.**

---

#### Düzgün ifadə nümunələri ✅

* **“Empirik paylanma nəzəri paylanmaya uyğun deyil.”**
* **“Seçilmiş əhəmiyyətlilik səviyyəsində nəzəri paylanma rədd edilir.”**
* **“Curve aproksimasiyası nəzəri paylanmanı izah etmir.”** ✔️ (bu halda doğrudur)

---

#### Niyə $χ²$ böyük olur?

Adətən bu səbəblərdən:

* Histogramda bəzi intervallarda fərqlər çox böyükdür
* Nəzəri model səhv seçilib (məs: normal əvəzinə başqa paylanma lazımdır)
* Parametrlər (orta, dispersiya) düzgün qiymətləndirilməyib
