```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Selfedu\Probability_Main\Pictures\Probability_8_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 8):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 1:
        plt.figure(figsize=(8.9, 4))
    elif i == 2:
        plt.figure(figsize=(8.9, 4))
    elif i == 4:
        plt.figure(figsize=(8.9, 4))
    elif i == 5:
        plt.figure(figsize=(8.9, 4))         
    else:     
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        
    if i < 1:
        plt.title('Bernoulli formulası', size='18')
    elif i < 2:
        plt.title('', size='18')        
    elif i < 3:
        plt.title('', size='18')
    elif i < 5:
        plt.title('Bernoulli formulası', size='18')
    elif i < 6:
        plt.title('', size='18')
    elif i < 7:
        plt.title('Bernoulli formulası - Ümumi tapşırıqlar', size='18')        
        
    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show()
```
