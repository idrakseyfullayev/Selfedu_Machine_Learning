```python
from PIL import Image

# Fayl yolunun sas hisssi
base_path = r'C:\Users\dell\Desktop\Selfedu\Probability_Main\Pictures\Probability_5_'

# killri saxlayacamz siyah
image_list = []

# 1-dn 5- qdr olan killri dvr il arq
for i in range(1, 5):  # 1-dn 5- qdr
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradrq
    img = Image.open(full_path)
    image_list.append(img)  # kili siyahya lav edirik

# Ntic: images siyahsnda btn yklnmi killr olacaq

import matplotlib.pyplot as plt

# killri gstrin
for i, image in enumerate(image_list):
    if i == 100:
        plt.figure(figsize=(8.9, 4))      
    else:     
        plt.figure(figsize=(25, 5))  # kilin ls
        
    if i < 4:
        plt.title('Monty Hall paradoksu', size='18')
    
    plt.imshow(image)
    plt.axis('off')  # Oxlar gizltmk n
    plt.show()    
```
