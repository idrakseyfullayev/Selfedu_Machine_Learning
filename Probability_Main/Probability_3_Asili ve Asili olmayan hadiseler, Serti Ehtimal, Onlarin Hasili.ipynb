# Asılı və Asılı olmayan hadisələr, Şərti Ehtimal, Onların Hasili

```python
from PIL import Image

# Fayl yolunun əsas hissəsi
base_path = r'C:\Users\dell\Desktop\Selfedu\Probability_Main\Pictures\Probability_3_'

# Şəkilləri saxlayacağımız siyahı
image_list = []

# 1-dən 5-ə qədər olan şəkilləri dövr ilə açırıq
for i in range(1, 12):  # 1-dən 5-ə qədər
    full_path = f'{base_path}{i}.jpg'  # Fayl yolunu dinamik olaraq yaradırıq
    img = Image.open(full_path)
    image_list.append(img)  # Şəkili siyahıya əlavə edirik

# Nəticə: images siyahısında bütün yüklənmiş şəkillər olacaq

import matplotlib.pyplot as plt

# Şəkilləri göstərin
for i, image in enumerate(image_list):
    if i == 0:
        plt.figure(figsize=(8.9, 4))
    elif i == 1:
        plt.figure(figsize=(8.9, 4))
    elif i == 2:
        plt.figure(figsize=(8.9, 4))
    elif i == 3:
        plt.figure(figsize=(8.9, 4))
    elif i == 5:
        plt.figure(figsize=(8.8, 4))
    elif i == 7:
        plt.figure(figsize=(8.8, 4))        
    else:     
        plt.figure(figsize=(25, 5))  # Şəkilin ölçüsü
        
    if i < 1:
        plt.title('Asılı və Asılı olmayan hadisələr', size='18')
    elif i < 4:
        plt.title('Asılı və Asılı olmayan hadisələr, Şərti Ehtimal', size='18')        
    elif i < 6:
        plt.title('Müstəqil İki hadisənin hasilinin ehtimalı', size='18')
    elif i < 7:
        plt.title('Asılı İki hadisənin hasilinin ehtimalı', size='18')        
    elif i == 7:
        plt.title('Yekun', size='18')
    elif i < 11:
        plt.title('Ümumi Tapşırıqlar', size='18')        
            
    plt.imshow(image)
    plt.axis('off')  # Oxları gizlətmək üçün
    plt.show()    
```
