#### Uniform (bərabərpaylanmış) paylanmanın orta qiyməti və dispersiyası



#### 1️⃣ Continuous Uniform [a, b] üçün formulası

* **Orta qiymət (expected value, μ):**
  $$
  \mu = \frac{a + b}{2}
  $$

* **Dispersiya (variance, σ²):**
  $$
  \sigma^2 = \frac{(b - a)^2}{12}
  $$

---

#### 2️⃣ Nümunə: [0, 1] intervallı Uniform

* Burada: **a = 0, b = 1**

1. **Orta qiymət:**
   $$
   \mu = \frac{0 + 1}{2} = \frac{1}{2}
   $$

2. **Dispersiya:**
   $$
   \sigma^2 = \frac{(1 - 0)^2}{12} = \frac{1}{12}
   $$

---

✅ Yəni **μ = 1/2** və **σ² = 1/12**
Bu standart nəticədir.



#### Uniform Pyalanmada niyə Dispersiya formulasında 12-ə bölünür: σ² = (b−a)² / 12



#### 1️⃣ Dispersiya tərifi

Dispersiya, **orta kvadrat fərq**dir:
$$
\sigma^2 = E[(X - \mu)^2]
$$

Uniform [a, b] üçün **PDF** belədir:
$$
f_X(x) = \frac{1}{b-a}, \quad x \in [a,b]
$$

Beləliklə:
$$
\sigma^2 = \int_a^b (x-\mu)^2 f_X(x) dx = \int_a^b (x-\mu)^2 \frac{1}{b-a} dx
$$

---

#### 2️⃣ İnterval [0,1] üçün

- * a = 0, b = 1  
- →  μ = 1/2
  $$
  \sigma^2 = \int_0^1 (x-1/2)^2 dx
  $$

* İnteqralı hesablayaq:
  $$
  (x - 1/2)^2 = x^2 - x + 1/4
  $$
  $$
  \int_0^1 (x^2 - x + 1/4) dx = \Big[\frac{x^3}{3} - \frac{x^2}{2} + \frac{x}{4}\Big]_0^1
  $$

* 0–1 intervalini tətbiq edək:
  $$
  \frac{1}{3} - \frac{1}{2} + \frac{1}{4} = \frac{4 - 6 + 3}{12} = \frac{1}{12}
  $$

---

#### 3️⃣ Nəticə

* Dispersiya **σ² = 1/12** çıxır, çünki:

1. Biz kvadrat fərqi inteqralladıq → (x−μ)²
2. [0,1] intervallı üçün hesablayanda tam nəticə **1/12** oldu

---

💡 Ümumi formula üçün: [a, b] intervalında da eyni qayda:
$$
\sigma^2 = \frac{(b-a)^2}{12}
$$

* (b−a)² → intervalın uzunluğu
* 12 → inteqralın riyazi nəticəsindən gəlir (yuxarıdakı hesablamadan)


### Uniform Paylanmada: Modelləşdirmə

- #### a və b hardan alınıb
- #### $Y = a + X(b-a)$ düsturu - Uniform Paylanma yaradan ədədlər çoxluğu əmələ gətirir



#### 1️⃣ Əsas düsturlar
$$
\begin{cases}
a = m_Y - \sqrt{3} \sigma_Y \\
b = m_Y + \sqrt{3} \sigma_Y
\end{cases}
$$

və

$$
Y = a + X (b - a)
$$

---

#### 2️⃣ Bu nə paylanmadır?

Bu **bərabər (Uniform) paylanma** üçündür.

👉 Yəni:
$$
Y \sim U(a,b)
$$

---

#### 3️⃣ Uniform paylanmanın əsas düsturları

Əgər
$$
Y \sim U(a,b)
$$

onda:

#### 📌 Riyazi gözləmə (orta qiymət)

$$
\mathbb{E}[Y] = \frac{a+b}{2}
$$

#### 📌 Dispersiya

$$
\mathrm{Var}(Y) = \frac{(b-a)^2}{12}
$$

#### 📌 Standart kənarlaşma

$$
\sigma_Y = \frac{b-a}{\sqrt{12}}
$$

---

#### 4️⃣ **a və b necə alınıb? (əsas hissə)**

Bizdə **verilib**:

* $m_Y = \mathbb{E}[Y]$
* $\sigma_Y$

Uniform paylanmada:

#### 1. Orta:

$$
m_Y = \frac{a+b}{2}
\Rightarrow
a+b = 2m_Y
$$

### 2. Standart kənarlaşma:

$$
\sigma_Y = \frac{b-a}{\sqrt{12}}
\Rightarrow
b-a = \sqrt{12} \sigma_Y = 2\sqrt{3} \sigma_Y
$$

---

#### 3️⃣ Sistemi həll edək

$$
\begin{cases}
a+b = 2m_Y \\
b-a = 2\sqrt{3} \sigma_Y
\end{cases}
$$

Toplayaq:
$$
2b = 2m_Y + 2\sqrt{3}\sigma_Y
\Rightarrow
b = m_Y + \sqrt{3}\sigma_Y
$$

Çıxaq:
$$
2a = 2m_Y - 2\sqrt{3}\sigma_Y
\Rightarrow
a = m_Y - \sqrt{3}\sigma_Y
$$

✅ **Şəkildəki düsturlar buradan gəlir**

---

#### 5️⃣ $Y = a + X(b-a)$ düsturu nədir?

Bu **standart uniform paylanmadan çevirmə düsturudur**.

#### 📌 Əgər:

$$
X \sim U(0,1)
$$

onda:

$$
Y = a + X(b-a) \sim U(a,b)
$$

👉 Bu **simulyasiya (Monte-Carlo)** üçün istifadə olunur.


Bu **uniform paylanma yaradan ədədlər çoxluğu**  əmələ gətirir:

* $X$ — **$U(0,1)$** paylanmalı təsadüfi ədədlər çoxluğudur
  $$
  X = \{x_1, x_2, x_3, \ldots\}, \quad 0 \le x_i \le 1
  $$

* $b-a$ — intervalın **uzunluğudur**

* $X(b-a)$ — bu çoxluqdakı hər bir ədədi $(b-a)$ qədər **miqyaslayır**:
  $$
  X(b-a) = \{x_1(b-a), x_2(b-a), \ldots\}
  $$
  Bu zaman ədədlər artıq:
  $$
  [0,, b-a]
  $$
  intervalında **bərabər sıxlıqla** paylanır.

* $+a$ əməliyyatı isə bu ədədlər çoxluğunu **sağa sürüşdürür**:
  $$
  Y = \{a + x_1(b-a),\ a + x_2(b-a),\ \ldots\}
  $$

---

#### 6️⃣ Məntiqi izah (intuisiya)

* $X$ → 0 ilə 1 arasında təsadüfi ədəd
* $b-a$ → intervalın uzunluğu
* $X(b-a)$ → bu intervalda təsadüfi məsafə
* $+a$ → intervalı (a)-dan başladır

---

#### 7️⃣ Qısa xülasə

| Düstur                       | Nə üçündür                      |
| ---------------------------- | ------------------------------- |
| $$a = m_Y - \sqrt{3}\sigma_Y$$ | Uniform paylanmanın sol sərhədi |
| $$b = m_Y + \sqrt{3}\sigma_Y$$ | Uniform paylanmanın sağ sərhədi |
| $$Y = a + X(b-a)$$             | $U(0,1)$ → $U(a,b)$ çevrilməsi  |


### Modelləmə - Python ilə experiment - Uniform paylanma üçün

#### Mean və Variance verilib a və b tapılır,  sonra $Y = a + X(b-a)$ düsturunda yerinə qoyulub Uniform ədədlər çoxluğu yaradılır. Sonra yenidən ədədlər çoxluğunun Mean və variance-ı tapılır və yoxlanılır ki, əvvəlcədən verilən Mean və Variance-a təxminən bərabərdir ya yox?

```python
import numpy as np

# Parametrlər
mean = 5        # E[X]
variance = 10  # Var(X)

a = mean - np.sqrt(3 * variance)
b = mean + np.sqrt(3 * variance)

N = 100_000

# Uniform paylanma
Y = np.random.rand(N) * (b - a) + a

# Riyazi gözləmə (simulyasiya)
estimated_mean = np.mean(Y)

# Dispersiya (simulyasiya)
estimated_variance = np.mean((Y - estimated_mean) ** 2)

print("E[X] ≈", estimated_mean)
print("Var(X) ≈", estimated_variance)
```

### Modelləmə - Python ilə experiment - Normal paylanma üçün

```python
# NumPy və matplotlib kitabxanalarını idxal edirik
import numpy as np
import matplotlib.pyplot as plt

# Varians (dispersiya) hər bir tək dəyişən üçün
variance_single = 10      
# Gözlənilən orta dəyər
mean_value = 15           

# Hər bir qrupda neçə nümunə olacağını təyin edirik
num_samples_per_group = 100
# Riyazi Gözlımə bərabərdir
mean = mean_value
# Qrupun ümumi dispersiyasını hesablayırıq
total_variance = num_samples_per_group * variance_single

# Uniform paylanmanın aşağı sərhədini təyin edirik
lower_bound = mean - np.sqrt(3 * total_variance)
# Uniform paylanmanın yuxarı sərhədini təyin edirik
upper_bound = mean + np.sqrt(3 * total_variance)

# Nümunə qruplarının ümumi sayını təyin edirik
num_groups = 10000  # N: neçə qrup nümunə götürəcəyik
# Hər qrup üçün ortalama dəyərləri saxlamaq üçün array yaradırıq
group_means = np.zeros(num_groups)  # nəticələri saxlamaq üçün array

# Hər bir qrup üçün nümunələri yaradırıq və ortalamasını hesablayırıq
for i in range(num_groups):
    # Hər qrup üçün num_samples_per_group sayda random ədəd əlavə edirik
    for j in range(num_samples_per_group):
        group_means[i] += lower_bound + np.random.rand() * (upper_bound - lower_bound)
    # Qrupdakı ədədlərin ortalamasını tapırıq
    group_means[i] /= num_samples_per_group  # Əsas Düstur Normal Paylanmada

# Bütün qrupların ortalamasını hesablayırıq
sample_mean = np.mean(group_means)

# Nümunə dispersiyasını hesablayırıq (N-1 ilə)
sample_variance = np.sum((group_means - sample_mean) ** 2) / (num_groups - 1)

# Hesablanan nümunə ortalamasını çap edirik
print("Sample Mean:", sample_mean)

# Hesablanan nümunə dispersiyasını çap edirik
print("Sample Variance:", sample_variance)

# Qrupların ortalamaları üçün histogram çəkirik
plt.hist(group_means, bins=20)

# X oxuna ad veririk
plt.xlabel("Group Mean Values")

# Y oxuna ad veririk
plt.ylabel("Frequency")

# Qrafik başlığını təyin edirik
plt.title("Histogram of Group Means")

# Qrafik üçün grid əlavə edirik
plt.grid(True)

# Qrafiki göstəririk
plt.show()


```

### Modelləmə - Python ilə experiment - Reley paylanma üçün

```python
import numpy as np          
import matplotlib.pyplot as plt  

# ---------------- Parametrlərin təyini ----------------
# Teoretik Z dəyişənin gözlənilən Variance
expected_variance_Z = 10                 

# Y dəyişənlərinin dispersiyası (Var(Y)), əvvəlki 'ey', riyazi formul əsasında
variance_Y = expected_variance_Z / (2 - np.pi / 2) 
# Y dəyişənlərinin standart sapması (σ_Y), əvvəlki 's_ey'
std_dev_Y = np.sqrt(variance_Y)       

# Teoretik Z dəyişənin orta qiyməti (E[Z] üçün yaxınlaşma), əvvəlki 'mz'
theoretical_mean_Z = np.sqrt(np.pi / 2) * std_dev_Y  

# ---------------- Simulyasiya parametrləri ----------------
# Simulyasiya üçün nümunə sayı, əvvəlki 'N'
num_samples = 10000
# Z üçün simulyasiya nəticələrini saxlamaq üçün boş array yaradılır
Z_samples = np.zeros(num_samples)
# Dispersiyanı hesablayarkən kvadrat fərqlərin cəmini toplamaq üçün dəyişən
sum_squared_diff = 0
# ---------------- Simulyasiya döngəsi ----------------
for i in range(num_samples):
    # Normal paylanmış Y1 dəyəri yarat, ortası 0, standart sapması std_dev_Y
    Y1 = np.random.randn() * std_dev_Y
    # Normal paylanmış Y2 dəyəri yarat, ortası 0, standart sapması std_dev_Y
    Y2 = np.random.randn() * std_dev_Y
    # Z = sqrt(Y1^2 + Y2^2) formulası ilə Z simulyasiya edilir
    Z_samples[i] = np.sqrt(Y1**2 + Y2**2) # Əsas düstur
    # Kvadrat fərqi toplayırıq: (Z_i - E[Z])^2, dispersiya üçün lazım
    sum_squared_diff += (Z_samples[i] - theoretical_mean_Z)**2
    

# ---------------- Eksperimental dispersiya ----------------
# Simulyasiya nəticəsində dispersiyanı hesablamaq (sample variance)
experimental_variance_Z = sum_squared_diff / (num_samples - 1)

# ---------------- Nəticələrin çapı ----------------
# Teoretik orta dəyəri ekrana çıxarır
print(f"Teoretik orta Z: {theoretical_mean_Z:.4f}")
# Simulyasiya nəticəsindən alınan orta dəyəri çap edir
print(f"Simulyasiya ilə orta Z: {np.mean(Z_samples):.4f}")
# Simulyasiya nəticəsindən alınan dispersiyanı çap edir
print(f"Simulyasiya ilə dispersiya Z: {experimental_variance_Z:.4f}")

# ---------------- Histogramın çəkilməsi ----------------
# Z simulyasiya nəticələrinin histogramı, 20 bin istifadə edilir
plt.hist(Z_samples, bins=20, edgecolor='black')
# Qrid əlavə edirik
plt.grid(True)
# X oxuna etiket əlavə edirik
plt.xlabel('Z')
# Y oxuna etiket əlavə edirik
plt.ylabel('Frequency')
# Histogramın başlığı
plt.title('Histogram of Z')
# Histogramı ekranda göstər
plt.show()
```

### Modelləmə - Python ilə experiment - Eksponensial paylanma üçün

```python
import numpy as np
import matplotlib.pyplot as plt

# Parametrlər
# Teoretik gözlənilən variance
expected_Y_variance = 10  
# Eksponential paylanmanın parametri λ = 1/√E[Var]
lambda_param = 1 / np.sqrt(expected_Y_variance)

# Teoretik riyazi gözləmə
theoretical_mean = 1 / lambda_param  # E[Y] = 1/λ

 # Simulyasiya üçün nümunə sayı
N = 10000 
# Exponential paylanma üzrə nümunələr üçün boş array yaradırıq
Y_samples = np.zeros(N)
# Hər bir nümunəni simulyasiya edirik
for i in range(N):
    # 0-1 aralığında uniform təsadüfi ədədlər yaradırıq
    X = np.random.rand()
    # Inverse transform metoduyla exponential nümunə
    Y_samples[i] = -1 / lambda_param * np.log(X)  # Əsas düstur

# Simulyasiya ilə orta hesablayırıq
simulated_mean = np.mean(Y_samples)  # Nümunələrin arifmetik ortası

# Simulyasiya ilə dispersiya hesablayırıq
simulated_variance = np.var(Y_samples, ddof=1)  # Nümunə dispersiyası (n-1 ilə)

# Nəticələri ekrana çap edirik
print(f"Teoretik Orta: {theoretical_mean}")  # Teoretik Orta
print(f"Simulyasiya ilə orta: {simulated_mean}")  # Simulyasiya nəticəsi ilə orta
print(f"Simulyasiya ilə dispersiya: {simulated_variance}")  # Simulyasiya nəticəsi ilə dispersiya

# Histogram çəkirik
plt.hist(Y_samples, bins=20, edgecolor='black')  # 20 interval ilə histogram
plt.grid(True)  # Şəbəkə xəttini əlavə edirik
plt.xlabel('Y dəyərləri')  # X oxunun adı
plt.ylabel('Tezlik')  # Y oxunun adı
plt.title('Exponential paylanma nümunələri')  # Qrafikin başlığı
plt.show()  # Qrafiki göstəririk
```
