## CRISP-DM

**CRISP-DM** modelinin **6 mərhələsini** **geniş və praktiki izahla**, həm **nəzəri**, həm də **real data analitika kontekstində** təqdim edilir.
Bu, həm **data science layihələrində**, həm də **analitik düşünmə strukturunda** əsas metodologiyadır.

---

## 🧭 1️⃣ **Business Understanding (Biznesin başa düşülməsi)**

### 🎯 Məqsəd:

Layihənin **biznes məqsədini**, **problemin mahiyyətini** və **gözlənilən nəticəni** aydın şəkildə müəyyənləşdirməkdir.
Data analitikləri çox vaxt texniki yanaşmaya fokuslanır, amma bu mərhələdə məqsəd — **biznesin nə istədiyini** başa düşməkdir.

### 📌 Əsas addımlar:

* Problemin təsviri: “Biz nəyi həll etmək istəyirik?”
  (məsələn: *müştərinin krediti ödəməyib-ödəməyəcəyini proqnozlaşdırmaq*).
* Layihənin məqsədləri və uğur meyarları (məsələn: *modelin 90% dəqiqliklə proqnoz verməsi*).
* Biznes sualının **analitik suala** çevrilməsi
  (məsələn: “Kim krediti ödəməyəcək?” → “Defolt ehtimalını proqnozlaşdırmaq”).
* Resurs və limitləri müəyyənləşdirmək (vaxt, data, alətlər və s.)

### 💡 Nümunə:

Bir bank rəhbərliyi deyir:

> “Kredit alan müştərilər arasında riskli olanları əvvəlcədən müəyyənləşdirək.”

Bu, analitik dilə çevrilir:

> “Məqsəd: müştərinin kredit ödəməməsini proqnozlaşdırmaq üçün klassifikasiya modeli qurmaq.”

---

## 📊 2️⃣ **Data Understanding (Datanın başa düşülməsi)**

### 🎯 Məqsəd:

Layihədə istifadə olunacaq datanı **toplamaq**, **ilkin analiz etmək**, və **data keyfiyyətini başa düşmək.**

### 📌 Əsas addımlar:

* Datanın toplanması (CSV, SQL, API və s. mənbələrdən)
* Data dictionary ilə tanışlıq (hər sütunun mənası)
* İlkin kəşfiyyat analizi (Exploratory Data Analysis — EDA)
* Data keyfiyyəti problemlərini müəyyənləşdirmək:

  * NaN dəyərlər
  * Uyğunsuz tiplər
  * Dublikatlar
  * Outlier-lər

### 🔍 İstifadə olunan alətlər:

`pandas`, `numpy`, `matplotlib`, `seaborn`, `df.info()`, `df.describe()`

### 💡 Nümunə:

Titanic dataset-də:

* `Age` sütununda 177 NaN var.
* `Sex` sütunu “male/female” kateqorikdir.
* `Fare` sütununda çox böyük fərqlər (outlier-lər) mövcuddur.

---

## 🧹 3️⃣ **Data Preparation (Datanın hazırlanması)**

### 🎯 Məqsəd:

Model üçün **uyğun və təmiz data** hazırlamaq.
Bu mərhələ ən çox vaxt aparan hissədir (bəzən 70-80% vaxt).

### 📌 Əsas addımlar:

* **NaN dəyərlərin** doldurulması və ya silinməsi
  (məsələn: `df['Age'].fillna(df['Age'].median(), inplace=True)`)
* **Outlier-lərin** aşkarlanması və düzəldilməsi
* **Feature Engineering** — yeni xüsusiyyətlər yaratmaq (məsələn: `FamilySize = SibSp + Parch + 1`)
* **Encoding** — kateqorik dəyişənləri sayısala çevirmək
  (məsələn: `pd.get_dummies(df['Sex'])`)
* **Scaling / Normalization** — fərqli miqyaslı dəyişənləri uyğunlaşdırmaq
* Train-test split — model üçün datanı bölmək

### 💡 Nümunə:

Titanic dataset-də:

* `Sex` → 0/1 (male/female)
* `Age` boşdursa, orta yaş ilə doldurulur
* `Fare` log-transformasiya olunur

---

## 🧠 4️⃣ **Modeling (Model qurulması)**

### 🎯 Məqsəd:

Hazırlanmış datadan istifadə edərək **proqnoz və ya təsnifat modeli** qurmaq.

### 📌 Əsas addımlar:

* Model növünü seçmək (Classification, Regression, Clustering və s.)
* Modelin hiperparametrlərini tənzimləmək (`GridSearchCV`, `BayesSearchCV`)
* Modeli təlimləndirmək (`fit()`)
* Modelin performansını ölçmək (`predict()` və metriklərlə)
* Ən yaxşı modeli seçmək

### 🔍 Mümkün modellər:

* **Classification:** Logistic Regression, Random Forest, XGBoost, SVM
* **Regression:** Decision Tree Regressor, LightGBM, CatBoost
* **Clustering:** K-Means, Hierarchical
* **Dimensionality Reduction:** PCA

### 💡 Nümunə:

Titanic datasında `Survived` sütunu üçün:

* `RandomForestClassifier` modeli qurulur.
* `accuracy_score` və `confusion_matrix` ilə nəticə qiymətləndirilir.

---

## ⚖️ 5️⃣ **Evaluation (Qiymətləndirmə)**

### 🎯 Məqsəd:

Modelin performansını **obyektiv ölçmək** və **biznes məqsədlərinə uyğunluğunu** yoxlamaq.

### 📌 Əsas addımlar:

* Model metriklərini analiz etmək:

  * **Classification:** Accuracy, Precision, Recall, F1-score, ROC-AUC
  * **Regression:** RMSE, MAE, R²
* Overfitting və underfitting vəziyyətini yoxlamaq.
* Cross-validation tətbiqi ilə sabitliyi test etmək.
* Nəticələri biznes məqsədləri ilə müqayisə etmək.

### 💡 Nümunə:

Əgər bank layihəsində modelin accuracy-si 95% olsa da, “riskli müştəriləri” düzgün tanımırsa, bu **biznes baxımından uğursuzdur.**
Ona görə **Precision və Recall** daha vacibdir.

---

## 🚀 6️⃣ **Deployment (Tətbiqetmə)**

### 🎯 Məqsəd:

Modeli **real mühitdə istifadə edilə bilən** hala gətirmək.

### 📌 Əsas addımlar:

* Modelin nəticələrini dashboard, API, və ya proqram sisteminə inteqrasiya etmək.
* Modelin periodik yenilənməsi üçün proses hazırlamaq.
* Performans monitorinqi (model drift, accuracy düşməsi və s.)
* İstifadəçilər üçün hesabat və vizuallar yaratmaq (məsələn: Power BI, Tableau)

### 💡 Nümunə:

* Model `Flask` və ya `FastAPI` ilə REST API şəklində serverdə yayımlanır.
* Tableau-da model nəticələri real vaxtda dashboard-da göstərilir.
* Müştəri gələndə sistem avtomatik onun risk səviyyəsini göstərir.

---

## 🔁 **Yekun fikir:**

CRISP-DM **xətti yox, dövri** bir prosesdir.
Əgər model zəifdirsə, “Evaluation”dan yenidən “Data Preparation”a qayıdılır.
Bu iterativ yanaşma data scientist-ə **davamlı təkmilləşmə** imkanı yaradır.
