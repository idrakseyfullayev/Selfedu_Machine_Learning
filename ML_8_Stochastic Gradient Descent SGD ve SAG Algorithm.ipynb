## Funksiyanın Gradienti (1)

### Funksiyanın Gradientinin Tərifi

>**Tərif**: Gradient, çoxdəyişənli bir funksiyanın hər bir dəyişənə görə qismən törəmələrinin vektor şəklində birləşməsidir. Bir funksiyanın gradienti, həmin funksiyanın dəyişmə sürətinin və istiqamətinin təsviridir. **Gradient həm də funksiyanın ən böyük artım istiqamətini göstərir.**

Əgər $f(x_1, x_2, ..., x_n)$ çoxdəyişənli bir skalyar funksiyadırsa, onun gradienti belə tərif olunur:

$$
\nabla f = \left[ \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n} \right]
$$

Burada:
- $\nabla f$: Funksiyanın gradienti.
- $\frac{\partial f}{\partial x_i}$: Funksiyanın $x_i$ dəyişəninə görə qismən törəməsi.

---

 >**Tərif:** Funksiyanın gradienti, çoxölçülü bir funksiyanın dəyişmə sürətini və istiqamətini göstərən, qismən törəmələrdən ibarət bir vektordur.

### Düstur:  
$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right)
$$ 
Daha ümumi halda:  
$$
\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \dots, \frac{\partial f}{\partial x_n} \right)
$$

---

### Gradientin Anlayışı

1. **Geometrik mənası**: Gradient bir vektordur və funksiyanın ən böyük artım istiqamətini göstərir. Gradientin uzunluğu isə həmin istiqamətdəki artımın sürətini ifadə edir.
2. **Tətbiqləri**:
   - **Optimallaşdırma**: Gradientlər optimallaşdırma prosesində, xüsusilə Gradient Descent alqoritmində istifadə olunur.
   - **Fizika və mühəndislik**: Temperatur, potensial və digər sahələrin dəyişmə sürətini analiz etmək üçün.
   - **Maşın öyrənməsi**: Neyron şəbəkələrdə itki funksiyasının minimumunu tapmaq üçün istifadə olunur.

---

### Gradientin Formal Misalları

#### Misal 1: İki dəyişənli bir funksiya
Verilən funksiya:

$$
f(x, y) = x^2 + y^2
$$

Gradienti tapmaq üçün dəyişənlər üzrə qismən törəmələr alınır:

$$
\frac{\partial f}{\partial x} = 2x, \quad \frac{\partial f}{\partial y} = 2y
$$

Beləliklə:

$$
\nabla f(x, y) = [2x, \ 2y]
$$

#### Misal 2: Kompleks bir funksiya
Verilən funksiya:

$$
f(x, y) = x^2y + 3xy^2
$$

Gradienti tapmaq üçün dəyişənlər üzrə qismən törəmələr alınır:

$$
\frac{\partial f}{\partial x} = 2xy + 3y^2, \quad \frac{\partial f}{\partial y} = x^2 + 6xy
$$

Beləliklə:

$$
\nabla f(x, y) = [2xy + 3y^2, \ x^2 + 6xy]
$$

---

### Gradientin Xüsusiyyətləri
1. **Sıfır Gradient**: Gradient sıfır olan nöqtələrdə ($\nabla f = 0$) funksiya maksimum, minimum və ya saddle nöqtədə ola bilər.
2. **Uzunluq və İstiqamət**: Gradientin norması ($|\nabla f|$) dəyişmə sürətini, vektorun istiqaməti isə funksiyanın ən böyük artım istiqamətini göstərir.

---

### Fiziki Misal

Bir potensial sahəni təsvir edən funksiya:

$$
f(x, y, z) = x^2 + y^2 + z^2
$$

Gradient:

$$
\nabla f(x, y, z) = [2x, 2y, 2z]
$$

Bu, həmin sahənin hər nöqtəsində dəyişmə sürətinin və istiqamətinin vektor şəklində təsviridir. Bu misalda gradient, funksiyanın həndəsi mənada bir kürə səthində necə dəyişdiyini göstərir.

Əlavə suallarınız varsa və ya nümunələri artırmaq istəyirsinizsə, mənə bildirin! 😊

## Tək dəyişənli bir funksiya üçün ən böyük artım istiqaməti

Budur, istədiyiniz kimi formatlanmış mətn:  

---

Bir funksiyanın **ən böyük artım istiqaməti** onun **gradient vektoru** ilə müəyyən edilir. Gradient vektor çoxdəyişənli funksiyalarda ən dik istiqaməti göstərir.  

Tək dəyişənli bir funksiya üçün isə sadəcə törəməni götürmək kifayətdir, çünki o, funksiyanın artım və azalma istiqamətini müəyyənləşdirir.  

### Məsələn:  
$ f(x) = x^2 $ funksiyasının gradientini (törəməsini) tapaq:

$$
\nabla f(x) = \frac{d}{dx} (x^2) = 2x
$$

Bu ifadə $ f(x) $-in **hansı istiqamətdə daha sürətli artdığını** göstərir:  
- Əgər $ x > 0 $ olarsa, $ \nabla f(x) = 2x > 0 $, yəni funksiya sağa doğru ($ +x $ istiqamətində) artır.
- Əgər $ x < 0 $ olarsa, $ \nabla f(x) = 2x < 0 $, yəni funksiya sola doğru ($ -x $ istiqamətində) azalır.
- Əgər $ x = 0 $ olarsa, $ \nabla f(x) = 0 $, yəni **kritik nöqtədir** (bu halda minimum nöqtədir).

**Nəticə:**  
Funksiya $ x^2 $ üçün **ən böyük artım istiqaməti müsbət $ x $ oxu istiqamətindədir**, çünki **gradientin işarəsi müsbətdir** və $ x $ böyüdükcə **gradientin dəyəri də artır**.

Əgər **çoxdəyişənli** bir funksiya olsa idi, məsələn $ f(x, y) = x^2 + y^2 $, onda gradient vektoru belə olardı:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = (2x, 2y)
$$

və ən böyük artım istiqaməti **(2x, 2y) vektorunun istiqamətində olurdu**.

## Ətraflı izah

Bəli, **$ x^2 $ funksiyası üçün ən böyük artım istiqaməti həmişə onun gradientinin istiqamətindədir**.  

### **Ətraflı izah**  
Verilmiş funksiya:  
$$
f(x) = x^2
$$

Bunun törəməsini tapaq:  
$$
\nabla f(x) = \frac{d}{dx} (x^2) = 2x
$$

Bu, **funksiyanın ən böyük artım istiqamətini göstərən gradient vektorudur**. Yəni:  
- **Əgər** $ x > 0 $ olarsa, **$ \nabla f(x) > 0 $** və funksiya **artır**.  
- **Əgər** $ x < 0 $ olarsa, **$ \nabla f(x) < 0 $** və funksiya **azalır**.  
- **Əgər** $ x = 0 $ olarsa, **$ \nabla f(x) = 0 $** və funksiya **stasionar nöqtədədir (minimum nöqtədir)**.  

Bu o deməkdir ki, **funksiya ən sürətli şəkildə 2x istiqamətində artır**.  

### **Alternativ izah**  
Gradient **ən böyük artım istiqamətini** göstərir, lakin bu **mütləq şəkildə** sabit bir vektor deyil. Çünki gradientin **modulu (böyüklüyü)** də dəyişir.  

$$
|\nabla f(x)| = |2x|
$$

Yəni:  
- $ x $ böyüdükcə gradient də artır → funksiya daha sürətli artır.  
- $ x $ kiçildikcə gradient azalır → funksiya daha yavaş artır və ya azalır.  

**Əsas nəticə:**  
✅ **Bəli, $ f(x) = x^2 $ funksiyası üçün ən böyük artım istiqaməti həmişə $ 2x $ istiqamətindədir.**  
$ x > 0 $ üçün müsbət istiqamətdə, $ x < 0 $ üçün mənfi istiqamətdə dəyişir.

## Gradient Descent (GD) 

### **Gradient Descent (GD) - Funksiyanın Törəməsinin Optimizasiyası:**


> **Tərif:** **Gradient Descent** (GD) funksiyanın minimumunu tapmaq üçün parametr üzərində **törəmənin əksi istiqamətində** addım-addım hərəkət etmə üsuludur. 

Düstur:
$$
\theta_{\text{new}} = \theta_{\text{old}} - \eta \cdot \nabla f(\theta)
$$
Burada:
- $ \theta $ - optimallaşdırılacaq parametrlər,
- $ \eta $ - öyrənmə sürəti (learning rate),
- $ \nabla f(\theta) $ - funksiyanın gradienti (törəməsi).

---

**Gradient Descent** (GD) optimallaşdırma və təlim metodudur. O, funksiyanı minimuma endirmək üçün istifadə edilir. Əsasən **itki funksiyası** və ya **cost function** ilə əlaqəli məsələlərdə, məsələn, **dərin öyrənmə** (deep learning), **lineer reqressiya** (linear regression) və ya **logistik reqressiya** (logistic regression) kimi modellərin optimallaşdırılmasında geniş istifadə olunur. Məqsəd, verilən bir funksiyanın minimum nöqtəsini tapmaqdır.

### **Gradient Descent-in Genis Anlayışı:**

Gradient Descent, funksiyanın minimum nöqtəsinə yaxınlaşmaq üçün parametrlərin (ağırlıqların və biasların) təkmilləşdirilməsi prosesidir. Bu proses, funksiyanın minimum nöqtəsinə doğru addım-addım hərəkət etməklə həyata keçirilir.

#### **Başlanğıcda:**
- Biz funksiyanın minimumunu tapmaq istəyirik.
- Funksiya $ f(\theta) $ bir neçə parametrdən asılıdır və məqsədimiz bu funksiyanı minimallaşdırmaqdır.
- **Gradient Descent**-də, biz hər dəfə funksiya üzrə **gradient** (törəmə) hesablayırıq və sonra həmin istiqamətdə hərəkət edirik.

Bu prosesi bir neçə dəfə təkrarlayaraq funksiyanı minimum nöqtəsinə yaxınlaşdırırıq.

#### **Gradient Descent-in İcra Addımları:**

1. **Başlanğıc Nöqtəsi Seçilir**:
   - İlk olaraq **parametrlər** $ \theta_0 $ təsadüfi və ya verilənlərə əsasən başlanğıc dəyərləri ilə seçilir.

2. **Törəmə Hesablanır**:
   - Hər iterasiyada, **itki funksiyasının törəməsi** hesablanır. Törəmə funksiyanı ən dik artdığı istiqamətdə göstərir.
   - Bu, parametrlərin necə dəyişdirilməli olduğunu müəyyən edir.

3. **Parametrlərin Yenilənməsi**:
   - Parametrlər, **gradientin əks istiqamətində** yenilənir. Beləliklə, funksiyanın minimumuna doğru hərəkət olunur.
   - Hər iterasiyada bu yeniləmə prosesi təkrarlanır.

4. **Iterasiya Davam Edir**:
   - Parametrlər yeniləndikcə, funksiyanı minimuma yaxınlaşdırırıq. Təkrarlama prosesi müəyyən bir ölçü ilə dayandırılır, məsələn, funksiyanın dəyişməsi çox az olduğu zaman.

### **Gradient Descent-in Riyazi Tərifi:**

Bir funksiyanı $ f(\theta)$ optimallaşdırmaq üçün **Gradient Descent** metodu belə işləyir:

$$
\theta_{\text{new}} = \theta_{\text{old}} - \eta \cdot \nabla f(\theta)
$$

Burada:
- $ \theta $ - funksiyanın optimallaşdırılacaq parametrləri (məsələn, ağırlıqlar və biaslar)
- $ \eta $ - **öyrənmə sürəti** (learning rate) - kiçik bir ədəd, parametrlərin nə qədər dəyişəcəyini təyin edir.
- $\nabla f(\theta) $ - itki funksiyasının **gradienti** (törəməsi), yəni funksiyanın parametrlər üzrə dəyişmə sürətini göstərir.

İlk başlanğıcda, parametrlər ($\theta_{\text{old}} $) müəyyən edilir və **gradient** hesablanır. Daha sonra, bu gradientin əks istiqamətində **öyrənmə sürəti** ($ \eta $) ilə çarpılır və **parametrlər yenilənir**.

### **İtkilərin Funksiyasının Törəməsi (Gradient):**

**Gradient** funksiyanın **törəməsi** kimi də təyin edilir və bu, funksiyanın müxtəlif parametrləri üzrə necə dəyişdiyini göstərir. Törəmə, funksiyanı ən sürətli artan istiqamətdə göstərir və Gradient Descent bu istiqamətin əksinə hərəkət edir. İtkilərin funksiyasının törəməsi (gradient) hesablama üçün aşağıdakı qaydada əldə edilə bilər:

$$
\nabla L(\theta) = \left[ \frac{\partial L}{\partial \theta_1}, \frac{\partial L}{\partial \theta_2}, \dots, \frac{\partial L}{\partial \theta_n} \right]
$$

### **Gradient Descent-in Növləri:**

1. **Batch Gradient Descent:**
   - **Təsviri:** Hər iterasiyada bütün təlim dəstini istifadə edərək gradient hesablanır.
   - **Üstünlüyü:** Çox doğru minimum nöqtəni tapır.
   - **Çatışmazlığı:** Böyük verilənlər üzərində çox vaxt aparır.

2. **Stochastic Gradient Descent (SGD):**
   - **Təsviri:** Hər iterasiyada yalnız bir nümunə üzərində gradient hesablanır.
   - **Üstünlüyü:** Çox sürətlidir və daha kiçik verilənlər üzərində işləyir.
   - **Çatışmazlığı:** Minimuma daha çox yanılmalarla yaxınlaşır.

3. **Mini-Batch Gradient Descent:**
   - **Təsviri:** Təlim dəstini kiçik partiyalara bölərək hər bir batch-də gradient hesablanır.
   - **Üstünlüyü:** Həm sürətli, həm də düzgün minimum tapmağa meyillidir.
   - **Çatışmazlığı:** Hələ də bəzi səhvlər ola bilər, amma daha optimallaşdırılmışdır.

### **Öyrənmə Sürəti (Learning Rate) - $ \eta $:**

**Öyrənmə sürəti** ($ \eta $) gradientin əks istiqamətində hərəkət etmək üçün nə qədər addım atılacağını müəyyən edir. **Kiçik öyrənmə sürəti** daha diqqətlə minimuma yaxınlaşır, amma proses daha yavaş olur. **Böyük öyrənmə sürəti** isə çox sürətli irəliləyir, amma çox yüksək olarsa, minimumu keçə bilər.

### **Misal:**

Deyək ki, bir **lineer reqressiya** funksiyası üzərində çalışırıq. Bu funksiya belədir:

$$
L(m, b) = \frac{1}{n} \sum_{i=1}^{n} (y_i - (mx_i + b))^2
$$

Burada:
- $ m $ və $ b $ modelin parametrləridir (ağırlıq və bias).
- $ x_i $ və $ y_i $ təlim məlumatlarının giriş və çıxış dəyişənləridir.

İlk başlanğıcda, $ m $ və $ b $ təsadüfi seçilir. Daha sonra bu parametrlər yenilənir. Gradientin hesablanması üçün aşağıdakı törəmələr istifadə olunur:

$$
\frac{\partial L}{\partial m} = \frac{-2}{n} \sum_{i=1}^{n} x_i(y_i - (mx_i + b))
$$
$$
\frac{\partial L}{\partial b} = \frac{-2}{n} \sum_{i=1}^{n} (y_i - (mx_i + b))
$$
Sonra yenilənmiş parametrlər əldə edilir:

$$
m_{\text{new}} = m_{\text{old}} - \eta \cdot \frac{\partial L}{\partial m}
$$
$$
b_{\text{new}} = b_{\text{old}} - \eta \cdot \frac{\partial L}{\partial b}
$$

Bu addımlar təkrarlanır və **itki funksiyası** minimuma doğru getdikcə modelin parametrləri optimallaşdırılır.

---

### **Nəticə:**
**Gradient Descent** metodu, funksiyanı minimallaşdırmaq üçün istifadə olunan əsas bir optimallaşdırma alətidir. **Törəmə** (gradient) hesablanaraq, funksiya minimuma doğru təkrarlanaraq yaxınlaşır. Bu metod **dərin öyrənmə**, **lineer reqressiya**, **logistik reqressiya** və digər optimallaşdırma məsələlərində geniş tətbiq olunur. **Öyrənmə sürəti** və **gradientin düzgün hesablanması** bu prosesin uğurlu olmasında mühüm rol oynayır.

## Batch Gradient Descent (BGD) nədir?

### **Batch Gradient Descent (BGD) nədir?**  
**Batch Gradient Descent (BGD)** — hər iterasiyada **bütün train datasetin** gradientini hesablayaraq çəkiləri yeniləyən optimallaşdırma üsuludur. Bu, ənənəvi gradient descent üsuludur və əsasən **hamar (smooth) və konveks (convex) funksiyalar** üçün istifadə olunur.  

---

### **BGD necə işləyir?**  
1. **Bütün nümunələr üzrə itki funksiyasının gradientini hesablayır:**  
   $$
   \nabla L(w) = \frac{1}{n} \sum_{i=1}^{n} \nabla L_i(w)
   $$  
   Burada:  
   - $ L(w) $ — itki funksiyası,  
   - $ n $ — train datasetdəki nümunələrin sayı,  
   - $ \nabla L_i(w) $ — hər nümunə üçün itkinin gradienti.  

2. **Çəkiləri (weights) yeniləyir:**  
   $$
   w = w - \eta \cdot \nabla L(w)
   $$  
   Burada $ \eta $ — learning rate-dir.  

3. **İterasiyalar davam edir** və model optimal dəyərlərə yaxınlaşır.  

---

### **BGD, SGD və SAG müqayisəsi**  

| Xüsusiyyətlər | **Batch Gradient Descent (BGD)** | **Stochastic Gradient Descent (SGD)** | **Stochastic Average Gradient (SAG)** |
|--------------|---------------------------------|---------------------------------|---------------------------------|
| **Gradient hesablanması** | Bütün dataset üzərində | Hər iterasiyada təsadüfi 1 nümunə üzərində | Əvvəlki gradientlərin orta qiymətini saxlayır və istifadə edir |
| **Konvergensiya sürəti** | Yavaş (daha çox iterasiya tələb edir) | Sürətli (tez nəticə verir, ancaq səs-küylüdür) | Sürətli və stabildir |
| **Yaddaş tələbi** | Çox (bütün datasetin gradientini saxlamalıdır) | Az (yalnız 1 nümunənin gradientini saxlayır) | Orta (bütün nümunələrin son gradientlərini saxlayır) |
| **Stabillik** | Stabil və hamar yaxınlaşma | Qeyri-stabil, çox dəyişkənlik var | Daha stabil və smooth yaxınlaşma |
| **Dərin öyrənmə üçün uyğundur?** | Yox, çünki çox yavaşdır | Bəli, çünki sürətlidir | Qismən, konveks problemlərdə yaxşıdır |
| **Overfitting riski** | Az (bütün dataya baxır) | Yüksək (səs-küylü yenilənmələr edir) | Az (orta gradientdən istifadə etdiyi üçün) |
| **Big Data üçün uyğunluq** | Yox (bütün datasetin gradientini hesablamalıdır) | Bəli (kiçik batch-lərlə işləyə bilir) | Orta səviyyədə (bütün gradientləri saxlamalıdır) |

---

### **Nəticə**  
- **BGD** – Stabil, lakin çox yavaş olduğu üçün **çox böyük datasetlərdə istifadə olunmur**.  
- **SGD** – Daha **sürətli**, lakin çox **səs-küylü** (noisy) olduğu üçün nəticələr çox dəyişkən ola bilər.  
- **SAG** – **SGD-dən daha stabil və sürətli**, ancaq əlavə yaddaş tələbi var.  

Əgər **çox böyük datasetiniz varsa, SGD daha uyğundur**. Əgər **stabil və sürətli konvergensiya istəyirsinizsə, SAG yaxşı seçimdir**.

## Stochastic Gradient Descent (SGD) nədir?

### **Stochastic Gradient Descent (SGD) nədir?**  
Stochastic Gradient Descent (SGD) — maşın öyrənməsi və dərin öyrənmə modellərində optimallaşdırma üçün istifadə olunan əsas metodlardan biridir. Əsas məqsədi itki funksiyasını minimallaşdırmaq üçün modelin çəkilərini (weights) optimallaşdırmaqdır.  

---

### **SGD-nin Əsas Prinsipi**  
SGD, Gradient Descent (GD) metodunun bir variantıdır, lakin ənənəvi Gradient Descent-dən fərqi ondadır ki, **hər iterasiyada bütün dataset əvəzinə yalnız bir nümunə (və ya mini-batch) üzərində yenilənmə aparır**.  

Əsas addımlar:  
1. **Gradientin hesablanması** – Modelin parametrlərinin hazırkı qiymətləri ilə itki funksiyasının (loss function) qismən törəmələri hesablanır.  
2. **Parametrlərin yenilənməsi** – Törəmələrə əsasən çəkilər yenilənir:  

   $$
   w = w - \eta \cdot \nabla L(w)
   $$  
   Burada:  
   - $w$ – modelin parametrləri  
   - $\eta$ – learning rate (öyrənmə sürəti)  
   - $\nabla L(w)$ – itki funksiyasının gradienti  

3. **Təkrarlanma** – Bu proses iterativ şəkildə təkrarlanır ki, model optimal parametrlərə çatsın.  

---

### **SGD ilə Batch Gradient Descent arasındakı fərq**  
| **Xüsusiyyətlər**         | **Batch Gradient Descent** | **Stochastic Gradient Descent (SGD)** |
|--------------------------|-------------------------|----------------------------------|
| **Hesablama üsulu**       | Bütün dataset üzərində hesablanır | Hər iterasiyada yalnız bir nümunə ilə yenilənir |
| **Sürət**                | Yavaş, çünki çoxlu hesablamalar lazımdır | Çox sürətli, çünki az məlumatla yenilənir |
| **Köçmə (Convergence)**   | Daha stabil, ancaq lokal minimumda ilişə bilər | Daha təsadüfi, lakin lokal minimumdan çıxa bilər |

---

### **SGD-nin Üstünlükləri və Çatışmazlıqları**  
✅ **Üstünlüklər:**  
- Böyük datasetlər üçün daha effektivdir.

- Learning rate düzgün tənzimlənməzsə, optimallaşma zəif nəticə verə bilər.  

---

## Stochastic Average Gradient (SAG) nədir?

### **Stochastic Average Gradient (SAG) nədir?**  
**Stochastic Average Gradient (SAG)** — **Stochastic Gradient Descent (SGD)** və **Batch Gradient Descent** metodlarının birləşməsi olan optimallaşdırma üsuludur. SAG əsasən **konveks** (convex) optimallaşdırma problemləri üçün istifadə olunur və **batch gradient descent** metodundan daha sürətli işləyə bilər.  

---

### **SAG-nin Əsas Prinsipi**  
SGD metodunda hər iterasiyada yalnız **bir nümunənin** gradienti istifadə edilir. SAG isə **bütün keçmiş gradientlərin orta qiymətini (average gradient) saxlayır və istifadə edir**.  

Əsas addımlar:  
1. **Gradientlərin yadda saxlanması** – Hər nümunənin son hesablanmış gradienti saxlanılır.  
2. **Orta gradientin hesablanması** – Model çəkilərinin yenilənməsində əvvəlki gradientlərin ortalaması istifadə olunur:  

   $$
   w = w - \eta \cdot \frac{1}{n} \sum_{i=1}^{n} \nabla L_i(w)
   $$  

   Burada:  
   - $ w $ – modelin çəkiləri  
   - $ \eta $ – learning rate  
   - $ \nabla L_i(w) $ – **i-ci nümunənin** gradienti  
   - $ n $ – ümumi nümunə sayı  

3. **İterativ yenilənmə** – Hər yeni nümunə gələndə yalnız onun gradienti yenilənir, digər gradientlər isə olduğu kimi saxlanılır.  

---

### **SAG ilə SGD arasındakı fərq**  
| **Xüsusiyyətlər**      | **SGD**                          | **SAG**                          |
|----------------------|--------------------------------|--------------------------------|
| **Gradient istifadə qaydası** | Yalnız bir nümunənin gradienti istifadə edilir | Əvvəlki gradientlərin orta qiyməti istifadə edilir |
| **Konvergensiya sürəti** | Daha yavaş və qeyri-stabil ola bilər | Daha sürətli və stabil optimallaşma təmin edir |
| **Hesablama yükü** | Azdır, ancaq səs-küylüdür (noisy) | Orta gradientlə daha stabil yenilənmələr aparır |
| **Dəyişikliklərə həssaslıq** | Təsadüfi dəyişikliklərdən çox təsirlənir | Dəyişikliklərə daha az həssasdır |

---

### **SAG-nin Üstünlükləri və Çatışmazlıqları**  
✅ **Üstünlüklər:**  
- **Daha sürətli konvergensiya** – SGD-dən daha stabil və sürətli yaxınlaşır.  
- **Gradientin effektiv istifadəsi** – Keçmiş gradientləri istifadə etdiyi üçün daha dəqiq nəticələr verir.  
- **Kiçik batch-lərlə işləyə bilir** – Tam batch gradient descent istifadə etmədən optimallaşma apara bilir.  

❌ **Çatışmazlıqlar:**  
- **Əlavə yaddaş tələbi** – Hər bir nümunənin gradientini saxlamaq lazımdır.  
- **Dərin öyrənmə üçün o qədər də uyğun deyil** – Çünki dərin neyron şəbəkələrdə qeyri-konveks optimallaşdırma problemləri var.  

---

### **SAG-nin Python-da tətbiqi (Scikit-Learn)**
Scikit-learn kitabxanasında **SAG** "solver" kimi **Logistic Regression** və digər modellərdə istifadə edilə bilər:

```python
from sklearn.linear_model import LogisticRegression

# SAG solver ilə logistik regresiya modeli
model = LogisticRegression(solver='sag', max_iter=1000)

# Modelin öyrədilməsi
model.fit(X_train, y_train)

# Proqnozlaşdırma
y_pred = model.predict(X_test)
```

---

### **SAG və Yeni Metodlar**  
SAG metodunun daha inkişaf etmiş versiyaları da var:  
- **SAGA (Stochastic Average Gradient with Adaptive updates)** – SAG-in bir az daha yaxşı versiyasıdır, qeyri-konveks problemlər üçün də işləyə bilir.  
- **SVRG (Stochastic Variance Reduced Gradient)** – SGD-dəki dəyişkənliyi azaltmaq üçün istifadə edilir.  

---

### **Nəticə**  
SAG metodu **SGD ilə Batch Gradient Descent arasında tarazlıq yaradan** bir yanaşmadır. O, **SGD-dən daha stabil və sürətli** işləyir, ancaq **əlavə yaddaş tələbi** var. Əgər **konveks optimallaşdırma problemi** üzərində işləyirsinizsə və yaddaş problemi yoxdursa, SAG yaxşı seçim ola bilər.

## Subgradient Nədir?

### **Subgradient Nədir?**


**Subgradientin** riyazi tərifi aşağıdakı kimi verilə bilər:

>**Tərif:** Bir funksiya $ f(x) $ konveksdirsə, onda onun hər hansı bir nöqtəsində subgradient $ g $ mövcud olur. Subgradient $ g $, funksiyanın **tangensial düz xəttinin** əvəzedicisidir və funksiyanın aşağıdakı şərti təmin etməlidir:

$$
f(y) \geq f(x) + g^T(y - x), \quad \forall y \in \mathbb{R}^n
$$

Burada:

- $ f(x) $ — funksiyanın qiyməti $ x $- nöqtəsində,
- $ g $ — subgradient vektoru (yəni tangensial düz xəttin əmsalı),
- $ x $ və $ y $ — funksiyanın nöqtələri,
- $ g^T(y - x) $ — vektorların daxili hasilidir, yəni subgradientin nöqtələr arasındakı düz xəttə uyğun olması şərtini təmin edir.

---

### **Subgradientin Əsas Anlayışı:**
Subgradient, funksiyanın **diferensialı** olmadığı halda, funksiyanı **təqribi optimallaşdırmaq** üçün istifadə edilən bir vasitədir. Həmçinin, bir funksiya **konveks** olduğunda onun bir neçə subgradienti mövcud ola bilər. Yəni, eyni nöqtədə bir neçə subgradient ola bilər, ancaq bu subgradientlər yuxarıdakı şərti təmin edirlər.

---

### **Subgradientin Tətbiqi:**
Əgər bir funksiya **konveks** deyil və diferensiallaşmazsa, məsələn, **L1 norması (Lasso cərimələmə)** funksiyasını istifadə edirik, o zaman bu tip funksiyalar üçün **subgradient** istifadə olunur. Subgradient Descent (SGD) metodu ilə bu funksiyaların optimallaşdırılması mümkündür.



---

**Subgradient** — **konveks olmayan funksiyalar** üçün istifadə olunan bir konseptdir və **gradient**-in əvəzidir. Konveks olmayan bir funksiyanın gradienti bütün nöqtələrdə mövcud olmaya bilər, buna görə də subgradient konsepti ortaya çıxır.

---

### **Subgradient haqqında əsas anlayışlar:**

1. **Subgradient, gradientin genişlənmiş bir formasıdır** və konveks olmayan funksiyaların təhlilində istifadə olunur. Konveks olmayan funksiyalar, məsələn, **L1 norması (Lasso cərimələməsi)** və ya **ən yaxşı uyğunlaşdırma problemlərində** yaranan funksiyalar bu kateqoriyaya aiddir.

2. **Subgradient, gradientin bənzəri bir vektordur, lakin hər zaman bir nöqtədə mövcud olmaya bilər.** O, birbaşa gradientin olmasının mümkün olmadığı hallarda funksiya ilə əlaqəli təqribi yanaşma təqdim edir.

3. **Subgradientin riyazi tərifi:**
   - Bir funksiya $ f(x) $ **konveks** olarsa, onda onun gradienti hər zaman mövcuddur.
   - Lakin bir funksiya **konveks deyil**sə, onda hər bir nöqtədə **subgradient** mövcud ola bilər.
   
   **Subgradientin tərifi:**  
   $ g $ vektoru, $ f(x) $ funksiyasının **subgradienti** olub, $ f(x) $ funksiyası ilə əlaqədar aşağıdakı şərtləri təmin etməlidir:

   $$
   f(y) \geq f(x) + g^T(y - x) \quad \text{hər hansı} \, y \, \text{üçün}
   $$

   Burada:
   - $ f(x) $ — funksiyanın qiyməti
   - $ g $ — subgradient vektoru
   - $ x $ və $ y $ — funksiyanın nöqtələri

---

### **Subgradientin İstifadə Olunduğu Hallar:**
Subgradientlər, **non-differentiable (diferensiallaşmayan)** funksiyaların optimallaşdırılmasında istifadə edilir. Bu, əsasən aşağıdakı hallarda olur:
- **Lasso regresiyası** (L1 cərimələmə) — Lasso-nun cərimə funksiyası **diferensiallaşmır**, buna görə də subgradientlər istifadə olunur.
- **SVM-də margin cəriməsi** — Margin tənlikləri çox vaxt **konveks olmayan** olur.
- **Çoxsaylı dəyişənlər arasında əlaqə tapma problemi** — bəzən **konveks olmayan** itki funksiyaları istifadə olunur.

---

### **Subgradientin xüsusiyyətləri:**

- **Subgradientin mövcudluğu:** Subgradient hər bir nöqtədə mövcud olmaya bilər, amma bir çox konveks olmayan funksiyalarda mövcuddur.
- **Çəkilən əmsallar:** Subgradientin dəqiq bir təyini olmadığından, çox vaxt **bir neçə subgradient** mövcud olur.
- **Optimal həllərin tapılması:** Subgradientlər **iterativ optimallaşdırma metodlarında** istifadə edilərək **optimal həllər** tapmaq üçün tətbiq olunur.

---

### **Subgradient ilə Gradient Arasındakı Fərq:**

| **Xüsusiyyət**      | **Gradient** | **Subgradient** |
|---------------------|--------------|-----------------|
| **Tətbiq sahəsi**    | Konveks funksiyalar | Konveks olmayan funksiyalar |
| **Mövcudluq**        | Hər zaman mövcuddur | Bəzi hallarda mövcud olur |
| **Diferensiallaşma** | Funksiya diferensialdır | Funksiya diferensial olmaya bilər |
| **Təqdim etmə**      | Tək bir vektor | Çox sayda subgradient ola bilər |
| **Əsas istifadə sahəsi** | Daha az mürəkkəb problemlər | Daha mürəkkəb və konveks olmayan problemlər |

---

### **Subgradientin tətbiqi:**
**Subgradient Descent (SGD)** — Konveks olmayan funksiyalar üçün **gradient descent** metodunun bir uzantısıdır. Subgradientlər istifadə edərək iterasiya ilə funksiyanı optimallaşdırmaq mümkündür.

---

### **Nümunə:**
Təsəvvür edin ki, aşağıdakı funksiya konveks olmayan bir funksiya olsun:

$$
f(x) = |x|
$$

Bu funksiya **diferensiallaşdırıla bilməz** $ x = 0 $-da, çünki burada kəsik (discontinuity) var. Lakin, **subgradient** $ g $ funksiyasının bu nöqtə üçün qəbul edilə bilən bir çox qiymətə malikdir:  
- $ g = -1 $,  
- $ g = 1 $,  
- $ g = 0 $.

Bu cür çoxsaylı subgradientlər mövcuddur və **sistemi optimallaşdırmaq** üçün istifadə edilə bilər.

---

### **Nəticə:**
**Subgradient** konveks olmayan funksiyaların optimallaşdırılmasında əsas rol oynayır. O, **gradient** metodunun qeyri-differensial funksiyalar üçün alternatividir və xüsusilə **L1 cərimələməsi** kimi problemlərdə istifadə olunur.

## Konveks Funksiya Nədir?

**Konveks funksiyalar** və **konvekslik** anlayışı **optimallaşdırma və analitik riyaziyyat** sahələrində əhəmiyyətli bir mövzudur.

### **Konveks Funksiya Nədir?**

Bir funksiya **konveks** (və ya konveks funksiyadır) əgər onun **qrafiki** hər iki nöqtə arasında olan düz xətt ilə **yuxarıya doğru** yerləşirsə. Yəni, hər hansı iki nöqtə arasında çəkilən düz xətt, funksiya qrafikinin üzərindən keçərək heç vaxt onu aşağıya doğru kəsmir. Bu, funksiya qrafikinin "aşağıya doğru açılan" olması deməkdir.

Riyazi olaraq, bir funksiya $ f(x) $ konveksdirsə, o zaman hər iki nöqtə $ x_1 $ və $ x_2 $ üçün və istənilən $ \lambda \in [0, 1] $ aşağıdakı şərt ödənilir:

$$
f(\lambda x_1 + (1-\lambda) x_2) \leq \lambda f(x_1) + (1-\lambda) f(x_2)
$$

Bu, deməkdir ki, funksiya $ f(x) $ hər iki nöqtə arasında çəkilən düz xəttdən **yuxarıda** və ya **ona paralel** yerləşir. Yəni, qrafikdəki nöqtələr bir-birini keçərkən, aralarındakı xətt funksiyanın yuxarısında qalacaq.

---

### **Konveks Funksiyanın Xüsusiyyətləri:**

1. **Daha Asan Optimallaşdırma:**  
   Konveks funksiyalar **unikal minimum** (və ya maksimum) əldə edir və bu, optimallaşdırma məsələlərini daha asanlaşdırır. Bu xüsusiyyətlər daha tez və effektiv nəticələr əldə etməyə imkan verir.

2. **İkinci Derivativ və Konvekslik:**  
   Əgər funksiya $ f(x) $ iki dəfə diferensialdırsa, və onun ikinci törəməsi (yəni $ f''(x) $) **hər zaman müsbətdirsə**, onda bu funksiya **konveks** olur. Yəni:
   $$
   f''(x) \geq 0 \quad \text{(hər \( x \)-də)}.
   $$

3. **Düzgünlük və Simmetriya:**  
   Konveks funksiyanın qrafiki **düzgün** və **simmetrik** olur. Həmçinin, qrafikin əyriliyi yuxarıya doğru açılır.

4. **Çoxsaylı Minimumlar:**  
   Konveks funksiyalar yalnız bir **minimum** nöqtəsinə sahib olur. Bu, funksiyanın optimallaşdırılmasını asanlaşdırır, çünki yalnız bir həll var.

---

### **Konvekslik və Qeyri-Konvekslik:**
- **Konveks funksiyalar**: Düzgün və bərabər əyriliyə malikdir, yəni **minimumu** tapmaq daha asandır. Məsələn, **kvadrat funksiyalar** (məsələn, $ f(x) = x^2 $) konveksdir.
- **Qeyri-konveks funksiyalar**: Bu funksiyalar, bəzən **bir neçə minimum** (yerinə yetirilməyən) və ya **maksimum** nöqtələri ilə mürəkkəb ola bilər. Məsələn, **Lasso cərimələmə funksiyası** və ya **ən yaxşı uyğunlaşdırma** kimi problemlər qeyri-konveksdir.

---

### **Konveks Funksiyalara Nümunələr:**

1. **Kvadrat Funksiya ($x^2$):**  
   $ f(x) = x^2 $ — Bu funksiyanın qrafiki bir **paraboladır** və çox mükəmməl şəkildə konveksdir.

2. **Exponential Funksiyası ($e^x$):**  
   $ f(x) = e^x $ — Bu funksiya da konveksdir, çünki qrafiki heç vaxt aşağıya doğru əymir.

3. **Lineer Funksiya:**  
   $ f(x) = ax + b $ — Xətti funksiyalar hər zaman konveksdir.

4. **Logaritmik Funksiya ($\ln(x)$):**  
   $ f(x) = \ln(x) $ ($x > 0$) — Bu funksiya konveksdir.

---

### **Konveks Funksiya ilə Qeyri-Konveks Funksiya Arasındakı Fərq:**

| **Xüsusiyyətlər**         | **Konveks Funksiya**                              | **Qeyri-Konveks Funksiya**                           |
|---------------------------|---------------------------------------------------|------------------------------------------------------|
| **Qrafik şəklində**       | Aşağıya doğru açılan, düzgün əyri                  | Çoxsaylı təpələr və ya dərələr, mürəkkəb əyri       |
| **İkinci törəmə**         | İkinci törəməsi müsbət və ya sıfır                | İkinci törəməsi mənfi və ya dəyişkən                |
| **Minimum nöqtəsi**       | Yalnız bir minimum nöqtəsi var                    | Çoxsaylı minimum və ya maksimum nöqtələri ola bilər  |
| **Optimallaşdırma**       | Daha asan, unikal minimum mövcuddur               | Çətin, çoxsaylı minimum və maksimum nöqtələr ola bilər |

---

### **Nəticə:**
Konveks funksiyalar **sadə** və **təmiz** optimallaşdırma problemləri üçün əsasdır. Bu funksiyaların qrafikləri yuxarıya doğru açıldığı üçün, optimallaşdırma prosesində daha asan minimum tapılır. **Qeyri-konveks funksiyalar** isə daha mürəkkəb ola bilər və optimallaşdırma prosesində daha çox çətinliklər yarada bilər.

## Eksponensial Hərəkətli Orta (EMA)

### **Eksponensial Hərəkətli Orta (EMA) Düsturunun Elde Edilməsi – Addım-Addım**  

Eksponensial hərəkətli orta (EMA) **ənənəvi hərəkətli orta (SMA)** ilə müqayisədə daha çevikdir və yeni məlumatlara daha tez reaksiya verir. Onun düsturu aşağıdakı addımlarla formalaşır.

---

## **1. Problem: Adi Hərəkətli Orta (SMA) Problemi**  
Ənənəvi **sürüşən ortalama (SMA)**, **bütün keçmiş məlumatlara eyni çəki** verir. Məsələn, **pəncərə uzunluğu 3 olan SMA** belə hesablanır:

$$
SMA_t = \frac{x_t + x_{t-1} + x_{t-2}}{3}
$$

Burada köhnə və yeni məlumatlar **bərabər çəki** alır, amma yeni məlumatların daha çox əhəmiyyət daşıması lazım ola bilər. Bu problemi həll etmək üçün **eksponensial hərəkətli orta (EMA)** istifadə olunur.

---

## **2. EMA Düsturunun Qurulması**  
> **Əsas prinsip**: **Son müşahidəyə daha çox çəki verib**, **köhnə müşahidələrin çəkisini eksponent şəkildə azaltmaqdır**. 

$$
EMA_t = \alpha \cdot x_t + (1 - \alpha) \cdot EMA_{t-1}
$$

Burada:  
- $ EMA_t $ — cari an üçün eksponensial hərəkətli orta,  
- $ x_t $ — cari an üçün müşahidə edilən real dəyər,  
- $ EMA_{t-1} $ — əvvəlki an üçün EMA dəyəri,  
- $ \alpha $ — düzləşdirmə əmsalı (smoothing factor), $ 0 < \alpha < 1 $ arasında olur.

**Sual:** Düzləşdirmə əmsalı **$ \alpha $** haradan gəlir?  
Bu, **pəncərə uzunluğu $ N $** olan zaman seriyası üçün belə təyin edilir:

$$
\alpha = \frac{2}{N+1}
$$

Burada:  
- $ N $ – nəzərə alınan zaman pəncərəsinin uzunluğudur.  

Bu seçim köhnə məlumatların təsirini eksponent olaraq azaltmağa imkan verir.

---

## **3. Rekursiv Düsturun Açılması**  
EMA **rekursiv** şəkildə hesablanır. Onun addım-addım açılışını edək:

Əvvəlki EMA düsturunu əvəz etsək:

$$
EMA_t = \alpha x_t + (1 - \alpha) (\alpha x_{t-1} + (1 - \alpha) EMA_{t-2})
$$

Daha da açsaq:

$$
EMA_t = \alpha x_t + (1 - \alpha) \alpha x_{t-1} + (1 - \alpha)^2 \alpha x_{t-2} + (1 - \alpha)^3 EMA_{t-3} + \dots
$$

Burada hər bir əvvəlki müşahidənin çəkisi **eksponent olaraq azalır**.

---

## **4. Başlanğıc Şərtin Müəyyən Edilməsi**  
EMA hesablanmağa başlaması üçün **ilk başlanğıc EMA dəyəri** lazımdır. O, adətən ilk $ N $ müşahidənin **sürüşən ortalaması (SMA)** kimi götürülür:

$$
EMA_0 = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

Bu dəyərdən sonra **rekursiv düstur** ilə hesablanır.

---

## **5. Addım-Addım Misal ilə Hesablama**  
Tutaq ki, aşağıdakı zaman seriyası verilmişdir:

| Gün | Qiymət ($ x_t $) |
|----|------|
| 1  | 10   |
| 2  | 12   |
| 3  | 15   |
| 4  | 14   |
| 5  | 13   |

Pəncərə uzunluğu **$ N = 3 $** götürək, onda:

$$
\alpha = \frac{2}{N+1} = \frac{2}{3+1} = 0.5
$$

İlk EMA **SMA kimi** götürülür:

$$
EMA_3 = \frac{10 + 12 + 15}{3} = 12.33
$$

İndi digər EMA-ları hesablayaq:

$$
EMA_4 = 0.5 \cdot 14 + (1 - 0.5) \cdot 12.33 = 0.5 \cdot 14 + 0.5 \cdot 12.33 = 13.165
$$

$$
EMA_5 = 0.5 \cdot 13 + (1 - 0.5) \cdot 13.165 = 13.08
$$

---

## **6. Yekun Düstur**  
Gördüyümüz kimi, **EMA tədricən formalaşır və eksponent olaraq çəkiləri azaldır**. Onun əsas düsturu belədir:

$$
EMA_t = \alpha \cdot x_t + (1 - \alpha) \cdot EMA_{t-1}
$$

və ya açılmış formada:

$$
EMA_t = \sum_{i=0}^{t} \alpha (1 - \alpha)^i x_{t-i}
$$

Bu üsul, **zaman seriyalarında tendensiyaları müəyyən etmək və anomaliyaları süzmək üçün istifadə olunur**.

## Eksponensial Hərəkətli Orta (**EMA**) **hesablamaların həcmini necə azaldır

Eksponensial Hərəkətli Orta (**EMA**) **hesablamaların həcmini necə azaldır?**  

Aydın olması üçün **ənənəvi üsullarla müqayisə edək.**  

---

## **1. Ənənəvi Yanaşma: Gradientlərin Tam Ortalaması**  
Əgər gradientləri hamarlamaq istəsək, **bütün əvvəlki gradientlərin ortasını tapmalı olarıq:**  

$$
v_t = \frac{1}{t} \sum_{i=1}^{t} g_i
$$

🔹 **Bu üsulun problemi:**  
   - Hər yeni gradient gəldikdə **bütün əvvəlki gradientlərin cəmini yenidən hesablamaq lazımdır**.  
   - **O(t) yaddaş və hesablama tələb edir**.  

💡 Yəni, **gradientlərin ortalamasını saxlamaq üçün hər dəfə bütün əvvəlki gradientləri yenidən hesablamaq böyük vaxt və yaddaş sərf edir.**  

---

## **2. EMA ilə Hesablamaların Azaldılması**  
EMA isə **yalnız son dəyəri istifadə edərək yeni dəyəri hesablayır**:  

$$
v_t = \beta v_{t-1} + (1 - \beta) g_t
$$

🔹 **Burada nə baş verir?**  
   - **Keçmiş bütün gradientləri saxlamağa ehtiyac yoxdur**!  
   - **Hər dəfə sadəcə son gradientlə əvvəlki dəyəri yeniləyirik.**  
   - **O(1) (sabit) hesablama və yaddaş tələb edir**.  

💡 **Yəni, əvvəlki üsulda bütün gradientləri saxlamaq lazım idi, EMA-da isə sadəcə bir ədəd saxlanılır və cəmi bir neçə riyazi əməliyyatla yeni dəyər tapılır.**  

---

## **3. EMA SGD-də Ümumi Hesablama Həcmini Necə Azaldır?**
### **✅ Yaddaşdan qənaət edir**  
- Əvvəlki gradientlərin **hamısını yadda saxlamaq əvəzinə** sadəcə bir ədəd saxlayır.  
- **Əvvəlki üsul O(t) yaddaş tələb edirdi, EMA isə O(1) yaddaşla işləyir.**  

### **✅ Daha az hesablamaya ehtiyac var**  
- Hər iterasiyada **bütün keçmiş gradientlərin ortalamasını yenidən hesablamaq əvəzinə, EMA bir toplama və vurma ilə işini bitirir.**  
- **Əvvəlki üsul O(t) hesablama tələb edirdi, EMA isə O(1) hesablama ilə işləyir.**  

---

## **Nəticə: EMA Niyə Hesablamaların Həcmini Azaldır?**  
| Yanaşma | Yaddaş tələbi | Hesablama mürəkkəbliyi |
|---------|--------------|------------------------|
| Ənənəvi ortalama | O(t) | O(t) |
| **EMA** | **O(1)** | **O(1)** |

✅ **EMA ilə yalnız son dəyəri saxlayaraq hesablamaları və yaddaşı ciddi şəkildə azaldırıq.**  
✅ **SGD və optimizasiya prosesində iterasiyaların daha səmərəli və sürətli işləməsini təmin edir.** 🚀

## Momentum (Gradient Descent üçün) və Eksponensial Hərəkətli Orta (EMA) Fərqləri

### **1. Momentum (Gradient Descent üçün) (Düzgün versiya)**  
$$
v_t = \beta v_{t-1} + \nabla J(\theta_t)
$$
$$
\theta_{t+1} = \theta_t - \alpha v_t
$$
**Formulda nələr var:**  
- **$v_t$**: Momentum dəyişəni (yəni, gradientin keçmişdəki təsirini saxlayan dəyişən)  
- **$\beta$**: Momentum faktoru (0 və 1 arasında dəyər alır, keçmiş gradientlərin təsirini nə qədər saxlayacağını müəyyən edir)  
- **$v_{t-1}$**: Keçmiş vaxtdakı momentum dəyişəni  
- **$\nabla J(\theta_t)$**: İndiki vaxtdakı gradient (cost funksiyasının $\theta_t$ parametrləri ilə bağlı gradienti)  
- **$\theta_{t+1}$**: Növbəti parametr qiyməti  
- **$\alpha$**: Learning rate (öyrənmə sürəti, gradientin ölçüsünü tənzimləyir)

### **2. Eksponensial Hərəkətli Orta (EMA) (Gradientlərin düzləşdirilməsi üçün)**  
$$
m_t = \beta m_{t-1} + (1 - \beta) \nabla J(\theta_t)
$$
**Formulda nələr var:**  
- **$m_t$**: Eksponensial hərəkətli ortalama (EMA)  
- **$\beta$**: Eksponensial ağırlıq faktoru (0 və 1 arasında dəyər alır, keçmiş gradientlərin nə qədər təsir edəcəyini müəyyən edir)  
- **$m_{t-1}$**: Keçmiş vaxtdakı eksponensial hərəkətli ortalama  
- **$\nabla J(\theta_t)$**: İndiki vaxtdakı gradient (cost funksiyasının $\theta_t$ parametrləri ilə bağlı gradienti)

---

### **Fərqlər:**  
✅ **Momentum** – **Keçmiş gradientlərin yığılmış təsirini saxlayır və yeniləməyə əlavə edir**. Burada **$1 - \beta$** yoxdur, çünki gradientlərə tam təsir etməsini istəyirik.  

✅ **EMA** – **Gradientlərin orta dəyərini eksponensial çəkili ortalama kimi hesablayır**. Burada **$1 - \beta$** var, çünki yeni gradientin nə qədər təsir edəcəyini təyin edir.  

---

🚀 **Nəticə:**  
- **Momentum** → Gradientin keçmişdəki təsirini saxlayır və **parametrləri yeniləmək üçün istifadə olunur**.  
- **EMA** → Gradientləri **hamarlamaq üçün istifadə olunur**, amma parametrlərə birbaşa təsir etmir.  

Bu səbəbdən **Adam optimizer** hər ikisini birləşdirir.

## Qeyd

✅ **Momentum** — **Gradientlərə tətbiq edilir, lakin parametrləri yeniləmək üçün istifadə olunur**. Yəni, Momentum əvvəlki gradientlərin təsirini saxlayaraq **parametrləri ($\theta$) daha hamar şəkildə dəyişməyə kömək edir**.  

✅ **EMA** — Əsasən **gradientin özünə tətbiq edilir** və **gradientin dəyişkənliyini (variability) azaltmaq** üçün istifadə olunur. Yəni, **gradientləri düzləşdirir** və daha stabil optimallaşdırma verir.  

---

### **Daha sadə izah:**  
- **Momentum** — Gradientlər əsasında **parametrləri daha stabil yeniləmək üçündür**.  
- **EMA** — Gradientlərin özünü düzləşdirərək **onların təsadüfi dəyişmələrini azaldır**.  

🚀 **Adam optimizer** isə **həm Momentum-u, həm də EMA-nı birləşdirir**.  
Momentum birinci momenti (ortalama gradient), EMA isə ikinci momenti (gradientlərin kvadratlarının ortalaması) idarə edir.
