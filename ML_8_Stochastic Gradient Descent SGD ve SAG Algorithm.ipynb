Bir funksiyanƒ± $ f(\theta)$ optimalla≈üdƒ±rmaq √º√ß√ºn **Gradient Descent** metodu bel…ô i≈ül…ôyir:

$$

### **√ñyr…ônm…ô S√ºr…ôti (Learning Rate) - $ \eta $:**

**√ñyr…ônm…ô s√ºr…ôti** ($ \eta $) gradientin …ôks istiqam…ôtind…ô h…ôr…ôk…ôt etm…ôk √º√ß√ºn n…ô q…ôd…ôr addƒ±m atƒ±lacaƒüƒ±nƒ± m√º…ôyy…ôn edir. **Ki√ßik √∂yr…ônm…ô s√ºr…ôti** daha diqq…ôtl…ô minimuma yaxƒ±nla≈üƒ±r, amma proses daha yava≈ü olur. **B√∂y√ºk √∂yr…ônm…ô s√ºr…ôti** is…ô √ßox s√ºr…ôtli ir…ôlil…ôyir, amma √ßox y√ºks…ôk olarsa, minimumu ke√ß…ô bil…ôr.

### **Misal:**

Dey…ôk ki, bir **lineer reqressiya** funksiyasƒ± √ºz…ôrind…ô √ßalƒ±≈üƒ±rƒ±q. Bu funksiya bel…ôdir:

$$

$$

Burada:

- $ x_i $ v…ô $ y_i $ t…ôlim m…ôlumatlarƒ±nƒ±n giri≈ü v…ô √ßƒ±xƒ±≈ü d…ôyi≈ü…ônl…ôridir.

ƒ∞lk ba≈ülanƒüƒ±cda, $ m $ v…ô $ b $ t…ôsad√ºfi se√ßilir. Daha sonra bu parametrl…ôr yenil…ônir. Gradientin hesablanmasƒ± √º√ß√ºn a≈üaƒüƒ±dakƒ± t√∂r…ôm…ôl…ôr istifad…ô olunur:

$$

$$

### **N…ôtic…ô:**

Bu metod **d…ôrin √∂yr…ônm…ô**, **lineer reqressiya**, **logistik reqressiya** v…ô dig…ôr optimalla≈üdƒ±rma m…ôs…ôl…ôl…ôrind…ô geni≈ü t…ôtbiq olunur. **√ñyr…ônm…ô s√ºr…ôti** v…ô **gradientin d√ºzg√ºn hesablanmasƒ±** bu prosesin uƒüurlu olmasƒ±nda m√ºh√ºm rol oynayƒ±r.

## Batch Gradient Descent (BGD) n…ôdir?

### **Batch Gradient Descent (BGD) n…ôdir?**  
**Batch Gradient Descent (BGD)** ‚Äî h…ôr iterasiyada **b√ºt√ºn train datasetin** gradientini hesablayaraq √ß…ôkil…ôri yenil…ôy…ôn optimalla≈üdƒ±rma √ºsuludur. Bu, …ôn…ôn…ôvi gradient descent √ºsuludur v…ô …ôsas…ôn **hamar (smooth) v…ô konveks (convex) funksiyalar** √º√ß√ºn istifad…ô olunur.  

---

### **BGD nec…ô i≈ül…ôyir?**  
1. **B√ºt√ºn n√ºmun…ôl…ôr √ºzr…ô itki funksiyasƒ±nƒ±n gradientini hesablayƒ±r:**  
   $$
   \nabla L(w) = \frac{1}{n} \sum_{i=1}^{n} \nabla L_i(w)
   $$  
   Burada:  
   - $ L(w) $ ‚Äî itki funksiyasƒ±,  
   - $ n $ ‚Äî train datasetd…ôki n√ºmun…ôl…ôrin sayƒ±,  
   - $ \nabla L_i(w) $ ‚Äî h…ôr n√ºmun…ô √º√ß√ºn itkinin gradienti.  

2. **√á…ôkil…ôri (weights) yenil…ôyir:**  
   $$
   w = w - \eta \cdot \nabla L(w)
   $$  
   Burada $ \eta $ ‚Äî learning rate-dir.  

3. **ƒ∞terasiyalar davam edir** v…ô model optimal d…ôy…ôrl…ôr…ô yaxƒ±nla≈üƒ±r.  

---

### **BGD, SGD v…ô SAG m√ºqayis…ôsi**

| **Konvergensiya s√ºr…ôti** | Yava≈ü (daha √ßox iterasiya t…ôl…ôb edir) | S√ºr…ôtli (tez n…ôtic…ô verir, ancaq s…ôs-k√ºyl√ºd√ºr) | S√ºr…ôtli v…ô stabildir |

| **Big Data √º√ß√ºn uyƒüunluq** | Yox (b√ºt√ºn datasetin gradientini hesablamalƒ±dƒ±r) | B…ôli (ki√ßik batch-l…ôrl…ô i≈ül…ôy…ô bilir) | Orta s…ôviyy…ôd…ô (b√ºt√ºn gradientl…ôri saxlamalƒ±dƒ±r) |

### **N…ôtic…ô**  
- **BGD** ‚Äì Stabil, lakin √ßox yava≈ü olduƒüu √º√ß√ºn **√ßox b√∂y√ºk datasetl…ôrd…ô istifad…ô olunmur**.  
- **SGD** ‚Äì Daha **s√ºr…ôtli**, lakin √ßox **s…ôs-k√ºyl√º** (noisy) olduƒüu √º√ß√ºn n…ôtic…ôl…ôr √ßox d…ôyi≈ük…ôn ola bil…ôr.  
- **SAG** ‚Äì **SGD-d…ôn daha stabil v…ô s√ºr…ôtli**, ancaq …ôlav…ô yadda≈ü t…ôl…ôbi var.

## Stochastic Average Gradient (SAG) n…ôdir?

### **Stochastic Average Gradient (SAG) n…ôdir?**  
**Stochastic Average Gradient (SAG)** ‚Äî **Stochastic Gradient Descent (SGD)** v…ô **Batch Gradient Descent** metodlarƒ±nƒ±n birl…ô≈üm…ôsi olan optimalla≈üdƒ±rma √ºsuludur.    $$  

   Burada:  
   - $ w $ ‚Äì modelin √ß…ôkil…ôri  
   - $ \eta $ ‚Äì learning rate  
   - $ \nabla L_i(w) $ ‚Äì **i-ci n√ºmun…ônin** gradienti

## Subgradient N…ôdir?

### **Subgradient N…ôdir?**


**Subgradientin** riyazi t…ôrifi a≈üaƒüƒ±dakƒ± kimi veril…ô bil…ôr:

>**T…ôrif:** Bir funksiya $ f(x) $ konveksdirs…ô, onda onun h…ôr hansƒ± bir n√∂qt…ôsind…ô subgradient $ g $ m√∂vcud olur. Subgradient $ g $, funksiyanƒ±n **tangensial d√ºz x…ôttinin** …ôv…ôzedicisidir v…ô funksiyanƒ±n a≈üaƒüƒ±dakƒ± ≈ü…ôrti t…ômin etm…ôlidir:

$$
f(y) \geq f(x) + g^T(y - x), \quad \forall y \in \mathbb{R}^n
$$

Burada:

- $ f(x) $ ‚Äî funksiyanƒ±n qiym…ôti $ x $- n√∂qt…ôsind…ô,
- $ g $ ‚Äî subgradient vektoru (y…ôni tangensial d√ºz x…ôttin …ômsalƒ±),
- $ x $ v…ô $ y $ ‚Äî funksiyanƒ±n n√∂qt…ôl…ôri,
- $ g^T(y - x) $ ‚Äî vektorlarƒ±n daxili hasilidir, y…ôni subgradientin n√∂qt…ôl…ôr arasƒ±ndakƒ± d√ºz x…ôtt…ô uyƒüun olmasƒ± ≈ü…ôrtini t…ômin edir.

---

### **Subgradientin ∆èsas Anlayƒ±≈üƒ±:**
Subgradient, funksiyanƒ±n **diferensialƒ±** olmadƒ±ƒüƒ± halda, funksiyanƒ± **t…ôqribi optimalla≈üdƒ±rmaq** √º√ß√ºn istifad…ô edil…ôn bir vasit…ôdir. H…ôm√ßinin, bir funksiya **konveks** olduƒüunda onun bir ne√ß…ô subgradienti m√∂vcud ola bil…ôr. Y…ôni, eyni n√∂qt…ôd…ô bir ne√ß…ô subgradient ola bil…ôr, ancaq bu subgradientl…ôr yuxarƒ±dakƒ± ≈ü…ôrti t…ômin edirl…ôr.

---

### **Subgradientin T…ôtbiqi:**
∆èg…ôr bir funksiya **konveks** deyil v…ô diferensialla≈ümazsa, m…ôs…ôl…ôn, **L1 normasƒ± (Lasso c…ôrim…ôl…ôm…ô)** funksiyasƒ±nƒ± istifad…ô edirik, o zaman bu tip funksiyalar √º√ß√ºn **subgradient** istifad…ô olunur. Subgradient Descent (SGD) metodu il…ô bu funksiyalarƒ±n optimalla≈üdƒ±rƒ±lmasƒ± m√ºmk√ºnd√ºr.



---

**Subgradient** ‚Äî **konveks olmayan funksiyalar** √º√ß√ºn istifad…ô olunan bir konseptdir v…ô **gradient**-in …ôv…ôzidir. Konveks olmayan bir funksiyanƒ±n gradienti b√ºt√ºn n√∂qt…ôl…ôrd…ô m√∂vcud olmaya bil…ôr, buna g√∂r…ô d…ô subgradient konsepti ortaya √ßƒ±xƒ±r.

---

### **Subgradient haqqƒ±nda …ôsas anlayƒ±≈ülar:**

1. **Subgradient, gradientin geni≈ül…ônmi≈ü bir formasƒ±dƒ±r** v…ô konveks olmayan funksiyalarƒ±n t…ôhlilind…ô istifad…ô olunur. Konveks olmayan funksiyalar, m…ôs…ôl…ôn, **L1 normasƒ± (Lasso c…ôrim…ôl…ôm…ôsi)** v…ô ya **…ôn yax≈üƒ± uyƒüunla≈üdƒ±rma probleml…ôrind…ô** yaranan funksiyalar bu kateqoriyaya aiddir.

2. **Subgradient, gradientin b…ônz…ôri bir vektordur, lakin h…ôr zaman bir n√∂qt…ôd…ô m√∂vcud olmaya bil…ôr.** O, birba≈üa gradientin olmasƒ±nƒ±n m√ºmk√ºn olmadƒ±ƒüƒ± hallarda funksiya il…ô …ôlaq…ôli t…ôqribi yana≈üma t…ôqdim edir.

## Momentum (Gradient Descent √º√ß√ºn) v…ô Eksponensial H…ôr…ôk…ôtli Orta (EMA) F…ôrql…ôri

### **1. Momentum (Gradient Descent √º√ß√ºn) (D√ºzg√ºn versiya)**  
$$
v_t = \beta v_{t-1} + \nabla J(\theta_t)
$$
$$
\theta_{t+1} = \theta_t - \alpha v_t
$$
**Formulda n…ôl…ôr var:**  
- **$v_t$**: Momentum d…ôyi≈ü…ôni (y…ôni, gradientin ke√ßmi≈üd…ôki t…ôsirini saxlayan d…ôyi≈ü…ôn)  
- **$\beta$**: Momentum faktoru (0 v…ô 1 arasƒ±nda d…ôy…ôr alƒ±r, ke√ßmi≈ü gradientl…ôrin t…ôsirini n…ô q…ôd…ôr saxlayacaƒüƒ±nƒ± m√º…ôyy…ôn edir)  
- **$v_{t-1}$**: Ke√ßmi≈ü vaxtdakƒ± momentum d…ôyi≈ü…ôni  
- **$\nabla J(\theta_t)$**: ƒ∞ndiki vaxtdakƒ± gradient (cost funksiyasƒ±nƒ±n $\theta_t$ parametrl…ôri il…ô baƒülƒ± gradienti)  
- **$\theta_{t+1}$**: N√∂vb…ôti parametr qiym…ôti  
- **$\alpha$**: Learning rate (√∂yr…ônm…ô s√ºr…ôti, gradientin √∂l√ß√ºs√ºn√º t…ônziml…ôyir)

### **2. Eksponensial H…ôr…ôk…ôtli Orta (EMA) (Gradientl…ôrin d√ºzl…ô≈üdirilm…ôsi √º√ß√ºn)**  
$$
m_t = \beta m_{t-1} + (1 - \beta) \nabla J(\theta_t)
$$
**Formulda n…ôl…ôr var:**  
- **$m_t$**: Eksponensial h…ôr…ôk…ôtli ortalama (EMA)  
- **$\beta$**: Eksponensial aƒüƒ±rlƒ±q faktoru (0 v…ô 1 arasƒ±nda d…ôy…ôr alƒ±r, ke√ßmi≈ü gradientl…ôrin n…ô q…ôd…ôr t…ôsir ed…ôc…ôyini m√º…ôyy…ôn edir)  
- **$m_{t-1}$**: Ke√ßmi≈ü vaxtdakƒ± eksponensial h…ôr…ôk…ôtli ortalama  
- **$\nabla J(\theta_t)$**: ƒ∞ndiki vaxtdakƒ± gradient (cost funksiyasƒ±nƒ±n $\theta_t$ parametrl…ôri il…ô baƒülƒ± gradienti)

---

### **F…ôrql…ôr:**  
‚úÖ **Momentum** ‚Äì **Ke√ßmi≈ü gradientl…ôrin yƒ±ƒüƒ±lmƒ±≈ü t…ôsirini saxlayƒ±r v…ô yenil…ôm…ôy…ô …ôlav…ô edir**. Burada **$1 - \beta$** yoxdur, √ß√ºnki gradientl…ôr…ô tam t…ôsir etm…ôsini ist…ôyirik.  

‚úÖ **EMA** ‚Äì **Gradientl…ôrin orta d…ôy…ôrini eksponensial √ß…ôkili ortalama kimi hesablayƒ±r**. Burada **$1 - \beta$** var, √ß√ºnki yeni gradientin n…ô q…ôd…ôr t…ôsir ed…ôc…ôyini t…ôyin edir.  

---

üöÄ **N…ôtic…ô:**  
- **Momentum** ‚Üí Gradientin ke√ßmi≈üd…ôki t…ôsirini saxlayƒ±r v…ô **parametrl…ôri yenil…ôm…ôk √º√ß√ºn istifad…ô olunur**.  
- **EMA** ‚Üí Gradientl…ôri **hamarlamaq √º√ß√ºn istifad…ô olunur**, amma parametrl…ôr…ô birba≈üa t…ôsir etmir.  

Bu s…ôb…ôbd…ôn **Adam optimizer** h…ôr ikisini birl…ô≈üdirir.

## Qeyd

‚úÖ **Momentum** ‚Äî **Gradientl…ôr…ô t…ôtbiq edilir, lakin parametrl…ôri yenil…ôm…ôk √º√ß√ºn istifad…ô olunur**. Y…ôni, Momentum …ôvv…ôlki gradientl…ôrin t…ôsirini saxlayaraq **parametrl…ôri ($\theta$) daha hamar ≈ü…ôkild…ô d…ôyi≈üm…ôy…ô k√∂m…ôk edir**.  

‚úÖ **EMA** ‚Äî ∆èsas…ôn **gradientin √∂z√ºn…ô t…ôtbiq edilir** v…ô **gradientin d…ôyi≈ük…ônliyini (variability) azaltmaq** √º√ß√ºn istifad…ô olunur. Y…ôni, **gradientl…ôri d√ºzl…ô≈üdirir** v…ô daha stabil optimalla≈üdƒ±rma verir.  

---

### **Daha sad…ô izah:**  
- **Momentum** ‚Äî Gradientl…ôr …ôsasƒ±nda **parametrl…ôri daha stabil yenil…ôm…ôk √º√ß√ºnd√ºr**.  
- **EMA** ‚Äî Gradientl…ôrin √∂z√ºn√º d√ºzl…ô≈üdir…ôr…ôk **onlarƒ±n t…ôsad√ºfi d…ôyi≈üm…ôl…ôrini azaldƒ±r**.  

üöÄ **Adam optimizer** is…ô **h…ôm Momentum-u, h…ôm d…ô EMA-nƒ± birl…ô≈üdirir**.  
Momentum birinci momenti (ortalama gradient), EMA is…ô ikinci momenti (gradientl…ôrin kvadratlarƒ±nƒ±n ortalamasƒ±) idar…ô edir.
