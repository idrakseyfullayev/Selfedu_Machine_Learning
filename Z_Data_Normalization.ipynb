## Data Normalization

> **Qeyd:** Sad…ôc…ô **data analitika** il…ô m…ô≈üƒüul olduƒüunda v…ô **model qurmadƒ±qda**, **datanƒ± normal paylanmaya g…ôtirm…ôk** z…ôruri deyil. Amma b…ôzi hallarda **vizual analiz** v…ô ya **statistik t…ôhlill…ôr** √º√ß√ºn normal paylanma **faydalƒ± ola bil…ôr**.

---

## üí° 1. **Normal paylanmaya g…ôtirm…ôk n…ô √º√ß√ºnd√ºr?**

Normal paylanma ‚Äì ortasƒ±nda √ßox, k…ônarlara doƒüru az d…ôy…ôr olan z…ôngvari (bell curve) paylanmadƒ±r.

Model quranda (m…ôs…ôl…ôn, Linear Regression, Logistic Regression, SVM, t-test v…ô s.) ‚Äì bu √ßox vacibdir. Amma sad…ôc…ô data analitika v…ô dashboard √º√ß√ºn a≈üaƒüƒ±dakƒ± s…ôb…ôbl…ôrl…ô faydalƒ± ola bil…ôr:

---

## ‚úÖ 2. **N…ô vaxt normal paylanmaya g…ôtirm…ôk faydalƒ±dƒ±r (analitikada)?**

### üìä a) **Vizualizasiyanƒ±n daha t…ômiz g√∂r√ºnm…ôsi √º√ß√ºn**

M…ôs…ôl…ôn, …ôg…ôr **histogram** v…ô ya **boxplot** √ß…ôkirs…ôns…ô v…ô d…ôyi≈ü…ônin d…ôy…ôrl…ôri √ßox "saƒüdan uzun quyruqludursa" (skewed), bu zaman grafikl…ôr √ßox balanssƒ±z g√∂r√ºn…ô bil…ôr. Log transformasiya v…ô ya ba≈üqa normalizasiya bu disbalansƒ± aradan qaldƒ±ra bil…ôr v…ô **vizual izahƒ± asanla≈üar.**

---

### üìê b) **Statistik t…ôhlill…ôr √º√ß√ºn (z-test, t-test v…ô s.)**

∆èg…ôr **t-test**, **ANOVA**, **correlation** (Pearson) kimi statistik metodlardan istifad…ô edirs…ôns…ô, bu testl…ôrin √ßoxu **normal paylanma f…ôrziyy…ôsin…ô …ôsaslanƒ±r.**

---

### üìà c) **ƒ∞stisna d…ôy…ôrl…ôri (outlier) daha asan m√º…ôyy…ôn etm…ôk √º√ß√ºn**

Outlier-l…ôrin t…ôsirini azaltmaq v…ô onlarƒ± **normal kontekstd…ô** analiz etm…ôk √º√ß√ºn log v…ô ya box-cox transformasiyalarƒ± il…ô datanƒ± normalla≈üdƒ±rmaq faydalƒ± ola bil…ôr.

---

## ‚ùå 3. **N…ô vaxt ehtiyac yoxdur?**

A≈üaƒüƒ±dakƒ± hallarda he√ß bir ≈ü…ôkild…ô datanƒ± normal paylanmaya g…ôtirm…ôy…ô ehtiyac **YOXDUR**:

* Sad…ôc…ô **descriptive statistics** (ortalama, median, min-max, count v…ô s.) aparƒ±rsansa,
* Dashboardda **…ôsas fikir verm…ôk ist…ôdiyin ≈üeyl…ôr sad…ôdir**: satƒ±≈ülarƒ±n trendi, istifad…ô√ßi sayƒ±, qruplara b√∂l√ºnm√º≈ü c…ôml…ôr v…ô s.,
* He√ß bir statistik test v…ô ya klassik ML modeli istifad…ô etmirs…ôns…ô.

---

| Sual                                           | Cavab                                                                                                            |

| Dashboardda g√∂st…ôr…ônd…ô m√ºtl…ôq lazƒ±mdƒ±rmƒ±?      | **Yox.** Dashboard √º√ß√ºn daha √ßox **a√ßƒ±qlƒ±q**, **faydalƒ± info** v…ô **d√ºzg√ºn interpretasiya** vacibdir, forma yox. |

## Datanƒ± **Normal paylanmaya g…ôtirm…ôk** (**"Normalizasiya" v…ô ya "Transformasiya"**)



## üîÑ 1. **Log Transform (Log10, ln)**

∆èn √ßox istifad…ô edil…ôn √ºsuldur. Skewed (qeyri-normal) paylanmanƒ± simmetrik hala g…ôtirir.

### üéØ Uyƒüundur: Saƒü quyruqlu (right-skewed) datalar √º√ß√ºn.

```python
import numpy as np
df['log_x'] = np.log(df['x'] + 1)  # +1 NaN v…ô ya 0-larƒ±n qar≈üƒ±sƒ±nƒ± almaq √º√ß√ºnd√ºr
```

üîπ **Excel-d…ô:**

* **Formula (natural log):** `=LN(A2+1)`
* **Formula (log10):** `=LOG10(A2+1)`

> `+1` sƒ±fƒ±r v…ô m…ônfi d…ôy…ôrl…ôr…ô qar≈üƒ± m√ºdafi…ôdir.

---

## üìà 2. **Square Root Transform**

D…ôy…ôrin k√∂k√ºn√º alƒ±r. Log q…ôd…ôr g√ºcl√º deyil, amma daha "yum≈üaq" d…ôyi≈üiklik verir.

```python
df['sqrt_x'] = np.sqrt(df['x'])
```

üîπ **Excel-d…ô:**

* **Formula:** `=SQRT(A2)`

> `A2` h√ºceyr…ôd…ôki saydƒ±r. M…ônfi d…ôy…ôrl…ôr √º√ß√ºn `#NUM!` s…ôhvi ver…ôc…ôk, …ôvv…ôlc…ô `=IF(A2<0,"",SQRT(A2))` ≈ü…ôklind…ô qoruma t…ôtbiq oluna bil…ôr.

---

## üìâ 3. **Box-Cox Transform**

Yarƒ±-avtomatik √ºsuldur. ∆èn uyƒüun lambda d…ôy…ôrini taparaq **…ôn yax≈üƒ± normallƒ±ƒüƒ±** t…ômin etm…ôy…ô √ßalƒ±≈üƒ±r.

### ≈û…ôrt: `x > 0` olmalƒ±dƒ±r.

```python
from scipy import stats
df['boxcox_x'], _ = stats.boxcox(df['x'])
```

üîπ **Excel-d…ô:**

Excel-d…ô **Box-Cox** avtomatik yoxdur, amma **t…ôxmini t…ôtbiqi** bu d√ºsturla m√ºmk√ºnd√ºr:

```
=IF(LAMBDA=0, LN(A2), (A2^LAMBDA - 1)/LAMBDA)
```

Burada `LAMBDA`‚Äìnƒ± √∂z√ºn tapmalƒ±san (m…ôs. `0.2`).
Alternativ olaraq, **Excel Add-in** v…ô ya **Power Query M dili** il…ô daha avtomatla≈üdƒ±rƒ±la bil…ôr.


---

## üßÆ 4. **Yeo-Johnson Transform**

Box-Cox-un inki≈üaf etmi≈ü versiyasƒ±dƒ±r. M…ônfi v…ô sƒ±fƒ±r d…ôy…ôrl…ôrl…ô d…ô i≈ül…ôy…ô bilir.

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
df['yj_x'] = pt.fit_transform(df[['x']])
```

üîπ **Excel-d…ô:**

Yeo-Johnson transformasiyasƒ± √ßox daha m√ºr…ôkk…ôbdir v…ô Excel-d…ô **birba≈üa d…ôst…ôkl…ônmir**. T…ôtbiq etm…ôk √º√ß√ºn:

1. **Python il…ô Yeo-Johnson d…ôy…ôrl…ôri hesabla**.
2. Excel-…ô **qaytar v…ô daxil et**.


---

## üîÉ 5. **Z-score Normalization (Standardization)**

Datanƒ± ortalama = 0 v…ô std = 1 ed…ôr…ôk normalla≈üdƒ±rƒ±r. Paylanmanƒ± d…ôyi≈üm…ôz, amma statistik testl…ôr √º√ß√ºn istifad…ô edilir.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['zscore_x'] = scaler.fit_transform(df[['x']])
```


üîπ **Excel-d…ô:**

```excel
=(A2 - AVERAGE($A$2:$A$100)) / STDEV.P($A$2:$A$100)
```

> `A2` ‚Äì cari h√ºceyr…ô, `A2:A100` ‚Äì s√ºt√ºnun tam sah…ôsi.

---

## üìä 6. **Min-Max Scaling**

Minimumu 0, maksimumu 1 edir. Paylanma formasƒ±nƒ± d…ôyi≈üm…ôz, sad…ôc…ô miqyaslandƒ±rƒ±r. ML √º√ß√ºn lazƒ±mdƒ±r, amma normal paylanma yaratmƒ±r.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['minmax_x'] = scaler.fit_transform(df[['x']])
```

üîπ **Excel-d…ô:**

```excel
=(A2 - MIN($A$2:$A$100)) / (MAX($A$2:$A$100) - MIN($A$2:$A$100))
```

## **Robust scaling** v…ô ya **Robust transformasiya**

> **Qeyd:**  ‚Äì x√ºsusil…ô **outlier-l…ôrin √ßox olduƒüu hallarda** istifad…ô olunan **dayanƒ±qlƒ± (robust)** normalla≈üdƒ±rma √ºsuludur. 

---

## üõ°Ô∏è **RobustScaler n…ôdir?**

**RobustScaler**, `median` v…ô `IQR` (interquartile range: Q3 ‚àí Q1) istifad…ô edir. Y…ôni:

$$
\text{RobustScaled}(x) = \frac{x - \text{median}}{IQR}
$$

Bu √ºsul **outlier-l…ôr…ô qar≈üƒ± h…ôssas deyil**, √ß√ºnki n…ô ortalama (mean), n…ô d…ô standart sapma (std) istifad…ô olunmur.

---

## üîé **Niy…ô istifad…ô edilir?**

| Klassik scalerl…ôr (StandardScaler, MinMaxScaler) | RobustScaler                  |
| ------------------------------------------------ | ----------------------------- |
| Mean v…ô Std istifad…ô edir                        | Median v…ô IQR istifad…ô edir   |
| Outlier-l…ôrl…ô √ßox h…ôssasdƒ±r                      | Outlier-l…ôr…ô qar≈üƒ± davamlƒ±dƒ±r |
| Normal paylanmanƒ± g√∂zl…ôyir                       | Normal paylanma vacib deyil   |

---

## üß™ **Python n√ºmun…ôsi:**

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
df['robust_x'] = scaler.fit_transform(df[['x']])
```

* **RobustScaler** is…ô median v…ô IQR istifad…ô etdiyin…ô g√∂r…ô `1000`-i **g√∂rm…ôzlikd…ôn g…ôlir**.

## **`RobustScaler`**-ƒ±n  Excel-d…ô t…ôtbiqi



## üß± **RobustScaler n…ôdir?**

`RobustScaler`, **orta** v…ô **standart sapma** …ôv…ôzin…ô:

* **Median** v…ô
* **IQR (interquartile range)** ‚Äì y…ôni `Q3 - Q1`

istifad…ô ed…ôr…ôk **normalizasiya** edir.

---

## üìê **Formulu:**

```text
x_scaled = (x - median) / IQR
```

Burada:

* `median` ‚Äì median d…ôy…ôr (orta yox!)
* `IQR = Q3 - Q1`

  * `Q1` ‚Äì 25-ci percentil
  * `Q3` ‚Äì 75-ci percentil

---

## üìä **Excel-d…ô t…ôtbiqi**

M…ôs…ôl…ôn, s√ºtun `A2:A100` olsun.

### üîπ 1. **Median hesabla:**

```excel
=MEDIAN(A2:A100)
```

### üîπ 2. **Q1 (25%) v…ô Q3 (75%) hesabla:**

```excel
=QUARTILE.INC(A2:A100, 1)   ‚Üí Q1  
=QUARTILE.INC(A2:A100, 3)   ‚Üí Q3
```

> Alternativ olaraq `QUARTILE.EXC` da istifad…ô ed…ô bil…ôrs…ôn.

### üîπ 3. **IQR hesabla:**

```excel
=QUARTILE.INC(A2:A100, 3) - QUARTILE.INC(A2:A100, 1)
```

---

### üîπ 4. **RobustScaler t…ôtbiqi h…ôr h√ºceyr…ôy…ô:**

Dey…ôk ki, bu d…ôy…ôrl…ôri ayrƒ±ca h√ºceyr…ôl…ôr…ô yazdƒ±n:

* `B1` ‚Äì Median
* `B2` ‚Äì IQR

**O zaman A2 √º√ß√ºn:**

```excel
=(A2 - $B$1) / $B$2
```

Bunu a≈üaƒüƒ± doƒüru `fill` ed…ôr…ôk b√ºt√ºn d…ôy…ôrl…ôri skal ed…ô bil…ôrs…ôn.

## Exponential Transformasiya 



## ‚ö° **Exponential Transformasiya n…ôdir?**

**Exponential transformasiya** ‚Äì bir d…ôyi≈ü…ônin h…ôr bir d…ôy…ôrini **eksponent (√ºst) ≈ü…ôklind…ô artƒ±rmaq** dem…ôkdir:

∆èn √ßox istifad…ô olunan variant:

```python
import numpy as np
df['exp_x'] = np.exp(df['x'])
```

## ‚ùì **N…ô vaxt istifad…ô olunur?**

∆èks…ôr hallarda **exponential transformasiya** datanƒ± **√ßox s√ºr…ôtl…ô b√∂y√ºd√ºr** v…ô bu s…ôb…ôbl…ô **normal paylanma yaratmaq √º√ß√ºn istifad…ô olunmur**. Amma b…ôzi hallar var ki:

### üìå 1. **Sol quyruqlu (left-skewed) datanƒ± saƒüa √ß…ôkm…ôk √º√ß√ºn**

## üîë ∆èsas A√ßar N√ºans:

## üß† Z-score n…ô edir?

Z-score formuludur:

$$

$$

* $x$: h…ôr bir m√º≈üahid…ô d…ôy…ôri
* $\mu$: d…ôyi≈ü…ônin orta qiym…ôti
* $\sigma$: standart sapmasƒ±

Bu transformasiya il…ô **h…ôr bir d…ôy…ôr orta qiym…ôtd…ôn n…ô q…ôd…ôr standart sapma uzaqda olduƒüunu** g√∂st…ôrir. Bel…ôlikl…ô:

* Orta: **0** olur
* Standart sapma: **1** olur
* Skewness v…ô ya daƒüƒ±lƒ±m formasƒ± d…ôyi≈ümir

---

## ‚ùóÔ∏èB…ôs niy…ô deyirl…ôr "Z-score il…ô normal paylanmaya g…ôtiririk"?

Bu, bir **yanlƒ±≈ü anlayƒ±≈ü**dan qaynaqlanƒ±r:

| Ger√ß…ôk m…ôqs…ôd                      | Qeyri-d…ôqiq anlayƒ±≈ü                   |
| ---------------------------------- | ------------------------------------- |
| Datanƒ± **standard scale**-…ô salmaq | Datanƒ± **normal paylanmaya √ßevirm…ôk** |

---

## üß™ N…ôtic…ô

* ∆èg…ôr datan **zaten t…ôxmin…ôn normal** paylanƒ±bsa, Z-score onu **daha s…ôliq…ôli v…ô m√ºqayis…ôy…ô hazƒ±r formaya** salƒ±r.
* Amma **skewed** v…ô ya **bimodal** datanƒ± sad…ôc…ô Z-score il…ô **normal etm…ô m√ºmk√ºn deyil**.
* Bu hallarda **log, box-cox, yeo-johnson, power transforms** lazƒ±mdƒ±r.

## üìå Yekun

> üîπ Normal paylanma √º√ß√ºn, …ôg…ôr data skewed-dursa, **transformasiya metodlarƒ±** t…ôtbiq olunmalƒ±dƒ±r (log, root, yeo-johnson v…ô s.).

## Box-Cox n…ôdir



## üì¶ Box-Cox n…ôdir v…ô n…ô edir?

**Box-Cox transformasiyasƒ±** ‚Äì statistikada **right-skewed** (v…ô bir q…ôd…ôr d…ô left-skewed) **paylanmalarƒ± daha normal formaya g…ôtirm…ôk √º√ß√ºn** istifad…ô olunur.

Bu, **log, sqrt, power v…ô s. transformasiyalarƒ±n √ºmumil…ô≈üdirilmi≈ü versiyasƒ±dƒ±r.**
Y…ôni, o avtomatik olaraq hansƒ± transformasiyanƒ±n daha uyƒüun olduƒüunu **√∂z i√ßind…ôki `Œª` (lambda) parametrin…ô …ôsas…ôn** tapƒ±r.

---

## üìê Formulu:

$$
x' =
\begin{cases}
\frac{x^\lambda - 1}{\lambda}, & \text{…ôg…ôr } \lambda \neq 0 \\
\log(x), & \text{…ôg…ôr } \lambda = 0
\end{cases}
$$

Burada:

* $x$: orijinal d…ôy…ôr (yalnƒ±z **m√ºsb…ôt olmalƒ±dƒ±r!** $x > 0$)
* $\lambda$: transformasiyanƒ±n **formasƒ±**nƒ± t…ôyin ed…ôn parametr

---

## üîÅ Œª (lambda) n…ô i≈ü g√∂r√ºr?

| Œª d…ôy…ôri | Effekt           | T…ôyin olunan transformasiya  |
| -------- | ---------------- | ---------------------------- |
| Œª = 1    | D…ôyi≈üiklik etmir | Original data qalƒ±r          |
| Œª = 0    | Log(x) transform | Saƒü …ôyilmi≈ü datalar √º√ß√ºn     |
| Œª = 0.5  | ‚àöx (square root) | Orta …ôyilm…ô                  |
| Œª < 0    | 1/x v…ô s.        | A≈üaƒüƒ±ya …ôyilmi≈ü datalar √º√ß√ºn |

**Sklearn v…ô SciPy** funksiyalar bu Œª d…ôy…ôrini **maksimum normal paylanma yaxƒ±nlƒ±ƒüƒ±na …ôsas…ôn avtomatik tapƒ±r.**

---

## üß† N…ô √º√ß√ºn bu q…ôd…ôr faydalƒ±dƒ±r?

√á√ºnki s…ôn …ôvv…ôlc…ôd…ôn bilmirs…ôn:

* `log(x)` mi yax≈üƒ± i≈ül…ôy…ôc…ôk?
* `sqrt(x)` mi?
* Yoxsa `x^0.3`?

Amma **Box-Cox avtomatik olaraq …ôn uyƒüun Œª** se√ß…ôr…ôk, datanƒ± **…ôn √ßox normal paylanmaya yaxƒ±nla≈üdƒ±ran d√∂n√º≈ü√ºm√º** t…ôtbiq edir.

---

## ‚ö†Ô∏è Qeyd:

* **Yalnƒ±z `x > 0` √º√ß√ºn i≈ül…ôyir.** M…ônfi v…ô sƒ±fƒ±r d…ôy…ôrl…ôrd…ô **`Yeo-Johnson`** istifad…ô edilm…ôlidir.

---

## ‚úÖ Python misalƒ±:

```python
from scipy.stats import boxcox
import numpy as np

data = np.array([1, 2, 3, 4, 5, 10, 50, 100])  # right-skewed
transformed_data, fitted_lambda = boxcox(data)

print("Optimal lambda:", fitted_lambda)
```

## üéØ Yekun

> **Box-Cox** transformasiyasƒ±:

* M√ºxt…ôlif power transformasiyalarƒ± **avtomatik sƒ±nayƒ±r**
* ∆èn uyƒüun `Œª` il…ô **normal paylanmaya yaxƒ±nla≈üdƒ±rƒ±r**
* Saƒü …ôyilmi≈ü datalarda **√ßox effektlidir**
* **Yalnƒ±z m√ºsb…ôt d…ôy…ôrl…ôrd…ô** i≈ül…ôyir


## Yeo-Johnson n…ôdir v…ô n…ô edir



## üìå **Yeo-Johnson n…ôdir v…ô n…ô edir?**

**Yeo-Johnson transformasiyasƒ±**, **Box-Cox** transformasiyasƒ±nƒ±n geni≈ül…ôndirilmƒ±≈ü versiyasƒ±dƒ±r.

> **∆èsas √ºst√ºnl√ºy√º:**
> **M…ônfi v…ô sƒ±fƒ±r d…ôy…ôrl…ôrd…ô d…ô i≈ül…ôyir!**
> (Box-Cox is…ô **yalnƒ±z m√ºsb…ôt d…ôy…ôrl…ôrd…ô** i≈ül…ôyir.)

---

## üìê R…ôsmi formul:

$$
x' =
\begin{cases}
\frac{(x + 1)^\lambda - 1}{\lambda}, & \text{…ôg…ôr } x \geq 0, \lambda \neq 0 \\
\log(x + 1), & \text{…ôg…ôr } x \geq 0, \lambda = 0 \\
\\
- \frac{(-x + 1)^{2 - \lambda} - 1}{2 - \lambda}, & \text{…ôg…ôr } x < 0, \lambda \neq 2 \\
- \log(-x + 1), & \text{…ôg…ôr } x < 0, \lambda = 2
\end{cases}
$$

B…ôli, bir az m√ºr…ôkk…ôb g√∂r√ºn√ºr, amma m…ôqs…ôd sad…ôdir:

---

## üéØ M…ôqs…ôd:

* H…ôm **saƒü**, h…ôm d…ô **sol …ôyilmi≈ü** datalarƒ± **simmetrikl…ô≈üdirm…ôk**,
* Datanƒ± **normal paylanmaya yaxƒ±nla≈üdƒ±rmaq**,
* V…ô bunu **h…ôm m√ºsb…ôt, h…ôm d…ô m…ônfi d…ôy…ôrl…ôrl…ô i≈ül…ôy…ôr…ôk** etm…ôk.

---

## ‚úÖ √úst√ºnl√ºkl…ôri:

| √ñz…ôllik           | Yeo-Johnson | Box-Cox |
| ----------------- | ----------- | ------- |
| M√ºsb…ôt d…ôy…ôrl…ôr   | ‚úÖ           | ‚úÖ       |
| M…ônfi d…ôy…ôrl…ôr    | ‚úÖ           | ‚ùå       |
| Sƒ±fƒ±r d…ôy…ôrl…ôr    | ‚úÖ           | ‚ùå       |
| Œª optimalla≈üdƒ±rma | ‚úÖ           | ‚úÖ       |

---

## üí° Praktik misal (Python):

```python
from sklearn.preprocessing import PowerTransformer
import numpy as np

data = np.array([[-100], [-5], [0], [1], [2], [10], [50]])

# Yeo-Johnson transform
pt = PowerTransformer(method='yeo-johnson', standardize=False)
transformed = pt.fit_transform(data)

print("Optimal lambda:", pt.lambdas_)
```
